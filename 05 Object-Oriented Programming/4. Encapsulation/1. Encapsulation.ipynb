# 🔒 Encapsulation in Python

**Encapsulation** is one of the four fundamental principles of **Object-Oriented Programming (OOP)**. It is the process of **combining data (variables) and the methods (functions) that operate on that data into a single unit called a class**, while also controlling how the data is accessed from outside the class.

Encapsulation helps protect an object's internal state by restricting direct access to its data and allowing controlled access through methods.

In Python, encapsulation is achieved using:

- Public members
- Protected members
- Private members
- Getter and setter methods (or the `@property` decorator)

# 📖 Topics Covered

## 1. What is Encapsulation?

**Encapsulation** is the process of:

- Combining data and methods into one class.
- Restricting direct access to object data.
- Providing controlled access through methods.

Instead of exposing variables directly, encapsulation encourages interaction through methods.

Example:

```python
class Student:

    def __init__(self):
        self.__name = "Alice"

    def get_name(self):
        return self.__name
```

Here:

- `__name` stores the data.
- `get_name()` provides controlled access.

---

## 2. Why Use Encapsulation?

Encapsulation helps us:

- Protect object data.
- Hide implementation details.
- Prevent accidental modification.
- Improve code security.
- Increase maintainability.
- Improve modularity.
- Make code easier to modify and extend.

---

## 3. How Encapsulation Works

Encapsulation combines:

- Variables (data)
- Methods (behavior)

inside a single class.

```python
class Employee:

    def __init__(self):
        self.salary = 50000

    def display_salary(self):
        print(self.salary)
```

The class keeps related data and operations together.

---

## 4. Data Hiding

One of the main goals of encapsulation is **data hiding**.

Instead of allowing direct access:

```python
employee.salary = -5000
```

We can control updates using methods.

```python
class Employee:

    def __init__(self):
        self.__salary = 50000

    def get_salary(self):
        return self.__salary
```

This prevents unwanted modification of internal data.

---

## 5. Controlled Access

Encapsulation allows data to be accessed in a controlled manner.

```python
class BankAccount:

    def __init__(self):
        self.__balance = 1000

    def get_balance(self):
        return self.__balance

account = BankAccount()

print(account.get_balance())
```

### Output

```text
1000
```

---

## 6. Encapsulation Using Methods

Getter methods retrieve data.

Setter methods modify data after validation.

```python
class Student:

    def __init__(self):
        self.__age = 20

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age

student = Student()

student.set_age(22)

print(student.get_age())
```

### Output

```text
22
```

## 8. Relationship with Access Modifiers

Python supports three levels of member access.

| Member Type | Naming Convention | Accessibility |
|--------------|------------------|---------------|
| Public | `name` | Accessible everywhere |
| Protected | `_name` | Intended for internal use and subclasses |
| Private | `__name` | Name-mangled to discourage direct access |

These access levels help implement encapsulation.

### Example 4: Product Price Validation

### Example 5: Temperature Validation

```python
class Temperature:

    def __init__(self):
        self.__celsius = 25

    def set_temperature(self, value):
        if value >= -273.15:
            self.__celsius = value

    def get_temperature(self):
        return self.__celsius

temperature = Temperature()

temperature.set_temperature(30)

print(temperature.get_temperature())
```

### Output

```text
30
```

## 11. Common Mistakes

Avoid these common mistakes while implementing encapsulation.

### Accessing Private Members Directly

Incorrect

```python
class Student:

    def __init__(self):
        self.__name = "Alice"

student = Student()

print(student.__name)
```

This raises an `AttributeError` because private members are name-mangled.

Correct

```python
print(student.get_name())
```

---

### Not Validating Input

Incorrect

```python
class Employee:

    def __init__(self):
        self.__salary = 50000

    def set_salary(self, salary):
        self.__salary = salary
```

This allows invalid values.

Better

```python
def set_salary(self, salary):

    if salary >= 0:
        self.__salary = salary
```

---

### Exposing Sensitive Data

Avoid making sensitive data public when it should be protected.

Use private members with controlled access methods instead.

---

### Ignoring Encapsulation

Instead of modifying internal data directly, interact through methods that enforce validation and business rules.

---

## 12. Best Practices

Follow these recommended practices when implementing encapsulation.

- Keep related data and methods together.
- Hide sensitive data using private members when appropriate.
- Validate input before updating object data.
- Provide getter and setter methods when controlled access is required.
- Keep implementation details hidden from users of the class.
- Expose only the functionality that users need.
- Use encapsulation to improve maintainability and flexibility.
