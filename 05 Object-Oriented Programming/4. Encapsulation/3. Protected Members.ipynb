# 🛡️ Protected Members in Python

**Protected members** are class members that are **intended to be accessed only within the class and its subclasses**. In Python, protected members are created by prefixing their names with a **single underscore (`_`)**.

Unlike some other programming languages, Python **does not strictly enforce protected access**. Instead, the single underscore serves as a **convention**, indicating that the member is intended for internal use and should not be accessed directly from outside the class.

Protected members help developers communicate that certain data or methods are meant to be used only by the class and its derived classes.

# 📖 Topics Covered

## 1. What are Protected Members?

A **protected member** is a variable or method whose name begins with a **single underscore (`_`)**.

Example:

```python
class Student:

    def __init__(self):
        self._name = "Alice"
```

Here,

- `_name` is a protected instance variable.

It indicates that the variable is intended for internal use.

---

## 2. Why Use Protected Members?

Protected members help us:

- Indicate internal implementation details.
- Support inheritance.
- Reduce accidental misuse.
- Improve code readability.
- Follow Python coding conventions.

Unlike private members, protected members can still be accessed if necessary.

---

## 3. Protected Variables

Protected variables are created using a single underscore.

```python
class Employee:

    def __init__(self):
        self._salary = 50000
```

Here,

- `_salary` is a protected variable.

---

## 4. Protected Methods

Methods can also be protected.

```python
class Student:

    def _display(self):
        print("Protected Method")
```

Although accessible, this method is intended for internal use.

---

## 5. Syntax

### Protected Variable

```python
class ClassName:

    def __init__(self):
        self._variable = value
```

---

### Protected Method

```python
class ClassName:

    def _method(self):
        # Method body
```

---

## 6. Accessing Protected Members

Python allows protected members to be accessed from outside the class, but this is discouraged.

```python
class Student:

    def __init__(self):
        self._name = "Alice"

student = Student()

print(student._name)
```

### Output

```text
Alice
```

Although this works, it is considered poor practice because `_name` is intended for internal use.

---

## 7. Protected Members in Inheritance

Protected members are mainly intended for subclasses.

```python
class Person:

    def __init__(self):
        self._name = "Alice"

class Student(Person):

    def display(self):
        print(self._name)

student = Student()

student.display()
```

### Output

```text
Alice
```

The subclass can access the protected variable directly.

---

## 8. Protected Methods in Inheritance

```python
class Animal:

    def _sound(self):
        print("Some Sound")

class Dog(Animal):

    def bark(self):
        self._sound()

dog = Dog()

dog.bark()
```

### Output

```text
Some Sound
```

Protected methods are commonly used to support inherited behavior.

---

## 9. Protected Members vs Public vs Private

| Member Type | Naming Convention | Accessibility |
|--------------|------------------|---------------|
| Public | `name` | Accessible everywhere |
| Protected | `_name` | Intended for internal use and subclasses |
| Private | `__name` | Name-mangled to discourage direct access |

Protected members rely on developer discipline rather than strict language enforcement.

## 10. Practical Examples

### Example 1: Student Information

```python
class Student:

    def __init__(self):
        self._name = "David"

    def display(self):
        print(self._name)

student = Student()

student.display()
```

### Example 2: Employee Salary

```python
class Employee:

    def __init__(self):
        self._salary = 45000

    def show_salary(self):
        print(self._salary)

employee = Employee()

employee.show_salary()
```

### Example 3: Vehicle Inheritance

```python
class Vehicle:

    def __init__(self):
        self._wheels = 4

class Car(Vehicle):

    def show(self):
        print(self._wheels)

car = Car()

car.show()
```

### Output

```text
4
```

## 12. Common Mistakes

Avoid these common mistakes while working with protected members.

### Assuming Protected Members Are Truly Private

Incorrect assumption:

```python
class Student:

    def __init__(self):
        self._name = "Alice"

student = Student()

print(student._name)
```

This code works because Python does **not** enforce protected access.

The underscore is only a convention.

---

### Using Protected Members as Public APIs

Avoid exposing protected members as part of your class's public interface.

Instead, provide public methods when external access is required.

---

### Ignoring the Naming Convention

Incorrect

```python
self.name
```

Protected variable

```python
self._name
```

The underscore communicates the intended usage to other developers.

---

### Modifying Protected Members Outside the Class

Although possible, directly changing protected members from outside the class can make code harder to maintain.

Prefer using public methods when modifications are needed.

---

## 13. Best Practices

Follow these recommended practices when using protected members.

- Use a single underscore for members intended for internal use.
- Access protected members primarily within the class and its subclasses.
- Do not rely on protected members as part of the public interface.
- Use public methods to expose necessary functionality.
- Keep implementation details hidden where appropriate.
- Follow Python naming conventions consistently.
- Use protected members to support inheritance and code reuse.
