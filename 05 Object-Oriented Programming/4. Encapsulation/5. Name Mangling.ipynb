# 🔑 Name Mangling in Python

**Name mangling** is a mechanism used by Python to **reduce accidental access to private members** of a class. When a variable or method is declared with **two leading underscores (`__`)**, Python automatically changes (or *mangles*) its name internally.

Name mangling is **not true access restriction**. Instead, it changes the member's name to make accidental access from outside the class less likely and to avoid naming conflicts in inheritance.

# 📖 Topics Covered

## 1. What is Name Mangling?

**Name mangling** is Python's process of automatically renaming private members.

When a member starts with **two leading underscores (`__`)**, Python internally changes its name.

Example:

```python
class Student:

    def __init__(self):
        self.__name = "Alice"
```

Internally, Python stores it as:

```python
_Student__name
```

This helps prevent accidental access and naming conflicts.

---

## 2. Why Does Python Use Name Mangling?

Python uses name mangling to:

- Reduce accidental access to private members.
- Avoid naming conflicts in inheritance.
- Support encapsulation.
- Protect internal implementation details.
- Encourage proper object-oriented design.

It is important to remember that name mangling **does not provide complete security**.

---

## 3. How Name Mangling Works

Python automatically converts:

```python
__variable
```

into

```python
_ClassName__variable
```

Example:

```python
class Employee:

    def __init__(self):
        self.__salary = 50000
```

Internally becomes:

```python
_Employee__salary
```

---

## 4. Accessing a Mangled Variable

Although private members cannot normally be accessed directly,

```python
class Student:

    def __init__(self):
        self.__name = "Alice"

student = Student()

print(student.__name)
```

### Output

```text
AttributeError
```

they can still be accessed using the mangled name.

```python
print(student._Student__name)
```

### Output

```text
Alice
```

This bypasses the intended privacy and should generally be avoided.

---

## 5. Name Mangling for Methods

Private methods are also mangled.

```python
class Calculator:

    def __add(self):
        print("Private Method")

calculator = Calculator()

calculator._Calculator__add()
```

### Output

```text
Private Method
```

---

## 6. Name Mangling in Inheritance

One of the primary reasons for name mangling is to avoid conflicts between parent and child classes.

```python
class Parent:

    def __init__(self):
        self.__value = 10

class Child(Parent):

    def __init__(self):
        super().__init__()
        self.__value = 20

child = Child()

print(child._Parent__value)
print(child._Child__value)
```

### Output

```text
10
20
```

Both variables exist independently because Python mangles their names differently.

---

## 7. Viewing Mangled Names

The `__dict__` attribute shows the actual attributes stored in an object.

```python
class Student:

    def __init__(self):
        self.__name = "Alice"

student = Student()

print(student.__dict__)
```

### Output

```text
{'_Student__name': 'Alice'}
```

Notice that `__name` has been converted to `_Student__name`.

---

## 8. Name Mangling is Not True Privacy

Unlike some programming languages, Python does **not** completely hide private members.

Example:

```python
class Test:

    def __init__(self):
        self.__number = 100

obj = Test()

print(obj._Test__number)
```

### Output

```text
100
```

This demonstrates that name mangling discourages direct access but does not prevent it.

## 11. Common Mistakes

Avoid these common mistakes while working with name mangling.

### Assuming Name Mangling Provides Complete Security

Incorrect assumption:

```python
self.__password
```

Private members can still be accessed using their mangled names.

Name mangling discourages access—it does not make data completely inaccessible.

---

### Accessing Mangled Names in Normal Code

```python
student._Student__name
```

Although valid, this should generally be avoided.

Prefer public methods or properties instead.

---

### Forgetting the Class Name

Incorrect

```python
student.__Student__name
```

Correct

```python
student._Student__name
```

The class name becomes part of the mangled name.

---

### Using Double Underscores Unnecessarily

Not every variable needs to be private.

Use:

- Public members for public data.
- Protected members for subclass access.
- Private members only for internal implementation details.

---

## 12. Best Practices

Follow these recommended practices when using name mangling.

- Use double underscores only when true internal implementation hiding is desired.
- Access private members through public methods or properties.
- Avoid directly using mangled names in application code.
- Use name mangling to prevent naming conflicts in inheritance.
- Follow Python naming conventions consistently.
- Keep implementation details hidden from users of your class.
- Choose the appropriate access level (public, protected, or private) based on your design requirements.
