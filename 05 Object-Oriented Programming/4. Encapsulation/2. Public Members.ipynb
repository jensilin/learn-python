# 🌐 Public Members in Python

**Public members** are the **default members** of a class in Python. They can be **accessed and modified from anywhere**, including:

- Inside the class
- Outside the class
- From subclasses
- From other modules (if imported)

Public members include both **variables (attributes)** and **methods**.

They are suitable for data and functionality that should be freely accessible.

# 📖 Topics Covered

## 1. What are Public Members?

A **public member** is a variable or method that can be accessed from anywhere.

In Python, members are **public by default**.

Example:

```python
class Student:

    def __init__(self):
        self.name = "Alice"
```

Here,

- `name` is a public instance variable.

Anyone can access it using an object.

---

## 2. Why Use Public Members?

Public members help us:

- Share information freely.
- Provide access to object data.
- Make objects easy to use.
- Improve code readability.
- Expose functionality intended for external use.

---

## 3. Public Variables

Public variables are created without using underscores.

```python
class Employee:

    def __init__(self):
        self.name = "John"
        self.salary = 50000
```

Both `name` and `salary` are public variables.

---

## 4. Public Methods

Public methods can also be accessed from anywhere.

```python
class Student:

    def greet(self):
        print("Welcome!")
```

Calling the method:

```python
student = Student()

student.greet()
```

### Output

```text
Welcome!
```

---

## 6. Accessing Public Members

Public members are accessed using the dot (`.`) operator.

```python
class Student:

    def __init__(self):
        self.name = "Alice"

student = Student()

print(student.name)
```

### Output

```text
Alice
```

---

## 7. Modifying Public Members

Public variables can be modified directly.

```python
class Student:

    def __init__(self):
        self.name = "Alice"

student = Student()

student.name = "Bob"

print(student.name)
```

### Output

```text
Bob
```

---

## 8. Public Members in Multiple Objects

Each object has its own copy of public instance variables.

```python
class Student:

    def __init__(self, name):
        self.name = name

student1 = Student("Alice")
student2 = Student("Bob")

print(student1.name)
print(student2.name)
```

### Output

```text
Alice
Bob
```

---

## 9. Public Members vs Protected vs Private

| Member Type | Naming Convention | Accessibility |
|--------------|------------------|---------------|
| Public | `name` | Accessible everywhere |
| Protected | `_name` | Intended for internal use and subclasses |
| Private | `__name` | Name-mangled to discourage direct access |

Public members have no leading underscores.

## 12. Common Mistakes

Avoid these common mistakes while working with public members.

### Exposing Sensitive Data

Incorrect

```python
class BankAccount:

    def __init__(self):
        self.password = "12345"
```

Sensitive information should not usually be stored as public members.

Instead, use private members with controlled access.

---

### Modifying Public Variables Without Validation

Incorrect

```python
class Product:

    def __init__(self):
        self.price = 100

product = Product()

product.price = -500
```

This may lead to invalid object states.

Use setter methods or properties if validation is required.

---

### Assuming Public Variables Are Protected

```python
class Student:

    def __init__(self):
        self.name = "Alice"
```

Anyone can freely access and modify `name`.

---

### Using Public Members for Internal Implementation

If a variable should only be used internally, consider making it protected (`_variable`) or private (`__variable`).

---

## 13. Best Practices

Follow these recommended practices when using public members.

- Use public members only for data intended to be externally accessible.
- Keep sensitive information private.
- Validate important data before modification.
- Use descriptive names for public variables and methods.
- Expose only the functionality users need.
- Keep public interfaces simple and easy to understand.
- Use encapsulation to hide implementation details when necessary.
