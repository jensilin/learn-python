# 🔐 Private Members in Python

**Private members** are class members that are **intended to be accessed only within the class in which they are defined**. In Python, private members are created by prefixing their names with **double underscores (`__`)**.

Python uses a mechanism called **name mangling** to make private members more difficult to access from outside the class. This helps protect an object's internal data and supports the principle of **encapsulation**.

Private members include both **variables (attributes)** and **methods**.

# 📖 Topics Covered

## 1. What are Private Members?

A **private member** is a variable or method whose name begins with **double underscores (`__`)**.

Example:

```python
class Student:

    def __init__(self):
        self.__name = "Alice"
```

Here,

- `__name` is a private instance variable.

It is intended to be accessed only inside the `Student` class.

---

## 2. Why Use Private Members?

Private members help us:

- Hide sensitive data.
- Protect object state.
- Prevent accidental modification.
- Support encapsulation.
- Improve maintainability.
- Reduce unintended dependencies.

---

## 3. Private Variables

Private variables are created using two leading underscores.

```python
class Employee:

    def __init__(self):
        self.__salary = 50000
```

Here,

- `__salary` is a private variable.

---

## 4. Private Methods

Methods can also be private.

```python
class Student:

    def __display(self):
        print("Private Method")

    def show(self):
        self.__display()

student = Student()

student.show()
```

### Output

```text
Private Method
```

Private methods are generally used as helper methods inside the class.

---

## 6. Accessing Private Members

Private members are designed to be accessed from within the class.

```python
class Student:

    def __init__(self):
        self.__name = "Alice"

    def display(self):
        print(self.__name)

student = Student()

student.display()
```

### Output

```text
Alice
```

Trying to access a private member directly from outside the class results in an error.

```python
print(student.__name)
```

### Output

```text
AttributeError
```

---

## 7. Name Mangling

Python protects private members using **name mangling**.

A member named:

```python
__name
```

is internally renamed to:

```python
_ClassName__name
```

Example:

```python
class Student:

    def __init__(self):
        self.__name = "Alice"

student = Student()

print(student._Student__name)
```

### Output

```text
Alice
```

Although this works, it bypasses the intended privacy and should generally be avoided.

---

## 8. Private Members in Inheritance

Private members are **not directly inherited** because of name mangling.

```python
class Person:

    def __init__(self):
        self.__name = "Alice"

class Student(Person):

    def display(self):
        print("Student")

student = Student()

student.display()
```

The subclass cannot directly access `__name`.

---

## 9. Private Methods Inside the Class

Private methods are commonly used for internal processing.

```python
class Calculator:

    def __add(self, a, b):
        return a + b

    def calculate(self):
        print(self.__add(10, 20))

calculator = Calculator()

calculator.calculate()
```

### Output

```text
30
```

---

## 10. Private Members vs Protected vs Public

| Member Type | Naming Convention | Accessibility |
|--------------|------------------|---------------|
| Public | `name` | Accessible everywhere |
| Protected | `_name` | Intended for internal use and subclasses |
| Private | `__name` | Name-mangled to discourage direct access |

## 11. Practical Examples

### Example 1: Student Information

```python
class Student:

    def __init__(self):
        self.__name = "David"

    def get_name(self):
        return self.__name

student = Student()

print(student.get_name())
```

### Output

```text
David
```

---

### Example 2: Bank Account

```python
class BankAccount:

    def __init__(self):
        self.__balance = 5000

    def show_balance(self):
        print(self.__balance)

account = BankAccount()

account.show_balance()
```

### Output

```text
5000
```

---

### Example 3: Employee Salary

```python
class Employee:

    def __init__(self):
        self.__salary = 45000

    def get_salary(self):
        return self.__salary

employee = Employee()

print(employee.get_salary())
```

### Example 4: Private Helper Method

```python
class Greeting:

    def __welcome(self):
        print("Welcome!")

    def display(self):
        self.__welcome()

greeting = Greeting()

greeting.display()
```

### Output

```text
Welcome!
```

### Example 5: Product Price

### Output

```text
100
```

## 13. Common Mistakes

Avoid these common mistakes while working with private members.

### Accessing Private Members Directly

Incorrect

```python
class Student:

    def __init__(self):
        self.__name = "Alice"

student = Student()

print(student.__name)
```

### Output

```text
AttributeError
```

Correct

```python
print(student.get_name())
```

---

### Using Name Mangling to Access Private Members

```python
print(student._Student__name)
```

Although this works, it bypasses encapsulation and should only be used in exceptional situations such as debugging.

---

### Forgetting Double Underscores

Incorrect

```python
self_name = "Alice"
```

or

```python
self._name = "Alice"
```

Private members require **two leading underscores**.

Correct

```python
self.__name = "Alice"
```

---

### Using Private Members When Protected Members Are More Appropriate

If subclasses need access to a member, consider using a **protected member (`_name`)** instead of a private member.

---

## 14. Best Practices

Follow these recommended practices when using private members.

- Use private members for sensitive or internal data.
- Access private data through public methods or properties.
- Avoid accessing private members using name mangling.
- Keep helper methods private when they are not part of the public interface.
- Use meaningful names for private members.
- Choose protected members if subclasses require access.
- Use encapsulation to protect important object data.
