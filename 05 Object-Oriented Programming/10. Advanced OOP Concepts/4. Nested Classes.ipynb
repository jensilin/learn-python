# 🏛️ Nested Classes in Python

A **nested class** is a class that is **defined inside another class**.

The class defined inside another class is called the **inner class**, while the enclosing class is called the **outer class**.

Nested classes are used to organize related classes together and indicate that the inner class is closely associated with the outer class.

They help improve:

- Code organization
- Readability
- Encapsulation
- Modularity
- Maintainability

Although nested classes are supported in Python, they are used less frequently than regular classes and are typically reserved for situations where an inner class is meaningful only within the context of its outer class.

# 📖 Topics Covered

## 1. What is a Nested Class?

A nested class is simply a class declared inside another class.

Example:

```python
class University:

    class Student:

        def display(self):
            return "Student Information"

student = University.Student()

print(student.display())
```

### Output

```text
Student Information
```

---

## 2. Why Use Nested Classes?

Nested classes help us:

- Group related classes together.
- Improve code organization.
- Reduce namespace pollution.
- Improve readability.
- Represent logical relationships.
- Hide implementation details.

---

## 3. Syntax

```python
class OuterClass:

    class InnerClass:

        # Methods and attributes
        pass
```

Creating an object:

```python
inner = OuterClass.InnerClass()
```

---

## 4. Accessing a Nested Class

The inner class is accessed using the outer class name.

```python
class Company:

    class Employee:

        def info(self):
            return "Employee Details"

employee = Company.Employee()

print(employee.info())
```

### Output

```text
Employee Details
```

---

## 5. Outer and Inner Classes

The outer class can create objects of the inner class.

```python
class Car:

    class Engine:

        def start(self):
            return "Engine Started"

    def __init__(self):
        self.engine = Car.Engine()

car = Car()

print(car.engine.start())
```

### Output

```text
Engine Started
```

---

## 6. Nested Class with Attributes

Nested classes can have their own attributes and methods.

```python
class Laptop:

    class Processor:

        def __init__(self, brand):
            self.brand = brand

processor = Laptop.Processor("Intel")

print(processor.brand)
```

### Output

```text
Intel
```

## 10. Limitations of Nested Classes

Nested classes also have some limitations.

- They are less commonly used in Python.
- They do not automatically access the outer class instance.
- They can make code harder to understand if overused.
- Deeply nested classes reduce readability.

## 11. Nested Classes vs Regular Classes

## 13. Common Mistakes

Avoid these common mistakes while using nested classes.

### Assuming the Inner Class Can Access the Outer Instance Automatically

Incorrect

```python
class Outer:

    value = 10

    class Inner:

        def show(self):
            print(value)
```

The inner class does **not** automatically have access to the outer class instance.

---

### Using Nested Classes Unnecessarily

Do not nest classes simply because they are related.

Only use nested classes when the inner class has meaning only within the outer class.

---

### Creating Deeply Nested Classes

Avoid structures like:

```python
class A:

    class B:

        class C:

            class D:
                pass
```

Too much nesting makes code difficult to read and maintain.

---

### Confusing Nested Classes with Inheritance

A nested class is **not** a subclass of the outer class.

The following relationship is incorrect:

```
Engine is a Car
```

Correct relationship:

```
Engine is defined inside Car.
```

---

## 14. Best Practices

Follow these recommended practices when using nested classes.

- Use nested classes only when there is a strong logical relationship.
- Keep nested classes small and focused.
- Avoid excessive levels of nesting.
- Use descriptive class names.
- Document the purpose of the inner class.
- Prefer regular classes if the inner class will be reused elsewhere.
- Keep responsibilities separated.

> **Note:** A **nested class** is a class defined inside another class to logically group closely related functionality. It improves code organization and encapsulation by keeping associated classes together. Unlike inheritance or composition, nesting does not imply an **"is-a"** or **"has-a"** relationship—it simply indicates that the inner class is conceptually tied to the outer class. Nested classes should be used thoughtfully, as excessive nesting can reduce code readability and maintainability.
