# 🤝 Aggregation in Python

**Aggregation** is an object-oriented programming (OOP) concept where one class **uses an object of another class**, but **both objects can exist independently**.

In simple terms:

> **Aggregation represents a weak "has-a" relationship.**

Unlike composition, the contained object is **not owned** by the container. If the container object is destroyed, the contained object can continue to exist.

Aggregation promotes:

- Code reusability
- Loose coupling
- Flexibility
- Better maintainability
- Independent object lifecycles

It is widely used when objects need to collaborate without depending on each other's existence.

# 💻 Quick Example

```python
class Teacher:

    def __init__(self, name):
        self.name = name

class Department:

    def __init__(self, teacher):
        self.teacher = teacher

teacher = Teacher("John")

department = Department(teacher)

print(department.teacher.name)
```

### Output

```text
John
```

The `Teacher` object exists independently and is simply used by the `Department`.

# 📖 Topics Covered

## 1. What is Aggregation?

Aggregation is an OOP relationship where one class refers to another class without owning its lifetime.

The contained object is created separately and passed into another object.

Example:

```python
class Student:

    def __init__(self, name):
        self.name = name

class Classroom:

    def __init__(self, student):
        self.student = student

student = Student("Alice")

classroom = Classroom(student)

print(classroom.student.name)
```

### Output

```text
Alice
```

---

## 2. Why Use Aggregation?

Aggregation helps us:

- Reuse existing objects.
- Reduce dependencies.
- Improve flexibility.
- Keep classes independent.
- Simplify maintenance.
- Promote loose coupling.

---

## 3. Understanding the Weak "Has-a" Relationship

Aggregation represents a relationship where one object uses another without owning it.

Examples:

- Department **has a** Teacher.
- Team **has** Players.
- Company **has** Employees.
- Library **has** Members.
- School **has** Teachers.

If the department is removed, the teacher still exists.

---

## 4. How Aggregation Works

The dependent object is usually passed as a parameter.

```python
class CPU:

    def process(self):
        return "Processing..."

class Computer:

    def __init__(self, cpu):
        self.cpu = cpu

cpu = CPU()

computer = Computer(cpu)

print(computer.cpu.process())
```

### Output

```text
Processing...
```

The CPU object exists independently of the Computer.

---

## 5. Aggregation vs Composition

| Aggregation | Composition |
|-------------|-------------|
| Weak "has-a" relationship | Strong "has-a" relationship |
| Objects exist independently | Contained object depends on owner |
| Object is passed from outside | Object is created inside the class |
| Loose coupling | Stronger coupling |
| Independent lifecycle | Shared lifecycle |

---

## 6. Object Lifecycles

In aggregation, destroying the container does **not** destroy the contained object.

```python
class Employee:

    def __init__(self, name):
        self.name = name

class Company:

    def __init__(self, employee):
        self.employee = employee

employee = Employee("David")

company = Company(employee)

del company

print(employee.name)
```

### Output

```text
David
```

The `Employee` object still exists even after the `Company` object is deleted.

---

## 7. Aggregation Encourages Reusability

The same object can be shared by multiple classes.

```python
class Printer:

    def print_document(self):
        return "Printing..."

class OfficeA:

    def __init__(self, printer):
        self.printer = printer

class OfficeB:

    def __init__(self, printer):
        self.printer = printer

printer = Printer()

office1 = OfficeA(printer)
office2 = OfficeB(printer)

print(office1.printer.print_document())
print(office2.printer.print_document())
```

### Output

```text
Printing...
Printing...
```

## 12. Common Mistakes

Avoid these common mistakes while using aggregation.

### Confusing Aggregation with Composition

Incorrect thinking:

```
Company creates and owns Employee.
```

Correct thinking:

```
Company uses an existing Employee.
```

---

### Creating the Object Inside the Class

Incorrect

```python
class Company:

    def __init__(self):
        self.employee = Employee()
```

This represents **composition**, not aggregation.

Correct

```python
employee = Employee()

company = Company(employee)
```

The object is created outside and passed into the class.

---

### Assuming Shared Lifetime

Deleting the container object does **not** delete the aggregated object.

Both have independent lifecycles.

---

### Tight Coupling

Avoid making aggregated classes overly dependent on each other's internal implementation.

---

## 13. Best Practices

Follow these recommended practices when using aggregation.

- Use aggregation when objects should exist independently.
- Pass dependent objects through constructors or methods.
- Keep classes loosely coupled.
- Reuse shared objects whenever appropriate.
- Avoid unnecessary ownership of external objects.
- Clearly model real-world relationships.
- Prefer aggregation when one object only needs to collaborate with another.
