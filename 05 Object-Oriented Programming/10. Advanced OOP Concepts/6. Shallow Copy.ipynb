# 📄 Shallow Copy in Python

A **shallow copy** creates a **new object**, but **does not create copies of the objects contained inside it**.

Instead, the new object **shares references** to the nested (inner) objects of the original object.

In simple words:

> **A shallow copy copies only the first level of an object. Nested objects are shared between the original and the copied object.**

Python provides several ways to create shallow copies, including:

- `copy.copy()`
- `list.copy()`
- Dictionary `.copy()`
- List slicing (`[:]`)

Understanding shallow copying is essential when working with mutable objects such as lists, dictionaries, and custom class instances.

# 📖 Topics Covered

## 1. What is a Shallow Copy?

A shallow copy creates a new outer object while sharing references to any nested objects.

Example:

```python
import copy

list1 = [1, 2, 3]

list2 = copy.copy(list1)

print(list1)
print(list2)
```

### Output

```text
[1, 2, 3]
[1, 2, 3]
```

The outer list is copied.

---

## 2. Why Use a Shallow Copy?

Shallow copying helps us:

- Duplicate simple objects.
- Prevent accidental modification of the original outer object.
- Improve performance.
- Save memory.
- Create lightweight copies.

## 4. Assignment vs Shallow Copy

Assignment does **not** create a new object.

```python
list1 = [10, 20, 30]

list2 = list1

list2.append(40)

print(list1)
```

### Output

```text
[10, 20, 30, 40]
```

Both variables refer to the same object.

---

Shallow copy creates a new object.

```python
import copy

list1 = [10, 20, 30]

list2 = copy.copy(list1)

list2.append(40)

print(list1)
print(list2)
```

### Output

```text
[10, 20, 30]
[10, 20, 30, 40]
```

---

## 5. Shallow Copy with Nested Objects

The outer object is copied, but inner objects are shared.

```python
import copy

list1 = [[1, 2], [3, 4]]

list2 = copy.copy(list1)

list2[0].append(100)

print(list1)
print(list2)
```

### Output

```text
[[1, 2, 100], [3, 4]]
[[1, 2, 100], [3, 4]]
```

Both lists share the same inner list.

---

## 6. Object References

Consider this example.

```python
import copy

data1 = [[10]]

data2 = copy.copy(data1)
```

Memory representation:

```text
data1
 │
 ▼
+------------------+
|  Reference A     |
+------------------+
        │
        ▼
      [10]

data2
 │
 ▼
+------------------+
|  Reference A     |
+------------------+
```

The outer lists are different, but both point to the same inner list.

---

## 7. Shallow Copy vs Deep Copy

| Shallow Copy | Deep Copy |
|--------------|-----------|
| Copies only the outer object | Copies the entire object hierarchy |
| Nested objects are shared | Nested objects are duplicated |
| Faster | Slower |
| Uses less memory | Uses more memory |
| `copy.copy()` | `copy.deepcopy()` |

## 8. Ways to Create a Shallow Copy

### Using `copy.copy()`

```python
import copy

numbers = [1, 2, 3]

copy_numbers = copy.copy(numbers)

print(copy_numbers)
```

### Output

```text
[1, 2, 3]
```

---

### Using `list.copy()`

```python
numbers = [5, 6, 7]

copy_numbers = numbers.copy()

print(copy_numbers)
```

### Output

```text
[5, 6, 7]
```

---

### Using List Slicing

```python
numbers = [10, 20, 30]

copy_numbers = numbers[:]

print(copy_numbers)
```

### Output

```text
[10, 20, 30]
```

---

### Using `dict.copy()`

```python
student = {
    "name": "Alice",
    "marks": 90
}

new_student = student.copy()

print(new_student)
```

### Output

```text
{'name': 'Alice', 'marks': 90}
```

## 11. Common Mistakes

Avoid these common mistakes while working with shallow copies.

### Confusing Assignment with Copying

Incorrect

```python
a = [1, 2]

b = a
```

This creates another reference.

Correct

```python
b = a.copy()
```

---

### Forgetting Nested Objects are Shared

```python
import copy

data1 = [[1]]

data2 = copy.copy(data1)
```

Changing the inner list affects both objects.

---

### Using Shallow Copy Instead of Deep Copy

For nested mutable objects, use:

```python
import copy

copy.deepcopy(data)
```

instead of:

```python
copy.copy(data)
```

when complete independence is required.

---

### Forgetting to Import the `copy` Module

Incorrect

```python
copy.copy(data)
```

Correct

```python
import copy

copy.copy(data)
```

---

## 12. Best Practices

Follow these recommended practices when using shallow copies.

- Use shallow copies for simple or one-level objects.
- Remember that nested mutable objects are shared.
- Use `deepcopy()` for independent nested objects.
- Prefer built-in `.copy()` methods when available.
- Understand the difference between references and copies.
- Test behavior with nested data structures.
- Choose the appropriate copying method for your application.

> **Note:** A **shallow copy** creates a new outer object while sharing references to nested objects. It is efficient in terms of memory and performance, making it suitable for simple data structures. However, when working with nested mutable objects, remember that changes made to shared inner objects will be reflected in both the original and the copied object. Use a **deep copy** instead when complete independence between objects is required.
