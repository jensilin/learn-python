# 📚 Deep Copy in Python

A **deep copy** creates a **completely independent copy** of an object, including **all of its nested (inner) objects**.

Unlike a **shallow copy**, where nested objects are shared, a deep copy recursively copies every object in the hierarchy.

In simple words:

> **A deep copy duplicates the entire object structure, so changes made to the copied object do not affect the original object.**

Python provides the **`deepcopy()`** function from the **`copy`** module to perform deep copying.

Deep copying is especially useful when working with complex objects containing nested lists, dictionaries, sets, or custom class instances.

# 💻 Quick Example

```python
import copy

numbers = [[10, 20], [30, 40]]

new_numbers = copy.deepcopy(numbers)

new_numbers[0].append(50)

print(numbers)
print(new_numbers)
```

### Output

```text
[[10, 20], [30, 40]]
[[10, 20, 50], [30, 40]]
```

The original list remains unchanged because all nested objects were copied.

# 📖 Topics Covered

## 1. What is a Deep Copy?

A deep copy creates a completely new object along with copies of all nested objects.

Example:

```python
import copy

list1 = [[1, 2], [3, 4]]

list2 = copy.deepcopy(list1)

print(list1)
print(list2)
```

### Output

```text
[[1, 2], [3, 4]]
[[1, 2], [3, 4]]
```

Both lists and their nested lists are different objects.

---

## 2. Why Use a Deep Copy?

Deep copying helps us:

- Create completely independent objects.
- Prevent unintended modifications.
- Work safely with nested mutable objects.
- Preserve original data.
- Avoid shared references.

---

## 3. Syntax

```python
import copy

new_object = copy.deepcopy(original_object)
```

The `deepcopy()` function recursively copies the entire object hierarchy.

---

## 4. Deep Copy vs Assignment

Assignment creates another reference.

```python
list1 = [[1, 2]]

list2 = list1

list2[0].append(100)

print(list1)
```

### Output

```text
[[1, 2, 100]]
```

Both variables point to the same object.

---

Deep copy creates an independent object.

```python
import copy

list1 = [[1, 2]]

list2 = copy.deepcopy(list1)

list2[0].append(100)

print(list1)
print(list2)
```

### Output

```text
[[1, 2]]
[[1, 2, 100]]
```

---

## 5. Deep Copy vs Shallow Copy

```python
import copy

original = [[1, 2]]

shallow = copy.copy(original)
deep = copy.deepcopy(original)

shallow[0].append(10)
deep[0].append(20)

print(original)
print(shallow)
print(deep)
```

### Output

```text
[[1, 2, 10]]
[[1, 2, 10]]
[[1, 2, 20]]
```

The shallow copy shares the inner list with the original, while the deep copy has its own independent inner list.

---

## 6. Memory Representation

Consider the following code:

```python
import copy

data1 = [[10]]

data2 = copy.deepcopy(data1)
```

Memory representation:

```text
data1
 │
 ▼
+-----------+
|  List A   |
+-----------+
     │
     ▼
 +---------+
 | [10]    |
 +---------+

data2
 │
 ▼
+-----------+
|  List B   |
+-----------+
     │
     ▼
 +---------+
 | [10]    |
 +---------+
```

Both the outer list and the inner list are copied independently.

## 7. Deep Copy of Dictionaries

```python
import copy

student1 = {
    "name": "Alice",
    "marks": {
        "math": 95
    }
}

student2 = copy.deepcopy(student1)

student2["marks"]["math"] = 100

print(student1)
print(student2)
```

### Output

```text
{'name': 'Alice', 'marks': {'math': 95}}
{'name': 'Alice', 'marks': {'math': 100}}
```

---

## 8. Deep Copy of Custom Objects

Custom class objects can also be deep copied.

```python
import copy

class Student:

    def __init__(self, name):
        self.name = name

student1 = Student("Alice")

student2 = copy.deepcopy(student1)

student2.name = "Bob"

print(student1.name)
print(student2.name)
```

### Output

```text
Alice
Bob
```

## 9. Practical Examples

### Example 1: Nested List

```python
import copy

numbers = [[10], [20]]

new_numbers = copy.deepcopy(numbers)

new_numbers[0].append(30)

print(numbers)
print(new_numbers)
```

### Example 2: Nested Dictionary

```python
import copy

employee = {
    "name": "John",
    "details": {
        "salary": 50000
    }
}

new_employee = copy.deepcopy(employee)

new_employee["details"]["salary"] = 60000

print(employee)
print(new_employee)
```

### Output

---

### Example 3: Matrix

```python
import copy

matrix = [
    [1, 2],
    [3, 4]
]

new_matrix = copy.deepcopy(matrix)

new_matrix[0][0] = 99

print(matrix)
print(new_matrix)
```

### Output

```text
[[1, 2], [3, 4]]
[[99, 2], [3, 4]]
```

---

### Example 4: Shopping Cart

```python
import copy

cart = [
    {"item": "Laptop", "price": 50000}
]

new_cart = copy.deepcopy(cart)

new_cart[0]["price"] = 60000

print(cart)
print(new_cart)
```

### Example 5: Custom Object

## 10. Deep Copy vs Shallow Copy

| Feature | Shallow Copy | Deep Copy |
|----------|--------------|-----------|
| Outer Object | Copied | Copied |
| Nested Objects | Shared | Copied |
| Independence | Partial | Complete |
| Memory Usage | Lower | Higher |
| Performance | Faster | Slower |
| Function | `copy.copy()` | `copy.deepcopy()` |

## 12. Common Mistakes

Avoid these common mistakes while working with deep copies.

### Forgetting to Import the `copy` Module

Incorrect

```python
copy.deepcopy(data)
```

Correct

```python
import copy

copy.deepcopy(data)
```

---

### Using Shallow Copy for Nested Objects

Incorrect

```python
copy.copy(data)
```

when complete independence is required.

Correct

```python
copy.deepcopy(data)
```

---

### Assuming Deep Copy is Always Necessary

Deep copying consumes more memory and processing time.

Use it only when nested mutable objects must be completely independent.

---

### Confusing Assignment with Copying

Incorrect

```python
b = a
```

This creates another reference.

Use:

```python
b = copy.deepcopy(a)
```

when an independent copy is required.

---

## 13. Best Practices

Follow these recommended practices when using deep copies.

- Use `deepcopy()` for nested mutable objects.
- Use shallow copies for simple objects when appropriate.
- Import the `copy` module before using `deepcopy()`.
- Consider memory usage before creating large deep copies.
- Test copied objects independently.
- Avoid unnecessary deep copies.
- Understand the difference between references, shallow copies, and deep copies.
