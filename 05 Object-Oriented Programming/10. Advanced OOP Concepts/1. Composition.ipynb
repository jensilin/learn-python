# 🧩 Composition in Python

**Composition** is an object-oriented programming (OOP) concept where one class **contains an object of another class** to reuse its functionality.

In simple terms:

> **Composition represents a "has-a" relationship.**

Instead of inheriting from another class, a class creates and uses objects of other classes as its members.

Composition promotes:

- Code reusability
- Modularity
- Flexibility
- Better encapsulation
- Loose coupling

It is often preferred over inheritance because it allows classes to be combined without creating rigid parent-child relationships.

# 💻 Quick Example

### Output

```text
Engine Started
```

# 📖 Topics Covered

## 1. What is Composition?

Composition is an OOP relationship where one object contains another object as one of its attributes.

The contained object helps perform part of the overall task.

Example:

```python
class CPU:

    def process(self):
        return "Processing..."

class Computer:

    def __init__(self):
        self.cpu = CPU()

computer = Computer()

print(computer.cpu.process())
```

### Output

```text
Processing...
```

---

## 2. Why Use Composition?

Composition helps us:

- Reuse existing classes.
- Reduce code duplication.
- Keep classes independent.
- Improve maintainability.
- Build modular applications.
- Prefer "has-a" relationships over unnecessary inheritance.

---

## 3. Understanding the "Has-a" Relationship

Composition represents ownership.

Examples:

- Car **has an** Engine.
- Computer **has a** CPU.
- House **has** Rooms.
- Library **has** Books.
- School **has** Students.

These are composition relationships.

---

## 4. How Composition Works

A class creates objects of other classes and stores them as instance variables.

```python
class Keyboard:

    def type(self):
        return "Typing..."

class Laptop:

    def __init__(self):
        self.keyboard = Keyboard()

laptop = Laptop()

print(laptop.keyboard.type())
```

### Output

```text
Typing...
```

---

## 5. Composition vs Inheritance

| Composition | Inheritance |
|-------------|-------------|
| Has-a relationship | Is-a relationship |
| One object contains another | One class extends another |
| More flexible | Less flexible |
| Promotes loose coupling | Creates tighter coupling |
| Easier to modify | Changes may affect subclasses |

Example:

- Car **has an** Engine → Composition
- Dog **is an** Animal → Inheritance

---

## 6. Composition Encourages Reusability

The same component can be shared across multiple classes.

```python
class Engine:

    def start(self):
        return "Engine Started"

class Car:

    def __init__(self):
        self.engine = Engine()

class Bike:

    def __init__(self):
        self.engine = Engine()

car = Car()
bike = Bike()

print(car.engine.start())
print(bike.engine.start())
```

### Output

```text
Engine Started
Engine Started
```

---

## 7. Nested Objects

Composition naturally creates nested objects.

```python
class Address:

    def __init__(self, city):
        self.city = city

class Person:

    def __init__(self, name, city):
        self.name = name
        self.address = Address(city)

person = Person("Alice", "Bangalore")

print(person.address.city)
```

### Output

```text
Bangalore
```

## 9. Practical Examples

### Example 1: Car and Engine

```python
class Engine:

    def start(self):
        return "Engine Started"

class Car:

    def __init__(self):
        self.engine = Engine()

car = Car()

print(car.engine.start())
```

### Output

```text
Engine Started
```

---

### Example 2: Computer and CPU

```python
class CPU:

    def execute(self):
        return "Executing Instructions"

class Computer:

    def __init__(self):
        self.cpu = CPU()

computer = Computer()

print(computer.cpu.execute())
```

### Output

```text
Executing Instructions
```

---

### Example 3: Library and Book

```python
class Book:

    def __init__(self, title):
        self.title = title

class Library:

    def __init__(self):
        self.book = Book("Python Basics")

library = Library()

print(library.book.title)
```

### Output

```text
Python Basics
```

---

### Example 4: Student and Address

```python
class Address:

    def __init__(self, city):
        self.city = city

class Student:

    def __init__(self, name, city):
        self.name = name
        self.address = Address(city)

student = Student("Alice", "Chennai")

print(student.address.city)
```

### Output

```text
Chennai
```

---

### Example 5: Mobile Phone

```python
class Battery:

    def charge(self):
        return "Battery Charging"

class Mobile:

    def __init__(self):
        self.battery = Battery()

mobile = Mobile()

print(mobile.battery.charge())
```

### Output

```text
Battery Charging
```

## 11. Common Mistakes

Avoid these common mistakes while using composition.

### Using Inheritance Instead of Composition

Incorrect thinking:

```
Car is an Engine
```

Correct thinking:

```
Car has an Engine
```

---

### Creating Unnecessary Dependencies

Avoid making one class depend on too many other classes.

Keep responsibilities focused.

---

### Mixing Responsibilities

Incorrect

```python
class Car:
    # Handles engine, payments,
    # customer details, insurance, etc.
```

Instead, separate responsibilities into multiple classes.

---

### Exposing Internal Objects Unnecessarily

Avoid allowing external code to directly modify internal objects unless required.

Keep encapsulation in mind.

---

## 12. Best Practices

Follow these recommended practices when using composition.

- Prefer composition when modeling "has-a" relationships.
- Keep each class responsible for one task.
- Reuse component classes whenever possible.
- Keep classes loosely coupled.
- Hide internal implementation details.
- Avoid deeply nested compositions unless necessary.
- Design small, reusable classes.

> **Note:** **Composition** is a fundamental object-oriented design technique that enables one class to build functionality by containing objects of other classes. By favoring composition where appropriate, developers can create flexible software designs that are easier to extend and less tightly coupled.
