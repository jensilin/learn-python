# 📋 Object Copying in Python

**Object copying** is the process of creating a new object from an existing object.

Instead of creating a completely new object manually, Python provides built-in ways to copy objects while preserving their data.

Python supports two types of object copying:

- **Shallow Copy**
- **Deep Copy**

Understanding object copying is important because assigning one object to another variable **does not create a copy**. Instead, both variables refer to the **same object**.

The `copy` module provides functions for creating copies safely.

# 📖 Topics Covered

## 1. What is Object Copying?

Object copying creates a new object using an existing object.

The copied object initially contains the same data as the original object.

Example:

```python
original = [1, 2, 3]

copied = original.copy()

print(original)
print(copied)
```

### Output

```text
[1, 2, 3]
[1, 2, 3]
```

---

## 2. Why is Object Copying Important?

Object copying helps us:

- Preserve original data.
- Avoid unintended modifications.
- Duplicate complex objects.
- Improve program reliability.
- Work safely with mutable objects.

---

## 3. Assignment is NOT Copying

Many beginners think assigning one variable to another creates a copy.

It does **not**.

```python
list1 = [10, 20, 30]

list2 = list1

list2.append(40)

print(list1)
print(list2)
```

### Output

```text
[10, 20, 30, 40]
[10, 20, 30, 40]
```

Both variables refer to the **same object**.

---

## 4. Creating a Shallow Copy

A shallow copy creates a new outer object, but nested objects are shared.

```python
import copy

list1 = [1, 2, 3]

list2 = copy.copy(list1)

print(list1)
print(list2)
```

### Output

```text
[1, 2, 3]
[1, 2, 3]
```

---

## 5. Shallow Copy with Nested Objects

Nested objects are shared between the original and the copied object.

```python
import copy

list1 = [[1, 2], [3, 4]]

list2 = copy.copy(list1)

list2[0].append(100)

print(list1)
print(list2)
```

### Output

```text
[[1, 2, 100], [3, 4]]
[[1, 2, 100], [3, 4]]
```

The inner list is shared by both objects.

---

## 6. Deep Copy

A deep copy duplicates the object **and all nested objects**.

```python
import copy

list1 = [[1, 2], [3, 4]]

list2 = copy.deepcopy(list1)

list2[0].append(100)

print(list1)
print(list2)
```

### Output

```text
[[1, 2], [3, 4]]
[[1, 2, 100], [3, 4]]
```

The original object remains unchanged.

---

## 7. Shallow Copy vs Deep Copy

| Shallow Copy | Deep Copy |
|--------------|-----------|
| Copies outer object | Copies entire object hierarchy |
| Nested objects are shared | Nested objects are duplicated |
| Faster | Slower |
| Uses less memory | Uses more memory |
| `copy.copy()` | `copy.deepcopy()` |

---

## 8. Using the `copy` Module

Python provides the `copy` module.

```python
import copy
```

Common functions:

| Function | Description |
|----------|-------------|
| `copy.copy()` | Creates a shallow copy |
| `copy.deepcopy()` | Creates a deep copy |

## 9. Copying Custom Objects

Custom class objects can also be copied.

```python
import copy

class Student:

    def __init__(self, name):
        self.name = name

student1 = Student("Alice")

student2 = copy.copy(student1)

print(student1.name)
print(student2.name)
```

### Output

```text
Alice
Alice
```

## 12. Common Mistakes

Avoid these common mistakes while working with object copying.

### Confusing Assignment with Copying

Incorrect

```python
a = [1, 2]

b = a
```

This creates another reference, not a new object.

Correct

```python
b = a.copy()
```

---

### Using Shallow Copy for Nested Objects

```python
import copy

nested = [[1]]

new_nested = copy.copy(nested)
```

Changes to inner objects affect both lists.

Use `deepcopy()` when nested objects should be independent.

---

### Forgetting to Import the `copy` Module

Incorrect

```python
copy.copy(data)
```

Correct

```python
import copy

copy.copy(data)
```

---

### Assuming All Objects Need Deep Copy

Deep copying consumes more memory and takes more time.

Use it only when necessary.

---

## 13. Best Practices

Follow these recommended practices when copying objects.

- Understand the difference between assignment and copying.
- Use shallow copy for simple objects.
- Use deep copy for nested mutable objects.
- Import the `copy` module when required.
- Avoid unnecessary deep copies.
- Test copied objects independently.
- Choose the appropriate copy method based on your application's needs.
