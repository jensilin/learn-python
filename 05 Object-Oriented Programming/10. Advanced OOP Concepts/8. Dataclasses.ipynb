# 🏗️ Dataclasses in Python

A **dataclass** is a special type of class introduced in **Python 3.7** through the **`dataclasses`** module.

It is designed to reduce the amount of boilerplate code needed when creating classes whose primary purpose is to store data.

By using the **`@dataclass`** decorator, Python automatically generates commonly used methods such as:

- `__init__()`
- `__repr__()`
- `__eq__()`

and many others.

Dataclasses make code:

- Shorter
- Cleaner
- Easier to read
- Easier to maintain

They are widely used in modern Python applications, including web development, APIs, machine learning, data processing, and configuration management.

# 💻 Quick Example

```python
from dataclasses import dataclass

@dataclass
class Student:
    name: str
    age: int

student = Student("Alice", 20)

print(student)
```

### Output

```text
Student(name='Alice', age=20)
```

Notice that we did not write an `__init__()` or `__repr__()` method. The `@dataclass` decorator generated them automatically.

# 📖 Topics Covered

## 1. What is a Dataclass?

A dataclass is a class that automatically generates common methods for storing data.

Example:

```python
from dataclasses import dataclass

@dataclass
class Employee:
    name: str
    salary: int

employee = Employee("John", 50000)

print(employee)
```

### Output

```text
Employee(name='John', salary=50000)
```

---

## 2. Why Use Dataclasses?

Dataclasses help us:

- Reduce boilerplate code.
- Automatically generate constructors.
- Automatically generate string representations.
- Automatically compare objects.
- Improve readability.
- Simplify maintenance.

---

## 3. The `@dataclass` Decorator

The `@dataclass` decorator tells Python to generate useful methods automatically.

Syntax:

```python
from dataclasses import dataclass

@dataclass
class ClassName:
    attribute: type
```

---

## 4. Automatically Generated Methods

By default, a dataclass generates:

| Method | Purpose |
|---------|----------|
| `__init__()` | Initializes objects |
| `__repr__()` | Returns a readable string representation |
| `__eq__()` | Compares objects based on their fields |

Example:

```python
from dataclasses import dataclass

@dataclass
class Book:
    title: str
    price: int

book = Book("Python", 500)

print(book)
```

### Output

```text
Book(title='Python', price=500)
```

---

## 5. Type Annotations

Dataclasses require **type annotations** for their fields.

```python
from dataclasses import dataclass

@dataclass
class Product:
    name: str
    price: float
    stock: int
```

Type annotations improve readability and support static type checking.

---

## 6. Default Values

Fields can have default values.

```python
from dataclasses import dataclass

@dataclass
class Student:
    name: str
    age: int = 18

student = Student("Alice")

print(student)
```

### Output

```text
Student(name='Alice', age=18)
```

---

## 7. Using `field()`

The `field()` function allows additional customization of dataclass fields.

```python
from dataclasses import dataclass, field

@dataclass
class Student:
    name: str
    marks: int = field(default=0)

student = Student("Alice")

print(student)
```

### Output

```text
Student(name='Alice', marks=0)
```

---

## 8. Using `default_factory`

Mutable default values should use `default_factory`.

Incorrect:

```python
from dataclasses import dataclass

@dataclass
class Student:
    marks: list = []
```

Correct:

```python
from dataclasses import dataclass, field

@dataclass
class Student:
    marks: list = field(default_factory=list)

student = Student()

print(student.marks)
```

### Output

```text
[]
```

Each object gets its own independent list.

---

## 9. Frozen Dataclasses

A frozen dataclass creates immutable objects.

```python
from dataclasses import dataclass

@dataclass(frozen=True)
class Employee:
    employee_id: int
    name: str

employee = Employee(101, "John")

print(employee)
```

### Output

```text
Employee(employee_id=101, name='John')
```

Trying to modify a field later raises a `FrozenInstanceError`.

---

## 10. Comparing Dataclass Objects

Objects are compared by their field values.

```python
from dataclasses import dataclass

@dataclass
class Student:
    name: str
    age: int

student1 = Student("Alice", 20)
student2 = Student("Alice", 20)

print(student1 == student2)
```

### Output

```text
True
```

## 14. Common Mistakes

Avoid these common mistakes while working with dataclasses.

### Forgetting the Decorator

Incorrect

```python
class Student:
    name: str
```

Correct

```python
from dataclasses import dataclass

@dataclass
class Student:
    name: str
```

---

### Using Mutable Default Values

Incorrect

```python
marks: list = []
```

Correct

```python
marks: list = field(default_factory=list)
```

---

### Assuming Dataclasses Replace Every Class

Dataclasses are best suited for classes whose primary purpose is storing data.

Classes with complex behavior may still be better implemented as regular classes.

---

### Forgetting Type Annotations

Dataclass fields should include type annotations.

Incorrect

```python
name = ""
```

Correct

```python
name: str
```

---

## 15. Best Practices

Follow these recommended practices when using dataclasses.

- Use dataclasses for data-centric classes.
- Always provide type annotations.
- Use `field(default_factory=...)` for mutable defaults.
- Keep dataclasses focused on storing data.
- Use `frozen=True` when immutability is required.
- Add methods only when they support the data model.
- Prefer dataclasses over manually writing repetitive boilerplate code.
