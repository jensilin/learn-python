# ✏️ Setter in Python

A **setter** is a method that is used to **modify (update) the value of an object's attribute**.

In modern Python, setters are most commonly implemented using the **`@property_name.setter`** decorator. A setter works together with a property (getter) to provide **controlled access** when assigning values to an attribute.

Instead of allowing direct modification of an object's internal data, setters enable developers to **validate**, **transform**, or **restrict** values before storing them.

Setters are an important part of **encapsulation**, helping maintain the integrity of an object's data.

# 💻 Quick Example

```python
class Student:

    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        self._name = value

student = Student("Alice")

student.name = "Bob"

print(student.name)
```

### Output

```text
Bob
```

Python automatically calls the setter when assigning a value to the property.

# 📖 Topics Covered

## 1. What is a Setter?

A setter is a method that updates the value of an attribute.

Instead of modifying the internal variable directly, Python calls the setter whenever a new value is assigned.

Example:

```python
class Employee:

    def __init__(self):
        self._salary = 50000

    @property
    def salary(self):
        return self._salary

    @salary.setter
    def salary(self, value):
        self._salary = value

employee = Employee()

employee.salary = 60000

print(employee.salary)
```

### Output

```text
60000
```

---

## 2. Why Use Setters?

Setters help us:

- Validate data before storing it.
- Prevent invalid values.
- Improve encapsulation.
- Hide implementation details.
- Keep objects in a valid state.
- Simplify future code maintenance.

---

## 3. Syntax

```python
class ClassName:

    @property
    def attribute(self):
        return self._attribute

    @attribute.setter
    def attribute(self, value):
        self._attribute = value
```

Python automatically calls the setter whenever:

```python
object.attribute = value
```

is executed.

---

## 4. Setter with Validation

One of the main advantages of setters is data validation.

```python
class Student:

    def __init__(self, marks):
        self._marks = marks

    @property
    def marks(self):
        return self._marks

    @marks.setter
    def marks(self, value):

        if value < 0:
            print("Marks cannot be negative.")
            return

        self._marks = value

student = Student(80)

student.marks = 95

print(student.marks)
```

### Output

```text
95
```

---

## 5. Rejecting Invalid Values

```python
class BankAccount:

    def __init__(self, balance):
        self._balance = balance

    @property
    def balance(self):
        return self._balance

    @balance.setter
    def balance(self, value):

        if value < 0:
            print("Invalid balance.")
            return

        self._balance = value

account = BankAccount(5000)

account.balance = -100

print(account.balance)
```

### Output

```text
Invalid balance.
5000
```

The invalid value is rejected.

## 6. Setter vs Direct Assignment

| Direct Assignment | Setter |
|-------------------|--------|
| Stores any value | Can validate values |
| No control | Controlled assignment |
| Internal implementation is exposed | Internal implementation is hidden |
| Less flexible | Highly flexible |

## 7. Python Calls the Setter Automatically

When assigning a value to a property, Python automatically calls the setter.

```python
class Product:

    def __init__(self, price):
        self._price = price

    @property
    def price(self):
        return self._price

    @price.setter
    def price(self, value):
        self._price = value

product = Product(500)

product.price = 700

print(product.price)
```

### Output

```text
700
```

Developers simply write:

```python
product.price = 700
```

instead of calling the setter directly.

### Example 3: Temperature

```python
class Temperature:

    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, value):

        if value < -273.15:
            print("Temperature below absolute zero is not allowed.")
            return

        self._celsius = value

temperature = Temperature(25)

temperature.celsius = 30

print(temperature.celsius)
```

## 10. Common Mistakes

Avoid these common mistakes while working with setters.

### Forgetting the Getter

A setter must be associated with an existing property.

Incorrect

```python
@name.setter
def name(self, value):
    self._name = value
```

Always define the property first.

---

### Assigning to the Property Inside the Setter

Incorrect

```python
@name.setter
def name(self, value):
    self.name = value
```

This causes infinite recursion.

Correct

```python
@name.setter
def name(self, value):
    self._name = value
```

---

### Skipping Validation

Instead of:

```python
self._marks = value
```

Validate the value when necessary.

---

### Accessing Internal Variables Directly

Instead of:

```python
student._marks = 95
```

Prefer:

```python
student.marks = 95
```

This ensures the setter is executed.

---

## 11. Best Practices

Follow these recommended practices when implementing setters.

- Validate input before storing it.
- Keep setters simple and efficient.
- Use meaningful error messages.
- Store values in internal variables (prefixed with `_`).
- Avoid recursive assignments inside setters.
- Use setters only when controlled modification is needed.
- Keep the public interface clean and intuitive.
