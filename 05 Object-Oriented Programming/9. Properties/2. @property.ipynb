# 🏷️ `@property` Decorator in Python

The **`@property`** decorator is a built-in feature in Python that allows a method to be accessed **like an attribute**.

Instead of calling a method using parentheses (`()`), the `@property` decorator lets you access it using simple dot (`.`) notation.

It is commonly used to:

- Create **read-only attributes**
- Compute values dynamically
- Hide implementation details
- Improve encapsulation
- Maintain backward compatibility when changing class implementations

The `@property` decorator is the foundation for creating **getter**, **setter**, and **deleter** methods in Python.

# 💻 Quick Example

```python
class Student:

    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

student = Student("Alice")

print(student.name)
```

### Output

```text
Alice
```

Notice that `name` is accessed like an attribute instead of a method.

# 📖 Topics Covered

## 1. What is `@property`?

The `@property` decorator converts a method into a property.

Without it:

```python
class Student:

    def name(self):
        return "Alice"

student = Student()

print(student.name())
```

### Output

```text
Alice
```

With `@property`:

```python
class Student:

    @property
    def name(self):
        return "Alice"

student = Student()

print(student.name)
```

### Output

```text
Alice
```

The method behaves like a normal attribute.

---

## 2. Why Use `@property`?

The `@property` decorator helps us:

- Hide implementation details.
- Improve encapsulation.
- Create read-only attributes.
- Compute values dynamically.
- Keep code simple and readable.
- Add validation later without changing external code.

## 3. Syntax

```python
class ClassName:

    @property
    def property_name(self):
        return value
```

The decorated method becomes accessible as:

```python
object.property_name
```

instead of:

```python
object.property_name()
```

---

## 4. Read-Only Property

A property with only a getter cannot be modified.

```python
class Employee:

    def __init__(self, employee_id):
        self._employee_id = employee_id

    @property
    def employee_id(self):
        return self._employee_id

employee = Employee(101)

print(employee.employee_id)
```

### Output

```text
101
```

Since no setter is defined, the property is read-only.

---

## 5. Computed Property

Properties can calculate values whenever they are accessed.

```python
class Rectangle:

    def __init__(self, length, width):
        self.length = length
        self.width = width

    @property
    def area(self):
        return self.length * self.width

rectangle = Rectangle(8, 5)

print(rectangle.area)
```

### Output

```text
40
```

The area is computed each time it is accessed.

## 6. Method vs Property

| Method | Property |
|---------|----------|
| Called with `()` | Accessed without `()` |
| Performs an action | Represents an attribute |
| `student.name()` | `student.name` |
| Usually executes logic | Can return stored or computed values |

Example:

```python
class Student:

    def get_name(self):
        return "Alice"

    @property
    def name(self):
        return "Alice"

student = Student()

print(student.get_name())
print(student.name)
```

### Output

```text
Alice
Alice
```

---

## 7. Dynamic Properties

A property can return values based on the current object state.

```python
class Temperature:

    def __init__(self, celsius):
        self.celsius = celsius

    @property
    def fahrenheit(self):
        return (self.celsius * 9 / 5) + 32

temperature = Temperature(30)

print(temperature.fahrenheit)
```

### Output

```text
86.0
```

## 10. Common Mistakes

Avoid these common mistakes while using `@property`.

### Forgetting the Decorator

Incorrect

```python
def name(self):
    return self._name
```

The method must be called with parentheses.

Correct

```python
@property
def name(self):
    return self._name
```

---

### Calling a Property Like a Method

Incorrect

```python
print(student.name())
```

Correct

```python
print(student.name)
```

Properties are accessed like attributes.

---

### Assigning to a Read-Only Property

```python
student.name = "Bob"
```

### Output

```text
AttributeError
```

A property without a setter cannot be modified.

---

### Performing Expensive Operations

Avoid writing properties that perform slow operations every time they are accessed.

---

## 11. Best Practices

Follow these recommended practices when using the `@property` decorator.

- Use `@property` for values that logically represent attributes.
- Keep property methods simple and efficient.
- Use properties for computed values.
- Use a single leading underscore (`_`) for internal variables.
- Avoid side effects inside property methods.
- Add setters only when modification should be allowed.
- Document computed properties clearly.
