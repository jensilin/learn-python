# 🏷️ Properties in Python

A **property** in Python provides a clean and controlled way to **access, modify, and manage an object's attributes**.

Instead of directly exposing instance variables, properties allow us to add **validation, computation, or custom behavior** while still accessing attributes using simple dot (`.`) notation.

Python provides the built-in **`property()`** function and the **`@property`**, `@<property>.setter`, and `@<property>.deleter` decorators to create properties.

Properties are an important feature of **encapsulation**, helping protect object data while keeping code readable and maintainable.

# 💻 Quick Example

```python
class Student:

    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

student = Student("Alice")

print(student.name)
```

### Output

```text
Alice
```

Notice that `name` is accessed like a normal attribute even though it is actually a method.

# 📖 Topics Covered

## 1. What is a Property?

A property is a special attribute whose value is controlled by one or more methods.

Instead of directly accessing an instance variable, Python automatically calls methods behind the scenes.

Example:

```python
class Employee:

    def __init__(self):
        self._salary = 50000

    @property
    def salary(self):
        return self._salary

employee = Employee()

print(employee.salary)
```

### Output

```text
50000
```

---

## 2. Why Use Properties?

Properties help us:

- Protect object data.
- Validate input values.
- Hide implementation details.
- Improve encapsulation.
- Compute values dynamically.
- Keep code clean and readable.
- Modify internal implementation without changing external code.

---

## 3. Attributes vs Properties

| Attribute | Property |
|-----------|----------|
| Stores data directly | Uses methods to control access |
| No validation | Can validate data |
| Direct access | Controlled access |
| Simple variable | Behaves like an attribute but executes methods |
| Less flexible | Highly flexible |

---

## 4. How Properties Work

A property looks like a normal attribute but executes a method.

```python
class Student:

    def __init__(self):
        self._age = 20

    @property
    def age(self):
        return self._age

student = Student()

print(student.age)
```

### Output

```text
20
```

Python internally calls:

```python
student.age()
```

conceptually, but the user simply writes:

```python
student.age
```

---

## 5. Why Use an Underscore (`_`)?

A single leading underscore indicates that an attribute is intended for internal use.

Example:

```python
class Product:

    def __init__(self):
        self._price = 500
```

The underscore is a naming convention that tells other developers:

> "This attribute is considered internal and should not be accessed directly."

It does **not** make the attribute private.

## 6. Property Components

A complete property may include three methods:

| Component | Purpose |
|-----------|---------|

| Deleter | Deletes a value |

These components will be covered in detail in the following notebooks.

## 7. Read-Only Property

A property with only a getter becomes read-only.

```python
class Circle:

    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius

circle = Circle(10)

print(circle.radius)
```

### Output

```text
10
```

Since no setter exists, the value cannot be modified through the property.

---

## 8. Computed Property

Properties can calculate values dynamically.

```python
class Rectangle:

    def __init__(self, length, width):
        self.length = length
        self.width = width

    @property
    def area(self):
        return self.length * self.width

rectangle = Rectangle(10, 5)

print(rectangle.area)
```

### Output

```text
50
```

The area is calculated every time it is accessed.

### Example 3: Temperature

```python
class Temperature:

    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def fahrenheit(self):
        return (self._celsius * 9 / 5) + 32

temperature = Temperature(25)

print(temperature.fahrenheit)
```

### Example 4: Rectangle

```python
class Rectangle:

    def __init__(self, length, width):
        self.length = length
        self.width = width

    @property
    def perimeter(self):
        return 2 * (self.length + self.width)

rectangle = Rectangle(10, 5)

print(rectangle.perimeter)
```

## 11. Common Mistakes

Avoid these common mistakes while working with properties.

### Accessing Internal Variables Directly

Instead of:

```python
print(student._name)
```

Prefer:

```python
print(student.name)
```

---

### Forgetting the `@property` Decorator

Incorrect

```python
def name(self):
    return self._name
```

Without `@property`, the method must be called with parentheses.

Correct

```python
@property
def name(self):
    return self._name
```

---

### Expecting Read-Only Properties to Be Writable

```python
student.name = "Bob"
```

### Output

```text
AttributeError
```

A property without a setter cannot be assigned a new value.

---

### Overusing Properties

Not every attribute needs to be a property.

Use properties only when controlled access or additional behavior is required.

---

## 12. Best Practices

Follow these recommended practices when using properties.

- Use properties for controlled attribute access.
- Keep property methods simple.
- Use properties for validation or computed values.
- Prefix internal variables with a single underscore (`_`).
- Avoid performing expensive operations inside properties.
- Expose a clean and intuitive interface.
- Use getters, setters, and deleters only when needed.
