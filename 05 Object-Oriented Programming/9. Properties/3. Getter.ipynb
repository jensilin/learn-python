# 📖 Getter in Python

A **getter** is a method that is used to **retrieve (read) the value of an object's attribute**.

In modern Python, getters are most commonly implemented using the **`@property`** decorator. This allows attributes to be accessed like normal variables while still providing controlled access to the underlying data.

Getters are an important part of **encapsulation**, helping protect an object's internal state while exposing only the information that should be accessible.

# 💻 Quick Example

```python
class Student:

    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

student = Student("Alice")

print(student.name)
```

### Output

```text
Alice
```

Although `name` is actually a method, it is accessed like an attribute.

# 📖 Topics Covered

## 1. What is a Getter?

A getter is a method that returns the value of an attribute.

Using a getter allows the class to control how data is accessed.

Example:

```python
class Employee:

    def __init__(self):
        self._salary = 50000

    @property
    def salary(self):
        return self._salary

employee = Employee()

print(employee.salary)
```

### Output

```text
50000
```

---

## 2. Why Use Getters?

Getters help us:

- Protect internal data.
- Improve encapsulation.
- Hide implementation details.
- Create read-only attributes.
- Compute values dynamically.
- Keep external code simple.

---

## 3. Traditional Getter vs `@property`

Before the `@property` decorator, getters were regular methods.

Traditional getter:

```python
class Student:

    def __init__(self):
        self._name = "Alice"

    def get_name(self):
        return self._name

student = Student()

print(student.get_name())
```

### Output

```text
Alice
```

Using `@property`:

```python
class Student:

    def __init__(self):
        self._name = "Alice"

    @property
    def name(self):
        return self._name

student = Student()

print(student.name)
```

### Output

```text
Alice
```

The second approach is cleaner and more Pythonic.

---

## 4. Read-Only Getter

If only a getter exists, the property becomes read-only.

```python
class Product:

    def __init__(self, price):
        self._price = price

    @property
    def price(self):
        return self._price

product = Product(1000)

print(product.price)
```

### Output

```text
1000
```

Without a setter, the value cannot be modified through the property.

---

## 5. Getters Can Compute Values

A getter does not have to return a stored value.

It can calculate values dynamically.

```python
class Rectangle:

    def __init__(self, length, width):
        self.length = length
        self.width = width

    @property
    def area(self):
        return self.length * self.width

rectangle = Rectangle(10, 5)

print(rectangle.area)
```

### Output

```text
50
```

The area is calculated every time it is accessed.

---

## 6. Getter vs Direct Attribute Access

| Direct Attribute | Getter |
|------------------|--------|
| Returns stored data | Can return stored or computed data |
| No validation | Can include validation or processing |
| Less flexible | More flexible |
| Internal implementation is exposed | Internal implementation is hidden |

---

## 7. Python Calls the Getter Automatically

When a property is accessed, Python automatically calls its getter method.

```python
class Book:

    def __init__(self, title):
        self._title = title

    @property
    def title(self):
        return self._title

book = Book("Python Basics")

print(book.title)
```

### Output

```text
Python Basics
```

Internally, Python invokes the getter method, but the user simply writes:

```python
book.title
```

### Example 4: Temperature Conversion

```python
class Temperature:

    def __init__(self, celsius):
        self.celsius = celsius

    @property
    def fahrenheit(self):
        return (self.celsius * 9 / 5) + 32

temperature = Temperature(25)

print(temperature.fahrenheit)
```

## 10. Common Mistakes

Avoid these common mistakes while working with getters.

### Forgetting `@property`

Incorrect

```python
def name(self):
    return self._name
```

Correct

```python
@property
def name(self):
    return self._name
```

---

### Calling a Getter Like a Method

Incorrect

```python
print(student.name())
```

Correct

```python
print(student.name)
```

Properties are accessed like attributes.

---

### Accessing Internal Variables Directly

Instead of:

```python
print(student._name)
```

Prefer:

```python
print(student.name)
```

---

### Performing Expensive Operations

Avoid performing slow database queries, network requests, or other expensive tasks inside a getter, as it may be called frequently.

---

## 11. Best Practices

Follow these recommended practices when implementing getters.

- Use `@property` instead of traditional getter methods whenever appropriate.
- Keep getters simple and efficient.
- Return meaningful values.
- Use a single leading underscore (`_`) for internal attributes.
- Avoid side effects inside getters.
- Use getters for computed values when appropriate.
- Expose a clean and intuitive interface.
