# 🗑️ Deleter in Python

A **deleter** is a method that is used to **delete an object's attribute** in a controlled manner.

In modern Python, deleters are implemented using the **`@property_name.deleter`** decorator. A deleter works together with a property's **getter** and **setter** to provide complete control over an attribute's lifecycle.

Instead of allowing attributes to be deleted directly, a deleter enables developers to perform **cleanup**, **validation**, **logging**, or **custom actions** before removing an attribute.

Deleters are an important part of **encapsulation**, allowing classes to safely manage their internal state.

# 💻 Quick Example

```python
class Student:

    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.deleter
    def name(self):
        print("Deleting student name...")
        del self._name

student = Student("Alice")

del student.name
```

### Output

```text
Deleting student name...
```

Python automatically calls the deleter when the property is deleted.

# 📖 Topics Covered

## 1. What is a Deleter?

A deleter is a method that controls how an attribute is deleted.

Instead of deleting the internal variable directly, Python calls the deleter whenever the `del` statement is used on a property.

Example:

```python
class Employee:

    def __init__(self):
        self._department = "IT"

    @property
    def department(self):
        return self._department

    @department.deleter
    def department(self):
        del self._department

employee = Employee()

del employee.department
```

The deleter executes before the attribute is removed.

---

## 2. Why Use Deleters?

Deleters help us:

- Perform cleanup operations.
- Control attribute deletion.
- Improve encapsulation.
- Log deletion events.
- Prevent accidental deletion.
- Keep objects in a consistent state.

---

## 3. Syntax

```python
class ClassName:

    @property
    def attribute(self):
        return self._attribute

    @attribute.deleter
    def attribute(self):
        del self._attribute
```

Python automatically calls the deleter whenever:

```python
del object.attribute
```

is executed.

---

## 4. Deleting an Attribute

```python
class Product:

    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.deleter
    def name(self):
        del self._name

product = Product("Laptop")

del product.name
```

The property is removed through the deleter.

---

## 5. Performing an Action Before Deletion

A deleter can perform custom actions before deleting an attribute.

```python
class BankAccount:

    def __init__(self, account_number):
        self._account_number = account_number

    @property
    def account_number(self):
        return self._account_number

    @account_number.deleter
    def account_number(self):
        print("Closing account...")
        del self._account_number

account = BankAccount("ACC123")

del account.account_number
```

### Output

```text
Closing account...
```

---

## 6. Python Calls the Deleter Automatically

Developers do not call the deleter method directly.

```python
class Book:

    def __init__(self, title):
        self._title = title

    @property
    def title(self):
        return self._title

    @title.deleter
    def title(self):
        print("Book removed.")
        del self._title

book = Book("Python Basics")

del book.title
```

### Output

```text
Book removed.
```

Python automatically invokes the deleter.

---

## 7. Deleting an Attribute Affects Future Access

Once an attribute is deleted, attempting to access it raises an error unless it is recreated.

```python
class Student:

    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.deleter
    def name(self):
        del self._name

student = Student("Alice")

del student.name

print(student.name)
```

### Output

```text
AttributeError
```

The attribute no longer exists.

---

## 8. Deleter vs Direct Deletion

| Direct Deletion | Deleter |
|-----------------|----------|
| Deletes the attribute immediately | Executes custom logic before deletion |
| No validation | Can validate deletion |
| No logging | Can log deletion events |
| Less flexible | Highly flexible |

## 11. Common Mistakes

Avoid these common mistakes while working with deleters.

### Forgetting the Property

A deleter must be associated with an existing property.

Incorrect

```python
@name.deleter
def name(self):
    del self._name
```

Always define the property first.

---

### Calling the Deleter Directly

Instead of:

```python
student.name()
```

or

```python
student.__delattr__("name")
```

Prefer:

```python
del student.name
```

Python automatically invokes the deleter.

---

### Deleting the Wrong Attribute

Incorrect

```python
del self.name
```

This may trigger the property again.

Correct

```python
del self._name
```

Delete the internal attribute.

---

### Accessing the Attribute After Deletion

```python
del student.name

print(student.name)
```

### Output

```text
AttributeError
```

Once deleted, the attribute no longer exists unless it is recreated.

---

## 12. Best Practices

Follow these recommended practices when implementing deleters.

- Use deleters only when controlled deletion is required.
- Perform cleanup tasks before removing attributes.
- Delete internal variables (prefixed with `_`).
- Keep deleter methods simple.
- Avoid deleting attributes unnecessarily.
- Log important deletion events when appropriate.
- Use `del object.property` instead of deleting internal variables directly.
