# 🧩 Cohesion in Python

**Cohesion** is a software design principle that describes **how closely related the responsibilities of a class or module are**.

A class is said to have **high cohesion** when all of its methods and attributes work together to accomplish a **single, well-defined purpose**.

In simple words:

> **A highly cohesive class focuses on one specific task and performs it well.**

Cohesion is one of the most important concepts in object-oriented design because it leads to code that is easier to understand, maintain, test, and reuse.

# 💻 Quick Example

Without high cohesion:

```python
class Student:

    def save(self):
        print("Saving student")

    def send_email(self):
        print("Sending email")

    def print_report(self):
        print("Printing report")
```

This class performs multiple unrelated tasks.

With high cohesion:

```python
class Student:

    def save(self):
        print("Saving student")


class EmailService:

    def send_email(self):
        print("Sending email")


class ReportPrinter:

    def print_report(self):
        print("Printing report")
```

Each class now has a single responsibility.

# 📖 Topics Covered

## 1. What is Cohesion?

Cohesion measures how closely the methods and attributes of a class are related.

A class with high cohesion focuses on a single purpose.

Example:

```python
class Calculator:

    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b
```

Both methods are related to mathematical calculations.

---

## 2. Why is Cohesion Important?

High cohesion helps us:

- Improve code readability.
- Reduce complexity.
- Simplify maintenance.
- Increase reusability.
- Make testing easier.
- Reduce bugs.
- Improve scalability.

---

## 3. High Cohesion

High cohesion means all members of a class work together toward a common goal.

Example:

```python
class BankAccount:

    def deposit(self, amount):
        print("Depositing", amount)

    def withdraw(self, amount):
        print("Withdrawing", amount)

    def check_balance(self):
        print("Displaying balance")
```

All methods relate to managing a bank account.

---

## 4. Low Cohesion

Low cohesion occurs when a class performs unrelated tasks.

Example:

```python
class Employee:

    def calculate_salary(self):
        print("Calculating salary")

    def send_email(self):
        print("Sending email")

    def print_report(self):
        print("Printing report")

    def connect_database(self):
        print("Connecting database")
```

This class mixes multiple unrelated responsibilities.

---

## 5. High Cohesion vs Low Cohesion

| High Cohesion | Low Cohesion |
|---------------|--------------|
| One responsibility | Multiple responsibilities |
| Easier to maintain | Difficult to maintain |
| Easier to understand | Hard to understand |
| Easier to test | Difficult to test |
| More reusable | Less reusable |

---

## 6. Cohesion and the Single Responsibility Principle (SRP)

High cohesion supports the **Single Responsibility Principle (SRP)**.

Example:

```python
class Invoice:

    def calculate_total(self):
        print("Calculating total")


class InvoicePrinter:

    def print_invoice(self):
        print("Printing invoice")
```

Each class has a single responsibility and therefore high cohesion.

---

## 7. Benefits of High Cohesion

High cohesion provides several benefits:

- Better organization
- Cleaner code
- Easier debugging
- Easier maintenance
- Better code reuse
- Simpler testing
- Lower coupling
- Improved scalability

## 10. Common Mistakes

Avoid these common mistakes while designing cohesive classes.

### Putting Too Many Responsibilities in One Class

Incorrect

```python
class Student:

    def save(self):
        ...

    def print_report(self):
        ...

    def send_email(self):
        ...

    def connect_database(self):
        ...
```

A single class should not perform unrelated tasks.

---

### Mixing Business Logic with Utility Functions

Keep utility functions in separate helper or service classes instead of combining them with business logic.

---

### Ignoring Class Responsibilities

Before adding a new method, ask:

> **Does this method belong to this class?**

If not, move it to a more appropriate class.

---

### Creating Large "God Classes"

Avoid classes that try to manage every aspect of an application.

Split them into smaller, focused classes.

---

## 11. Best Practices

Follow these recommended practices when designing highly cohesive classes.

- Give each class a single, well-defined purpose.
- Keep related methods together.
- Separate unrelated functionality into different classes.
- Follow the Single Responsibility Principle.
- Keep classes small and focused.
- Refactor large classes regularly.
- Prefer composition over large monolithic classes when appropriate.
- Write unit tests to verify each class independently.
