# 📖 Topics Covered

## 1. Follow the Single Responsibility Principle (SRP)

Each class should have **only one responsibility**.

Good Example:

```python
class Invoice:

    def calculate_total(self):
        print("Calculating total")


class InvoicePrinter:

    def print_invoice(self):
        print("Printing invoice")
```

Benefits:

- Easier maintenance
- Better readability
- Simpler testing

---

## 2. Keep Classes Small

Small classes are easier to understand and maintain.

Instead of:

```python
class Employee:

    def calculate_salary(self):
        ...

    def print_report(self):
        ...

    def send_email(self):
        ...

    def save_database(self):
        ...
```

Split responsibilities into separate classes.

---

## 3. Use Meaningful Class Names

Choose names that clearly describe the purpose of the class.

Good names:

```python
BankAccount
Student
Invoice
Product
ShoppingCart
```

Avoid vague names such as:

```python
Data
Manager
Object
Thing
```

---

## 4. Use Meaningful Method Names

Method names should describe exactly what they do.

Good examples:

```python
deposit()

withdraw()

calculate_total()

generate_report()
```

Avoid:

```python
do()

run()

process()

work()
```

unless their purpose is obvious from context.

---

## 5. Prefer Composition Over Inheritance

Inheritance should be used only when there is a true **"is-a"** relationship.

Whenever possible, use composition.

Example:

```python
class Engine:

    def start(self):
        print("Engine Started")


class Car:

    def __init__(self, engine):
        self.engine = engine

    def start(self):
        self.engine.start()

engine = Engine()

car = Car(engine)

car.start()
```

Composition provides greater flexibility than deep inheritance hierarchies.

---

## 6. Encapsulate Data

Keep data protected and expose it through methods or properties when necessary.

Example:

```python
class BankAccount:

    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

account = BankAccount()

account.deposit(500)

print(account.get_balance())
```

### Output

```text
500
```

---

## 7. Avoid Code Duplication

If the same code appears multiple times, move it into a reusable method or class.

Instead of repeating logic:

```python
total = price * quantity
```

in several places, create:

```python
class Calculator:

    def calculate_total(self, price, quantity):
        return price * quantity
```

---

## 8. Keep Methods Short

Each method should perform **one task**.

Good example:

```python
class Calculator:

    def add(self, a, b):
        return a + b
```

Avoid methods with hundreds of lines of code.

---

## 9. Write Reusable Classes

Design classes that can be used in different parts of an application.

Example:

```python
class Logger:

    def log(self, message):
        print(message)
```

This logger can be reused by many different classes.

---

## 10. Handle Exceptions Properly

Raise and handle meaningful exceptions.

```python
class BankAccount:

    def withdraw(self, balance, amount):

        if amount > balance:
            raise ValueError("Insufficient balance.")

account = BankAccount()

try:
    account.withdraw(1000, 1500)

except ValueError as error:
    print(error)
```

### Output

```text
Insufficient balance.
```

---

## 11. Write Readable Code

Readable code is easier to maintain than clever code.

Example:

```python
student_name = "Alice"

student_age = 20
```

Instead of:

```python
a = "Alice"

b = 20
```

Use descriptive variable names.

---

## 12. Write Comments Only When Necessary

Good code should explain itself through meaningful names.

Use comments only to explain:

- Complex algorithms
- Business rules
- Non-obvious logic

Avoid unnecessary comments such as:

```python
# Add two numbers
total = a + b
```

## 13. Test Your Classes

Every important class should be tested independently.

Example:

```python
class Calculator:

    def multiply(self, a, b):
        return a * b

calculator = Calculator()

print(calculator.multiply(5, 6))
```

### Output

```text
30
```

Testing helps detect bugs early.

## 14. Follow SOLID Principles

The SOLID principles improve software quality.

- Single Responsibility Principle
- Open/Closed Principle
- Liskov Substitution Principle
- Interface Segregation Principle
- Dependency Inversion Principle

Applying SOLID results in cleaner and more maintainable software.

## 16. Common Mistakes

Avoid these common mistakes while designing object-oriented applications.

### Creating Large Classes

Large classes become difficult to understand and maintain.

Split them into smaller, focused classes.

---

### Ignoring Encapsulation

Do not expose internal data unnecessarily.

Protect sensitive attributes whenever appropriate.

---

### Overusing Inheritance

Avoid deep inheritance hierarchies when composition is a better choice.

---

### Duplicating Code

Repeated code increases maintenance effort and the chance of bugs.

Extract common functionality into reusable methods or helper classes.

---

### Using Poor Names

Avoid unclear names such as:

```python
x

temp

obj

data
```

Use meaningful names that describe their purpose.
