# 🏛️ SOLID Principles in Python

**SOLID** is a set of **five object-oriented design principles** that help developers write software that is:

- Easy to understand
- Easy to maintain
- Easy to extend
- Easy to test
- Less prone to bugs

The SOLID principles were introduced by **Robert C. Martin (Uncle Bob)** and are considered one of the foundations of modern object-oriented software design.

Rather than focusing on syntax, SOLID focuses on **how classes and objects should be designed**.

# 💻 Quick Example

Without SOLID:

```python
class Report:

    def generate(self):
        print("Generating report")

    def save(self):
        print("Saving report")

    def email(self):
        print("Sending report by email")
```

This class has multiple responsibilities.

Using SOLID:

```python
class Report:

    def generate(self):
        print("Generating report")


class ReportSaver:

    def save(self):
        print("Saving report")


class ReportEmailer:

    def email(self):
        print("Sending report by email")
```

Each class now has a single responsibility.

# 📖 Topics Covered

## 1. What is SOLID?

SOLID is an acronym representing five software design principles.

| Letter | Principle |
|---------|-----------|
| **S** | Single Responsibility Principle (SRP) |
| **O** | Open/Closed Principle (OCP) |
| **L** | Liskov Substitution Principle (LSP) |
| **I** | Interface Segregation Principle (ISP) |
| **D** | Dependency Inversion Principle (DIP) |

Together, these principles help create flexible and maintainable object-oriented applications.

# 3. Single Responsibility Principle (SRP)

**Definition**

> **A class should have only one reason to change.**

A class should perform **one responsibility** only.

### Example

```python
class Invoice:

    def calculate_total(self):
        print("Calculating total")


class InvoicePrinter:

    def print_invoice(self):
        print("Printing invoice")
```

The `Invoice` class calculates totals, while `InvoicePrinter` handles printing.

---

# 4. Open/Closed Principle (OCP)

**Definition**

> **Software entities should be open for extension but closed for modification.**

Instead of changing existing code, extend it by adding new classes or behavior.

### Example

```python
class Payment:

    def pay(self):
        pass


class CreditCardPayment(Payment):

    def pay(self):
        print("Paid using credit card")


class UpiPayment(Payment):

    def pay(self):
        print("Paid using UPI")
```

New payment methods can be added without modifying existing classes.

---

# 5. Liskov Substitution Principle (LSP)

**Definition**

> **Objects of a derived class should be replaceable with objects of the base class without changing program correctness.**

A child class should behave like its parent.

### Example

```python
class Bird:

    def make_sound(self):
        print("Bird makes a sound")


class Sparrow(Bird):

    def make_sound(self):
        print("Chirp")
```

`Sparrow` can be used wherever a `Bird` is expected.

---

# 6. Interface Segregation Principle (ISP)

**Definition**

> **Clients should not be forced to depend on methods they do not use.**

Instead of creating one large interface, create smaller, focused interfaces.

### Example

```python
class Printer:

    def print_document(self):
        print("Printing...")


class Scanner:

    def scan_document(self):
        print("Scanning...")
```

Separate classes have separate responsibilities.

---

# 7. Dependency Inversion Principle (DIP)

**Definition**

> **High-level modules should not depend on low-level modules. Both should depend on abstractions.**

Depend on interfaces or abstract classes rather than concrete implementations.

### Example

```python
class Keyboard:

    def type(self):
        print("Typing...")


class Computer:

    def __init__(self, keyboard):
        self.keyboard = keyboard

    def work(self):
        self.keyboard.type()

keyboard = Keyboard()

computer = Computer(keyboard)

computer.work()
```

The `Computer` depends on an abstraction (a keyboard object) instead of creating one internally.

---

# 8. Overview of SOLID Principles

| Principle | Purpose |
|-----------|---------|
| SRP | One responsibility per class |
| OCP | Extend behavior without modifying existing code |
| LSP | Child classes should correctly replace parent classes |
| ISP | Small, focused interfaces |
| DIP | Depend on abstractions, not concrete classes |

```python
class Notification:

    def send(self):
        pass


class EmailNotification(Notification):

    def send(self):
        print("Email Sent")


class SMSNotification(Notification):

    def send(self):
        print("SMS Sent")
```

## Example 4: Shapes

```python
class Shape:

    def area(self):
        pass


class Rectangle(Shape):

    def area(self):
        return 20


class Circle(Shape):

    def area(self):
        return 50

print(Rectangle().area())
print(Circle().area())
```

# 12. Common Mistakes

Avoid these common mistakes while applying SOLID principles.

### Putting Too Much Logic in One Class

Incorrect

```python
class Student:

    def save(self):
        ...

    def print(self):
        ...

    def email(self):
        ...
```

Instead, split responsibilities into separate classes.

---

### Modifying Existing Classes Repeatedly

Rather than changing existing classes whenever new functionality is needed, extend them through inheritance or composition.

---

### Ignoring Abstractions

Avoid tightly coupling classes to specific implementations.

Prefer depending on abstract classes or interfaces.

---

### Creating Large Interfaces

Keep interfaces small and focused so that classes only implement methods they actually need.

---

# 13. Best Practices

Follow these recommended practices when applying SOLID principles.

- Give each class a single responsibility.
- Extend existing code instead of modifying it whenever possible.
- Ensure subclasses can replace their parent classes without unexpected behavior.
- Design small, focused interfaces.
- Depend on abstractions instead of concrete implementations.
- Prefer composition over inheritance when appropriate.
- Refactor large classes into smaller, reusable components.
- Write unit tests to verify class behavior.
- Keep classes loosely coupled and highly cohesive.
