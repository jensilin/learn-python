# 🔗 Coupling in Python

**Coupling** is a software design principle that describes **how strongly one class or module depends on another**.

In object-oriented programming, the goal is to achieve **low coupling**, where classes interact with each other while remaining as independent as possible.

In simple words:

> **Coupling measures the level of dependency between two classes or modules.**

Classes with **low coupling** are easier to maintain, modify, test, and reuse than classes with **high coupling**.

Coupling is one of the most important concepts in software engineering and is closely related to the **SOLID Principles**, especially the **Dependency Inversion Principle (DIP)**.

# 💻 Quick Example

Without low coupling (High Coupling):

```python
class Engine:

    def start(self):
        print("Engine started")


class Car:

    def __init__(self):
        self.engine = Engine()

    def start(self):
        self.engine.start()

car = Car()

car.start()
```

Here, the `Car` class creates its own `Engine` object, making it tightly dependent on that specific implementation.

With low coupling:

```python
class Engine:

    def start(self):
        print("Engine started")


class Car:

    def __init__(self, engine):
        self.engine = engine

    def start(self):
        self.engine.start()

engine = Engine()

car = Car(engine)

car.start()
```

The `Car` class now depends on an external object, making it more flexible.

# 📖 Topics Covered

## 1. What is Coupling?

Coupling measures the dependency between classes or modules.

When one class relies heavily on another class, they are said to be tightly coupled.

Example:

```python
class Printer:

    def print_document(self):
        print("Printing...")


class Office:

    def __init__(self):
        self.printer = Printer()
```

The `Office` class is tightly coupled to the `Printer` class.

---

## 2. Why is Coupling Important?

Low coupling helps us:

- Improve flexibility.
- Simplify maintenance.
- Reduce dependencies.
- Increase code reuse.
- Improve testing.
- Make software easier to extend.
- Reduce bugs.

---

## 3. High Coupling

High coupling occurs when classes depend heavily on each other.

Example:

```python
class Database:

    def connect(self):
        print("Database Connected")


class Application:

    def __init__(self):
        self.database = Database()

    def start(self):
        self.database.connect()

application = Application()

application.start()
```

### Output

```text
Database Connected
```

If the database implementation changes, the `Application` class may also need to change.

---

## 4. Low Coupling

Low coupling reduces dependencies between classes.

Example:

```python
class Database:

    def connect(self):
        print("Database Connected")


class Application:

    def __init__(self, database):
        self.database = database

    def start(self):
        self.database.connect()

database = Database()

application = Application(database)

application.start()
```

### Output

```text
Database Connected
```

The `Application` class can now work with any object that provides a `connect()` method.

---

## 5. High Coupling vs Low Coupling

| High Coupling | Low Coupling |
|---------------|--------------|
| Strong dependency | Weak dependency |
| Difficult to maintain | Easier to maintain |
| Difficult to test | Easier to test |
| Less reusable | More reusable |
| Changes affect multiple classes | Changes are isolated |

---

## 6. Coupling and Cohesion

Coupling and cohesion are closely related.

| Cohesion | Coupling |
|----------|----------|
| Measures relationships **within** a class | Measures dependencies **between** classes |
| High cohesion is desirable | Low coupling is desirable |
| Focuses on a single responsibility | Focuses on reducing dependencies |

A good software design aims for:

- **High Cohesion**
- **Low Coupling**

---

## 7. Dependency Injection

One common way to achieve low coupling is through **Dependency Injection**.

Instead of creating dependent objects inside a class, pass them from outside.

```python
class Logger:

    def log(self):
        print("Logging...")


class Application:

    def __init__(self, logger):
        self.logger = logger

    def run(self):
        self.logger.log()

logger = Logger()

application = Application(logger)

application.run()
```

### Output

```text
Logging...
```

## 8. Practical Examples

### Example 1: Payment Gateway

```python
class PaymentGateway:

    def pay(self):
        print("Payment Successful")


class ShoppingCart:

    def __init__(self, gateway):
        self.gateway = gateway

    def checkout(self):
        self.gateway.pay()

gateway = PaymentGateway()

cart = ShoppingCart(gateway)

cart.checkout()
```

### Output

```text
Payment Successful
```

---

### Example 2: Email Service

```python
class EmailService:

    def send(self):
        print("Email Sent")


class Notification:

    def __init__(self, email_service):
        self.email_service = email_service

    def notify(self):
        self.email_service.send()

service = EmailService()

notification = Notification(service)

notification.notify()
```

### Output

```text
Email Sent
```

---

### Example 3: Authentication

```python
class Authentication:

    def login(self):
        print("Login Successful")


class Website:

    def __init__(self, auth):
        self.auth = auth

    def start(self):
        self.auth.login()

auth = Authentication()

website = Website(auth)

website.start()
```

### Output

```text
Login Successful
```

---

### Example 4: Logger

```python
class Logger:

    def write(self):
        print("Writing Log")


class Server:

    def __init__(self, logger):
        self.logger = logger

    def execute(self):
        self.logger.write()

logger = Logger()

server = Server(logger)

server.execute()
```

### Output

```text
Writing Log
```

### Example 5: Printer

```python
class Printer:

    def print_document(self):
        print("Printing...")


class Employee:

    def __init__(self, printer):
        self.printer = printer

    def print_report(self):
        self.printer.print_document()

printer = Printer()

employee = Employee(printer)

employee.print_report()
```

### Output

## 10. Common Mistakes

Avoid these common mistakes while designing loosely coupled systems.

### Creating Objects Inside Classes

Incorrect

```python
class Car:

    def __init__(self):
        self.engine = Engine()
```

This tightly couples `Car` to `Engine`.

Better

```python
class Car:

    def __init__(self, engine):
        self.engine = engine
```

---

### Depending on Concrete Classes

Avoid depending directly on specific implementations.

Instead, depend on abstractions or objects with the required behavior.

---

### Sharing Too Much Knowledge

Classes should communicate only through well-defined interfaces or methods.

Avoid exposing unnecessary internal details.

---

### Ignoring Dependency Injection

Passing dependencies through constructors or methods usually results in more flexible and testable code.

---

## 11. Best Practices

Follow these recommended practices when designing loosely coupled software.

- Prefer low coupling between classes.
- Use dependency injection whenever appropriate.
- Depend on abstractions instead of concrete implementations.
- Keep communication between classes simple.
- Minimize unnecessary dependencies.
- Design reusable components.
- Write independent, testable classes.
- Combine low coupling with high cohesion for better software design.
