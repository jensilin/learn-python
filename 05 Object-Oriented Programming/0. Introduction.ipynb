# 💻 Quick Example

Without OOP:

```python
name = "Alice"
balance = 5000

balance += 1000

print(balance)
```

### Output

```text
6000
```

Using OOP:

```python
class BankAccount:

    def __init__(self, name, balance):

        self.name = name
        self.balance = balance

    def deposit(self, amount):

        self.balance += amount


account = BankAccount("Alice", 5000)

account.deposit(1000)

print(account.balance)
```

### Output

```text
6000
```

The OOP version groups related data and behavior together, making the code easier to understand and maintain.

# 📖 What is Object-Oriented Programming?

Object-Oriented Programming (OOP) is a programming paradigm that models software using **objects**.

An object combines:

- **Data (Attributes)** → Information about the object.
- **Behavior (Methods)** → Actions the object can perform.

For example:

A **Car** object may contain:

Attributes:

- Brand
- Model
- Color
- Speed

Methods:

- Start()
- Stop()
- Accelerate()
- Brake()

This closely resembles how objects exist in the real world.

# 🧩 Core Concepts of OOP

The four fundamental pillars of Object-Oriented Programming are:

### 1. Encapsulation

Bundling data and methods together while protecting internal data from direct access.

---

### 2. Inheritance

Creating new classes from existing classes to reuse functionality.

---

### 3. Polymorphism

Allowing the same interface to perform different behaviors depending on the object.

---

### 4. Abstraction

Hiding implementation details and exposing only the necessary functionality.

These four principles form the foundation of object-oriented software design.

# ⚠️ Common Mistakes

Beginners often make these mistakes while learning OOP.

### Thinking Everything Must Be a Class

Not every problem requires object-oriented programming. Use OOP where it improves code organization and reusability.

---

### Creating Large Classes

Avoid placing too many responsibilities inside a single class.

Keep classes focused on one purpose.

---

### Ignoring Encapsulation

Protect important data instead of allowing unrestricted access.

---

### Overusing Inheritance

Use inheritance only when there is a true **"is-a"** relationship.

Otherwise, prefer composition.

---

### Using Poor Names

Choose descriptive names for classes, methods, and variables.

Good examples:

```python
Student

BankAccount

ShoppingCart

Employee
```

Avoid names like:

```python
Data

Object

Temp

Test
```
