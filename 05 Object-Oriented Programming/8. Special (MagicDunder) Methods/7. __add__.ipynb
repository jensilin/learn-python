# ➕ `__add__()` Method in Python

The **`__add__()`** method is a special **magic (dunder) method** that defines **how the addition operator (`+`) behaves for objects of a custom class**.

Whenever the `+` operator is used between two objects, Python automatically calls the `__add__()` method.

By implementing `__add__()`, developers can customize how objects are combined, making user-defined classes behave like Python's built-in data types.

# 💻 Quick Example

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __str__(self):
        return str(self.value)

num1 = Number(10)
num2 = Number(20)

print(num1 + num2)
```

### Output

```text
30
```

Python automatically calls the `__add__()` method when the `+` operator is used.

# 📖 Topics Covered

## 1. What is `__add__()`?

The `__add__()` method defines how objects should behave when the addition (`+`) operator is used.

When Python encounters:

```python
object1 + object2
```

it internally performs:

```python
object1.__add__(object2)
```

---

## 2. Why Use `__add__()`?

The `__add__()` method helps us:

- Customize the `+` operator.
- Improve code readability.
- Support operator overloading.
- Make custom classes behave like built-in types.
- Build reusable object-oriented applications.

---

## 3. Default Behavior Without `__add__()`

If `__add__()` is not implemented, using the `+` operator on custom objects raises an error.

```python
class Number:

    def __init__(self, value):
        self.value = value

num1 = Number(10)
num2 = Number(20)

print(num1 + num2)
```

### Output

```text
TypeError:
unsupported operand type(s) for +:
'Number' and 'Number'
```

---

## 4. Implementing `__add__()`

By implementing `__add__()`, we define how objects are added.

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __str__(self):
        return str(self.value)

num1 = Number(15)
num2 = Number(25)

print(num1 + num2)
```

### Output

```text
40
```

---

## 5. Returning a New Object

It is common for `__add__()` to return a **new object** instead of modifying existing ones.

```python
class Point:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(2, 3)
p2 = Point(4, 5)

print(p1 + p2)
```

### Output

```text
(6, 8)
```

---

## 6. Type Checking Inside `__add__()`

It is good practice to verify that the second operand is of the expected type.

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):

        if not isinstance(other, Number):
            return NotImplemented

        return Number(self.value + other.value)

    def __str__(self):
        return str(self.value)

num1 = Number(10)
num2 = Number(5)

print(num1 + num2)
```

### Output

```text
15
```

Returning `NotImplemented` allows Python to handle unsupported operands correctly.

## 7. Python Calls `__add__()` Automatically

There is no need to call `__add__()` directly.

```python
class Demo:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Demo(self.value + other.value)

    def __str__(self):
        return str(self.value)

demo1 = Demo(7)
demo2 = Demo(8)

print(demo1 + demo2)
```

### Output

```text
15
```

## 10. Common Mistakes

Avoid these common mistakes while implementing `__add__()`.

### Forgetting to Return a Value

Incorrect

```python
def __add__(self, other):
    self.value + other.value
```

Correct

```python
def __add__(self, other):
    return self.value + other.value
```

---

### Ignoring Type Checking

Incorrect

```python
def __add__(self, other):
    return self.value + other.value
```

If `other` is not the expected type, an `AttributeError` may occur.

Better:

```python
if not isinstance(other, Number):
    return NotImplemented
```

---

### Modifying the Current Object

Avoid modifying the existing object unless that behavior is intended.

Prefer returning a new object.

---

### Calling `__add__()` Directly

Instead of:

```python
num1.__add__(num2)
```

Prefer:

```python
num1 + num2
```

Python automatically invokes `__add__()`.

---

## 11. Best Practices

Follow these recommended practices when implementing `__add__()`.

- Return a new object whenever appropriate.
- Keep addition behavior intuitive.
- Use `isinstance()` for type checking.
- Return `NotImplemented` for unsupported operand types.
- Avoid modifying existing objects unnecessarily.
- Ensure the returned object is meaningful.
- Use the `+` operator instead of calling `__add__()` directly.
