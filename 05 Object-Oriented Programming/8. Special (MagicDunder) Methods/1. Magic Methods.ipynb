# ✨ Magic Methods (Dunder Methods) in Python

**Magic Methods**, also known as **Dunder Methods (Double Underscore Methods)**, are special methods in Python that begin and end with **double underscores (`__`)**.

Examples include:

- `__init__()`
- `__str__()`
- `__repr__()`
- `__len__()`
- `__add__()`
- `__eq__()`

These methods allow developers to **customize the behavior of objects** and integrate user-defined classes seamlessly with Python's built-in features.

Magic methods are automatically invoked by Python in response to specific operations, such as creating objects, printing objects, comparing objects, performing arithmetic operations, or calculating the length of an object.

# 💻 Quick Example

```python
class Student:

    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student: {self.name}"

student = Student("Alice")

print(student)
```

### Output

```text
Student: Alice
```

Python automatically calls the `__str__()` method when the object is printed.

# 📖 Topics Covered

## 1. What are Magic Methods?

Magic methods are predefined methods that allow Python objects to respond to built-in operations.

For example:

- Creating an object
- Printing an object
- Comparing objects
- Performing arithmetic
- Finding object length

Instead of calling these methods directly, Python calls them automatically.

Example:

```python
class Demo:

    def __str__(self):
        return "Hello from Demo"

obj = Demo()

print(obj)
```

### Output

```text
Hello from Demo
```

---

## 2. Why Use Magic Methods?

Magic methods help us:

- Customize object behavior.

- Integrate with Python's object model.

## 3. Why Are They Called "Dunder" Methods?

"Dunder" is short for **Double UNDERscore**.

Every magic method starts and ends with two underscores.

Examples:

```python
__init__()
__str__()
__repr__()
__len__()
__add__()
__eq__()
```

---

## 4. How Magic Methods Work

Python automatically calls magic methods when specific operations occur.

Example:

```python
class Book:

    def __init__(self, title):
        self.title = title

book = Book("Python Basics")
```

Internally, Python executes:

```python
book = Book.__new__(Book)
Book.__init__(book, "Python Basics")
```

Normally, developers do not call these methods directly.

---

## 5. Commonly Used Magic Methods

| Magic Method | Purpose |
|--------------|---------|
| `__init__()` | Initialize an object |
| `__str__()` | User-friendly string representation |
| `__repr__()` | Official object representation |
| `__len__()` | Return object length |
| `__add__()` | Overload the `+` operator |
| `__sub__()` | Overload the `-` operator |
| `__mul__()` | Overload the `*` operator |
| `__eq__()` | Compare objects using `==` |
| `__lt__()` | Compare objects using `<` |
| `__gt__()` | Compare objects using `>` |

These methods make custom objects behave like Python's built-in objects.

## 6. Python Calls Magic Methods Automatically

Example:

```python
class Student:

    def __len__(self):
        return 5

student = Student()

print(len(student))
```

### Output

```text
5
```

Python automatically calls `__len__()` when `len()` is used.

## 9. Practical Examples

### Example 1: Object Initialization

```python
class Student:

    def __init__(self, name):
        self.name = name

student = Student("John")

print(student.name)
```

### Output

```text
John
```

---

### Example 2: Printing an Object

```python
class Student:

    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name

student = Student("Alice")

print(student)
```

### Output

```text
Alice
```

---

### Example 3: Finding Object Length

```python
class Playlist:

    def __len__(self):
        return 12

playlist = Playlist()

print(len(playlist))
```

### Output

```text
12
```

---

### Example 4: Operator Overloading

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return self.value + other.value

n1 = Number(10)
n2 = Number(15)

print(n1 + n2)
```

### Output

```text
25
```

---

### Example 5: Comparing Objects

```python
class Student:

    def __init__(self, marks):
        self.marks = marks

    def __eq__(self, other):
        return self.marks == other.marks

s1 = Student(90)
s2 = Student(90)

print(s1 == s2)
```

### Output

```text
True
```

## 11. Common Mistakes

Avoid these common mistakes while working with magic methods.

### Calling Magic Methods Directly

Instead of:

```python
student.__len__()
```

Prefer:

```python
len(student)
```

Python automatically invokes the appropriate magic method.

---

### Returning the Wrong Type

Incorrect

```python
def __len__(self):
    return "5"
```

`__len__()` must return an integer.

Correct

```python
def __len__(self):
    return 5
```

---

### Forgetting to Return a Value

Incorrect

```python
def __str__(self):
    print("Student")
```

`__str__()` should return a string, not print it.

Correct

```python
def __str__(self):
    return "Student"
```

---

### Overloading Operators Unnecessarily

Only overload operators when the behavior is intuitive and meaningful.

---

## 12. Best Practices

Follow these recommended practices when working with magic methods.

- Implement only the magic methods your class needs.
- Keep behavior intuitive and consistent with built-in types.
- Return the correct data types.
- Use built-in functions instead of calling magic methods directly.
- Write clear and maintainable implementations.
- Follow Python's data model conventions.
- Document custom behavior when appropriate.
