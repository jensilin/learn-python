# 🖨️ `__str__()` Method in Python

The **`__str__()`** method is a special **magic (dunder) method** that defines the **human-readable string representation** of an object.

Whenever an object is passed to functions such as:

- `print()`
- `str()`

Python automatically calls the `__str__()` method.

Without a `__str__()` method, Python displays the object's default representation, which usually includes the class name and memory address.

Implementing `__str__()` makes objects easier to read and understand.

# 💻 Quick Example

```python
class Student:

    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student Name: {self.name}"

student = Student("Alice")

print(student)
```

### Output

```text
Student Name: Alice
```

Python automatically calls `__str__()` when printing the object.

# 📖 Topics Covered

## 1. What is `__str__()`?

The `__str__()` method returns a **user-friendly string representation** of an object.

It is mainly intended for end users.

Example:

```python
class Book:

    def __str__(self):
        return "Python Programming"

book = Book()

print(book)
```

### Output

```text
Python Programming
```

---

## 2. Why Use `__str__()`?

The `__str__()` method helps us:

- Display meaningful object information.
- Improve readability.
- Simplify debugging.
- Make printed objects easier to understand.
- Produce user-friendly output.

---

## 3. Default Object Representation

Without `__str__()`, Python displays the default object representation.

```python
class Student:
    pass

student = Student()

print(student)
```

### Output

```text
<__main__.Student object at 0x...>
```

The memory address will differ each time the program runs.

---

## 4. Using `__str__()`

By implementing `__str__()`, we control what is displayed.

```python
class Student:

    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name

student = Student("John")

print(student)
```

### Output

```text
John
```

---

## 5. Python Calls `__str__()` Automatically

Python automatically invokes `__str__()` in many situations.

```python
class Car:

    def __str__(self):
        return "Toyota Corolla"

car = Car()

print(car)
print(str(car))
```

### Output

```text
Toyota Corolla
Toyota Corolla
```

---

## 6. `__str__()` Must Return a String

The return value must always be a string.

```python
class Number:

    def __str__(self):
        return "100"

number = Number()

print(number)
```

### Output

```text
100
```

Returning any non-string value raises a `TypeError`.

---

## 7. `__str__()` vs `__repr__()`

| `__str__()` | `__repr__()` |
|-------------|--------------|
| User-friendly representation | Developer-oriented representation |
| Used by `print()` and `str()` | Used by `repr()` and the interactive interpreter |
| Focuses on readability | Focuses on debugging and object recreation when possible |
| Optional | Optional but recommended |

If `__str__()` is not defined, Python may fall back to `__repr__()`.

## 8. Practical Examples

### Example 1: Student

```python
class Student:

    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student: {self.name}"

student = Student("Alice")

print(student)
```

### Output

```text
Student: Alice
```

---

### Example 2: Employee

```python
class Employee:

    def __init__(self, name, department):
        self.name = name
        self.department = department

    def __str__(self):
        return f"{self.name} - {self.department}"

employee = Employee("John", "IT")

print(employee)
```

### Output

```text
John - IT
```

---

### Example 3: Book

```python
class Book:

    def __init__(self, title):
        self.title = title

    def __str__(self):
        return f"Book: {self.title}"

book = Book("Python Basics")

print(book)
```

### Output

```text
Book: Python Basics
```

---

### Example 4: Bank Account

```python
class BankAccount:

    def __init__(self, account_number):
        self.account_number = account_number

    def __str__(self):
        return f"Account Number: {self.account_number}"

account = BankAccount("ACC123")

print(account)
```

### Output

```text
Account Number: ACC123
```

---

### Example 5: Rectangle

```python
class Rectangle:

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def __str__(self):
        return f"Rectangle({self.length} × {self.width})"

rectangle = Rectangle(10, 5)

print(rectangle)
```

### Output

```text
Rectangle(10 × 5)
```

## 10. Common Mistakes

Avoid these common mistakes while using `__str__()`.

### Returning a Non-String Value

Incorrect

```python
class Student:

    def __str__(self):
        return 100
```

### Output

```text
TypeError
```

Always return a string.

---

### Printing Instead of Returning

Incorrect

```python
def __str__(self):
    print("Student")
```

Correct

```python
def __str__(self):
    return "Student"
```

---

### Forgetting to Implement `__str__()`

Without `__str__()`, object output becomes less informative.

```text
<__main__.Student object at 0x...>
```

---

### Calling `__str__()` Directly

Instead of:

```python
student.__str__()
```

Prefer:

```python
print(student)
```

Python automatically calls `__str__()`.

---

## 11. Best Practices

Follow these recommended practices when implementing `__str__()`.

- Always return a string.
- Keep the output concise and meaningful.
- Include the most important object information.
- Make the representation easy for users to read.
- Avoid exposing sensitive information.
- Use f-strings for clean formatting.
- Let Python call `__str__()` automatically.
