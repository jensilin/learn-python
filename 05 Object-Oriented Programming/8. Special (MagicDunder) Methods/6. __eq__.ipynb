# ⚖️ `__eq__()` Method in Python

The **`__eq__()`** method is a special **magic (dunder) method** that defines **how two objects are compared for equality (`==`)**.

Whenever the equality operator (`==`) is used between two objects, Python automatically calls the `__eq__()` method.

By default, Python compares whether two object references point to the **same object in memory**. By implementing `__eq__()`, we can compare objects based on **their data (attributes)** instead of their memory addresses.

# 💻 Quick Example

```python
class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __eq__(self, other):
        return self.marks == other.marks

student1 = Student("Alice", 90)
student2 = Student("Bob", 90)

print(student1 == student2)
```

### Output

```text
True
```

Python automatically calls the `__eq__()` method when the `==` operator is used.

# 📖 Topics Covered

## 1. What is `__eq__()`?

The `__eq__()` method defines how objects should be compared for equality.

Whenever Python encounters:

```python
object1 == object2
```

it internally calls:

```python
object1.__eq__(object2)
```

---

## 2. Why Use `__eq__()`?

The `__eq__()` method helps us:

- Compare objects by their values.
- Improve readability.
- Simplify business logic.
- Support object-oriented design.
- Make custom classes behave like built-in types.

---

## 3. Default Equality Comparison

Without `__eq__()`, Python compares object identities.

```python
class Student:

    def __init__(self, name):
        self.name = name

student1 = Student("Alice")
student2 = Student("Alice")

print(student1 == student2)
```

### Output

```text
False
```

Even though both objects contain the same data, they are different objects in memory.

---

## 4. Implementing `__eq__()`

By implementing `__eq__()`, objects can be compared based on their attributes.

```python
class Student:

    def __init__(self, name):
        self.name = name

    def __eq__(self, other):
        return self.name == other.name

student1 = Student("Alice")
student2 = Student("Alice")

print(student1 == student2)
```

### Output

```text
True
```

---

## 5. Difference Between `==` and `is`

| `==` | `is` |
|------|------|
| Compares values | Compares object identity |
| Uses `__eq__()` | Checks memory location |
| Can be customized | Cannot be customized |

Example:

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]

print(list1 == list2)
print(list1 is list2)
```

### Output

```text
True
False
```

---

## 6. Type Checking Inside `__eq__()`

It is good practice to verify that the other object is of the expected type.

```python
class Student:

    def __init__(self, marks):
        self.marks = marks

    def __eq__(self, other):

        if not isinstance(other, Student):
            return NotImplemented

        return self.marks == other.marks

student = Student(90)

print(student == Student(90))
```

### Output

```text
True
```

Returning `NotImplemented` allows Python to handle unsupported comparisons correctly.

---

## 7. Python Calls `__eq__()` Automatically

There is no need to call `__eq__()` directly.

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __eq__(self, other):
        return self.value == other.value

number1 = Number(10)
number2 = Number(10)

print(number1 == number2)
```

### Output

```text
True
```

## 10. Common Mistakes

Avoid these common mistakes while implementing `__eq__()`.

### Forgetting to Return a Boolean

Incorrect

```python
def __eq__(self, other):
    print(self.name == other.name)
```

Correct

```python
def __eq__(self, other):
    return self.name == other.name
```

---

### Ignoring Type Checking

Incorrect

```python
def __eq__(self, other):
    return self.name == other.name
```

If `other` is not the expected type, an `AttributeError` may occur.

Better:

```python
if not isinstance(other, Student):
    return NotImplemented
```

---

### Comparing Object Identity Instead of Values

Incorrect

```python
return self is other
```

This compares memory addresses rather than object data.

---

### Calling `__eq__()` Directly

Instead of:

```python
student1.__eq__(student2)
```

Prefer:

```python
student1 == student2
```

Python automatically invokes `__eq__()`.

---

## 11. Best Practices

Follow these recommended practices when implementing `__eq__()`.

- Compare meaningful object attributes.
- Return a Boolean value or `NotImplemented`.
- Use `isinstance()` for type checking.
- Keep comparisons simple and efficient.
- Ensure equality logic is consistent throughout the class.
- Avoid comparing unrelated object types.
- Use the `==` operator instead of calling `__eq__()` directly.

> **Note:** The **`__eq__()`** method allows developers to define what it means for two objects to be considered equal. Instead of relying on Python's default identity comparison, `__eq__()` enables objects to be compared based on their attributes or state. When implemented with proper type checking and meaningful comparison logic, it makes custom classes behave naturally with the `==` operator and improves the readability, correctness, and maintainability of object-oriented programs.
