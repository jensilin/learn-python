# 🔍 `__repr__()` Method in Python

The **`__repr__()`** method is a special **magic (dunder) method** that defines the **official string representation** of an object.

It is primarily intended for:

- Developers
- Debugging
- Logging
- Interactive Python sessions

Whenever an object is passed to:

- `repr()`
- The Python interpreter (interactive shell)

Python automatically calls the `__repr__()` method.

A good `__repr__()` implementation should return a string that is **clear, unambiguous, and, whenever possible, capable of recreating the object**.

# 💻 Quick Example

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"Student(name='{self.name}', age={self.age})"

student = Student("Alice", 20)

print(repr(student))
```

### Output

```text
Student(name='Alice', age=20)
```

Python automatically calls `__repr__()` when `repr()` is used.

# 📖 Topics Covered

## 1. What is `__repr__()`?

The `__repr__()` method returns the **official string representation** of an object.

It is mainly intended for developers and debugging.

Example:

```python
class Book:

    def __repr__(self):
        return "Book('Python Basics')"

book = Book()

print(repr(book))
```

### Output

```text
Book('Python Basics')
```

---

## 2. Why Use `__repr__()`?

The `__repr__()` method helps us:

- Improve debugging.
- Display detailed object information.
- Create meaningful log messages.
- Represent objects clearly.
- Make objects easier to inspect.

---

## 3. Default Object Representation

Without `__repr__()`, Python displays a default representation.

```python
class Student:
    pass

student = Student()

print(repr(student))
```

### Output

```text
<__main__.Student object at 0x...>
```

The memory address changes each time the program runs.

---

## 4. Using `__repr__()`

By implementing `__repr__()`, we control the object's official representation.

```python
class Student:

    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f"Student('{self.name}')"

student = Student("John")

print(repr(student))
```

### Output

```text
Student('John')
```

---

## 5. Python Calls `__repr__()` Automatically

Python automatically invokes `__repr__()` in several situations.

```python
class Car:

    def __repr__(self):
        return "Car('Toyota')"

car = Car()

print(repr(car))
```

### Output

```text
Car('Toyota')
```

In the Python interactive interpreter, simply typing:

```python
car
```

also displays:

```text
Car('Toyota')
```

---

## 6. `__repr__()` Should Return a String

The return value must always be a string.

```python
class Number:

    def __repr__(self):
        return "100"

number = Number()

print(repr(number))
```

### Output

```text
100
```

Returning any non-string value raises a `TypeError`.

---

## 7. `__repr__()` vs `__str__()`

| `__repr__()` | `__str__()` |
|---------------|-------------|
| Official representation | User-friendly representation |
| Used for debugging | Used for display |
| Called by `repr()` | Called by `print()` and `str()` |
| Should be unambiguous | Should be readable |
| May recreate the object | Focuses on presentation |

Example:

```python
class Student:

    def __repr__(self):
        return "Student('Alice')"

    def __str__(self):
        return "Alice"

student = Student()

print(repr(student))
print(student)
```

### Output

```text
Student('Alice')
Alice
```

---

## 8. If `__str__()` Is Missing

If `__str__()` is not implemented, Python falls back to `__repr__()`.

```python
class Book:

    def __repr__(self):
        return "Book('Python')"

book = Book()

print(book)
```

### Output

```text
Book('Python')
```

## 9. Practical Examples

### Example 1: Student

```python
class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __repr__(self):
        return f"Student(name='{self.name}', marks={self.marks})"

student = Student("Alice", 95)

print(repr(student))
```

### Output

```text
Student(name='Alice', marks=95)
```

---

### Example 2: Employee

```python
class Employee:

    def __init__(self, name, department):
        self.name = name
        self.department = department

    def __repr__(self):
        return f"Employee('{self.name}', '{self.department}')"

employee = Employee("John", "IT")

print(repr(employee))
```

### Output

```text
Employee('John', 'IT')
```

---

### Example 3: Rectangle

```python
class Rectangle:

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def __repr__(self):
        return f"Rectangle({self.length}, {self.width})"

rectangle = Rectangle(10, 5)

print(repr(rectangle))
```

### Output

```text
Rectangle(10, 5)
```

---

### Example 4: Product

```python
class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __repr__(self):
        return f"Product('{self.name}', {self.price})"

product = Product("Laptop", 65000)

print(repr(product))
```

### Output

```text
Product('Laptop', 65000)
```

---

### Example 5: Bank Account

```python
class BankAccount:

    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def __repr__(self):
        return (
            f"BankAccount('{self.account_number}', {self.balance})"
        )

account = BankAccount("ACC123", 5000)

print(repr(account))
```

### Output

```text
BankAccount('ACC123', 5000)
```

## 11. Common Mistakes

Avoid these common mistakes while using `__repr__()`.

### Returning a Non-String Value

Incorrect

```python
class Student:

    def __repr__(self):
        return 100
```

### Output

```text
TypeError
```

Always return a string.

---

### Printing Instead of Returning

Incorrect

```python
def __repr__(self):
    print("Student")
```

Correct

```python
def __repr__(self):
    return "Student()"
```

---

### Returning an Ambiguous Representation

Poor

```python
def __repr__(self):
    return "Student"
```

Better

```python
def __repr__(self):
    return "Student('Alice', 20)"
```

Include enough information to identify the object's state.

---

### Calling `__repr__()` Directly

Instead of:

```python
student.__repr__()
```

Prefer:

```python
repr(student)
```

Python automatically calls `__repr__()`.

---

## 12. Best Practices

Follow these recommended practices when implementing `__repr__()`.

- Always return a string.
- Make the representation unambiguous.
- Include important object attributes.
- Use a format similar to a constructor call when practical.
- Keep the output concise but informative.
- Avoid exposing sensitive information.
- Let Python invoke `__repr__()` automatically.
