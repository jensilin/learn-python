# 📏 `__len__()` Method in Python

The **`__len__()`** method is a special **magic (dunder) method** that defines the **length of an object**.

Whenever the built-in **`len()`** function is called on an object, Python automatically invokes the object's `__len__()` method.

Many built-in Python data types already implement this method:

- Lists
- Tuples
- Strings
- Dictionaries
- Sets

By implementing `__len__()` in custom classes, we can make our own objects work naturally with the `len()` function.

# 💻 Quick Example

```python
class Playlist:

    def __init__(self, songs):
        self.songs = songs

    def __len__(self):
        return len(self.songs)

playlist = Playlist(["Song A", "Song B", "Song C"])

print(len(playlist))
```

### Output

```text
3
```

Python automatically calls the `__len__()` method when `len()` is used.

# 📖 Topics Covered

## 1. What is `__len__()`?

The `__len__()` method returns the **number of items contained in an object**.

It is automatically called whenever `len()` is used.

Example:

```python
class Numbers:

    def __len__(self):
        return 5

numbers = Numbers()

print(len(numbers))
```

### Output

```text
5
```

---

The `__len__()` method helps us:

- Define the size of custom objects.
- Integrate custom classes with Python.
- Improve code readability.

## 4. Implementing `__len__()` in a Custom Class

```python
class Library:

    def __init__(self, books):
        self.books = books

    def __len__(self):
        return len(self.books)

library = Library(["Book1", "Book2", "Book3"])

print(len(library))
```

---

## 5. Python Calls `__len__()` Automatically

There is no need to call `__len__()` directly.

```python
class Basket:

    def __len__(self):
        return 4

basket = Basket()

print(len(basket))
```

### Output

```text
4
```

Internally, Python performs:

```python
basket.__len__()
```

However, developers should always use:

```python
len(basket)
```

---

## 6. `__len__()` Must Return an Integer

The `__len__()` method must return a **non-negative integer**.

```python
class Team:

    def __len__(self):
        return 8

team = Team()

print(len(team))
```

### Output

```text
8
```

Returning any other type raises a `TypeError`.

## 10. Common Mistakes

Avoid these common mistakes while implementing `__len__()`.

### Returning a String

Incorrect

```python
class Demo:

    def __len__(self):
        return "5"
```

### Output

```text
TypeError
```

Always return an integer.

---

### Returning a Negative Value

Incorrect

```python
class Demo:

    def __len__(self):
        return -1
```

### Output

```text
ValueError
```

The returned value must be a **non-negative integer**.

---

### Printing Instead of Returning

Incorrect

```python
def __len__(self):
    print(5)
```

Correct

```python
def __len__(self):
    return 5
```

---

### Calling `__len__()` Directly

Instead of:

```python
object.__len__()
```

Prefer:

```python
len(object)
```

Python automatically calls `__len__()`.

---

## 11. Best Practices

Follow these recommended practices when implementing `__len__()`.

- Always return a non-negative integer.
- Base the returned value on meaningful object data.
- Keep the implementation simple.
- Use `len()` inside `__len__()` when appropriate.
- Avoid performing expensive computations.
- Use `len(object)` instead of calling `__len__()` directly.
- Ensure the returned length accurately reflects the object's state.
