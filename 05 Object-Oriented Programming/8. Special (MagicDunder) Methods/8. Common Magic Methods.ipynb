# ✨ Common Magic (Dunder) Methods in Python

Python provides many **magic methods** (also called **dunder methods**, short for *double underscore methods*) that allow custom classes to work naturally with Python's built-in functions, operators, and language features.

These methods begin and end with double underscores (`__`), such as:

- `__init__`
- `__str__`
- `__repr__`
- `__len__`
- `__eq__`
- `__add__`

By implementing these methods, your custom objects can behave like built-in Python objects such as strings, lists, dictionaries, and numbers.

# 💻 Quick Example

```python
class Student:

    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name

student = Student("Alice")

print(student)
```

### Output

```text
Alice
```

Python automatically calls the `__str__()` method when `print()` is used.

## 4. Object Initialization

The `__init__()` method initializes an object.

```python
class Student:

    def __init__(self, name):
        self.name = name

student = Student("Alice")

print(student.name)
```

### Output

```text
Alice
```

---

## 5. String Representation

The `__str__()` method returns a user-friendly representation.

```python
class Book:

    def __str__(self):
        return "Python Programming"

book = Book()

print(book)
```

### Output

```text
Python Programming
```

---

## 6. Official Representation

The `__repr__()` method returns a developer-friendly representation.

```python
class Book:

    def __repr__(self):
        return "Book('Python Programming')"

book = Book()

print(repr(book))
```

### Output

```text
Book('Python Programming')
```

---

## 7. Length Calculation

The `__len__()` method allows objects to work with `len()`.

```python
class Team:

    def __len__(self):
        return 11

team = Team()

print(len(team))
```

### Output

```text
11
```

---

## 8. Equality Comparison

The `__eq__()` method defines how two objects are compared using `==`.

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __eq__(self, other):
        return self.value == other.value

num1 = Number(10)
num2 = Number(10)

print(num1 == num2)
```

### Output

```text
True
```

---

## 9. Operator Overloading

The `__add__()` method customizes the `+` operator.

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __str__(self):
        return str(self.value)

num1 = Number(20)
num2 = Number(30)

print(num1 + num2)
```

### Output

```text
50
```

---

## 10. Membership Testing

The `__contains__()` method defines how the `in` operator works.

```python
class Fruits:

    def __init__(self):
        self.items = ["Apple", "Banana", "Orange"]

    def __contains__(self, item):
        return item in self.items

fruits = Fruits()

print("Apple" in fruits)
```

### Output

```text
True
```

---

## 11. Making Objects Callable

The `__call__()` method allows an object to behave like a function.

```python
class Greeting:

    def __call__(self, name):
        return f"Hello, {name}!"

greet = Greeting()

print(greet("Alice"))
```

### Output

```text
Hello, Alice!
```

---

## 12. Boolean Evaluation

The `__bool__()` method defines the truth value of an object.

```python
class Wallet:

    def __init__(self, balance):
        self.balance = balance

    def __bool__(self):
        return self.balance > 0

wallet = Wallet(100)

print(bool(wallet))
```

### Output

```text
True
```

### Example 2: Shopping Cart

```python
class Cart:

    def __init__(self, items):
        self.items = items

    def __len__(self):
        return len(self.items)

cart = Cart(["Laptop", "Mouse"])

print(len(cart))
```

### Example 3: Product Comparison

```python
class Product:

    def __init__(self, code):
        self.code = code

    def __eq__(self, other):
        return self.code == other.code

print(Product("P101") == Product("P101"))
```

### Output

```text
True
```

### Example 4: Vector Addition

```python
class Vector:

    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return Vector(self.x + other.x)

    def __str__(self):
        return str(self.x)

vector1 = Vector(15)
vector2 = Vector(25)

print(vector1 + vector2)
```

### Output

```text
40
```

---

### Example 5: Callable Object

```python
class Calculator:

    def __call__(self, a, b):
        return a + b

calculator = Calculator()

print(calculator(10, 20))
```

### Output

```text
30
```

## 15. Common Mistakes

Avoid these common mistakes while working with magic methods.

### Calling Magic Methods Directly

Instead of:

```python
object.__len__()
```

Prefer:

```python
len(object)
```

---

Instead of:

```python
object.__str__()
```

Prefer:

```python
print(object)
```

---

Instead of:

```python
object.__eq__(other)
```

Prefer:

```python
object == other
```

---

### Returning Incorrect Types

Each magic method should return the expected data type.

Examples:

- `__len__()` → integer
- `__str__()` → string
- `__bool__()` → boolean
- `__eq__()` → boolean or `NotImplemented`

---

### Implementing Unnecessary Magic Methods

Only implement magic methods that make sense for your class.

For example, a `Student` class may not need `__add__()` unless adding two students has a meaningful interpretation.

---

### Ignoring Type Checking

For methods such as `__eq__()` and `__add__()`, use `isinstance()` and return `NotImplemented` when appropriate.

---

## 16. Best Practices

Follow these recommended practices when implementing magic methods.

- Implement only the methods your class needs.
- Follow Python's expected behavior for each method.
- Always return the correct data type.
- Keep implementations simple and efficient.
- Use meaningful logic for operator overloading.
- Return `NotImplemented` for unsupported operations.
- Let Python invoke magic methods automatically instead of calling them directly.
