# 🏗️ `__init__()` Method in Python

The **`__init__()`** method is one of the most important **magic (dunder) methods** in Python. It is known as the **constructor** because it is **automatically called whenever a new object is created**.

The primary purpose of the `__init__()` method is to **initialize an object's attributes with initial values**.

Unlike many programming languages, `__init__()` **does not create the object itself**. Object creation is handled by another special method called `__new__()`. After the object is created, Python automatically calls `__init__()` to initialize it.

# 💻 Quick Example

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

student = Student("Alice", 20)

print(student.name)
print(student.age)
```

### Output

```text
Alice
20
```

The `__init__()` method automatically initializes the object's attributes when the object is created.

# 📖 Topics Covered

## 1. What is `__init__()`?

The `__init__()` method is a special method that Python automatically calls immediately after creating an object.

Its primary purpose is to initialize the object's attributes.

Example:

```python
class Person:

    def __init__(self):
        print("Object Initialized")

person = Person()
```

### Output

```text
Object Initialized
```

---

## 2. Why Use `__init__()`?

The `__init__()` method helps us:

- Initialize object data.
- Assign instance variables.
- Avoid repetitive code.
- Improve code readability.
- Ensure every object starts with valid data.
- Simplify object creation.

---

## 3. Syntax

```python
class ClassName:

    def __init__(self, parameters):
        self.attribute = parameters
```

Here:

- `__init__` is the constructor.
- `self` refers to the current object.
- Additional parameters provide initial values.

---

## 4. Using `self` Inside `__init__()`

The `self` parameter refers to the current object.

It is used to store values as instance variables.

```python
class Student:

    def __init__(self, name):
        self.name = name

student = Student("John")

print(student.name)
```

### Output

```text
John
```

---

## 5. Constructor with Multiple Parameters

```python
class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

employee = Employee("Alice", 50000)

print(employee.name)
print(employee.salary)
```

### Output

```text
Alice
50000
```

---

## 6. Constructor with Default Values

Default values make parameters optional.

```python
class Student:

    def __init__(self, name="Guest"):
        self.name = name

student1 = Student()
student2 = Student("Bob")

print(student1.name)
print(student2.name)
```

### Output

```text
Guest
Bob
```

---

## 7. Every Object Gets Its Own Data

Each object stores its own instance variables.

```python
class Student:

    def __init__(self, name):
        self.name = name

student1 = Student("Alice")
student2 = Student("Bob")

print(student1.name)
print(student2.name)
```

### Output

```text
Alice
Bob
```

---

## 8. `__init__()` is Called Automatically

There is no need to call `__init__()` manually.

```python
class Demo:

    def __init__(self):
        print("Constructor Called")

demo = Demo()
```

### Output

```text
Constructor Called
```

Python automatically invokes the constructor during object creation.

---

## 9. Difference Between `__new__()` and `__init__()`

| `__new__()` | `__init__()` |
|-------------|--------------|
| Creates a new object | Initializes the object |
| Called before `__init__()` | Called after `__new__()` |
| Returns the object | Returns `None` |
| Rarely overridden | Frequently used |

## 12. Common Mistakes

Avoid these common mistakes while using `__init__()`.

### Forgetting `self`

Incorrect

```python
class Student:

    def __init__(name):
        self.name = name
```

Always include `self` as the first parameter.

---

### Forgetting `self.`

Incorrect

```python
class Student:

    def __init__(self, name):
        name = name
```

Correct

```python
self.name = name
```

---

### Returning a Value

Incorrect

```python
def __init__(self):
    return 10
```

### Output

```text
TypeError
```

The `__init__()` method must always return `None` implicitly.

---

### Calling `__init__()` Directly

Incorrect

```python
student.__init__("John")
```

Create objects normally:

```python
student = Student("John")
```
