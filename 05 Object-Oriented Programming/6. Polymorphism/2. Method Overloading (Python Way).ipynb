# ➕ Method Overloading (Python Way)

**Method overloading** is the ability to define **multiple methods with the same name but different parameter lists**.

Unlike languages such as **Java** and **C++**, **Python does not support traditional method overloading**. If multiple methods with the same name are defined in a class, **only the last definition is retained**, and the earlier ones are overwritten.

Instead, Python achieves similar behavior using:

- Default arguments
- Variable-length positional arguments (`*args`)
- Variable-length keyword arguments (`**kwargs`)
- Argument type checking (when necessary)

This approach is often referred to as **Method Overloading (Python Way)**.

# 📖 Topics Covered

## 1. What is Method Overloading?

Method overloading means having **multiple methods with the same name** but **different numbers or types of parameters**.

Example in languages like Java:

```java
add(int a, int b)
add(int a, int b, int c)
add(double a, double b)
```

The compiler automatically selects the correct method.

Python does **not** work this way.

---

## 2. Why Doesn't Python Support Traditional Method Overloading?

Python uses **dynamic typing** and method names are simply assigned in the class namespace.

If multiple methods have the same name, the newest definition replaces the previous one.

Example:

```python
class Demo:

    def show(self):
        print("First Method")

    def show(self):
        print("Second Method")

demo = Demo()

demo.show()
```

### Output

```text
Second Method
```

The first `show()` method is overwritten.

---

## 3. Using Default Arguments

One way to simulate method overloading is by using default parameter values.

```python
class Calculator:

    def add(self, a, b=0):
        return a + b

calc = Calculator()

print(calc.add(5))
print(calc.add(5, 8))
```

### Output

```text
5
13
```

---

## 4. Using Multiple Default Arguments

```python
class Calculator:

    def add(self, a, b=0, c=0):
        return a + b + c

calc = Calculator()

print(calc.add(5))
print(calc.add(5, 10))
print(calc.add(5, 10, 15))
```

### Output

```text
5
15
30
```

---

## 5. Using `*args`

The `*args` parameter allows a method to accept **any number of positional arguments**.

```python
class Calculator:

    def add(self, *numbers):
        return sum(numbers)

calc = Calculator()

print(calc.add(10))
print(calc.add(10, 20))
print(calc.add(10, 20, 30))
print(calc.add(10, 20, 30, 40))
```

### Output

```text
10
30
60
100
```

---

## 6. Using `**kwargs`

The `**kwargs` parameter allows a method to accept **any number of keyword arguments**.

```python
class Student:

    def display(self, **details):

        for key, value in details.items():
            print(f"{key}: {value}")

student = Student()

student.display(name="Alice", age=20, course="Python")
```

### Output

```text
name: Alice
age: 20
course: Python
```

---

## 7. Argument Type Checking

Sometimes behavior changes depending on the argument type.

```python
class Printer:

    def print_data(self, value):

        if isinstance(value, int):
            print("Integer:", value)

        elif isinstance(value, str):
            print("String:", value)

        else:
            print("Unknown Type")

printer = Printer()

printer.print_data(100)
printer.print_data("Python")
```

### Output

```text
Integer: 100
String: Python
```

## 9. Practical Examples

### Example 1: Calculator

```python
class Calculator:

    def multiply(self, a, b=1):
        return a * b

calc = Calculator()

print(calc.multiply(5))
print(calc.multiply(5, 10))
```

### Output

```text
5
50
```

---

### Example 2: Sum Using `*args`

```python
class Calculator:

    def total(self, *numbers):
        return sum(numbers)

calc = Calculator()

print(calc.total(1, 2))
print(calc.total(1, 2, 3))
print(calc.total(1, 2, 3, 4))
```

### Output

```text
3
6
10
```

---

### Example 3: Employee Details

```python
class Employee:

    def display(self, **details):

        for key, value in details.items():
            print(key, ":", value)

employee = Employee()

employee.display(name="John", department="IT", salary=50000)
```

### Output

```text
name : John
department : IT
salary : 50000
```

---

### Example 4: Greeting Function

```python
class Greeting:

    def greet(self, name="Guest"):
        print("Hello", name)

greeting = Greeting()

greeting.greet()
greeting.greet("Alice")
```

### Output

```text
Hello Guest
Hello Alice
```

---

### Example 5: Shape Area

```python
class Area:

    def calculate(self, length, width=None):

        if width is None:
            return length * length

        return length * width

area = Area()

print(area.calculate(5))
print(area.calculate(5, 10))
```

### Output

```text
25
50
```

## 11. Common Mistakes

Avoid these common mistakes while implementing method overloading in Python.

### Defining Multiple Methods with the Same Name

Incorrect

```python
class Demo:

    def show(self):
        pass

    def show(self):
        pass
```

Only the last method remains.

---

### Forgetting Default Values

Incorrect

```python
def add(a, b):
    return a + b

add(5)
```

### Output

```text
TypeError
```

Provide default values if optional arguments are expected.

---

### Misusing `*args`

Incorrect

```python
sum(*numbers)
```

Correct

```python
sum(numbers)
```

`numbers` is already a tuple.

---

### Overcomplicating Argument Checking

Avoid writing long chains of `if` statements when a simpler design can solve the problem.

---

## 12. Best Practices

Follow these recommended practices when implementing method overloading in Python.

- Use default arguments for optional parameters.
- Use `*args` when the number of arguments is unknown.
- Use `**kwargs` for flexible keyword arguments.
- Keep methods simple and readable.
- Validate arguments only when necessary.
- Avoid defining multiple methods with the same name.
- Document accepted arguments clearly.
