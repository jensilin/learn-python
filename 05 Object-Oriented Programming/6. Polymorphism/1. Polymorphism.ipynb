# 🎭 Polymorphism in Python

**Polymorphism** is one of the four fundamental principles of **Object-Oriented Programming (OOP)**.

The word **Polymorphism** comes from two Greek words:

- **Poly** → Many
- **Morph** → Forms

Therefore, **Polymorphism** means **"Many Forms."**

In Python, polymorphism allows **the same method, function, or operator to behave differently depending on the object it is working with**.

Polymorphism makes programs more **flexible**, **reusable**, and **easy to extend**.

# 📖 Topics Covered

## 1. What is Polymorphism?

**Polymorphism** allows one interface to represent different implementations.

Different objects can respond differently to the same method call.

Example:

```python
class Bird:

    def sound(self):
        print("Bird Sound")

class Sparrow(Bird):

    def sound(self):
        print("Chirp")

class Crow(Bird):

    def sound(self):
        print("Caw")

birds = [Sparrow(), Crow()]

for bird in birds:
    bird.sound()
```

### Output

```text
Chirp
Caw
```

## 3. Types of Polymorphism in Python

Python supports several forms of polymorphism.

### Runtime Polymorphism

Achieved using **method overriding**.

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        print("Dog Bark")
```

---

### Compile-Time Polymorphism (Python Style)

Python does not support traditional method overloading.

Instead, similar behavior is achieved using:

- Default arguments
- `*args`
- `**kwargs`

Example:

```python
class Calculator:

    def add(self, a, b=0):
        return a + b

calc = Calculator()

print(calc.add(5))
print(calc.add(5, 10))
```

### Output

```text
5
15
```

---

### Built-in Polymorphism

Many built-in Python functions are polymorphic.

Example:

```python
print(len("Python"))
print(len([1, 2, 3]))
print(len({"a": 1, "b": 2}))
```

### Output

```text
6
3
2
```

The same `len()` function works with different object types.

---

## 4. Runtime Polymorphism

Runtime polymorphism occurs when a child class overrides a parent method.

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        print("Dog Bark")

class Cat(Animal):

    def sound(self):
        print("Cat Meow")

animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```

### Output

```text
Dog Bark
Cat Meow
```

The actual method executed depends on the object at runtime.

---

## 5. Polymorphism with Functions

A function can work with different object types as long as they provide the required method.

```python
class Dog:

    def speak(self):
        print("Dog Barks")

class Cat:

    def speak(self):
        print("Cat Meows")

def make_sound(animal):
    animal.speak()

make_sound(Dog())
make_sound(Cat())
```

### Output

```text
Dog Barks
Cat Meows
```

This is an example of Python's dynamic nature.

## 6. Polymorphism with Operators

Operators also exhibit polymorphism.

```python
print(10 + 20)
print("Hello " + "Python")
print([1, 2] + [3, 4])
```

### Output

```text
30
Hello Python
[1, 2, 3, 4]
```

The `+` operator performs different operations depending on the operands.

## 8. Practical Examples

### Example 1: Animals

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        print("Dog Bark")

class Cat(Animal):

    def sound(self):
        print("Cat Meow")

animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```

### Output

```text
Dog Bark
Cat Meow
```

---

### Example 2: Payment System

```python
class Payment:

    def pay(self):
        print("Processing Payment")

class CreditCard(Payment):

    def pay(self):
        print("Paid using Credit Card")

class UPI(Payment):

    def pay(self):
        print("Paid using UPI")

payments = [CreditCard(), UPI()]

for payment in payments:
    payment.pay()
```

### Output

```text
Paid using Credit Card
Paid using UPI
```

---

### Example 3: Employees

```python
class Employee:

    def work(self):
        print("Employee Working")

class Developer(Employee):

    def work(self):
        print("Writing Code")

class Tester(Employee):

    def work(self):
        print("Testing Software")

employees = [Developer(), Tester()]

for employee in employees:
    employee.work()
```

### Output

```text
Writing Code
Testing Software
```

---

### Example 4: Shapes

```python
class Shape:

    def draw(self):
        print("Drawing Shape")

class Circle(Shape):

    def draw(self):
        print("Drawing Circle")

class Rectangle(Shape):

    def draw(self):
        print("Drawing Rectangle")

shapes = [Circle(), Rectangle()]

for shape in shapes:
    shape.draw()
```

### Output

```text
Drawing Circle
Drawing Rectangle
```

---

### Example 5: Notifications

```python
class Notification:

    def send(self):
        print("Sending Notification")

class Email(Notification):

    def send(self):
        print("Sending Email")

class SMS(Notification):

    def send(self):
        print("Sending SMS")

notifications = [Email(), SMS()]

for notification in notifications:
    notification.send()
```

### Output

```text
Sending Email
Sending SMS
```

## 10. Common Mistakes

Avoid these common mistakes while working with polymorphism.

### Confusing Polymorphism with Inheritance

Inheritance enables polymorphism, but they are different concepts.

Polymorphism focuses on **different behavior through a common interface**.

---

### Writing Long Conditional Statements

Instead of

```python
if animal == "dog":
    ...
elif animal == "cat":
    ...
```

prefer polymorphism.

Each class should implement its own behavior.

---

### Forgetting Method Overriding

If child classes do not override the parent method, all objects may behave the same.

---

### Assuming Python Supports Traditional Method Overloading

Python achieves similar functionality using:

- Default arguments
- `*args`
- `**kwargs`

Traditional compile-time method overloading is not supported.

---

## 11. Best Practices

Follow these recommended practices when using polymorphism.

- Design classes around a common interface.
- Override methods only when behavior differs.
- Prefer polymorphism over long conditional statements.
- Write reusable functions that work with different object types.
- Follow the Open/Closed Principle by extending behavior without modifying existing code.
- Keep related classes consistent in their method names and purpose.
- Use polymorphism to simplify application logic.
