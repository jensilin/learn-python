# 🔄 Method Overriding in Polymorphism

**Method overriding** is one of the most important ways to achieve **runtime polymorphism** in Python.

It occurs when a **child class provides its own implementation of a method that already exists in its parent class**. When the overridden method is called through a child object, Python executes the **child class version** instead of the parent version.

Method overriding enables different subclasses to provide their own behavior while sharing a common interface.

# 📖 Topics Covered

## 1. What is Method Overriding?

Method overriding happens when a child class defines a method with the **same name and parameters** as a method in its parent class.

The child implementation replaces the inherited implementation for objects of the child class.

Example:

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        print("Dog Bark")
```

Here, `Dog` overrides the `sound()` method of `Animal`.

---

Method overriding allows us to:

- Customize inherited behavior.
- Achieve runtime polymorphism.
- Reuse parent class designs.
- Improve flexibility.

- Follow object-oriented design principles.

## 3. Syntax

```python
class Parent:

    def method(self):
        # Parent implementation

class Child(Parent):

    def method(self):
        # Child implementation
```

The method name and parameter list should remain compatible.

---

## 4. Runtime Polymorphism Using Method Overriding

Python decides which method to execute **at runtime** based on the object's actual type.

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        print("Dog Bark")

class Cat(Animal):

    def sound(self):
        print("Cat Meow")

animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```

### Output

```text
Dog Bark
Cat Meow
```

The same method call produces different behavior depending on the object.

---

## 5. Parent Method is Replaced

If the child class overrides a method, the parent version is not called automatically.

```python
class Person:

    def introduce(self):
        print("I am a Person.")

class Student(Person):

    def introduce(self):
        print("I am a Student.")

student = Student()

student.introduce()
```

### Output

```text
I am a Student.
```

---

## 6. Calling the Parent Method Using `super()`

The parent implementation can still be accessed using `super()`.

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        super().sound()
        print("Dog Bark")

dog = Dog()

dog.sound()
```

### Output

```text
Animal Sound
Dog Bark
```

This allows the child class to extend, rather than completely replace, the parent's behavior.

---

## 7. Method Overriding and Polymorphism

Overriding is the primary mechanism for implementing runtime polymorphism.

```python
class Shape:

    def draw(self):
        print("Drawing Shape")

class Circle(Shape):

    def draw(self):
        print("Drawing Circle")

class Rectangle(Shape):

    def draw(self):
        print("Drawing Rectangle")

shapes = [Circle(), Rectangle()]

for shape in shapes:
    shape.draw()
```

### Output

```text
Drawing Circle
Drawing Rectangle
```

Each subclass provides its own implementation of `draw()`.

---

## 8. Method Overriding vs Method Overloading

| Feature | Method Overriding | Method Overloading |
|---------|-------------------|--------------------|
| Occurs In | Parent and child classes | Same class |
| Requires Inheritance | Yes | No |
| Purpose | Replace inherited behavior | Support different parameter combinations |
| Python Support | Yes | Achieved using default arguments, `*args`, or `**kwargs` |

## 11. Common Mistakes

Avoid these common mistakes while using method overriding.

### Forgetting to Inherit

Incorrect

```python
class Dog:

    def sound(self):
        print("Dog Bark")
```

This is not method overriding because `Dog` does not inherit from another class.

---

### Changing the Method Name

Incorrect

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def bark(self):
        print("Dog Bark")
```

The parent method is not overridden because the method names are different.

---

### Forgetting to Use `super()` When Needed

If the parent method performs important initialization or processing, call it using `super()`.

```python
class Dog(Animal):

    def sound(self):
        super().sound()
        print("Dog Bark")
```

---

### Confusing Overriding with Overloading

Method overriding replaces inherited behavior.

Method overloading provides different ways to call a method.

These are separate concepts.

---

## 12. Best Practices

Follow these recommended practices when implementing method overriding.

- Override methods only when behavior actually changes.
- Keep method signatures compatible with the parent class.
- Use `super()` when parent behavior should also execute.
- Avoid duplicating parent logic.
- Use meaningful method names.
- Follow the Liskov Substitution Principle.
- Keep overridden methods focused and easy to understand.
