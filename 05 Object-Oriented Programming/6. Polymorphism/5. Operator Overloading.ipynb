# ⚙️ Operator Overloading in Python

**Operator overloading** is an Object-Oriented Programming (OOP) feature that allows **operators to have different meanings depending on the objects they operate on**.

Python already overloads many operators for built-in data types. For example, the `+` operator performs addition for numbers but concatenation for strings and lists.

Python also allows developers to **define the behavior of operators for their own classes** by implementing **special methods (magic/dunder methods)** such as `__add__()`, `__sub__()`, and `__eq__()`.

Operator overloading makes custom objects behave like built-in Python objects.

# 💻 Quick Example

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __str__(self):
        return str(self.value)

num1 = Number(10)
num2 = Number(20)

print(num1 + num2)
```

### Output

```text
30
```

The `+` operator now works with `Number` objects.

## 1. What is Operator Overloading?

Operator overloading means **giving existing operators new behavior for user-defined classes**.

Instead of writing:

```python
num1.add(num2)
```

we can write:

```python
num1 + num2
```

This makes code cleaner and easier to read.

Python uses special methods to overload operators.

Example:

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __str__(self):
        return str(self.value)

a = Number(5)
b = Number(7)

print(a + b)
```

## 6. Overloading the `+` Operator

```python
class Employee:

    def __init__(self, salary):
        self.salary = salary

    def __add__(self, other):
        return self.salary + other.salary

emp1 = Employee(40000)
emp2 = Employee(50000)

print(emp1 + emp2)
```

### Output

```text
90000
```

## 7. Overloading the `==` Operator

```python
class Student:

    def __init__(self, marks):
        self.marks = marks

    def __eq__(self, other):
        return self.marks == other.marks

s1 = Student(90)
s2 = Student(90)

print(s1 == s2)
```

### Output

```text
True
```

---

## 8. Overloading the `<` Operator

```python
class Product:

    def __init__(self, price):
        self.price = price

    def __lt__(self, other):
        return self.price < other.price

p1 = Product(500)
p2 = Product(700)

print(p1 < p2)
```

### Output

```text
True
```

## 9. Practical Examples

### Example 1: Vector Addition

```python
class Vector:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)

print(v1 + v2)
```

### Output

```text
(6, 8)
```

---

### Example 2: Bank Account

```python
class Account:

    def __init__(self, balance):
        self.balance = balance

    def __add__(self, other):
        return self.balance + other.balance

a1 = Account(1000)
a2 = Account(2500)

print(a1 + a2)
```

### Output

```text
3500
```

### Example 3: Book Comparison

```python
class Book:

    def __init__(self, pages):
        self.pages = pages

    def __gt__(self, other):
        return self.pages > other.pages

book1 = Book(250)
book2 = Book(180)

print(book1 > book2)
```

### Output

```text
True
```

### Example 4: Rectangle Comparison

```python
class Rectangle:

    def __init__(self, area):
        self.area = area

    def __eq__(self, other):
        return self.area == other.area

r1 = Rectangle(100)
r2 = Rectangle(100)

print(r1 == r2)
```

### Example 5: Temperature Addition

```python
class Temperature:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Temperature(self.value + other.value)

    def __str__(self):
        return f"{self.value}°C"

t1 = Temperature(20)
t2 = Temperature(15)

print(t1 + t2)
```

## 11. Common Mistakes

Avoid these common mistakes while implementing operator overloading.

### Forgetting to Return a Value

Incorrect

```python
def __add__(self, other):
    self.value + other.value
```

The method should return the result.

Correct

```python
def __add__(self, other):
    return self.value + other.value
```

---

### Assuming the Other Object Has the Same Attributes

Incorrect

```python
return self.value + other.value
```

If `other` is not the expected type, this may raise an `AttributeError`.

Consider validating the type when appropriate.

---

### Forgetting `__str__()`

Without a `__str__()` method, printing custom objects displays a less informative representation.

```text
<__main__.Number object at 0x...>
```

Implementing `__str__()` improves readability.

---

### Overloading Operators with Unexpected Behavior

Operators should behave in a way that users naturally expect.

For example, `+` should generally combine or add values rather than perform unrelated operations.

---

## 12. Best Practices

Follow these recommended practices when implementing operator overloading.

- Overload operators only when it improves readability.
- Keep operator behavior intuitive.
- Return appropriate values or objects.
- Implement `__str__()` for readable output.
- Validate operand types when necessary.
- Avoid surprising or inconsistent operator behavior.
- Follow Python's data model conventions.
