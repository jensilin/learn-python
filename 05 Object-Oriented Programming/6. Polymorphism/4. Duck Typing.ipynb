# 🦆 Duck Typing in Python

**Duck Typing** is a programming concept used in Python where **the type of an object is less important than the methods and behaviors it provides**.

The idea comes from the famous saying:

> **"If it walks like a duck and quacks like a duck, then it is probably a duck."**

In Python, if an object has the required methods or attributes, it can be used, **regardless of its actual class**.

Duck typing is one of the key reasons why Python is considered a **dynamically typed** and **flexible** programming language.

# 💻 Quick Example

```python
class Dog:

    def speak(self):
        print("Dog barks.")

class Cat:

    def speak(self):
        print("Cat meows.")

def make_sound(animal):
    animal.speak()

make_sound(Dog())
make_sound(Cat())
```

### Output

```text
Dog barks.
Cat meows.
```

The function does not care whether the object is a `Dog` or a `Cat`. It only requires that the object has a `speak()` method.

# 📖 Topics Covered

## 1. What is Duck Typing?

Duck typing means that **Python focuses on an object's behavior rather than its actual type**.

If an object provides the required methods or attributes, it can be used.

Example:

```python
class Bird:

    def fly(self):
        print("Flying")

class Airplane:

    def fly(self):
        print("Airplane Flying")

def start_flying(obj):
    obj.fly()

start_flying(Bird())
start_flying(Airplane())
```

### Output

```text
Flying
Airplane Flying
```

Both objects work because they implement the `fly()` method.

---

## 2. Why Use Duck Typing?

Duck typing helps us:

- Write flexible code.
- Reduce dependency on specific classes.
- Improve code reuse.
- Support polymorphism.
- Simplify application design.
- Encourage behavior-based programming.

---

## 3. How Duck Typing Works

Python checks whether the required method exists **when the code runs**.

It does not require objects to inherit from the same parent class.

```python
class Car:

    def start(self):
        print("Car Started")

class Bike:

    def start(self):
        print("Bike Started")

def start_vehicle(vehicle):
    vehicle.start()

start_vehicle(Car())
start_vehicle(Bike())
```

### Output

```text
Car Started
Bike Started
```

---

## 4. Duck Typing Without Inheritance

Duck typing works even when classes are completely unrelated.

```python
class Teacher:

    def work(self):
        print("Teaching")

class Doctor:

    def work(self):
        print("Treating Patients")

def perform_work(person):
    person.work()

perform_work(Teacher())
perform_work(Doctor())
```

### Output

```text
Teaching
Treating Patients
```

No inheritance is involved.

---

## 5. Duck Typing vs Inheritance

Inheritance:

```python
Animal
   │
  Dog
```

The child inherits from the parent.

Duck Typing:

```text
Dog -----> speak()
Cat -----> speak()
Robot ---> speak()
```

Objects only need to provide the required behavior.

---

## 6. Duck Typing vs `isinstance()`

Instead of checking object types:

```python
if isinstance(animal, Dog):
    animal.speak()
```

Prefer:

```python
animal.speak()
```

This makes the code more flexible.

Only use `isinstance()` when type validation is genuinely required.

## 10. Common Mistakes

Avoid these common mistakes while using duck typing.

### Assuming an Object Has a Method

Incorrect

```python
class Dog:
    pass

dog = Dog()

dog.speak()
```

### Output

```text
AttributeError
```

The method does not exist.

---

### Overusing `isinstance()`

Incorrect

```python
if isinstance(vehicle, Car):
    vehicle.start()
```

If multiple object types support `start()`, checking the type is often unnecessary.

---

### Ignoring Missing Methods

If an object lacks the expected method, Python raises an exception.

```python
class Person:
    pass

def work(obj):
    obj.do_work()

work(Person())
```

### Output

```text
AttributeError
```

Ensure the object provides the required behavior.

---

### Confusing Duck Typing with Inheritance

Duck typing does **not** require inheritance.

It only requires compatible behavior.

---

## 11. Best Practices

Follow these recommended practices when using duck typing.

- Program to behavior rather than specific types.
- Avoid unnecessary type checking.
- Design classes with consistent method names.
- Use inheritance only when an "is-a" relationship exists.
- Keep interfaces simple and predictable.
- Handle missing methods gracefully when appropriate.
- Document expected object behavior clearly.
