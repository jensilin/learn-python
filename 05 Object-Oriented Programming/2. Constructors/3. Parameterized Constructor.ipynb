# 🏗️ Parameterized Constructor in Python

A **parameterized constructor** is a constructor that accepts **one or more arguments (in addition to `self`)**. It allows you to **initialize an object with different values at the time of object creation**.

In Python, a parameterized constructor is implemented using the **`__init__()`** method with parameters.

Parameterized constructors make objects more flexible because each object can store different data.

# 📖 Topics Covered

## 1. What is a Parameterized Constructor?

A **parameterized constructor** accepts one or more parameters to initialize an object with user-provided values.

Unlike a default constructor, each object can be initialized differently.

Example:

```python
class Car:

    def __init__(self, brand):
        self.brand = brand

car = Car("Toyota")

print(car.brand)
```

### Output

```text
Toyota
```

---

## 2. Why Use a Parameterized Constructor?

Parameterized constructors help us:

- Initialize objects with different values.
- Reduce repetitive code.
- Create flexible and reusable classes.
- Improve readability.
- Avoid assigning values manually after object creation.

---

## 3. Syntax

### Syntax

```python
class ClassName:

    def __init__(self, parameter1, parameter2):
        self.parameter1 = parameter1
        self.parameter2 = parameter2
```

### Explanation

- `__init__` → Constructor method.
- `self` → Refers to the current object.
- Remaining parameters receive values during object creation.

---

## 4. Creating an Object with Parameters

Arguments are supplied when creating the object.

```python
class Student:

    def __init__(self, name):
        self.name = name

student = Student("John")

print(student.name)
```

### Output

```text
John
```

---

## 5. Initializing Multiple Attributes

A parameterized constructor can initialize multiple attributes.

```python
class Employee:

    def __init__(self, name, age):
        self.name = name
        self.age = age

employee = Employee("Alice", 25)

print(employee.name)
print(employee.age)
```

### Output

```text
Alice
25
```

---

## 6. Every Object Can Have Different Values

Each object stores its own data.

```python
class Laptop:

    def __init__(self, brand):
        self.brand = brand

laptop1 = Laptop("Dell")
laptop2 = Laptop("HP")

print(laptop1.brand)
print(laptop2.brand)
```

### Output

```text
Dell
HP
```

---

## 7. Constructor Executes Automatically

The constructor runs automatically whenever a new object is created.

```python
class Book:

    def __init__(self, title):
        self.title = title
        print("Book object created")

book = Book("Python Basics")
```

### Output

```text
Book object created
```

## 10. Common Mistakes

Avoid these common mistakes while using parameterized constructors.

### Forgetting Required Arguments

Incorrect

```python
class Student:

    def __init__(self, name):
        self.name = name

student = Student()
```

### Output

```text
TypeError
```

Correct

```python
student = Student("Alice")
```

---

### Passing Too Many Arguments

Incorrect

```python
class Student:

    def __init__(self, name):
        self.name = name

student = Student("Alice", 20)
```

### Output

```text
TypeError
```

Pass only the required number of arguments.

---

### Forgetting to Assign Parameters to Attributes

Incorrect

```python
class Student:

    def __init__(self, name):
        name = name
```

Correct

```python
class Student:

    def __init__(self, name):
        self.name = name
```

---

### Forgetting `self`

Incorrect

```python
class Student:

    def __init__(name):
        self.name = name
```

Correct

```python
class Student:

    def __init__(self, name):
        self.name = name
```

---

## 11. Best Practices

Follow these recommended practices when using parameterized constructors.

- Initialize all required attributes inside the constructor.
- Use meaningful parameter names.
- Keep constructors simple and focused.
- Validate input values when necessary.
- Avoid performing lengthy operations inside constructors.
- Assign parameters using `self.attribute = parameter`.
- Use descriptive object names.
