# 🏗️ Default Constructor in Python

A **default constructor** is a constructor that **does not require any additional arguments (other than `self`)**. It is automatically called whenever an object is created and is commonly used to initialize objects with **default values**.

In Python, a default constructor is implemented using the **`__init__()`** method without any parameters except `self`.

Default constructors are useful when every object should start with the same initial state.

# 📖 Topics Covered

## 1. What is a Default Constructor?

A **default constructor** is a constructor that takes **only one parameter**, `self`.

It is automatically executed whenever an object is created.

It is commonly used to:

- Initialize default values.
- Display initialization messages.
- Prepare an object for use.

Example:

```python
class Car:

    def __init__(self):
        print("Car Created")
```

Creating an object:

```python
car = Car()
```

### Output

```text
Car Created
```

---

## 2. Why Use a Default Constructor?

Default constructors help us:

- Assign default values.
- Reduce repetitive code.
- Ensure every object starts consistently.
- Improve readability.
- Simplify object initialization.

---

## 3. Syntax

### Syntax

```python
class ClassName:

    def __init__(self):
        # Initialization code
```

### Explanation

- `__init__` → Constructor method.
- `self` → Refers to the current object.
- No additional parameters are required.

---

## 4. Initializing Default Values

A default constructor can assign default values to object attributes.

```python
class Student:

    def __init__(self):
        self.name = "Alice"
        self.age = 20

student = Student()

print(student.name)
print(student.age)
```

### Output

```text
Alice
20
```

---

## 5. Constructor Executes Automatically

The constructor is called automatically when an object is created.

```python
class Employee:

    def __init__(self):
        print("Employee Created")

employee = Employee()
```

### Output

```text
Employee Created
```

---

## 6. Every Object Gets Default Values

Each object created receives the initialized values.

```python
class Laptop:

    def __init__(self):
        self.brand = "Dell"

laptop1 = Laptop()
laptop2 = Laptop()

print(laptop1.brand)
print(laptop2.brand)
```

### Output

```text
Dell
Dell
```

---

## 7. Default Constructor Runs for Every Object

```python
class Animal:

    def __init__(self):
        print("Animal Object Created")

dog = Animal()
cat = Animal()
```

### Output

```text
Animal Object Created
Animal Object Created
```

The constructor runs once for each object.

## 10. Common Mistakes

Avoid these common mistakes while using default constructors.

### Misspelling `__init__`

Incorrect

```python
class Student:

    def init(self):
        print("Hello")
```

Correct

```python
class Student:

    def __init__(self):
        print("Hello")
```

---

### Forgetting `self`

Incorrect

```python
class Student:

    def __init__():
        print("Hello")
```

Correct

```python
class Student:

    def __init__(self):
        print("Hello")
```

---

### Expecting the Constructor to Run Without Creating an Object

```python
class Test:

    def __init__(self):
        print("Created")
```

Nothing happens until an object is created.

```python
obj = Test()
```

---

### Calling the Constructor Manually

Less preferred

```python
obj = Student()

obj.__init__()
```

Preferred

```python
obj = Student()
```

The constructor is automatically called during object creation.

- Avoid lengthy computations inside constructors.
