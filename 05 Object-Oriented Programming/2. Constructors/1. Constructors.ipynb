# 🏗️ Constructors in Python

A **constructor** is a special method in a class that is **automatically called whenever an object is created**. It is primarily used to **initialize the object's attributes** and perform any setup required during object creation.

In Python, the constructor is implemented using the **`__init__()`** method.

Constructors make it easy to assign initial values to objects, ensuring that every object starts in a valid and consistent state.

# 📖 Topics Covered

## 1. What is a Constructor?

A **constructor** is a special method that is automatically executed whenever an object is created.

In Python, the constructor is named:

```python
__init__()
```

Its main purpose is to:

- Initialize object attributes.
- Assign default values.
- Prepare the object for use.

---

## 2. Why Use Constructors?

Constructors help us:

- Initialize object data automatically.
- Reduce repetitive code.
- Ensure every object starts with valid values.
- Improve code readability.
- Simplify object creation.

Without constructors, attributes would need to be assigned manually after creating every object.

---

## 3. The `__init__()` Method

The `__init__()` method is Python's constructor.

It is automatically called whenever a new object is created.

Example:

```python
class Student:

    def __init__(self):
        print("Constructor Executed")

student = Student()
```

### Output

```text
Constructor Executed
```

---

## 4. Syntax

### Syntax

```python
class ClassName:

    def __init__(self):
        # Initialization code
```

### Explanation

- `__init__` → Special constructor method.
- `self` → Refers to the current object.
- Runs automatically during object creation.

---

## 5. Constructor Execution

Notice that we never call `__init__()` directly.

```python
class Car:

    def __init__(self):
        print("Car Created")

car = Car()
```

### Output

```text
Car Created
```

Python automatically calls the constructor when `Car()` creates an object.

---

## 6. Initializing Object Attributes

Constructors are commonly used to initialize attributes.

```python
class Student:

    def __init__(self):
        self.name = "Alice"
        self.age = 20

student = Student()

print(student.name)
print(student.age)
```

### Output

```text
Alice
20
```

---

## 7. Constructor Runs for Every Object

Each new object calls the constructor.

```python
class Employee:

    def __init__(self):
        print("Employee Created")

emp1 = Employee()
emp2 = Employee()
emp3 = Employee()
```

### Output

```text
Employee Created
Employee Created
Employee Created
```

The constructor executes once for each object.

## 11. Common Mistakes

Avoid these common mistakes while working with constructors.

### Misspelling `__init__`

Incorrect

```python
class Student:

    def init(self):
        print("Hello")
```

The method above is **not** a constructor.

Correct

```python
class Student:

    def __init__(self):
        print("Hello")
```

---

### Forgetting `self`

Incorrect

```python
class Student:

    def __init__():
        print("Hello")
```

Correct

```python
class Student:

    def __init__(self):
        print("Hello")
```

---

### Calling `__init__()` Directly

Incorrect

```python
student = Student()

student.__init__()
```

Usually, there is no need to call the constructor manually.

Simply create the object:

```python
student = Student()
```

---

### Forgetting to Create an Object

```python
class Test:

    def __init__(self):
        print("Created")
```

Nothing happens until an object is created.

```python
obj = Test()
```

## 12. Best Practices

Follow these recommended practices when using constructors.

- Use constructors to initialize object attributes.
- Keep constructors simple and focused.
- Avoid performing lengthy operations inside constructors.
- Always include `self` as the first parameter.
- Use meaningful attribute names.
- Initialize every required attribute inside the constructor.
- Keep initialization logic readable and organized.
