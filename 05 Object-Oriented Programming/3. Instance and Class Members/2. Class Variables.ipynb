# 🏫 Class Variables in Python

**Class variables** are variables that belong to the **class itself**, rather than to individual objects. A single copy of a class variable is **shared by all objects** created from that class.

Class variables are defined **inside the class but outside any methods**.

They are commonly used to store values that should be the same for every object, such as company names, tax rates, school names, or counters.

# 📖 Topics Covered

## 1. What are Class Variables?

A **class variable** is a variable that belongs to the class itself.

Only **one copy** of the variable exists, and all objects share it.

Example:

```python
class Student:

    college = "ABC College"
```

Here,

- `college` is a class variable.
- Every `Student` object shares the same value.

---

## 2. Why Use Class Variables?

Class variables help us:

- Store common information.
- Reduce memory usage.
- Share data among all objects.
- Avoid duplicate values.
- Keep common settings in one place.

---

## 3. Creating Class Variables

Class variables are created directly inside the class.

```python
class Employee:

    company = "ABC Technologies"
```

Notice that `company` is **not** inside the constructor or any method.

---

## 4. Syntax

### Syntax

```python
class ClassName:

    class_variable = value
```

### Explanation

- Defined inside the class.
- Outside all methods.
- Shared by every object.

---

## 5. Accessing Class Variables

Class variables can be accessed using an object.

```python
class Student:

    college = "ABC College"

student = Student()

print(student.college)
```

### Output

```text
ABC College
```

---

They can also be accessed using the class name.

```python
class Student:

    college = "ABC College"

print(Student.college)
```

### Output

```text
ABC College
```

Using the class name is generally preferred because it clearly indicates that the variable belongs to the class.

---

## 6. Class Variables are Shared

Every object accesses the same class variable.

```python
class Student:

    college = "ABC College"

student1 = Student()
student2 = Student()

print(student1.college)
print(student2.college)
```

### Output

```text
ABC College
ABC College
```

---

## 7. Modifying a Class Variable

A class variable should usually be modified using the class name.

```python
class Student:

    college = "ABC College"

Student.college = "XYZ College"

student = Student()

print(student.college)
```

### Output

```text
XYZ College
```

Since the variable belongs to the class, all objects see the updated value.

---

## 8. Class Variable vs Instance Variable

```python
class Student:

    college = "ABC College"

    def __init__(self, name):
        self.name = name

student1 = Student("Alice")
student2 = Student("Bob")

print(student1.college)
print(student2.college)

print(student1.name)
print(student2.name)
```

### Output

```text
ABC College
ABC College
Alice
Bob
```

Here:

- `college` → Class variable (shared)
- `name` → Instance variable (unique)

## 10. Class Variables vs Instance Variables

| Class Variables | Instance Variables |
|-----------------|--------------------|
| Belong to the class | Belong to each object |
| Shared by all objects | Separate copy for every object |
| Created outside methods | Usually created inside `__init__()` |
| Accessed using class or object | Accessed using the object |
| Used for common data | Used for object-specific data |

## 12. Common Mistakes

Avoid these common mistakes while working with class variables.

### Creating a Class Variable Inside the Constructor

Incorrect

```python
class Student:

    def __init__(self):
        self.college = "ABC College"
```

This creates an **instance variable**, not a class variable.

Correct

```python
class Student:

    college = "ABC College"
```

---

### Modifying a Class Variable Through an Object

```python
class Student:

    college = "ABC College"

student = Student()

student.college = "XYZ College"

print(student.college)
print(Student.college)
```

### Output

```text
XYZ College
ABC College
```

The assignment creates a **new instance variable**, leaving the original class variable unchanged.

Correct way:

```python
Student.college = "XYZ College"
```

---

### Assuming Each Object Has Its Own Copy

```python
class Test:

    value = 100

obj1 = Test()
obj2 = Test()
```

Both objects share the same class variable.

---

### Forgetting the Difference Between Class and Instance Variables

Choose:

- Class variables for **shared data**
- Instance variables for **unique object data**

---

## 13. Best Practices

Follow these recommended practices when using class variables.

- Use class variables only for data shared by all objects.
- Access class variables using the class name whenever possible.
- Use meaningful names for shared data.
- Avoid modifying class variables through objects.
- Keep constants as class variables.
- Use class variables for counters and configuration values.
- Use instance variables for object-specific information.
