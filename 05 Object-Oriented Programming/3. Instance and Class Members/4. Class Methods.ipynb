# 🏫 Class Methods in Python

**Class methods** are methods that **belong to the class rather than individual objects**. They can **access and modify class variables**, but they **cannot directly access instance variables**.

A class method is defined using the **`@classmethod`** decorator, and its first parameter is **`cls`**, which refers to the class itself.

Class methods are commonly used to work with data shared by all objects or to create alternative constructors.

# 📖 Topics Covered

## 1. What are Class Methods?

A **class method** is a method that works with the **class itself**, not with individual objects.

Unlike instance methods:

- It receives `cls` instead of `self`.
- It works with class variables.
- It can be called using either the class or an object.

Example:

```python
class Student:

    @classmethod
    def display(cls):
        print("Class Method")
```

Here, `display()` is a class method.

---

## 2. Why Use Class Methods?

Class methods help us:

- Access class variables.
- Modify shared data.
- Create alternative constructors.
- Perform operations related to the class.
- Keep class-level logic organized.

---

## 3. The `@classmethod` Decorator

To create a class method, use the `@classmethod` decorator.

```python
class Student:

    @classmethod
    def display(cls):
        print("Hello")
```

Without this decorator, Python treats the method as a regular instance method.

---

## 4. The `cls` Parameter

The first parameter of every class method is `cls`.

```python
class Student:

    @classmethod
    def display(cls):
        print(cls)
```

Calling the method:

```python
Student.display()
```

### Output

```text
<class '__main__.Student'>
```

`cls` refers to the class itself.

---

## 5. Syntax

### Syntax

```python
class ClassName:

    @classmethod
    def method_name(cls):
        # Method body
```

### Explanation

- `@classmethod` converts the method into a class method.
- `cls` refers to the class.
- The method can access class variables.

---

## 6. Accessing Class Variables

Class methods commonly access class variables.

```python
class Student:

    college = "ABC College"

    @classmethod
    def display(cls):
        print(cls.college)

Student.display()
```

### Output

```text
ABC College
```

---

## 7. Modifying Class Variables

Class methods can modify shared data.

```python
class Student:

    college = "ABC College"

    @classmethod
    def change_college(cls, name):
        cls.college = name

Student.change_college("XYZ College")

print(Student.college)
```

### Output

```text
XYZ College
```

---

## 8. Calling a Class Method

A class method can be called using either the class name or an object.

```python
class Student:

    college = "ABC College"

    @classmethod
    def show(cls):
        print(cls.college)

Student.show()

student = Student()

student.show()
```

### Output

```text
ABC College
ABC College
```

Although both work, calling a class method using the class name is generally preferred.

---

## 9. Class Method as an Alternative Constructor

Class methods are often used to create objects in different ways.

```python
class Student:

    def __init__(self, name):
        self.name = name

    @classmethod
    def create_default(cls):
        return cls("Unknown")

student = Student.create_default()

print(student.name)
```

### Output

```text
Unknown
```

This technique is commonly known as an **alternative constructor**.

---

## 10. Class Methods vs Instance Methods

| Class Method | Instance Method |
|--------------|-----------------|
| Uses `cls` | Uses `self` |
| Works with class variables | Works with instance variables |
| Uses `@classmethod` | No decorator required |
| Belongs to the class | Belongs to an object |
| Called using class or object | Usually called using an object |

## 11. Practical Examples

### Example 1: Company Information

```python
class Employee:

    company = "ABC Technologies"

    @classmethod
    def show_company(cls):
        print(cls.company)

Employee.show_company()
```

### Output

```text
ABC Technologies
```

---

### Example 2: Updating a Shared Value

```python
class Product:

    tax = 18

    @classmethod
    def update_tax(cls, value):
        cls.tax = value

Product.update_tax(20)

print(Product.tax)
```

### Output

```text
20
```

---

### Example 3: Counting Objects

```python
class Student:

    count = 0

    def __init__(self):
        Student.count += 1

    @classmethod
    def total_students(cls):
        print(cls.count)

Student()
Student()
Student()

Student.total_students()
```

### Output

```text
3
```

---

### Example 4: Alternative Constructor

```python
class Car:

    def __init__(self, brand):
        self.brand = brand

    @classmethod
    def default_car(cls):
        return cls("Toyota")

car = Car.default_car()

print(car.brand)
```

### Output

```text
Toyota
```

---

### Example 5: School Name

```python
class Student:

    school = "Green Valley School"

    @classmethod
    def display_school(cls):
        print(cls.school)

Student.display_school()
```

### Output

```text
Green Valley School
```

## 13. Common Mistakes

Avoid these common mistakes while working with class methods.

### Forgetting the `@classmethod` Decorator

Incorrect

```python
class Student:

    def display(cls):
        print("Hello")
```

Correct

```python
class Student:

    @classmethod
    def display(cls):
        print("Hello")
```

---

### Using `self` Instead of `cls`

Incorrect

```python
class Student:

    @classmethod
    def display(self):
        print(self)
```

Preferred

```python
class Student:

    @classmethod
    def display(cls):
        print(cls)
```

---

### Trying to Access Instance Variables

Incorrect

```python
class Student:

    @classmethod
    def display(cls):
        print(cls.name)
```

A class method cannot directly access instance variables because they belong to individual objects.

---

### Modifying Instance Data Inside a Class Method

Use instance methods to work with object-specific data.

Use class methods only for class-level operations.

---

## 14. Best Practices

Follow these recommended practices when using class methods.

- Use `@classmethod` for class-level operations.
- Always use `cls` as the first parameter.
- Access class variables through `cls`.
- Use class methods to create alternative constructors.
- Keep class methods focused on shared data.
- Avoid accessing instance variables directly.
- Call class methods using the class name whenever possible.
