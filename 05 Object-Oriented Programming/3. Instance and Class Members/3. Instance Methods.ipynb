# ⚙️ Instance Methods in Python

**Instance methods** are methods that **belong to an individual object (instance)** of a class. They can **access and modify the instance variables** of that object using the `self` parameter.

Instance methods are the **most commonly used methods** in Object-Oriented Programming because they define the behavior of objects.

# 📖 Topics Covered

## 1. What are Instance Methods?

An **instance method** is a method that operates on a specific object.

It can:

- Access instance variables.
- Modify instance variables.
- Perform operations related to the object.

Every instance method must have **`self` as its first parameter**.

Example:

```python
class Student:

    def display(self):
        print("Student Information")
```

Here, `display()` is an instance method.

---

## 2. Why Use Instance Methods?

Instance methods help us:

- Access object data.
- Modify object attributes.
- Define object behavior.
- Keep code organized.
- Represent real-world actions.

---

## 3. Creating an Instance Method

Instance methods are created like normal methods but include `self` as the first parameter.

```python
class Student:

    def display(self):
        print("Hello")
```

---

## 4. Syntax

### Syntax

```python
class ClassName:

    def method_name(self):
        # Method body
```

### Explanation

- `self` refers to the current object.
- The method belongs to each object.
- It can access instance variables.

---

## 5. Calling an Instance Method

Instance methods are called using an object.

```python
class Student:

    def greet(self):
        print("Welcome!")

student = Student()

student.greet()
```

### Output

```text
Welcome!
```

---

## 6. Accessing Instance Variables

Instance methods can access instance variables using `self`.

```python
class Student:

    def __init__(self, name):
        self.name = name

    def display(self):
        print(self.name)

student = Student("Alice")

student.display()
```

### Output

```text
Alice
```

---

## 7. Modifying Instance Variables

Instance methods can also update object data.

```python
class Employee:

    def __init__(self):
        self.salary = 50000

    def increment(self):
        self.salary += 5000

employee = Employee()

employee.increment()

print(employee.salary)
```

### Output

```text
55000
```

---

## 8. Using Multiple Instance Variables

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(self.name)
        print(self.age)

student = Student("Bob", 21)

student.display()
```

### Output

```text
Bob
21
```

---

## 9. Multiple Objects Using the Same Instance Method

The same instance method can be used by different objects.

```python
class Car:

    def __init__(self, brand):
        self.brand = brand

    def show_brand(self):
        print(self.brand)

car1 = Car("Toyota")
car2 = Car("Honda")

car1.show_brand()
car2.show_brand()
```

### Output

```text
Toyota
Honda
```

Each object uses the same method but operates on its own data.

## 10. Practical Examples

### Example 1: Student Details

```python
class Student:

    def __init__(self, name):
        self.name = name

    def display(self):
        print("Student:", self.name)

student = Student("David")

student.display()
```

### Output

```text
Student: David
```

---

### Example 2: Bank Account

```python
class BankAccount:

    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

account = BankAccount(5000)

account.deposit(2000)

print(account.balance)
```

### Output

```text
7000
```

---

### Example 3: Rectangle Area

```python
class Rectangle:

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

rectangle = Rectangle(10, 5)

print(rectangle.area())
```

### Output

```text
50
```

---

### Example 4: Product Information

```python
class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print(self.name)
        print(self.price)

product = Product("Laptop", 65000)

product.display()
```

### Output

```text
Laptop
65000
```

---

### Example 5: Updating Object Data

```python
class Counter:

    def __init__(self):
        self.count = 0

    def increase(self):
        self.count += 1

counter = Counter()

counter.increase()
counter.increase()

print(counter.count)
```

### Output

```text
2
```

## 12. Common Mistakes

Avoid these common mistakes while working with instance methods.

### Forgetting `self`

Incorrect

```python
class Student:

    def display():
        print("Hello")
```

Correct

```python
class Student:

    def display(self):
        print("Hello")
```

---

### Calling an Instance Method Without an Object

Incorrect

```python
class Student:

    def display(self):
        print("Hello")

Student.display()
```

This raises a `TypeError` because no object is provided for `self`.

Correct

```python
student = Student()

student.display()
```

---

### Accessing Instance Variables Without `self`

Incorrect

```python
class Student:

    def __init__(self):
        self.name = "Alice"

    def display(self):
        print(name)
```

Correct

```python
class Student:

    def __init__(self):
        self.name = "Alice"

    def display(self):
        print(self.name)
```

---

### Forgetting to Create an Object

```python
class Test:

    def show(self):
        print("Hello")
```

The method cannot be called until an object is created.

```python
obj = Test()

obj.show()
```

## 13. Best Practices

Follow these recommended practices when using instance methods.

- Always include `self` as the first parameter.
- Use instance methods to work with object-specific data.
- Keep methods focused on a single responsibility.
- Access instance variables using `self`.
- Use descriptive method names.
- Avoid writing excessively long methods.
- Group related operations within the same class.
