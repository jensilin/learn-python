# 👤 Instance Variables in Python

**Instance variables** are variables that belong to an **individual object (instance)** of a class. Each object has its **own copy** of these variables, allowing different objects to store different data.

Instance variables are usually created inside the **constructor (`__init__()`)** using the `self` keyword.

They are one of the most important concepts in Object-Oriented Programming because they store the unique data associated with each object.

# 📖 Topics Covered

## 1. What are Instance Variables?

An **instance variable** is a variable that belongs to a specific object.

Each object maintains its own separate copy of instance variables.

Example:

```python
class Student:

    def __init__(self):
        self.name = "Alice"
```

Here,

- `name` is an instance variable.
- It belongs to each `Student` object individually.

---

## 2. Why Use Instance Variables?

Instance variables help us:

- Store object-specific data.
- Keep each object's information separate.
- Represent real-world entities.
- Avoid data sharing between objects.
- Make classes reusable.

---

## 3. Creating Instance Variables

Instance variables are commonly created inside the constructor using `self`.

```python
class Employee:

    def __init__(self):
        self.name = "John"
        self.salary = 50000
```

Here,

- `name`
- `salary`

are instance variables.

---

## 4. Syntax

### Syntax

```python
class ClassName:

    def __init__(self):
        self.variable_name = value
```

### Explanation

- `self` refers to the current object.
- `variable_name` becomes an instance variable.
- Every object gets its own copy.

---

## 5. Accessing Instance Variables

Instance variables are accessed using the dot (`.`) operator.

```python
class Student:

    def __init__(self):
        self.name = "Alice"

student = Student()

print(student.name)
```

### Output

```text
Alice
```

---

## 6. Initializing Different Values

Each object can have different instance variable values.

```python
class Student:

    def __init__(self, name):
        self.name = name

student1 = Student("Alice")
student2 = Student("Bob")

print(student1.name)
print(student2.name)
```

### Output

```text
Alice
Bob
```

---

## 7. Modifying Instance Variables

Instance variables can be modified after object creation.

```python
class Car:

    def __init__(self):
        self.brand = "Toyota"

car = Car()

car.brand = "Honda"

print(car.brand)
```

### Output

```text
Honda
```

---

## 8. Instance Variables Belong to Objects

Each object stores its own data independently.

```python
class Laptop:

    def __init__(self, brand):
        self.brand = brand

laptop1 = Laptop("Dell")
laptop2 = Laptop("HP")

print(laptop1.brand)
print(laptop2.brand)
```

### Output

```text
Dell
HP
```

Changing one object's data does not affect the other.

---

## 9. Instance Variables vs Class Variables

```python
class Student:

    college = "ABC College"

    def __init__(self, name):
        self.name = name

student1 = Student("Alice")
student2 = Student("Bob")

print(student1.college)
print(student2.college)

print(student1.name)
print(student2.name)
```

### Output

```text
ABC College
ABC College
Alice
Bob
```

Here:

- `college` is a **class variable** (shared).
- `name` is an **instance variable** (unique).

## 10. Practical Examples

### Example 1: Student Information

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

student = Student("David", 20)

print(student.name)
print(student.age)
```

### Output

```text
David
20
```

---

### Example 2: Bank Account

```python
class BankAccount:

    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

account = BankAccount(1001, 5000)

print(account.account_number)
print(account.balance)
```

### Output

```text
1001
5000
```

---

### Example 3: Product Details

```python
class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

product = Product("Laptop", 65000)

print(product.name)
print(product.price)
```

### Output

```text
Laptop
65000
```

---

### Example 4: Vehicle

```python
class Vehicle:

    def __init__(self, brand, wheels):
        self.brand = brand
        self.wheels = wheels

car = Vehicle("Toyota", 4)

print(car.brand)
print(car.wheels)
```

### Output

```text
Toyota
4
```

---

### Example 5: Updating an Instance Variable

```python
class Employee:

    def __init__(self):
        self.salary = 40000

employee = Employee()

employee.salary = 45000

print(employee.salary)
```

### Output

```text
45000
```

## 12. Common Mistakes

Avoid these common mistakes while working with instance variables.

### Forgetting `self`

Incorrect

```python
class Student:

    def __init__(self, name):
        name = name
```

Correct

```python
class Student:

    def __init__(self, name):
        self.name = name
```

---

### Accessing an Instance Variable Without an Object

Incorrect

```python
class Student:

    def __init__(self):
        self.name = "Alice"

print(Student.name)
```

This raises an `AttributeError` because `name` belongs to an object, not the class.

Correct

```python
student = Student()

print(student.name)
```

---

### Assuming All Objects Share Instance Variables

```python
class Test:

    def __init__(self, value):
        self.value = value

obj1 = Test(10)
obj2 = Test(20)

print(obj1.value)
print(obj2.value)
```

### Output

```text
10
20
```

Each object has its own copy of the instance variable.

---

### Forgetting to Initialize an Instance Variable

Incorrect

```python
class Student:

    def display(self):
        print(self.name)
```

Calling `display()` before `self.name` is assigned will raise an `AttributeError`.

Initialize all required instance variables before using them.

---

## 13. Best Practices

Follow these recommended practices when using instance variables.

- Initialize instance variables inside the constructor whenever possible.
- Always access instance variables using `self`.
- Use meaningful and descriptive variable names.
- Keep object-specific data in instance variables.
- Avoid unnecessary modification of instance variables.
- Initialize every required attribute before using it.
- Keep related data grouped within the same class.
