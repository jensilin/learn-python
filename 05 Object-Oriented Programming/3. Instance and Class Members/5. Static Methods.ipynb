# ⚡ Static Methods in Python

**Static methods** are methods that **belong to a class but do not operate on either the class (`cls`) or an object (`self`)**. They are independent methods placed inside a class because they are logically related to that class.

A static method is defined using the **`@staticmethod`** decorator.

Static methods are commonly used for **utility functions**, **helper methods**, **validation logic**, and **mathematical operations** that do not need to access or modify class or instance data.

# 📖 Topics Covered

## 1. What are Static Methods?

A **static method** is a method that belongs to a class but does **not** require:

- `self`
- `cls`

It works like a regular function but is placed inside a class because it is related to that class.

Example:

```python
class Student:

    @staticmethod
    def welcome():
        print("Welcome!")
```

Here, `welcome()` is a static method.

---

## 2. Why Use Static Methods?

Static methods help us:

- Group related utility functions.
- Improve code organization.
- Perform calculations.
- Validate input.
- Avoid unnecessary object creation.

---

## 3. The `@staticmethod` Decorator

To create a static method, use the `@staticmethod` decorator.

```python
class Test:

    @staticmethod
    def show():
        print("Hello")
```

Without this decorator, Python expects the method to receive `self` when called through an object.

---

## 4. Syntax

### Syntax

```python
class ClassName:

    @staticmethod
    def method_name(parameters):
        # Method body
```

### Explanation

- `@staticmethod` converts the method into a static method.
- No `self` parameter.
- No `cls` parameter.
- Behaves like a normal function inside the class.

---

## 5. Calling Static Methods

Static methods can be called using the class name.

```python
class Calculator:

    @staticmethod
    def add(a, b):
        return a + b

print(Calculator.add(5, 10))
```

### Output

```text
15
```

---

They can also be called using an object.

```python
class Calculator:

    @staticmethod
    def multiply(a, b):
        return a * b

calculator = Calculator()

print(calculator.multiply(4, 5))
```

### Output

```text
20
```

Although both approaches work, calling static methods using the class name is generally preferred.

---

## 6. Static Methods Cannot Access Instance Variables

```python
class Student:

    @staticmethod
    def display():
        print("Static Method")
```

Since there is no `self`, a static method cannot directly access instance variables.

---

## 7. Static Methods Cannot Access Class Variables Directly

```python
class Student:

    college = "ABC College"

    @staticmethod
    def show():
        print(Student.college)

Student.show()
```

### Output

```text
ABC College
```

A static method can access class variables only by explicitly using the class name.

---

## 8. Static Methods vs Class Methods vs Instance Methods

| Feature | Instance Method | Class Method | Static Method |
|---------|-----------------|--------------|---------------|
| First Parameter | `self` | `cls` | None |
| Decorator | None | `@classmethod` | `@staticmethod` |
| Access Instance Variables | ✅ Yes | ❌ No | ❌ No |
| Access Class Variables | ✅ Yes | ✅ Yes | Only using class name |
| Called Using | Object | Class or object | Class or object |
| Main Purpose | Work with object data | Work with class data | Utility/helper functions |

---

## 9. Practical Examples

### Example 1: Addition

```python
class Calculator:

    @staticmethod
    def add(a, b):
        return a + b

print(Calculator.add(15, 10))
```

### Output

```text
25
```

---

### Example 2: Finding Maximum

```python
class Utility:

    @staticmethod
    def maximum(a, b):
        return max(a, b)

print(Utility.maximum(30, 20))
```

### Output

```text
30
```

---

### Example 3: Even Number Check

```python
class Number:

    @staticmethod
    def is_even(number):
        return number % 2 == 0

print(Number.is_even(12))
print(Number.is_even(7))
```

### Output

```text
True
False
```

---

### Example 4: Temperature Conversion

```python
class Temperature:

    @staticmethod
    def celsius_to_fahrenheit(celsius):
        return (celsius * 9 / 5) + 32

print(Temperature.celsius_to_fahrenheit(25))
```

### Output

```text
77.0
```

---

### Example 5: Area of a Circle

```python
class Geometry:

    @staticmethod
    def area_of_circle(radius):
        return 3.14 * radius * radius

print(Geometry.area_of_circle(5))
```

### Output

```text
78.5
```

## 11. Common Mistakes

Avoid these common mistakes while working with static methods.

### Forgetting the `@staticmethod` Decorator

Incorrect

```python
class Test:

    def show():
        print("Hello")
```

Correct

```python
class Test:

    @staticmethod
    def show():
        print("Hello")
```

---

### Using `self` Inside a Static Method

Incorrect

```python
class Student:

    @staticmethod
    def display():
        print(self.name)
```

`self` does not exist inside a static method.

---

### Using `cls` Inside a Static Method

Incorrect

```python
class Student:

    @staticmethod
    def display():
        print(cls.college)
```

`cls` is not available in static methods.

---

### Using a Static Method for Object-Specific Logic

If a method needs to access instance variables, it should be an **instance method**, not a static method.

---

## 12. Best Practices

Follow these recommended practices when using static methods.

- Use static methods for utility or helper functions.
- Do not use `self` or `cls` inside static methods.
- Keep static methods independent of object state.
- Call static methods using the class name whenever possible.
- Use descriptive method names.
- Keep methods short and focused.
- Use static methods only when object or class data is not required.
