# 🛠️ Creating Objects in Python

After defining a **class**, the next step is to **create objects** from that class. An object is an **instance** of a class that can access the attributes and methods defined within it.

Creating objects allows you to represent multiple real-world entities using the same class blueprint.

For example, one `Student` class can be used to create many student objects, each representing a different student.

# 📖 Topics Covered

## 1. What is Object Creation?

Object creation is the process of creating an **instance** of a class.

When an object is created:

- Memory is allocated.
- A new object is initialized.
- The object can access the class's attributes and methods.

Example:

```python
class Car:
    pass

car = Car()
```

Here:

- `Car` is the class.
- `car` is the object.

---

## 2. Syntax for Creating an Object

### Syntax

```python
object_name = ClassName()
```

### Explanation

- `object_name` → Variable that stores the object.
- `ClassName()` → Creates a new object of the class.

Example:

```python
class Student:
    pass

student = Student()
```

---

## 3. Creating a Single Object

```python
class Employee:
    pass

employee = Employee()

print(employee)
```

### Sample Output

```text
<__main__.Employee object at 0x7f8c5d9b8fd0>
```

> **Note:** The memory address will be different each time the program runs.

---

## 4. Creating Multiple Objects

One class can create many independent objects.

```python
class Student:
    pass

student1 = Student()
student2 = Student()
student3 = Student()
```

Each object occupies its own memory location.

---

## 5. Accessing Class Attributes

Objects can access class attributes using the dot (`.`) operator.

```python
class Car:

    brand = "Toyota"

car = Car()

print(car.brand)
```

### Output

```text
Toyota
```

---

## 6. Calling Methods Using Objects

Objects can call methods defined inside the class.

```python
class Dog:

    def bark(self):
        print("Woof!")

dog = Dog()

dog.bark()
```

### Output

```text
Woof!
```

---

## 7. Creating Independent Objects

Although objects are created from the same class, each one is a separate instance.

```python
class Laptop:
    pass

laptop1 = Laptop()
laptop2 = Laptop()

print(laptop1 is laptop2)
```

### Output

```text
False
```

Each object has its own identity.

---

## 8. Checking Object Identity

Use the `id()` function to view the unique identity of an object.

```python
class Book:
    pass

book1 = Book()
book2 = Book()

print(id(book1))
print(id(book2))
```

### Sample Output

```text
140542478760848
140542478760976
```

The values will differ because the objects are different.

---

## 9. Creating Objects from the Same Class

```python
class Mobile:
    pass

phone1 = Mobile()
phone2 = Mobile()

print(type(phone1))
print(type(phone2))
```

### Output

```text
<class '__main__.Mobile'>
<class '__main__.Mobile'>
```

Both objects belong to the same class but are different instances.

## 10. Practical Examples

### Example 1: Student Objects

```python
class Student:

    college = "ABC College"

student1 = Student()
student2 = Student()

print(student1.college)
print(student2.college)
```

### Output

```text
ABC College
ABC College
```

---

### Example 2: Vehicle Objects

```python
class Vehicle:

    wheels = 4

car = Vehicle()
truck = Vehicle()

print(car.wheels)
print(truck.wheels)
```

### Output

```text
4
4
```

---

### Example 3: Calculator Object

```python
class Calculator:

    def add(self):
        print(25 + 15)

calc = Calculator()

calc.add()
```

### Output

```text
40
```

---

### Example 4: Greeting Objects

```python
class Greeting:

    def welcome(self):
        print("Welcome to Python!")

greet = Greeting()

greet.welcome()
```

### Output

```text
Welcome to Python!
```

---

### Example 5: Comparing Objects

```python
class Test:
    pass

obj1 = Test()
obj2 = Test()

print(obj1 == obj2)
print(obj1 is obj2)
```

### Output

```text
False
False
```

Both objects are different instances.

## 12. Common Mistakes

Avoid these common mistakes while creating objects.

### Forgetting Parentheses

Incorrect

```python
class Student:
    pass

student = Student
```

Correct

```python
student = Student()
```

---

### Creating an Object Before Defining the Class

Incorrect

```python
student = Student()

class Student:
    pass
```

Correct

```python
class Student:
    pass

student = Student()
```

---

### Calling Methods Without an Object

Incorrect

```python
class Dog:

    def bark(self):
        print("Woof!")

Dog.bark()
```

Correct

```python
dog = Dog()

dog.bark()
```

---

### Assuming Different Objects Are Identical

```python
class Car:
    pass

car1 = Car()
car2 = Car()

print(car1 is car2)
```

### Output

```text
False
```

Each object has its own identity.

---

## 13. Best Practices

Follow these recommended practices when creating objects.

- Create objects only after defining the class.
- Use meaningful object names.
- Create separate objects for separate entities.
- Access members using the dot (`.`) operator.
- Avoid creating unnecessary objects.
- Use objects to represent real-world entities.
- Keep your code readable and well organized.
