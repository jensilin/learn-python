# ⚖️ Class vs Object in Python

A **class** and an **object** are two of the most fundamental concepts in **Object-Oriented Programming (OOP)**.

A **class** acts as a **blueprint or template**, while an **object** is the **actual instance created from that blueprint**.

Understanding the difference between a class and an object is essential before learning advanced OOP concepts such as constructors, inheritance, encapsulation, and polymorphism.

# 📖 Topics Covered

## 1. What is a Class?

A **class** is a blueprint or template used to create objects.

It defines:

- Attributes (data)
- Methods (behavior)

A class itself does not represent a real entity. It only describes how objects should be created.

Example:

```python
class Car:
    pass
```

Here, `Car` is a class.

---

## 2. What is an Object?

An **object** is an instance of a class.

It is the real entity created using the class.

Example:

```python
class Car:
    pass

car1 = Car()
```

Here:

- `Car` → Class
- `car1` → Object

---

## 3. Relationship Between Class and Object

A class defines the structure.

Objects use that structure.

One class can create many objects.

```python
class Student:
    pass

student1 = Student()
student2 = Student()
student3 = Student()
```

All three objects belong to the same class.

---

## 4. Real-World Example

Imagine a **house blueprint**.

- Blueprint → Class
- Actual house → Object

One blueprint can be used to build many houses.

Similarly,

- Student → Class
- John, Alice, David → Objects

---

## 5. Class vs Object Comparison

| Class | Object |
|--------|--------|
| Blueprint | Instance |
| Logical entity | Physical entity |
| Defines structure | Uses the structure |
| Created using `class` keyword | Created using `ClassName()` |
| Does not occupy memory like an object instance | Occupies memory |
| One class can create many objects | Belongs to one class |

---

## 6. Creating a Class and an Object

```python
class Employee:
    pass

employee = Employee()
```

Here,

- `Employee` is the class.
- `employee` is the object.

---

## 7. Multiple Objects from One Class

```python
class Laptop:
    pass

laptop1 = Laptop()
laptop2 = Laptop()
laptop3 = Laptop()
```

Each object is different.

```python
print(laptop1 is laptop2)
```

### Output

```text
False
```

---

## 8. Accessing Class Members Through Objects

```python
class Car:

    brand = "Toyota"

car = Car()

print(car.brand)
```

### Output

```text
Toyota
```

Objects use the dot (`.`) operator to access attributes and methods.

---

## 9. Checking Types

```python
class Animal:
    pass

dog = Animal()

print(type(Animal))
print(type(dog))
```

### Output

```text
<class 'type'>
<class '__main__.Animal'>
```

The class itself is an object of Python's built-in `type` class, while `dog` is an instance of `Animal`.

## 10. Practical Examples

## 12. Common Mistakes

Avoid these common mistakes while learning classes and objects.

### Thinking a Class Is an Object

Incorrect understanding:

```
Student = Object
```

Correct understanding:

```
Student → Class
student1 → Object
```

---

### Forgetting to Create an Object

Incorrect

```python
class Student:
    pass

Student.show()
```

Correct

```python
class Student:

    def show(self):
        print("Hello")

student = Student()

student.show()
```

---

### Forgetting Parentheses

Incorrect

```python
student = Student
```

Correct

```python
student = Student()
```

---

### Assuming All Objects Are the Same

```python
class Car:
    pass

car1 = Car()
car2 = Car()

print(car1 == car2)
```

### Output

```text
False
```

Each object is a separate instance.

---

## 13. Best Practices

Follow these recommended practices when working with classes and objects.

- Use meaningful class names in PascalCase.
- Use descriptive object names.
- Create separate objects for separate entities.
- Keep each class focused on a single responsibility.
- Access members using the dot (`.`) operator.
- Reuse classes to create multiple objects instead of duplicating code.
- Design classes to represent real-world concepts.

# 📊 Summary

| Feature | Class | Object |
|---------|-------|--------|
| Definition | Blueprint or template | Instance of a class |
| Purpose | Defines structure and behavior | Represents a real entity |
| Creation | Using the `class` keyword | Using `ClassName()` |
| Memory | Defines the blueprint | Occupies memory |
| Quantity | One class can create many objects | Each object belongs to one class |

> **Note:** A class defines **what an object should look like**, while an object is the **actual instance created from that definition**. You typically create one class and many objects from it. Understanding the relationship between classes and objects is the foundation of Object-Oriented Programming and prepares you for advanced concepts such as constructors, inheritance, encapsulation, and polymorphism.
