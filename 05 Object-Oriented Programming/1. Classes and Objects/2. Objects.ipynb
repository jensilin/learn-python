# 🎯 Objects in Python

An **object** is an **instance of a class**. When a class is created, it serves as a blueprint. An object is the actual entity created from that blueprint.

Objects contain their own **attributes (data)** and can perform **methods (functions)** defined inside their class.

In Object-Oriented Programming (OOP), objects represent real-world entities such as students, cars, employees, bank accounts, and many more.

# 📖 Topics Covered

## 1. What is an Object?

An **object** is an instance of a class.

A class defines the structure, while objects are the actual entities created from that structure.

Example:

```python
class Car:
    pass

car1 = Car()
```

Here:

- `Car` is a class.
- `car1` is an object of the `Car` class.

---

## 2. Why Use Objects?

Objects help us:

- Represent real-world entities.
- Store individual data.
- Call methods defined in a class.
- Create multiple independent instances.
- Organize programs efficiently.

---

## 3. Syntax for Creating an Object

### Syntax

```python
object_name = ClassName()
```

Example:

```python
class Student:
    pass

student = Student()
```

### Explanation

- `Student` → Class
- `Student()` → Constructor call that creates an object
- `student` → Object reference

---

## 4. Creating Multiple Objects

A single class can create many objects.

```python
class Student:
    pass

student1 = Student()
student2 = Student()
student3 = Student()
```

Each object is independent.

---

## 5. Checking the Type of an Object

Use the `type()` function to determine the class of an object.

```python
class Employee:
    pass

emp = Employee()

print(type(emp))
```

### Output

```text
<class '__main__.Employee'>
```

---

## 6. Checking Object Identity

Every object has its own identity.

```python
class Laptop:
    pass

laptop1 = Laptop()
laptop2 = Laptop()

print(id(laptop1))
print(id(laptop2))
```

### Sample Output

```text
140543260798112
140543260798304
```

Each object has a unique identity.

---

## 7. Multiple Objects of the Same Class

```python
class Animal:
    pass

dog = Animal()
cat = Animal()

print(type(dog))
print(type(cat))
```

### Output

```text
<class '__main__.Animal'>
<class '__main__.Animal'>
```

Although both belong to the same class, they are different objects.

---

## 8. Objects Can Access Methods

Objects use the dot (`.`) operator to call methods.

```python
class Student:

    def greet(self):
        print("Welcome!")

student = Student()

student.greet()
```

### Output

```text
Welcome!
```

---

## 9. Objects Can Access Attributes

```python
class Car:

    brand = "Toyota"

car = Car()

print(car.brand)
```

### Output

```text
Toyota
```

---

## 10. Practical Examples

### Example 1: Student Object

```python
class Student:

    college = "ABC College"

student = Student()

print(student.college)
```

### Output

```text
ABC College
```

---

### Example 2: Bank Account

```python
class BankAccount:

    bank_name = "XYZ Bank"

account = BankAccount()

print(account.bank_name)
```

### Output

```text
XYZ Bank
```

---

### Example 3: Calculator

```python
class Calculator:

    def add(self):
        print(50 + 20)

calc = Calculator()

calc.add()
```

### Output

```text
70
```

---

### Example 4: Vehicle

```python
class Vehicle:

    wheels = 4

car = Vehicle()

print(car.wheels)
```

### Output

```text
4
```

## 12. Common Mistakes

Avoid these common mistakes while creating and using objects.

### Forgetting Parentheses

Incorrect

```python
class Student:
    pass

student = Student
```

Correct

```python
student = Student()
```

---

### Calling an Instance Method Without an Object

Incorrect

```python
class Dog:

    def bark(self):
        print("Woof!")

Dog.bark()
```

Correct

```python
dog = Dog()

dog.bark()
```

---

### Assuming All Objects Are the Same

```python
class Book:
    pass

book1 = Book()
book2 = Book()

print(book1 == book2)
```

### Output

```text
False
```

Different objects are not equal unless explicitly defined.

---

### Forgetting That Each Object Has Its Own Identity

```python
class Test:
    pass

a = Test()
b = Test()

print(id(a))
print(id(b))
```

The identities are different because they are different objects.

---

## 13. Best Practices

Follow these recommended practices when working with objects.

- Use meaningful object names.
- Create separate objects for separate entities.
- Access attributes and methods using the dot (`.`) operator.
- Avoid creating unnecessary objects.
- Keep object names descriptive and readable.
- Use objects to represent real-world entities.
