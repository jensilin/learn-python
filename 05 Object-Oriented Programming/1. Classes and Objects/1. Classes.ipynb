# 🏛️ Classes in Python

A **class** is a **blueprint or template** for creating objects in Python. It defines the **attributes (data)** and **methods (functions)** that the objects created from the class will have.

Think of a class as a **design or blueprint**, while an object is the **actual item created from that design**.

Classes are the foundation of **Object-Oriented Programming (OOP)** and help organize code into reusable and maintainable components.

# 📖 Topics Covered

## 1. What is a Class?

A **class** is a user-defined data type that acts as a blueprint for creating objects.

A class defines:

- Attributes (variables)
- Methods (functions)
- Behavior of objects

Once a class is created, multiple objects can be created from it.

Example:

```python
class Car:
    pass
```

The class `Car` is only a blueprint. No actual car object exists until one is created.

---

## 2. Why Use Classes?

Classes help us:

- Organize code.
- Reuse code.
- Model real-world entities.
- Improve readability.
- Reduce code duplication.
- Build scalable applications.

Without classes, managing large applications becomes difficult.

---

## 3. Syntax of a Class

### Syntax

```python
class ClassName:
    # Class body
```

Example:

```python
class Student:
    pass
```

### Explanation

- `class` → Keyword used to define a class.
- `Student` → Class name.
- `:` → Begins the class body.
- `pass` → Placeholder indicating an empty class.

---

## 4. Creating a Simple Class

```python
class Person:
    pass
```

The class is created successfully, but it currently contains no attributes or methods.

---

## 5. Creating a Class with Attributes

Class attributes are variables defined inside a class.

```python
class Car:
    brand = "Toyota"
    wheels = 4
```

Accessing class attributes:

```python
print(Car.brand)
print(Car.wheels)
```

### Output

```text
Toyota
4
```

---

## 6. Creating a Class with Methods

Methods define the behavior of a class.

```python
class Dog:

    def bark(self):
        print("Woof!")
```

Creating an object:

```python
dog = Dog()

dog.bark()
```

### Output

```text
Woof!
```

---

## 7. The `self` Parameter

The first parameter of an instance method is usually named `self`.

It refers to the **current object** that calls the method.

Example:

```python
class Student:

    def show(self):
        print("Welcome to Python")
```

Creating an object:

```python
student = Student()

student.show()
```

### Output

```text
Welcome to Python
```

---

## 8. Class Naming Conventions

According to **PEP 8**, class names should use **PascalCase** (also called CapWords).

Good examples:

```python
class Student:
    pass

class BankAccount:
    pass

class EmployeeDetails:
    pass
```

Avoid:

```python
class student:
    pass

class bank_account:
    pass
```

## 12. Common Mistakes

Avoid these common mistakes while creating classes.

### Forgetting Parentheses When Creating an Object

Incorrect

```python
class Student:
    pass

student = Student
```

Correct

```python
student = Student()
```

---

### Forgetting `self` in Methods

Incorrect

```python
class Test:

    def display():
        print("Hello")
```

Correct

```python
class Test:

    def display(self):
        print("Hello")
```

---

### Incorrect Class Naming

Incorrect

```python
class student:
    pass
```

Preferred

```python
class Student:
    pass
```

---

### Using a Class Without Creating an Object

Incorrect

```python
class Dog:

    def bark(self):
        print("Woof!")

Dog.bark()
```

This raises an error because `bark()` is an instance method.

Correct

```python
dog = Dog()

dog.bark()
```

---

## 13. Best Practices

Follow these recommended practices when creating classes.

- Use meaningful class names.
- Follow PascalCase naming conventions.
- Keep each class focused on a single responsibility.
- Define methods that represent the behavior of objects.
- Use `self` correctly in instance methods.
- Write reusable and modular classes.
- Add documentation for larger classes.
