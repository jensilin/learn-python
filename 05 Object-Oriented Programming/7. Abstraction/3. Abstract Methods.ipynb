# 📝 Abstract Methods in Python

An **Abstract Method** is a method that is **declared in an abstract class but does not contain a complete implementation**. Instead, it defines a **contract** that every concrete child class must follow.

Abstract methods ensure that subclasses provide their own implementation while maintaining a consistent interface across related classes.

In Python, abstract methods are created using the **`@abstractmethod`** decorator from the **`abc`** module.

# 💻 Quick Example

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):

    def sound(self):
        print("Dog barks.")

dog = Dog()

dog.sound()
```

### Output

```text
Dog barks.
```

The `sound()` method is declared in the abstract class and implemented in the child class.

# 📖 Topics Covered

## 1. What is an Abstract Method?

An **abstract method** is a method that is declared without a complete implementation.

It specifies **what a child class must do**, but not **how it should do it**.

Example:

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass
```

Every concrete subclass of `Shape` must implement the `area()` method.

## 3. Creating an Abstract Method

To create an abstract method:

1. Import `abstractmethod`.
2. Create an abstract class by inheriting from `ABC`.
3. Decorate the method with `@abstractmethod`.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass
```

---

## 4. Implementing an Abstract Method

Every concrete child class must provide an implementation.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):

    def start(self):
        print("Car Started")

car = Car()

car.start()
```

### Output

```text
Car Started
```

---

## 5. Child Class Must Implement Every Abstract Method

If a subclass does not implement all inherited abstract methods, Python prevents object creation.

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    pass

dog = Dog()
```

### Output

```text
TypeError:
Can't instantiate abstract class Dog
with abstract method sound
```

---

## 6. Multiple Abstract Methods

An abstract class can contain multiple abstract methods.

```python
from abc import ABC, abstractmethod

class Device(ABC):

    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

class Computer(Device):

    def start(self):
        print("Computer Started")

    def stop(self):
        print("Computer Stopped")

computer = Computer()

computer.start()
computer.stop()
```

### Output

```text
Computer Started
Computer Stopped
```

---

## 7. Abstract Methods Can Coexist with Regular Methods

Abstract classes may contain both abstract methods and normal methods.

```python
from abc import ABC, abstractmethod

class Employee(ABC):

    @abstractmethod
    def work(self):
        pass

    def login(self):
        print("Employee Logged In")

class Developer(Employee):

    def work(self):
        print("Writing Code")

developer = Developer()

developer.login()
developer.work()
```

### Output

```text
Employee Logged In
Writing Code
```

---

## 8. Abstract Methods and Runtime Polymorphism

Abstract methods work naturally with polymorphism.

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):

    def sound(self):
        print("Dog Bark")

class Cat(Animal):

    def sound(self):
        print("Cat Meow")

animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```

### Output

```text
Dog Bark
Cat Meow
```

Each subclass provides its own implementation while sharing the same interface.

## 12. Common Mistakes

Avoid these common mistakes while working with abstract methods.

### Forgetting the `@abstractmethod` Decorator

Incorrect

```python
class Animal(ABC):

    def sound(self):
        pass
```

Without the decorator, `sound()` becomes a regular method.

---

### Forgetting to Implement an Abstract Method

Incorrect

```python
class Dog(Animal):
    pass
```

This prevents object creation.

---

### Creating Objects of an Abstract Class

Incorrect

```python
animal = Animal()
```

Abstract classes cannot be instantiated.

---

### Forgetting to Import `abstractmethod`

Always import:

```python
from abc import ABC, abstractmethod
```

## 13. Best Practices

Follow these recommended practices when working with abstract methods.

- Use abstract methods to define required behavior.
- Keep abstract methods focused and meaningful.
- Implement every abstract method in concrete subclasses.
- Use descriptive method names.
- Combine abstract and regular methods where appropriate.
- Avoid unnecessary abstraction in small projects.
- Design interfaces that are simple and extensible.
