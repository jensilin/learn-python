# 🎯 Abstraction in Python

**Abstraction** is one of the four fundamental principles of **Object-Oriented Programming (OOP)**.

Abstraction means **hiding the internal implementation details of an object and exposing only the essential features to the user**.

In simple words:

> **Show what an object can do, but hide how it does it.**

Python supports abstraction using the **`abc` (Abstract Base Class)** module, which allows developers to define **abstract classes** and **abstract methods**.

Abstraction helps build applications that are **simple to use, easier to maintain, and more secure**.

# 💻 Quick Example

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):

    def sound(self):
        print("Dog barks.")

dog = Dog()

dog.sound()
```

### Output

```text
Dog barks.
```

The user only calls `sound()`. The internal implementation is provided by the child class.

## 1. What is Abstraction?

Abstraction means **hiding unnecessary implementation details and exposing only the required functionality**.

Users interact with an object's interface without knowing how it is implemented internally.

Example:

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):

    def start(self):
        print("Car Started")

car = Car()

car.start()
```

### Output

```text
Car Started
```

## 4. Abstraction in Python

Python implements abstraction using the **`abc`** module.

The `abc` module provides:

- `ABC`
- `@abstractmethod`

These allow developers to define methods that must be implemented by subclasses.

---

## 5. Abstract Class

An **abstract class** is a class that cannot be instantiated directly.

It acts as a blueprint for other classes.

Example:

```python
from abc import ABC

class Animal(ABC):
    pass
```

Objects cannot be created directly from an abstract class.

---

## 6. Abstract Method

An **abstract method** is declared inside an abstract class but has no implementation.

Every child class must provide its own implementation.

Example:

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass
```

---

## 7. Implementing an Abstract Method

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):

    def sound(self):
        print("Dog Bark")

dog = Dog()

dog.sound()
```

### Output

```text
Dog Bark
```

---

## 8. What Happens If a Child Doesn't Implement an Abstract Method?

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    pass

dog = Dog()
```

### Output

```text
TypeError:
Can't instantiate abstract class Dog
with abstract method sound
```

Python prevents creating an object until all abstract methods are implemented.

## 10. Practical Examples

### Example 1: Vehicle

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Bike(Vehicle):

    def start(self):
        print("Bike Started")

bike = Bike()

bike.start()
```

### Output

```text
Bike Started
```

### Example 2: Payment System

```python
from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def pay(self):
        pass

class CreditCard(Payment):

    def pay(self):
        print("Paid using Credit Card")

payment = CreditCard()

payment.pay()
```

### Output

```text
Paid using Credit Card
```

### Example 3: Employee

```python
from abc import ABC, abstractmethod

class Employee(ABC):

    @abstractmethod
    def work(self):
        pass

class Developer(Employee):

    def work(self):
        print("Writing Code")

developer = Developer()

developer.work()
```

### Example 4: Shape

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def draw(self):
        pass

class Circle(Shape):

    def draw(self):
        print("Drawing Circle")

circle = Circle()

circle.draw()
```

## 12. Common Mistakes

Avoid these common mistakes while working with abstraction.

### Creating Objects of an Abstract Class

Incorrect

```python
animal = Animal()
```

Abstract classes cannot be instantiated.

---

### Forgetting to Implement Abstract Methods

Incorrect

```python
class Dog(Animal):
    pass
```

All abstract methods must be implemented before creating objects.

---

### Forgetting the `@abstractmethod` Decorator

Incorrect

```python
class Animal(ABC):

    def sound(self):
        pass
```

Without `@abstractmethod`, the method is treated as a normal method.

---

### Importing the Wrong Module

Always import:

```python
from abc import ABC, abstractmethod
```

## 13. Best Practices

Follow these recommended practices when using abstraction.

- Use abstraction to define common interfaces.
- Keep abstract classes focused on shared behavior.
- Implement every abstract method in child classes.
- Hide implementation details from users.
- Use meaningful method names.
- Avoid unnecessary abstraction in small programs.
- Design abstract classes to be reusable and extensible.
