# 🏛️ Abstract Classes in Python

An **Abstract Class** is a special type of class that **cannot be instantiated directly**. It serves as a **blueprint** for other classes by defining common attributes and methods that child classes should have.

Abstract classes may contain:

- Abstract methods (methods without implementation)
- Regular methods (methods with implementation)
- Constructors
- Variables

In Python, abstract classes are created using the **`ABC`** class from the **`abc`** module.

# 💻 Quick Example

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):

    def sound(self):
        print("Dog barks.")

dog = Dog()

dog.sound()
```

### Output

```text
Dog barks.
```

The `Animal` class acts as an abstract blueprint, while `Dog` provides the actual implementation.

# 📖 Topics Covered

## 1. What is an Abstract Class?

An **abstract class** is a class that **cannot be instantiated directly** and is intended to be inherited by other classes.

It defines a common interface that all child classes must follow.

Example:

```python
from abc import ABC

class Animal(ABC):
    pass
```

The `Animal` class is now an abstract class because it inherits from `ABC`.

---

## 2. Why Use Abstract Classes?

Abstract classes help us:

- Define a common blueprint.
- Enforce consistent method implementation.
- Reduce code duplication.
- Improve code organization.
- Support polymorphism.
- Build maintainable applications.

---

## 3. Creating an Abstract Class

To create an abstract class:

1. Import `ABC`.
2. Inherit from `ABC`.
3. Define one or more abstract methods.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass
```

---

## 4. Abstract Classes Cannot Be Instantiated

Trying to create an object of an abstract class results in an error.

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

animal = Animal()
```

### Output

```text
TypeError:
Can't instantiate abstract class Animal
with abstract method sound
```

---

## 5. Child Classes Must Implement Abstract Methods

Every concrete child class must implement all inherited abstract methods.

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):

    def sound(self):
        print("Dog Bark")

dog = Dog()

dog.sound()
```

### Output

```text
Dog Bark
```

---

## 6. Abstract Classes Can Have Regular Methods

Abstract classes are not limited to abstract methods. They can also contain fully implemented methods.

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

    def sleep(self):
        print("Animal Sleeping")

class Dog(Animal):

    def sound(self):
        print("Dog Bark")

dog = Dog()

dog.sound()
dog.sleep()
```

### Output

```text
Dog Bark
Animal Sleeping
```

---

## 7. Abstract Classes Can Have Constructors

Abstract classes can define constructors that are inherited by child classes.

```python
from abc import ABC, abstractmethod

class Person(ABC):

    def __init__(self, name):
        self.name = name

    @abstractmethod
    def work(self):
        pass

class Developer(Person):

    def work(self):
        print(self.name, "is coding.")

developer = Developer("Alice")

developer.work()
```

### Output

```text
Alice is coding.
```

---

## 8. Abstract Classes Can Store Variables

Like regular classes, abstract classes can contain instance and class variables.

```python
from abc import ABC, abstractmethod

class Employee(ABC):

    company = "ABC Technologies"

    @abstractmethod
    def work(self):
        pass

class Manager(Employee):

    def work(self):
        print("Managing Team")

manager = Manager()

print(manager.company)
manager.work()
```

### Output

```text
ABC Technologies
Managing Team
```

## 10. Practical Examples

### Example 1: Vehicle

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):

    def start(self):
        print("Car Started")

car = Car()

car.start()
```

### Output

```text
Car Started
```

### Example 2: Payment System

```python
from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def pay(self):
        pass

class UPI(Payment):

    def pay(self):
        print("Paid using UPI")

payment = UPI()

payment.pay()
```

### Output

```text
Paid using UPI
```

### Example 3: Shape

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Square(Shape):

    def area(self):
        print("Calculating Square Area")

square = Square()

square.area()
```

## 12. Common Mistakes

Avoid these common mistakes while working with abstract classes.

### Creating an Object of an Abstract Class

Incorrect

```python
animal = Animal()
```

Abstract classes cannot be instantiated.

---

### Forgetting to Implement an Abstract Method

Incorrect

```python
class Dog(Animal):
    pass
```

Concrete child classes must implement every inherited abstract method.

---

### Forgetting to Inherit from `ABC`

Incorrect

```python
class Animal:

    @abstractmethod
    def sound(self):
        pass
```

Without inheriting from `ABC`, the class is not treated as an abstract class.

---

### Assuming Abstract Classes Cannot Have Normal Methods

Abstract classes can contain regular methods, constructors, and variables in addition to abstract methods.

---

## 13. Best Practices

Follow these recommended practices when working with abstract classes.

- Use abstract classes to define shared interfaces.
- Keep abstract classes focused on common behavior.
- Implement all abstract methods in child classes.
- Reuse common logic through regular methods.
- Use meaningful method names.
- Avoid unnecessary abstraction for simple programs.
- Design abstract classes to be extensible.
