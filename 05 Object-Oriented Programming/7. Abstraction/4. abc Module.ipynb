# 📦 The `abc` Module in Python

The **`abc` (Abstract Base Classes)** module is a built-in Python module that provides the tools required to implement **abstraction**.

It allows developers to create:

- Abstract classes
- Abstract methods
- Standardized interfaces

By using the `abc` module, Python ensures that subclasses implement required methods before objects can be created.

The `abc` module is commonly used in object-oriented programming to build **scalable, maintainable, and reusable applications**.

# 💻 Quick Example

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):

    def sound(self):
        print("Dog barks.")

dog = Dog()

dog.sound()
```

### Output

```text
Dog barks.
```

The `abc` module ensures that every concrete subclass implements the `sound()` method.

# 📖 Topics Covered

## 1. What is the `abc` Module?

The **`abc`** module is a built-in Python module used to create **Abstract Base Classes (ABCs)**.

It provides a way to define common interfaces that subclasses must follow.

Importing the module:

```python
from abc import ABC, abstractmethod
```

---

## 2. Why Use the `abc` Module?

The `abc` module helps us:

- Implement abstraction.
- Create abstract classes.
- Create abstract methods.
- Enforce method implementation.
- Build reusable interfaces.
- Improve code consistency.
- Support polymorphism.

---

## 3. Main Components of the `abc` Module

The two most commonly used components are:

| Component | Purpose |
|-----------|---------|
| `ABC` | Creates an abstract base class |
| `@abstractmethod` | Declares an abstract method |

These components work together to implement abstraction.

---

## 4. Creating an Abstract Base Class

An abstract base class is created by inheriting from `ABC`.

```python
from abc import ABC

class Animal(ABC):
    pass
```

The `Animal` class is now an abstract base class.

---

## 5. Creating an Abstract Method

Abstract methods are created using the `@abstractmethod` decorator.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass
```

Every concrete subclass must implement the `start()` method.

---

## 6. Complete Example Using the `abc` Module

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Square(Shape):

    def area(self):
        print("Calculating Square Area")

square = Square()

square.area()
```

### Output

```text
Calculating Square Area
```

---

## 7. Preventing Object Creation

The `abc` module prevents objects from being created if abstract methods are not implemented.

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

animal = Animal()
```

### Output

```text
TypeError:
Can't instantiate abstract class Animal
with abstract method sound
```

---

## 8. Multiple Abstract Methods

The `abc` module allows multiple abstract methods in the same class.

```python
from abc import ABC, abstractmethod

class Device(ABC):

    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

class Computer(Device):

    def start(self):
        print("Computer Started")

    def stop(self):
        print("Computer Stopped")

computer = Computer()

computer.start()
computer.stop()
```

### Output

```text
Computer Started
Computer Stopped
```

---

## 9. Mixing Abstract and Regular Methods

Abstract classes can contain both abstract methods and normal methods.

```python
from abc import ABC, abstractmethod

class Employee(ABC):

    @abstractmethod
    def work(self):
        pass

    def login(self):
        print("Employee Logged In")

class Developer(Employee):

    def work(self):
        print("Writing Code")

developer = Developer()

developer.login()
developer.work()
```

### Output

```text
Employee Logged In
Writing Code
```

## 12. Common Mistakes

Avoid these common mistakes while working with the `abc` module.

### Forgetting to Import `ABC`

Incorrect

```python
class Animal(ABC):
    pass
```

Always import:

```python
from abc import ABC
```

---

### Forgetting the `@abstractmethod` Decorator

Incorrect

```python
class Animal(ABC):

    def sound(self):
        pass
```

Without the decorator, `sound()` is treated as a normal method.

---

### Instantiating an Abstract Class

Incorrect

```python
animal = Animal()
```

Abstract classes cannot be instantiated.

---

### Forgetting to Implement Abstract Methods

Incorrect

```python
class Dog(Animal):
    pass
```

Concrete subclasses must implement all inherited abstract methods before objects can be created.

## 13. Best Practices

Follow these recommended practices when using the `abc` module.

- Use `ABC` to define abstract base classes.
- Mark required methods with `@abstractmethod`.
- Keep abstract classes focused on shared behavior.
- Implement every abstract method in concrete subclasses.
- Combine abstract and regular methods where appropriate.
- Avoid unnecessary abstraction for simple programs.
- Design reusable and extensible interfaces.
