## 1. Designing the Employee Class

The `Employee` class stores information about each employee.

```python
class Employee:

    def __init__(self, employee_id, name, department, designation, salary):

        self.employee_id = employee_id
        self.name = name
        self.department = department
        self.designation = designation
        self.salary = salary

    def display(self):

        print(f"Employee ID : {self.employee_id}")
        print(f"Name        : {self.name}")
        print(f"Department  : {self.department}")
        print(f"Designation : {self.designation}")
        print(f"Salary      : ₹{self.salary}")
```

## 3. Adding an Employee

```python
def add_employee(self, employee):

    self.employees.append(employee)

    print("Employee added successfully.")
```

---

## 4. Viewing All Employees

```python
def view_employees(self):

    if not self.employees:

        print("No employees available.")
        return

    for employee in self.employees:

        employee.display()

        print("-" * 40)
```

---

## 5. Searching for an Employee

```python
def search_employee(self, employee_id):

    for employee in self.employees:

        if employee.employee_id == employee_id:

            return employee

    return None
```

---

## 6. Updating Employee Salary

```python
def update_salary(self, employee_id, salary):

    employee = self.search_employee(employee_id)

    if employee:

        employee.salary = salary

        print("Salary updated successfully.")

    else:

        print("Employee not found.")
```

---

## 7. Deleting an Employee

```python
def delete_employee(self, employee_id):

    employee = self.search_employee(employee_id)

    if employee:

        self.employees.remove(employee)

        print("Employee deleted successfully.")

    else:

        print("Employee not found.")
```

## 8. Complete Project

```python
class Employee:

    def __init__(self, employee_id, name, department, designation, salary):

        self.employee_id = employee_id
        self.name = name
        self.department = department
        self.designation = designation
        self.salary = salary

    def display(self):

        print(f"Employee ID : {self.employee_id}")
        print(f"Name        : {self.name}")
        print(f"Department  : {self.department}")
        print(f"Designation : {self.designation}")
        print(f"Salary      : ₹{self.salary}")


class EmployeeManagementSystem:

    def __init__(self):

        self.employees = []

    def add_employee(self, employee):

        self.employees.append(employee)

        print("Employee added successfully.")

    def view_employees(self):

        if not self.employees:

            print("No employees available.")
            return

        for employee in self.employees:

            employee.display()

            print("-" * 40)

    def search_employee(self, employee_id):

        for employee in self.employees:

            if employee.employee_id == employee_id:
                return employee

        return None

    def update_salary(self, employee_id, salary):

        employee = self.search_employee(employee_id)

        if employee:

            employee.salary = salary

            print("Salary updated successfully.")

        else:

            print("Employee not found.")

    def delete_employee(self, employee_id):

        employee = self.search_employee(employee_id)

        if employee:

            self.employees.remove(employee)

            print("Employee deleted successfully.")

        else:

            print("Employee not found.")


system = EmployeeManagementSystem()

employee1 = Employee(101, "Alice", "HR", "Manager", 65000)
employee2 = Employee(102, "Bob", "IT", "Software Engineer", 75000)

system.add_employee(employee1)
system.add_employee(employee2)

print("\nEmployee Records")

system.view_employees()

print("\nUpdating Salary")

system.update_salary(102, 80000)

print("\nUpdated Employee Details")

employee = system.search_employee(102)

if employee:
    employee.display()

print("\nDeleting Employee")

system.delete_employee(101)

print("\nRemaining Employees")

system.view_employees()
```

## 12. Common Mistakes

Avoid these common mistakes while building an employee management system.

### Allowing Duplicate Employee IDs

Always ensure that every employee has a unique employee ID.

---

### Allowing Invalid Salaries

Validate salary before storing it.

Example:

```python
if salary < 0:
    raise ValueError("Salary cannot be negative.")
```

---

### Not Checking Whether an Employee Exists

Always verify that an employee exists before updating or deleting their record.

---

### Mixing User Interface with Business Logic

Keep menu-driven input/output separate from the classes responsible for managing employee records.

---

## 13. Best Practices

Follow these recommended practices while building the project.

- Create separate classes for employee data and employee management.
- Validate all user inputs.
- Prevent duplicate employee IDs.
- Use meaningful class and method names.
- Keep methods short and focused.
- Protect employee data using encapsulation.
- Handle exceptions gracefully.
- Avoid duplicate code.
- Follow the SOLID principles where appropriate.
- Write clean, modular, and maintainable code.
