## 1. Designing the Student Class

The `Student` class stores information about each student.

```python
class Student:

    def __init__(self, roll_no, name, age, marks):

        self.roll_no = roll_no
        self.name = name
        self.age = age
        self.marks = marks

    def display(self):

        print(f"Roll No : {self.roll_no}")
        print(f"Name    : {self.name}")
        print(f"Age     : {self.age}")
        print(f"Marks   : {self.marks}")
```

## 3. Adding a Student

```python
class StudentManagementSystem:

    def __init__(self):
        self.students = []

    def add_student(self, student):

        self.students.append(student)

        print("Student added successfully.")
```

## 4. Viewing All Students

```python
def view_students(self):

    if not self.students:

        print("No students available.")
        return

    for student in self.students:

        student.display()

        print("-" * 30)
```

---

## 5. Searching for a Student

```python
def search_student(self, roll_no):

    for student in self.students:

        if student.roll_no == roll_no:

            return student

    return None
```

---

## 6. Updating Student Information

```python
def update_student(self, roll_no, marks):

    student = self.search_student(roll_no)

    if student:

        student.marks = marks

        print("Student updated.")

    else:

        print("Student not found.")
```

---

## 7. Deleting a Student

```python
def delete_student(self, roll_no):

    student = self.search_student(roll_no)

    if student:

        self.students.remove(student)

        print("Student deleted.")

    else:

        print("Student not found.")
```

## 8. Complete Project

```python
class Student:

    def __init__(self, roll_no, name, age, marks):

        self.roll_no = roll_no
        self.name = name
        self.age = age
        self.marks = marks

    def display(self):

        print(f"Roll No : {self.roll_no}")
        print(f"Name    : {self.name}")
        print(f"Age     : {self.age}")
        print(f"Marks   : {self.marks}")


class StudentManagementSystem:

    def __init__(self):

        self.students = []

    def add_student(self, student):

        self.students.append(student)

        print("Student added successfully.")

    def view_students(self):

        if not self.students:

            print("No students available.")
            return

        for student in self.students:

            student.display()
            print("-" * 30)

    def search_student(self, roll_no):

        for student in self.students:

            if student.roll_no == roll_no:
                return student

        return None

    def update_student(self, roll_no, marks):

        student = self.search_student(roll_no)

        if student:

            student.marks = marks

            print("Student updated.")

        else:

            print("Student not found.")

    def delete_student(self, roll_no):

        student = self.search_student(roll_no)

        if student:

            self.students.remove(student)

            print("Student deleted.")

        else:

            print("Student not found.")


system = StudentManagementSystem()

student1 = Student(101, "Alice", 20, 92)
student2 = Student(102, "Bob", 21, 85)

system.add_student(student1)
system.add_student(student2)

print("\nAll Students")

system.view_students()

print("\nUpdating Student")

system.update_student(102, 90)

print("\nSearching Student")

student = system.search_student(102)

if student:
    student.display()

print("\nDeleting Student")

system.delete_student(101)

print("\nRemaining Students")

system.view_students()
```

## 12. Common Mistakes

Avoid these common mistakes while building a student management system.

### Using Dictionaries Instead of Objects

While dictionaries work, using classes provides better organization and demonstrates OOP concepts.

---

### Not Checking for Duplicate Roll Numbers

Always validate that each roll number is unique before adding a new student.

---

### Ignoring Input Validation

Ensure that values such as age and marks are within valid ranges.

Example:

```python
if marks < 0 or marks > 100:
    raise ValueError("Marks must be between 0 and 100.")
```

---

### Mixing User Interface with Business Logic

Keep the menu-driven interface separate from the classes that manage student data.
