## 1. Designing the Book Class

The `Book` class stores information about each book.

```python
class Book:

    def __init__(self, book_id, title, author):

        self.book_id = book_id
        self.title = title
        self.author = author
        self.is_issued = False

    def display(self):

        status = "Issued" if self.is_issued else "Available"

        print(f"Book ID : {self.book_id}")
        print(f"Title   : {self.title}")
        print(f"Author  : {self.author}")
        print(f"Status  : {status}")
```

## 2. Creating the Library Management Class

This class manages all library books.

```python
class LibraryManagementSystem:

    def __init__(self):

        self.books = []
```

The `books` list stores all book objects.

## 3. Adding a Book

```python
def add_book(self, book):

    self.books.append(book)

    print("Book added successfully.")
```

---

## 4. Viewing All Books

```python
def view_books(self):

    if not self.books:

        print("No books available.")
        return

    for book in self.books:

        book.display()

        print("-" * 35)
```

---

## 5. Searching for a Book

```python
def search_book(self, book_id):

    for book in self.books:

        if book.book_id == book_id:

            return book

    return None
```

---

## 6. Issuing a Book

```python
def issue_book(self, book_id):

    book = self.search_book(book_id)

    if book:

        if book.is_issued:

            print("Book is already issued.")

        else:

            book.is_issued = True

            print("Book issued successfully.")

    else:

        print("Book not found.")
```

---

## 7. Returning a Book

```python
def return_book(self, book_id):

    book = self.search_book(book_id)

    if book:

        if not book.is_issued:

            print("Book was not issued.")

        else:

            book.is_issued = False

            print("Book returned successfully.")

    else:

        print("Book not found.")
```

---

## 8. Removing a Book

```python
def remove_book(self, book_id):

    book = self.search_book(book_id)

    if book:

        self.books.remove(book)

        print("Book removed successfully.")

    else:

        print("Book not found.")
```

## 9. Complete Project

```python
class Book:

    def __init__(self, book_id, title, author):

        self.book_id = book_id
        self.title = title
        self.author = author
        self.is_issued = False

    def display(self):

        status = "Issued" if self.is_issued else "Available"

        print(f"Book ID : {self.book_id}")
        print(f"Title   : {self.title}")
        print(f"Author  : {self.author}")
        print(f"Status  : {status}")


class LibraryManagementSystem:

    def __init__(self):

        self.books = []

    def add_book(self, book):

        self.books.append(book)

        print("Book added successfully.")

    def view_books(self):

        if not self.books:

            print("No books available.")
            return

        for book in self.books:

            book.display()

            print("-" * 35)

    def search_book(self, book_id):

        for book in self.books:

            if book.book_id == book_id:
                return book

        return None

    def issue_book(self, book_id):

        book = self.search_book(book_id)

        if book:

            if book.is_issued:

                print("Book is already issued.")

            else:

                book.is_issued = True

                print("Book issued successfully.")

        else:

            print("Book not found.")

    def return_book(self, book_id):

        book = self.search_book(book_id)

        if book:

            if not book.is_issued:

                print("Book was not issued.")

            else:

                book.is_issued = False

                print("Book returned successfully.")

        else:

            print("Book not found.")

    def remove_book(self, book_id):

        book = self.search_book(book_id)

        if book:

            self.books.remove(book)

            print("Book removed successfully.")

        else:

            print("Book not found.")


library = LibraryManagementSystem()

book1 = Book(101, "Python Basics", "John Smith")
book2 = Book(102, "Data Structures", "Alice Brown")

library.add_book(book1)
library.add_book(book2)

print("\nLibrary Books")

library.view_books()

print("\nIssuing Book")

library.issue_book(101)

print("\nLibrary Status")

library.view_books()

print("\nReturning Book")

library.return_book(101)

print("\nRemoving Book")

library.remove_book(102)

print("\nRemaining Books")

library.view_books()
```

## 13. Common Mistakes

Avoid these common mistakes while building a library management system.

### Allowing Duplicate Book IDs

Always ensure that every book has a unique ID before adding it to the library.

---

### Issuing an Already Issued Book

Check the book's availability before issuing it.

Example:

```python
if book.is_issued:
    print("Book is already issued.")
```

---

Validate the issue status before processing a return.

Example:

```python
if not book.is_issued:
    print("Book was not issued.")
```

---

### Mixing User Interface with Business Logic

Keep menu-driven input/output separate from the classes that manage books.
