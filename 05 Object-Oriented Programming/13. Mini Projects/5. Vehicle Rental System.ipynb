## 1. Designing the Vehicle Class

The `Vehicle` class stores information about each vehicle.

```python
class Vehicle:

    def __init__(self, vehicle_id, brand, model, rent_per_day):

        self.vehicle_id = vehicle_id
        self.brand = brand
        self.model = model
        self.rent_per_day = rent_per_day
        self.is_rented = False

    def display(self):

        status = "Rented" if self.is_rented else "Available"

        print(f"Vehicle ID  : {self.vehicle_id}")
        print(f"Brand       : {self.brand}")
        print(f"Model       : {self.model}")
        print(f"Rent/Day    : ₹{self.rent_per_day}")
        print(f"Status      : {status}")
```

---

## 2. Creating the Vehicle Rental Management Class

This class manages all rental vehicles.

```python
class VehicleRentalSystem:

    def __init__(self):

        self.vehicles = []
```

The `vehicles` list stores all vehicle objects.

---

## 3. Adding a Vehicle

```python
def add_vehicle(self, vehicle):

    self.vehicles.append(vehicle)

    print("Vehicle added successfully.")
```

---

## 4. Viewing All Vehicles

```python
def view_vehicles(self):

    if not self.vehicles:

        print("No vehicles available.")
        return

    for vehicle in self.vehicles:

        vehicle.display()

        print("-" * 40)
```

---

## 5. Searching for a Vehicle

```python
def search_vehicle(self, vehicle_id):

    for vehicle in self.vehicles:

        if vehicle.vehicle_id == vehicle_id:

            return vehicle

    return None
```

---

## 6. Renting a Vehicle

```python
def rent_vehicle(self, vehicle_id):

    vehicle = self.search_vehicle(vehicle_id)

    if vehicle:

        if vehicle.is_rented:

            print("Vehicle is already rented.")

        else:

            vehicle.is_rented = True

            print("Vehicle rented successfully.")

    else:

        print("Vehicle not found.")
```

---

## 7. Returning a Vehicle

```python
def return_vehicle(self, vehicle_id):

    vehicle = self.search_vehicle(vehicle_id)

    if vehicle:

        if not vehicle.is_rented:

            print("Vehicle was not rented.")

        else:

            vehicle.is_rented = False

            print("Vehicle returned successfully.")

    else:

        print("Vehicle not found.")
```

---

## 8. Removing a Vehicle

```python
def remove_vehicle(self, vehicle_id):

    vehicle = self.search_vehicle(vehicle_id)

    if vehicle:

        self.vehicles.remove(vehicle)

        print("Vehicle removed successfully.")

    else:

        print("Vehicle not found.")
```

## 9. Complete Project

```python
class Vehicle:

    def __init__(self, vehicle_id, brand, model, rent_per_day):

        self.vehicle_id = vehicle_id
        self.brand = brand
        self.model = model
        self.rent_per_day = rent_per_day
        self.is_rented = False

    def display(self):

        status = "Rented" if self.is_rented else "Available"

        print(f"Vehicle ID  : {self.vehicle_id}")
        print(f"Brand       : {self.brand}")
        print(f"Model       : {self.model}")
        print(f"Rent/Day    : ₹{self.rent_per_day}")
        print(f"Status      : {status}")


class VehicleRentalSystem:

    def __init__(self):

        self.vehicles = []

    def add_vehicle(self, vehicle):

        self.vehicles.append(vehicle)

        print("Vehicle added successfully.")

    def view_vehicles(self):

        if not self.vehicles:

            print("No vehicles available.")
            return

        for vehicle in self.vehicles:

            vehicle.display()

            print("-" * 40)

    def search_vehicle(self, vehicle_id):

        for vehicle in self.vehicles:

            if vehicle.vehicle_id == vehicle_id:
                return vehicle

        return None

    def rent_vehicle(self, vehicle_id):

        vehicle = self.search_vehicle(vehicle_id)

        if vehicle:

            if vehicle.is_rented:

                print("Vehicle is already rented.")

            else:

                vehicle.is_rented = True

                print("Vehicle rented successfully.")

        else:

            print("Vehicle not found.")

    def return_vehicle(self, vehicle_id):

        vehicle = self.search_vehicle(vehicle_id)

        if vehicle:

            if not vehicle.is_rented:

                print("Vehicle was not rented.")

            else:

                vehicle.is_rented = False

                print("Vehicle returned successfully.")

        else:

            print("Vehicle not found.")

    def remove_vehicle(self, vehicle_id):

        vehicle = self.search_vehicle(vehicle_id)

        if vehicle:

            self.vehicles.remove(vehicle)

            print("Vehicle removed successfully.")

        else:

            print("Vehicle not found.")


system = VehicleRentalSystem()

vehicle1 = Vehicle(101, "Toyota", "Corolla", 2500)
vehicle2 = Vehicle(102, "Honda", "City", 2200)

system.add_vehicle(vehicle1)
system.add_vehicle(vehicle2)

print("\nAvailable Vehicles")

system.view_vehicles()

print("\nRenting Vehicle")

system.rent_vehicle(101)

print("\nVehicle Status")

system.view_vehicles()

print("\nReturning Vehicle")

system.return_vehicle(101)

print("\nRemoving Vehicle")

system.remove_vehicle(102)

print("\nRemaining Vehicles")

system.view_vehicles()
```

## 13. Common Mistakes

Avoid these common mistakes while building a vehicle rental system.

### Allowing Duplicate Vehicle IDs

Always ensure that each vehicle has a unique ID before adding it to the system.

---

### Renting an Already Rented Vehicle

Always verify the rental status before renting a vehicle.

Example:

```python
if vehicle.is_rented:
    print("Vehicle is already rented.")
```

---

### Returning a Vehicle That Was Never Rented

Validate the rental status before processing a return.

Example:

```python
if not vehicle.is_rented:
    print("Vehicle was not rented.")
```

---

### Allowing Invalid Rental Charges

Ensure that the rent per day is greater than zero.

Example:

```python
if rent_per_day <= 0:
    raise ValueError("Rent per day must be greater than zero.")
```

---

### Mixing User Interface with Business Logic

Keep menu-driven input/output separate from the classes responsible for managing vehicle records.
