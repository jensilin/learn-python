# 👨‍👩‍👦 Single Inheritance in Python

**Single inheritance** is the simplest and most commonly used type of inheritance in Python. In single inheritance, **one child class inherits from one parent class**.

The child class automatically gains access to the accessible variables and methods of the parent class while also being able to define its own variables and methods.

Single inheritance promotes **code reuse**, **maintainability**, and **clean object-oriented design**.

# 📖 Topics Covered

## 1. What is Single Inheritance?

**Single inheritance** is a type of inheritance where **one child class inherits from one parent class**.

It establishes an **"is-a"** relationship.

Example:

```python
class Animal:

    def eat(self):
        print("Eating")

class Dog(Animal):
    pass
```

Here:

- `Animal` → Parent class
- `Dog` → Child class

`Dog` inherits everything accessible from `Animal`.

---

## 2. Why Use Single Inheritance?

Single inheritance helps us:

- Reuse existing code.
- Reduce code duplication.
- Improve readability.
- Build organized class hierarchies.
- Extend existing classes.
- Simplify application development.

---

## 3. Syntax

```python
class ParentClass:

    # Parent members

class ChildClass(ParentClass):

    # Child members
```

Example:

```python
class Vehicle:

    pass

class Car(Vehicle):

    pass
```

---

## 4. Parent and Child Relationship

The parent class contains common functionality.

The child class automatically inherits that functionality.

```python
class Person:

    def introduce(self):
        print("I am a Person.")

class Student(Person):
    pass

student = Student()

student.introduce()
```

### Output

```text
I am a Person.
```

---

## 5. Child Class Can Add New Members

The child class can define additional variables and methods.

```python
class Animal:

    def eat(self):
        print("Eating")

class Dog(Animal):

    def bark(self):
        print("Barking")

dog = Dog()

dog.eat()
dog.bark()
```

### Output

```text
Eating
Barking
```

---

## 6. Child Class Can Access Parent Variables

```python
class Person:

    def __init__(self):
        self.name = "Alice"

class Student(Person):
    pass

student = Student()

print(student.name)
```

### Output

```text
Alice
```

---

## 7. Reusing Parent Class Methods

One parent class can provide common functionality for multiple child objects.

```python
class Employee:

    def work(self):
        print("Working")

class Developer(Employee):
    pass

developer = Developer()

developer.work()
```

### Output

```text
Working
```

## 9. Class Hierarchy

A simple single inheritance hierarchy looks like this:

```text
        Animal
           │
           │
         Dog
```

Another example:

```text
       Vehicle
           │
           │
          Car
```

Each child has exactly **one parent**.

## 10. Practical Examples

### Example 1: Animal and Dog

```python
class Animal:

    def eat(self):
        print("Animal is eating.")

class Dog(Animal):
    pass

dog = Dog()

dog.eat()
```

### Output

### Example 2: Vehicle and Car

```python
class Vehicle:

    def start(self):
        print("Vehicle Started")

class Car(Vehicle):
    pass

car = Car()

car.start()
```

### Output

### Example 3: Person and Student

```python
class Person:

    def introduce(self):
        print("Hello!")

class Student(Person):

    def study(self):
        print("Studying")

student = Student()

student.introduce()
student.study()
```

### Example 4: Employee and Manager

```python
class Employee:

    def work(self):
        print("Employee Working")

class Manager(Employee):

    def manage(self):
        print("Managing Team")

manager = Manager()

manager.work()
manager.manage()
```

### Output

```text
Employee Working
Managing Team
```

### Example 5: Shape and Circle

```python
class Shape:

    def draw(self):
        print("Drawing Shape")

class Circle(Shape):

    def area(self):
        print("Calculating Area")

circle = Circle()

circle.draw()
circle.area()
```

## 12. Common Mistakes

Avoid these common mistakes while working with single inheritance.

### Forgetting to Inherit

Incorrect

```python
class Dog:

    pass
```

Correct

```python
class Dog(Animal):

    pass
```

---

### Duplicating Parent Methods

Incorrect

```python
class Dog:

    def eat(self):
        print("Eating")
```

Instead, inherit the method from the parent whenever possible.

---

### Creating Incorrect Relationships

Inheritance should represent an **"is-a"** relationship.

Correct

```text
Dog → Animal
```

Incorrect

```text
Car → Engine
```

A car **has an** engine; it is not an engine.

---

### Accessing Parent Private Members Directly

Private members (`__member`) are not directly accessible in child classes due to name mangling.

---

## 13. Best Practices

Follow these recommended practices when using single inheritance.

- Use inheritance only when an **"is-a"** relationship exists.
- Keep parent classes generic and reusable.
- Extend functionality rather than duplicate it.
- Avoid placing unrelated features in the parent class.
- Use meaningful class names.
- Design simple inheritance hierarchies.
- Favor composition over inheritance when appropriate.
