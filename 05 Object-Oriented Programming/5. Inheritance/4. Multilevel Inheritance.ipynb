# 🏛️ Multilevel Inheritance in Python

**Multilevel inheritance** is a type of inheritance where **a class inherits from another class, which itself inherits from another class**. This creates a chain of inheritance where each child class becomes the parent of the next class.

It allows classes to inherit properties and methods through multiple levels, promoting **code reuse**, **logical hierarchy**, and **incremental extension** of functionality.

# 📖 Topics Covered

## 1. What is Multilevel Inheritance?

**Multilevel inheritance** is an inheritance structure where a child class inherits from another child class.

Example:

```python
class A:
    pass

class B(A):
    pass

class C(B):
    pass
```

Here:

- `A` → Grandparent class
- `B` → Parent class
- `C` → Child class

The class `C` inherits members from both `A` and `B`.

---

## 2. Why Use Multilevel Inheritance?

Multilevel inheritance helps us:

- Reuse code across multiple levels.
- Build logical class hierarchies.
- Extend functionality gradually.
- Reduce code duplication.
- Improve maintainability.
- Support scalable application design.

---

## 3. Syntax

```python
class GrandParent:

    # Grandparent members

class Parent(GrandParent):

    # Parent members

class Child(Parent):

    # Child members
```

Each class inherits from the class immediately above it.

---

## 4. Inheritance Chain

Inheritance flows from top to bottom.

```text
GrandParent
      │
      ▼
   Parent
      │
      ▼
    Child
```

The child class can access members inherited from every level.

---

## 5. Accessing Grandparent Members

A child class automatically inherits accessible members from all ancestor classes.

```python
class Animal:

    def eat(self):
        print("Eating")

class Mammal(Animal):
    pass

class Dog(Mammal):
    pass

dog = Dog()

dog.eat()
```

### Output

```text
Eating
```

Although `Dog` directly inherits from `Mammal`, it can also access members inherited from `Animal`.

---

## 6. Each Class Can Add New Features

Every level can introduce additional variables and methods.

```python
class Animal:

    def eat(self):
        print("Eating")

class Mammal(Animal):

    def walk(self):
        print("Walking")

class Dog(Mammal):

    def bark(self):
        print("Barking")

dog = Dog()

dog.eat()
dog.walk()
dog.bark()
```

### Output

```text
Eating
Walking
Barking
```

---

## 7. Reusing Code Across Multiple Levels

Each class builds upon the previous one.

```python
class Person:

    def introduce(self):
        print("I am a Person.")

class Employee(Person):

    def work(self):
        print("Working")

class Manager(Employee):

    def manage(self):
        print("Managing Team")

manager = Manager()

manager.introduce()
manager.work()
manager.manage()
```

### Output

```text
I am a Person.
Working
Managing Team
```

## 10. Practical Examples

### Example 1: Animal Hierarchy

```python
class Animal:

    def eat(self):
        print("Animal is eating.")

class Mammal(Animal):

    def walk(self):
        print("Mammal is walking.")

class Dog(Mammal):

    def bark(self):
        print("Dog is barking.")

dog = Dog()

dog.eat()
dog.walk()
dog.bark()
```

### Output

```text
Animal is eating.
Mammal is walking.
Dog is barking.
```

---

### Example 2: Educational System

```python
class Person:

    def introduce(self):
        print("I am a Person.")

class Student(Person):

    def study(self):
        print("Studying")

class CollegeStudent(Student):

    def attend_class(self):
        print("Attending Class")

student = CollegeStudent()

student.introduce()
student.study()
student.attend_class()
```

### Output

```text
I am a Person.
Studying
Attending Class
```

---

### Example 3: Employee Hierarchy

```python
class Employee:

    def work(self):
        print("Working")

class TeamLead(Employee):

    def guide(self):
        print("Guiding Team")

class Manager(TeamLead):

    def manage(self):
        print("Managing Project")

manager = Manager()

manager.work()
manager.guide()
manager.manage()
```

### Output

```text
Working
Guiding Team
Managing Project
```

---

### Example 4: Vehicle Hierarchy

```python
class Vehicle:

    def start(self):
        print("Vehicle Started")

class Car(Vehicle):

    def drive(self):
        print("Driving")

class SportsCar(Car):

    def turbo(self):
        print("Turbo Mode Enabled")

car = SportsCar()

car.start()
car.drive()
car.turbo()
```

### Output

```text
Vehicle Started
Driving
Turbo Mode Enabled
```

---

### Example 5: Device Hierarchy

```python
class Device:

    def power_on(self):
        print("Power On")

class Computer(Device):

    def boot(self):
        print("Booting System")

class Laptop(Computer):

    def charge(self):
        print("Charging Battery")

laptop = Laptop()

laptop.power_on()
laptop.boot()
laptop.charge()
```

### Output

```text
Power On
Booting System
Charging Battery
```

## 12. Common Mistakes

Avoid these common mistakes while working with multilevel inheritance.

### Creating Very Deep Inheritance Chains

Avoid creating too many inheritance levels.

Example:

```text
A
│
B
│
C
│
D
│
E
│
F
```

Deep hierarchies can make code difficult to understand and maintain.

---

### Duplicating Parent Functionality

Do not rewrite methods that are already available through inheritance unless customization is required.

---

### Using Inheritance Without an "Is-a" Relationship

Correct:

```text
Dog → Mammal → Animal
```

Incorrect:

```text
Car → Engine
```

A car **has an** engine; it is not an engine.

---

### Forgetting Private Members

Private members (`__member`) of ancestor classes are not directly accessible because of name mangling.

---

## 13. Best Practices

Follow these recommended practices when using multilevel inheritance.

- Build logical inheritance hierarchies.
- Keep inheritance chains reasonably short.
- Reuse existing functionality instead of duplicating code.
- Use meaningful class names.
- Override methods only when behavior needs to change.
- Prefer composition if inheritance becomes overly complex.
- Document inheritance relationships clearly.
