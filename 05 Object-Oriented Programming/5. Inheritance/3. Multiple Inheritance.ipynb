# 👨‍👩‍👧‍👦 Multiple Inheritance in Python

**Multiple inheritance** is a type of inheritance in which **one child class inherits from two or more parent classes**.

The child class gains access to the accessible variables and methods of **all its parent classes**, allowing it to combine functionality from multiple sources.

Python fully supports multiple inheritance and provides the **Method Resolution Order (MRO)** mechanism to determine which parent's method should be executed when multiple parents define methods with the same name.

# 📖 Topics Covered

## 1. What is Multiple Inheritance?

**Multiple inheritance** allows a child class to inherit from **more than one parent class**.

Example:

```python
class Father:
    pass

class Mother:
    pass

class Child(Father, Mother):
    pass
```

Here,

- `Father` → Parent class
- `Mother` → Parent class
- `Child` → Child class

The `Child` class inherits from both parents.

---

Multiple inheritance helps us:

- Combine features from multiple classes.
- Reuse code efficiently.
- Avoid duplicate implementations.
- Create flexible class hierarchies.
- Build modular applications.

## 4. Accessing Parent Members

The child class can access methods from every parent.

```python
class Camera:

    def click(self):
        print("Taking Photo")

class Phone:

    def call(self):
        print("Calling")

class SmartPhone(Camera, Phone):
    pass

phone = SmartPhone()

phone.click()
phone.call()
```

### Output

```text
Taking Photo
Calling
```

---

## 5. Child Class Can Add Its Own Members

The child class can also define new methods.

```python
class Printer:

    def print_document(self):
        print("Printing")

class Scanner:

    def scan_document(self):
        print("Scanning")

class AllInOne(Printer, Scanner):

    def fax(self):
        print("Sending Fax")

device = AllInOne()

device.print_document()
device.scan_document()
device.fax()
```

### Output

```text
Printing
Scanning
Sending Fax
```

---

## 6. Method Resolution Order (MRO)

When two parent classes contain a method with the same name, Python follows the **Method Resolution Order (MRO)** to determine which method to execute.

```python
class Father:

    def show(self):
        print("Father")

class Mother:

    def show(self):
        print("Mother")

class Child(Father, Mother):
    pass

child = Child()

child.show()
```

### Output

```text
Father
```

Python searches the parent classes from **left to right** according to the MRO.

---

## 7. Viewing the MRO

The MRO can be viewed using the `mro()` method.

```python
class Father:
    pass

class Mother:
    pass

class Child(Father, Mother):
    pass

print(Child.mro())
```

### Sample Output

```text
[<class '__main__.Child'>,
 <class '__main__.Father'>,
 <class '__main__.Mother'>,
 <class 'object'>]
```

This shows the order Python follows when searching for attributes and methods.

## 12. Common Mistakes

Avoid these common mistakes while working with multiple inheritance.

### Ignoring Method Resolution Order

If multiple parent classes define the same method, always understand which implementation will be called.

```python
class A:

    def show(self):
        print("A")

class B:

    def show(self):
        print("B")

class C(A, B):
    pass
```

`C().show()` calls `A.show()` because `A` appears first.

---

### Creating Unnecessary Multiple Inheritance

Do not inherit from multiple classes unless the child truly requires functionality from all of them.

---

### Forgetting About Name Conflicts

Two parent classes may contain methods or variables with identical names, which can lead to unexpected behavior if the MRO is not understood.

---

### Building Complex Class Hierarchies

Large multiple inheritance hierarchies can become difficult to understand and maintain.

Keep designs as simple as possible.

---

## 13. Best Practices

Follow these recommended practices when using multiple inheritance.

- Use multiple inheritance only when it models the problem naturally.
- Understand Python's Method Resolution Order (MRO).
- Avoid unnecessary parent classes.
- Keep parent classes focused on a single responsibility.
- Consider mixins for reusable functionality.
- Document inheritance hierarchies clearly.
- Prefer composition over multiple inheritance when appropriate.
