# 🌳 Hierarchical Inheritance in Python

**Hierarchical inheritance** is a type of inheritance where **multiple child classes inherit from a single parent class**.

The parent class contains common variables and methods that are shared among all child classes. Each child class can also define its own unique features while reusing the functionality provided by the parent class.

Hierarchical inheritance promotes **code reuse**, **maintainability**, and **logical organization** in object-oriented programming.

# 📖 Topics Covered

## 1. What is Hierarchical Inheritance?

**Hierarchical inheritance** is an inheritance structure where **multiple child classes inherit from one parent class**.

Example:

```python
class Parent:
    pass

class Child1(Parent):
    pass

class Child2(Parent):
    pass
```

Here,

- `Parent` → Parent class
- `Child1` → Child class
- `Child2` → Child class

Both child classes inherit from the same parent class.

---

## 2. Why Use Hierarchical Inheritance?

Hierarchical inheritance helps us:

- Reuse common code.
- Reduce duplication.
- Keep related classes organized.
- Improve maintainability.
- Extend parent functionality in different ways.
- Build scalable applications.

## 3. Syntax

---

## 4. Parent and Child Relationship

The parent class contains common functionality.

Each child class inherits those members independently.

```python
class Person:

    def introduce(self):
        print("I am a Person.")

class Student(Person):
    pass

class Teacher(Person):
    pass

student = Student()
teacher = Teacher()

student.introduce()
teacher.introduce()
```

### Output

```text
I am a Person.
I am a Person.
```

---

## 5. Each Child Can Have Its Own Members

Each child class can define additional variables and methods.

```python
class Animal:

    def eat(self):
        print("Eating")

class Dog(Animal):

    def bark(self):
        print("Barking")

class Cat(Animal):

    def meow(self):
        print("Meowing")

dog = Dog()
cat = Cat()

dog.eat()
dog.bark()

cat.eat()
cat.meow()
```

### Output

```text
Eating
Barking
Eating
Meowing
```

---

## 6. Reusing Parent Class Methods

A single parent class can provide common functionality for multiple child classes.

```python
class Vehicle:

    def start(self):
        print("Vehicle Started")

class Car(Vehicle):
    pass

class Bike(Vehicle):
    pass

car = Car()
bike = Bike()

car.start()
bike.start()
```

### Output

```text
Vehicle Started
Vehicle Started
```

## 9. Practical Examples

### Example 1: Animal Hierarchy

```python
class Animal:

    def eat(self):
        print("Animal is eating.")

class Dog(Animal):

    def bark(self):
        print("Dog is barking.")

class Cat(Animal):

    def meow(self):
        print("Cat is meowing.")

dog = Dog()
cat = Cat()

dog.eat()
dog.bark()

cat.eat()
cat.meow()
```

### Output

```text
Animal is eating.
Dog is barking.
Animal is eating.
Cat is meowing.
```

---

### Example 2: Employee Management

```python
class Employee:

    def work(self):
        print("Working")

class Manager(Employee):

    def manage(self):
        print("Managing Team")

class Developer(Employee):

    def code(self):
        print("Writing Code")

manager = Manager()
developer = Developer()

manager.work()
manager.manage()

developer.work()
developer.code()
```

### Output

```text
Working
Managing Team
Working
Writing Code
```

---

### Example 3: Educational System

```python
class Person:

    def introduce(self):
        print("Hello!")

class Student(Person):

    def study(self):
        print("Studying")

class Teacher(Person):

    def teach(self):
        print("Teaching")

student = Student()
teacher = Teacher()

student.introduce()
student.study()

teacher.introduce()
teacher.teach()
```

### Output

```text
Hello!
Studying
Hello!
Teaching
```

---

### Example 4: Vehicle System

```python
class Vehicle:

    def fuel(self):
        print("Fuel Required")

class Car(Vehicle):

    def drive(self):
        print("Driving Car")

class Bike(Vehicle):

    def ride(self):
        print("Riding Bike")

car = Car()
bike = Bike()

car.fuel()
car.drive()

bike.fuel()
bike.ride()
```

### Output

```text
Fuel Required
Driving Car
Fuel Required
Riding Bike
```

---

### Example 5: Banking System

```python
class Account:

    def login(self):
        print("Login Successful")

class SavingsAccount(Account):

    def interest(self):
        print("Interest Added")

class CurrentAccount(Account):

    def overdraft(self):
        print("Overdraft Available")

savings = SavingsAccount()
current = CurrentAccount()

savings.login()
savings.interest()

current.login()
current.overdraft()
```

### Output

```text
Login Successful
Interest Added
Login Successful
Overdraft Available
```

## 11. Common Mistakes

Avoid these common mistakes while working with hierarchical inheritance.

### Duplicating Parent Code

Incorrect

```python
class Dog:

    def eat(self):
        print("Eating")
```

If the functionality already exists in the parent class, inherit it instead of rewriting it.

---

### Using Inheritance Without an "Is-a" Relationship

Correct

```text
Dog → Animal
Cat → Animal
```

Incorrect

```text
Car → Engine
Bike → Engine
```

A car and a bike **have an** engine; they are not engines.

---

### Adding Unrelated Features to the Parent Class

The parent class should only contain functionality that is common to all child classes.

Avoid placing child-specific methods in the parent class.

---

### Misunderstanding Private Members

Private members (`__member`) are not directly inherited due to name mangling.

---

## 12. Best Practices

Follow these recommended practices when using hierarchical inheritance.

- Place only shared functionality in the parent class.
- Keep child classes focused on their unique responsibilities.
- Avoid duplicate code across child classes.
- Use meaningful class names.
- Keep inheritance hierarchies simple.
- Override parent methods only when behavior needs to change.
- Prefer composition if inheritance is not the best fit.
