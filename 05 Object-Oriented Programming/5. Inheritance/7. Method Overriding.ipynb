# 🔄 Method Overriding in Python

**Method overriding** is an Object-Oriented Programming (OOP) concept in which a **child class provides its own implementation of a method that already exists in its parent class**.

When a child class overrides a method, the **child class version is executed instead of the parent class version** whenever that method is called using a child object.

Method overriding allows child classes to **customize or extend inherited behavior** without modifying the parent class.

# 📖 Topics Covered

## 1. What is Method Overriding?

**Method overriding** occurs when a child class defines a method with the **same name and parameters** as a method in its parent class.

The child class method replaces the inherited version for objects of the child class.

Example:

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        print("Dog Bark")
```

Here, `Dog` overrides the `sound()` method of `Animal`.

---

## 2. Why Use Method Overriding?

Method overriding helps us:

- Customize inherited behavior.
- Support polymorphism.
- Extend existing functionality.
- Improve flexibility.
- Build reusable class hierarchies.
- Avoid modifying parent classes.

---

## 3. Syntax

```python
class Parent:

    def method(self):
        # Parent implementation

class Child(Parent):

    def method(self):
        # Child implementation
```

The child method must have the same name as the parent method.

---

## 4. How Method Overriding Works

When a method is called using a child object, Python first checks the child class.

If the method exists there, Python executes it instead of the parent's version.

```python
class Vehicle:

    def start(self):
        print("Vehicle Started")

class Car(Vehicle):

    def start(self):
        print("Car Started")

car = Car()

car.start()
```

### Output

```text
Car Started
```

---

## 5. Parent Method is Hidden

After overriding, the parent method is not called automatically.

```python
class Person:

    def introduce(self):
        print("I am a Person.")

class Student(Person):

    def introduce(self):
        print("I am a Student.")

student = Student()

student.introduce()
```

### Output

```text
I am a Student.
```

---

## 6. Calling the Parent Method

The child class can call the parent method using `super()`.

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        super().sound()
        print("Dog Bark")

dog = Dog()

dog.sound()
```

### Output

```text
Animal Sound
Dog Bark
```

The `super()` function will be discussed in detail in the next notebook.

---

## 7. Rules for Method Overriding

For proper overriding:

- The child class must inherit from the parent class.
- The method name should be the same.
- The method should accept compatible parameters.
- The child method replaces the parent implementation for child objects.

---

## 8. Method Overriding vs Method Overloading

| Feature | Method Overriding | Method Overloading |
|---------|-------------------|--------------------|
| Occurs In | Child class | Same class |
| Purpose | Replace parent behavior | Multiple ways to call a method |
| Inheritance Required | Yes | No |
| Python Support | Yes | Uses default arguments, `*args`, or `**kwargs` instead of traditional overloading |

## 11. Common Mistakes

Avoid these common mistakes while working with method overriding.

### Forgetting to Inherit

Incorrect

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog:

    def sound(self):
        print("Dog Bark")
```

This is **not** method overriding because `Dog` does not inherit from `Animal`.

---

### Changing the Method Name

Incorrect

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def bark(self):
        print("Dog Bark")
```

Since the method names differ, `sound()` is not overridden.

---

### Forgetting to Call the Parent Method

Sometimes the parent implementation performs important work.

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        super().sound()
        print("Dog Bark")
```

Use `super()` when both parent and child behavior are needed.

---

### Confusing Overriding with Overloading

Method overriding replaces a parent method.

Method overloading provides multiple ways to call a method.

They are different concepts.

---

## 12. Best Practices

Follow these recommended practices when using method overriding.

- Override methods only when behavior needs to change.
- Keep overridden methods compatible with the parent method.
- Use `super()` when parent behavior should be preserved.
- Avoid duplicating parent code unnecessarily.
- Use meaningful method names.
- Document overridden behavior if it differs significantly.
- Follow the Liskov Substitution Principle by ensuring child classes can be used wherever parent classes are expected.

> **Note:** Method overriding allows a child class to provide its own implementation of a method inherited from a parent class. It is one of the key mechanisms that enables **runtime polymorphism**, allowing different subclasses to respond differently to the same method call. When both the parent and child implementations are needed, the parent method can be invoked using the **`super()`** function.
