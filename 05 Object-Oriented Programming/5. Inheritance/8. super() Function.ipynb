# 🚀 `super()` Function in Python

The **`super()`** function is a built-in Python function that allows a **child class to access the methods and constructors of its parent class**.

It is most commonly used to:

- Call the parent class constructor (`__init__()`).
- Call overridden methods from the parent class.
- Reuse parent class functionality without duplicating code.
- Support multiple inheritance by following Python's **Method Resolution Order (MRO)**.

Using `super()` makes inheritance cleaner, easier to maintain, and more flexible.

# 📖 Topics Covered

## 1. What is `super()`?

`super()` is a built-in function that returns a temporary object representing the parent class.

It allows the child class to call methods and constructors defined in its parent class.

Example:

```python
class Parent:

    def display(self):
        print("Parent Method")

class Child(Parent):

    def show(self):
        super().display()

child = Child()

child.show()
```

### Output

```text
Parent Method
```

---

## 2. Why Use `super()`?

The `super()` function helps us:

- Reuse parent class code.
- Avoid duplicate code.
- Call overridden methods.
- Initialize parent class attributes.
- Improve maintainability.
- Support multiple inheritance through MRO.

## 4. Calling the Parent Constructor

One of the most common uses of `super()` is calling the parent class constructor.

```python
class Person:

    def __init__(self):
        print("Person Constructor")

class Student(Person):

    def __init__(self):
        super().__init__()
        print("Student Constructor")

student = Student()
```

### Output

```text
Person Constructor
Student Constructor
```

The parent constructor executes before the child constructor.

---

## 5. Passing Arguments to the Parent Constructor

```python
class Person:

    def __init__(self, name):
        self.name = name

class Student(Person):

    def __init__(self, name, course):
        super().__init__(name)
        self.course = course

student = Student("Alice", "Python")

print(student.name)
print(student.course)
```

### Output

```text
Alice
Python
```

---

## 6. Calling an Overridden Parent Method

When a child overrides a parent method, `super()` allows both implementations to run.

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        super().sound()
        print("Dog Bark")

dog = Dog()

dog.sound()
```

### Output

```text
Animal Sound
Dog Bark
```

---

## 7. `super()` in Multiple Inheritance

Python uses the **Method Resolution Order (MRO)** when `super()` is used with multiple inheritance.

```python
class A:

    def display(self):
        print("Class A")

class B(A):

    def display(self):
        super().display()
        print("Class B")

class C(B):

    def display(self):
        super().display()
        print("Class C")

obj = C()

obj.display()
```

### Output

```text
Class A
Class B
Class C
```

Each class calls the next class in the MRO.

---

## 8. Viewing the Method Resolution Order

You can view the MRO using the `mro()` method.

```python
class A:
    pass

class B(A):
    pass

class C(B):
    pass

print(C.mro())
```

### Sample Output

```text
[<class '__main__.C'>,
 <class '__main__.B'>,
 <class '__main__.A'>,
 <class 'object'>]
```

This is the order Python follows when resolving methods.

## 9. Practical Examples

### Example 1: Employee Constructor

```python
class Employee:

    def __init__(self):
        print("Employee Created")

class Manager(Employee):

    def __init__(self):
        super().__init__()
        print("Manager Created")

manager = Manager()
```

### Example 2: Vehicle

```python
class Vehicle:

    def start(self):
        print("Vehicle Started")

class Car(Vehicle):

    def start(self):
        super().start()
        print("Car Ready")

car = Car()

car.start()
```

### Output

```text
Vehicle Started
Car Ready
```

### Example 3: Student Information

```python
class Person:

    def __init__(self, name):
        self.name = name

class Student(Person):

    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

student = Student("David", 20)

print(student.name)
print(student.age)
```

## 11. Common Mistakes

Avoid these common mistakes while working with `super()`.

### Forgetting to Call the Parent Constructor

Incorrect

```python
class Student(Person):

    def __init__(self):
        print("Student")
```

If the parent constructor initializes important attributes, omitting `super().__init__()` may leave the object in an incomplete state.

Correct

```python
class Student(Person):

    def __init__(self):
        super().__init__()
        print("Student")
```

---

### Duplicating Parent Code

Incorrect

```python
class Student(Person):

    def __init__(self):
        self.name = "Alice"
```

Instead of rewriting the parent's initialization logic, call it using `super()`.

---

### Forgetting Parent Constructor Arguments

Incorrect

```python
super().__init__()
```

when the parent expects parameters.

Correct

```python
super().__init__(name)
```

---

### Calling the Parent Class Directly

Instead of

```python
Person.__init__(self, name)
```

prefer

```python
super().__init__(name)
```

This automatically follows the Method Resolution Order and works correctly in multiple inheritance.

---

## 12. Best Practices

Follow these recommended practices when using `super()`.

- Use `super()` to call parent constructors.
- Use `super()` when overriding methods and the parent behavior should also execute.
- Prefer `super()` over directly calling the parent class.
- Understand the Method Resolution Order (MRO), especially in multiple inheritance.
- Avoid duplicating initialization code.
- Keep parent classes reusable and generic.
- Use `super()` consistently in inheritance hierarchies.
