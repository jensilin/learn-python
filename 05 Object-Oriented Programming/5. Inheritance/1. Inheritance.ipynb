# 🧬 Inheritance in Python

**Inheritance** is one of the four fundamental principles of **Object-Oriented Programming (OOP)**. It allows a **new class (child class)** to acquire the **properties (variables)** and **behaviors (methods)** of an **existing class (parent class)**.

Inheritance promotes **code reuse**, reduces duplication, and helps create logical relationships between classes.

In Python, inheritance is implemented by specifying the parent class inside parentheses when defining the child class.

# 📖 Topics Covered

## 1. What is Inheritance?

**Inheritance** is the mechanism through which one class acquires the attributes and methods of another class.

The existing class is called the **parent class** (or **base class**).

The new class is called the **child class** (or **derived class**).

Example:

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):
    pass
```

Here:

- `Animal` → Parent class
- `Dog` → Child class

`Dog` automatically inherits the `sound()` method.

---

## 2. Why Use Inheritance?

Inheritance helps us:

- Reuse existing code.
- Reduce code duplication.
- Improve maintainability.
- Build logical class hierarchies.
- Extend existing functionality.
- Support polymorphism.
- Simplify large applications.

---

## 3. Parent Class and Child Class

A parent class contains common functionality.

A child class inherits and can extend or modify that functionality.

```python
class Vehicle:

    def start(self):
        print("Vehicle Started")

class Car(Vehicle):
    pass
```

`Car` automatically gets the `start()` method.

---

## 4. Syntax

### Syntax

```python
class ParentClass:

    # Parent members

class ChildClass(ParentClass):

    # Child members
```

### Explanation

- `ParentClass` contains reusable code.
- `ChildClass` inherits from `ParentClass`.
- The child class can use all accessible members of the parent class.

---

## 5. Accessing Inherited Members

A child object can directly access inherited methods.

```python
class Person:

    def greet(self):
        print("Hello!")

class Student(Person):
    pass

student = Student()

student.greet()
```

### Output

```text
Hello!
```

---

## 6. Child Class Can Have Its Own Members

A child class can define additional variables and methods.

```python
class Animal:

    def eat(self):
        print("Eating")

class Dog(Animal):

    def bark(self):
        print("Barking")

dog = Dog()

dog.eat()
dog.bark()
```

### Output

```text
Eating
Barking
```

---

## 7. Reusing Parent Class Code

Instead of writing the same code multiple times, inheritance allows child classes to reuse parent class functionality.

```python
class Employee:

    def work(self):
        print("Working")

class Manager(Employee):
    pass

class Developer(Employee):
    pass

manager = Manager()
developer = Developer()

manager.work()
developer.work()
```

### Output

```text
Working
Working
```

## 8. Types of Inheritance in Python

Python supports several types of inheritance.

| Type | Description |
|------|-------------|
| Single Inheritance | One parent, one child |
| Multiple Inheritance | Multiple parents, one child |
| Multilevel Inheritance | Grandparent → Parent → Child |
| Hierarchical Inheritance | One parent, multiple children |
| Hybrid Inheritance | Combination of multiple inheritance types |

Each type will be covered in the following notebooks.

### Example 3: Employee and Manager

```python
class Employee:

    def work(self):
        print("Employee Working")

class Manager(Employee):

    def manage(self):
        print("Managing Team")

manager = Manager()

manager.work()
manager.manage()
```

## 12. Common Mistakes

Avoid these common mistakes while working with inheritance.

### Forgetting to Specify the Parent Class

Incorrect

```python
class Dog:

    pass
```

Correct

```python
class Dog(Animal):

    pass
```

---

### Creating Unnecessary Inheritance

Do not inherit from a class unless there is a true **"is-a"** relationship.

For example:

- `Dog` **is an** `Animal` ✅
- `Car` **is not a** `Engine` ❌

---

### Duplicating Parent Code

Incorrect

```python
class Dog:

    def eat(self):
        print("Eating")
```

If the same functionality already exists in the parent class, inherit it instead of rewriting it.

---

### Misunderstanding Private Members

Private members (`__member`) of the parent class are **not directly accessible** in the child class because of name mangling.

---

## 13. Best Practices

Follow these recommended practices when using inheritance.

- Use inheritance only when an **"is-a"** relationship exists.
- Keep parent classes general and reusable.
- Extend functionality instead of duplicating code.
- Avoid deep inheritance hierarchies unless necessary.
- Use meaningful class names.
- Favor composition over inheritance when classes do not naturally fit an inheritance relationship.
- Keep inherited classes simple and maintainable.
