# 🧩 Hybrid Inheritance in Python

**Hybrid inheritance** is a type of inheritance that combines **two or more inheritance types** into a single inheritance hierarchy.

Since it is a combination of inheritance models such as **single**, **multiple**, **multilevel**, and **hierarchical inheritance**, hybrid inheritance provides flexibility for designing complex object-oriented systems.

Python supports hybrid inheritance because it supports **multiple inheritance**. When multiple parent classes contain methods with the same name, Python uses the **Method Resolution Order (MRO)** to determine which method should be executed.

## 4. Hybrid Inheritance Structure

One common hybrid structure is:

```text
          A
         / \
        /   \
       B     C
        \   /
         \ /
          D
```

Here,

- `B` and `C` inherit from `A` (Hierarchical Inheritance).
- `D` inherits from both `B` and `C` (Multiple Inheritance).

Together they form **Hybrid Inheritance**.

---

## 5. Method Resolution Order (MRO)

When multiple parent classes define the same method, Python follows the **Method Resolution Order (MRO)**.

```python
class A:

    def show(self):
        print("A")

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass

d = D()

d.show()
```

### Output

```text
A
```

Python searches for methods according to the MRO.

---

## 6. Viewing the MRO

The MRO can be displayed using the `mro()` method.

```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass

print(D.mro())
```

### Sample Output

```text
[<class '__main__.D'>,
 <class '__main__.B'>,
 <class '__main__.C'>,
 <class '__main__.A'>,
 <class 'object'>]
```

This order determines how Python searches for attributes and methods.

## 7. Diamond Problem

The **Diamond Problem** occurs when a class inherits from two classes that both inherit from the same parent.

```text
          A
         / \
        /   \
       B     C
        \   /
         \ /
          D
```

If all classes define the same method, Python resolves the conflict using the **Method Resolution Order (MRO)**.

Example:

```python
class A:

    def show(self):
        print("A")

class B(A):

    def show(self):
        print("B")

class C(A):

    def show(self):
        print("C")

class D(B, C):
    pass

d = D()

d.show()
```

### Output

```text
B
```

Since `B` appears before `C`, Python executes `B.show()`.

## 9. Practical Examples

### Example 1: Diamond Structure

```python
class A:

    def show(self):
        print("Class A")

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass

d = D()

d.show()
```

### Output

### Example 2: Employee Management

```python
class Person:

    def introduce(self):
        print("I am a Person")

class Employee(Person):

    def work(self):
        print("Working")

class Student(Person):

    def study(self):
        print("Studying")

class Intern(Employee, Student):
    pass

intern = Intern()

intern.introduce()
intern.work()
intern.study()
```

### Output

```text
I am a Person
Working
Studying
```

---

### Example 3: Smart Device

```python
class Device:

    def power_on(self):
        print("Power On")

class Phone(Device):

    def call(self):
        print("Calling")

class Camera(Device):

    def photo(self):
        print("Taking Photo")

class SmartPhone(Phone, Camera):
    pass

phone = SmartPhone()

phone.power_on()
phone.call()
phone.photo()
```

### Output

```text
Power On
Calling
Taking Photo
```

---

### Example 4: University System

```python
class Person:

    def login(self):
        print("Login Successful")

class Teacher(Person):

    def teach(self):
        print("Teaching")

class Researcher(Person):

    def research(self):
        print("Researching")

class Professor(Teacher, Researcher):
    pass

professor = Professor()

professor.login()
professor.teach()
professor.research()
```

### Output

```text
Login Successful
Teaching
Researching
```

---

### Example 5: Multimedia Device

```python
class Device:

    def start(self):
        print("Starting Device")

class AudioPlayer(Device):

    def play_audio(self):
        print("Playing Audio")

class VideoPlayer(Device):

    def play_video(self):
        print("Playing Video")

class MediaCenter(AudioPlayer, VideoPlayer):
    pass

media = MediaCenter()

media.start()
media.play_audio()
media.play_video()
```

### Output

```text
Starting Device
Playing Audio
Playing Video
```

## 11. Common Mistakes

Avoid these common mistakes while working with hybrid inheritance.

### Ignoring Method Resolution Order

Always understand the MRO when multiple parent classes define the same method.

---

### Building Complex Inheritance Trees

Very large hybrid hierarchies become difficult to understand and maintain.

Keep the design as simple as possible.

---

### Creating Unnecessary Multiple Inheritance

Use multiple inheritance only when the child class genuinely requires functionality from multiple parent classes.

---

### Forgetting About the Diamond Problem

If multiple inheritance forms a diamond-shaped hierarchy, understand how Python resolves method calls using the MRO.

---

## 12. Best Practices

Follow these recommended practices when using hybrid inheritance.

- Use hybrid inheritance only when it naturally models the problem.
- Understand Python's Method Resolution Order (MRO).
- Keep inheritance hierarchies simple.
- Design parent classes with a single responsibility.
- Prefer composition if inheritance becomes overly complex.
- Use meaningful class names.
- Document inheritance relationships clearly.
