# 🔍 `isinstance()` and `issubclass()` in Python

Python provides two useful built-in functions for working with **classes**, **objects**, and **inheritance**:

- **`isinstance()`** – Checks whether an object is an instance of a specified class (or its subclasses).
- **`issubclass()`** – Checks whether one class is derived from another class.

These functions are widely used in **Object-Oriented Programming (OOP)** to perform type checking, validate objects, and work safely with inheritance hierarchies.

# 📖 Topics Covered

## 1. What is `isinstance()`?

The `isinstance()` function checks whether an **object** belongs to a specified class.

It also returns `True` if the object belongs to any subclass of that class.

### Syntax

```python
isinstance(object, class)
```

or

```python
isinstance(object, (class1, class2, class3))
```

### Return Value

- `True` → Object belongs to the specified class (or subclass).
- `False` → Otherwise.

---

## 2. Example of `isinstance()`

```python
class Student:
    pass

student = Student()

print(isinstance(student, Student))
```

### Output

```text
True
```

---

## 3. `isinstance()` with Inheritance

```python
class Animal:
    pass

class Dog(Animal):
    pass

dog = Dog()

print(isinstance(dog, Dog))
print(isinstance(dog, Animal))
```

### Output

```text
True
True
```

Although `dog` is created from `Dog`, it is also considered an instance of `Animal` because `Dog` inherits from `Animal`.

---

## 4. Checking Multiple Classes

You can check multiple classes using a tuple.

```python
print(isinstance(5, (int, float)))
```

### Output

```text
True
```

Another example:

```python
print(isinstance("Python", (list, str)))
```

### Output

```text
True
```

---

## 5. What is `issubclass()`?

The `issubclass()` function checks whether one **class** inherits from another class.

### Syntax

```python
issubclass(child_class, parent_class)
```

or

```python
issubclass(child_class, (parent1, parent2))
```

### Return Value

- `True` → The class inherits from the specified parent.
- `False` → Otherwise.

---

## 6. Example of `issubclass()`

```python
class Animal:
    pass

class Dog(Animal):
    pass

print(issubclass(Dog, Animal))
```

### Output

```text
True
```

---

## 7. `issubclass()` with Multiple Classes

```python
class Vehicle:
    pass

class Car(Vehicle):
    pass

print(issubclass(Car, (Vehicle, object)))
```

### Output

```text
True
```

---

## 8. Difference Between `isinstance()` and `issubclass()`

| Feature | `isinstance()` | `issubclass()` |
|---------|----------------|----------------|
| Checks | Objects | Classes |
| Input | Object | Class |
| Returns | `True` or `False` | `True` or `False` |
| Inheritance Aware | Yes | Yes |
| Common Use | Object type checking | Class hierarchy checking |

## 11. Common Mistakes

Avoid these common mistakes while working with `isinstance()` and `issubclass()`.

### Passing an Object to `issubclass()`

Incorrect

```python
dog = Dog()

issubclass(dog, Animal)
```

### Output

```text
TypeError
```

`issubclass()` expects **classes**, not objects.

Correct

```python
issubclass(Dog, Animal)
```

---

### Passing a Class to `isinstance()`

Incorrect

```python
isinstance(Dog, Animal)
```

This checks a **class object**, not an instance.

Correct

```python
dog = Dog()

isinstance(dog, Animal)
```

---

### Using `type()` Instead of `isinstance()`

Incorrect

```python
type(dog) == Animal
```

This returns `False` because `dog` is an instance of `Dog`, not exactly `Animal`.

Better

```python
isinstance(dog, Animal)
```

`isinstance()` correctly considers inheritance.

---

### Forgetting That Every Class Inherits from `object`

```python
class Student:
    pass

print(issubclass(Student, object))
```

### Output

```text
True
```

In Python, every new-style class ultimately inherits from `object`.

## 12. Best Practices

Follow these recommended practices when using `isinstance()` and `issubclass()`.

- Use `isinstance()` to validate object types.
- Use `issubclass()` to inspect inheritance relationships.
- Prefer `isinstance()` over `type()` when inheritance should be considered.
- Use tuples when checking against multiple classes.
- Avoid unnecessary type checking when polymorphism can solve the problem.
- Write code that works with behavior (methods) rather than relying heavily on exact types.
- Use these functions to improve code safety and readability.
