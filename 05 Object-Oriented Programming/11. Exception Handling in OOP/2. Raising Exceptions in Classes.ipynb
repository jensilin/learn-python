# 🚀 Raising Exceptions in Classes in Python

Raising exceptions inside classes is an essential part of **object-oriented programming (OOP)**. It allows a class to **detect invalid data, prevent incorrect operations, and enforce business rules**.

Instead of allowing an object to enter an invalid state, a class can **raise an exception** whenever an invalid operation is attempted.

Examples include:

- Preventing a negative bank balance.
- Rejecting an invalid student age.
- Preventing invalid product prices.
- Rejecting empty usernames.
- Preventing division by zero.

Using exceptions inside classes helps create **robust, secure, and reliable applications**.

# 📖 Topics Covered

## 1. What Does Raising Exceptions in Classes Mean?

A class raises an exception whenever invalid data or an invalid operation is detected.

Example:

```python
class Student:

    def __init__(self, age):

        if age < 0:
            raise ValueError("Age cannot be negative.")

        self.age = age

student = Student(-5)
```

### Output

```text
Traceback (most recent call last):
...
ValueError: Age cannot be negative.
```

---

## 2. Why Raise Exceptions in Classes?

Raising exceptions helps us:

- Prevent invalid objects.
- Enforce business rules.
- Improve data validation.
- Detect programming errors early.
- Improve application reliability.
- Simplify debugging.

---

## 3. Raising Exceptions in Constructors

The constructor (`__init__`) is a common place for validation.

```python
class Product:

    def __init__(self, price):

        if price <= 0:
            raise ValueError("Price must be greater than zero.")

        self.price = price

product = Product(-500)
```

### Output

```text
Traceback (most recent call last):
...
ValueError: Price must be greater than zero.
```

---

## 4. Raising Exceptions in Methods

Methods can validate input before performing operations.

```python
class Calculator:

    def divide(self, a, b):

        if b == 0:
            raise ZeroDivisionError("Cannot divide by zero.")

        return a / b

calculator = Calculator()

calculator.divide(10, 0)
```

### Output

```text
Traceback (most recent call last):
...
ZeroDivisionError: Cannot divide by zero.
```

---

## 5. Using Built-in Exceptions

Python provides many built-in exception classes.

```python
class User:

    def __init__(self, username):

        if username == "":
            raise ValueError("Username cannot be empty.")

        self.username = username

user = User("")
```

### Output

```text
Traceback (most recent call last):
...
ValueError: Username cannot be empty.
```

---

## 6. Using Custom Exceptions

Custom exceptions make business logic more meaningful.

```python
class InvalidBalanceError(Exception):
    pass

class BankAccount:

    def __init__(self, balance):

        if balance < 0:
            raise InvalidBalanceError("Invalid account balance.")

        self.balance = balance

account = BankAccount(-500)
```

### Output

```text
Traceback (most recent call last):
...
InvalidBalanceError: Invalid account balance.
```

---

## 7. Handling Raised Exceptions

Exceptions raised by classes can be handled using `try-except`.

```python
class Employee:

    def __init__(self, salary):

        if salary < 0:
            raise ValueError("Salary cannot be negative.")

        self.salary = salary

try:

    employee = Employee(-1000)

except ValueError as error:
    print(error)
```

### Output

```text
Salary cannot be negative.
```

## 8. Practical Examples

### Example 1: Student Age

```python
class Student:

    def __init__(self, age):

        if age < 18:
            raise ValueError("Student must be at least 18 years old.")

        self.age = age

student = Student(16)
```

### Output

```text
Traceback (most recent call last):
...
ValueError: Student must be at least 18 years old.
```

---

### Example 2: Bank Withdrawal

```python
class BankAccount:

    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):

        if amount > self.balance:
            raise ValueError("Insufficient balance.")

        self.balance -= amount

account = BankAccount(1000)

account.withdraw(1500)
```

### Output

```text
Traceback (most recent call last):
...
ValueError: Insufficient balance.
```

---

### Example 3: Product Quantity

```python
class Product:

    def __init__(self, quantity):

        if quantity < 0:
            raise ValueError("Quantity cannot be negative.")

        self.quantity = quantity

product = Product(-10)
```

### Output

```text
Traceback (most recent call last):
...
ValueError: Quantity cannot be negative.
```

---

### Example 4: Login System

```python
class Login:

    def authenticate(self, password):

        if len(password) < 6:
            raise ValueError("Password is too short.")

        return "Login Successful"

login = Login()

login.authenticate("123")
```

### Output

```text
Traceback (most recent call last):
...
ValueError: Password is too short.
```

## 11. Common Mistakes

Avoid these common mistakes while raising exceptions in classes.

### Allowing Invalid Objects

Incorrect

```python
class Student:

    def __init__(self, age):
        self.age = age
```

This allows invalid ages.

Correct

```python
if age < 0:
    raise ValueError(...)
```

---

### Raising the Wrong Exception Type

Incorrect

```python
raise Exception("Invalid price")
```

Better

```python
raise ValueError("Price must be positive.")
```

Choose the most appropriate exception type.

---

### Ignoring Validation

Do not trust user input.

Always validate values before storing them.

---

### Catching Every Exception

Avoid

```python
except:
    pass
```

Instead

```python
except ValueError:
    ...
```

Handle only the exceptions you expect.

---

## 12. Best Practices

Follow these recommended practices when raising exceptions in classes.

- Validate data inside constructors and methods.
- Raise exceptions as early as possible.
- Use meaningful exception messages.
- Prefer built-in exceptions when appropriate.
- Create custom exceptions for application-specific rules.
- Keep validation logic simple and readable.
- Document the conditions under which exceptions are raised.
