# 🚨 Custom Exceptions in Python

A **custom exception** is a user-defined exception created by inheriting from Python's built-in **`Exception`** class (or one of its subclasses).

Custom exceptions allow you to define **meaningful error types** that are specific to your application's business logic.

Instead of raising generic exceptions like `ValueError` or `TypeError`, you can create exceptions with descriptive names such as:

- `InvalidAgeError`
- `InsufficientBalanceError`
- `InvalidPasswordError`
- `StudentNotFoundError`

Custom exceptions make programs:

- Easier to understand
- Easier to debug
- Easier to maintain
- More professional
- Better organized

# 📖 Topics Covered

## 1. What is a Custom Exception?

A custom exception is an exception class created by the programmer.

It inherits from the built-in `Exception` class.

Example:

```python
class InvalidMarksError(Exception):
    pass
```

Now this exception can be used like any built-in exception.

---

## 2. Why Use Custom Exceptions?

Custom exceptions help us:

- Represent application-specific errors.
- Improve code readability.
- Make debugging easier.
- Separate different types of errors.
- Provide meaningful error messages.
- Improve program maintainability.

## 4. Creating a Simple Custom Exception

```python
class InvalidAgeError(Exception):
    pass

raise InvalidAgeError("Invalid age entered.")
```

### Output

```text
Traceback (most recent call last):
...
InvalidAgeError: Invalid age entered.
```

---

## 5. Raising a Custom Exception

```python
class NegativeNumberError(Exception):
    pass

number = -10

if number < 0:
    raise NegativeNumberError("Negative numbers are not allowed.")
```

### Output

```text
Traceback (most recent call last):
...
NegativeNumberError: Negative numbers are not allowed.
```

---

## 6. Handling a Custom Exception

```python
class InvalidPasswordError(Exception):
    pass

try:

    password = "123"

    if len(password) < 6:
        raise InvalidPasswordError("Password is too short.")

except InvalidPasswordError as error:
    print(error)
```

### Output

```text
Password is too short.
```

---

## 7. Custom Exception with Additional Information

Custom exceptions can store additional data.

```python
class InvalidAgeError(Exception):

    def __init__(self, age):
        self.age = age
        super().__init__(f"{age} is not a valid age.")

try:
    raise InvalidAgeError(-5)

except InvalidAgeError as error:
    print(error)
```

### Output

```text
-5 is not a valid age.
```

---

## 8. Inheriting from Other Exception Classes

Custom exceptions can inherit from other built-in exceptions.

```python
class InvalidInputError(ValueError):
    pass

raise InvalidInputError("Input is invalid.")
```

### Output

```text
Traceback (most recent call last):
...
InvalidInputError: Input is invalid.
```

## 9. Practical Examples

### Example 1: Bank Account

```python
class InsufficientBalanceError(Exception):
    pass

balance = 1000
withdraw = 1500

try:

    if withdraw > balance:
        raise InsufficientBalanceError("Insufficient balance.")

except InsufficientBalanceError as error:
    print(error)
```

### Output

```text
Insufficient balance.
```

---

### Example 2: Student Marks

```python
class InvalidMarksError(Exception):
    pass

marks = 120

try:

    if marks > 100:
        raise InvalidMarksError("Marks cannot exceed 100.")

except InvalidMarksError as error:
    print(error)
```

### Output

```text
Marks cannot exceed 100.
```

---

### Example 3: Login System

```python
class LoginFailedError(Exception):
    pass

password = "abc"

try:

    if password != "python123":
        raise LoginFailedError("Invalid credentials.")

except LoginFailedError as error:
    print(error)
```

### Output

```text
Invalid credentials.
```

---

### Example 4: Voting System

```python
class UnderAgeError(Exception):
    pass

age = 16

try:

    if age < 18:
        raise UnderAgeError("Not eligible to vote.")

except UnderAgeError as error:
    print(error)
```

### Output

```text
Not eligible to vote.
```

### Example 5: Product Price

```python
class InvalidPriceError(Exception):
    pass

price = -100

try:

    if price < 0:
        raise InvalidPriceError("Price cannot be negative.")

except InvalidPriceError as error:
    print(error)
```

### Output

```text
Price cannot be negative.
```

## 12. Common Mistakes

Avoid these common mistakes while working with custom exceptions.

### Not Inheriting from `Exception`

Incorrect

```python
class MyError:
    pass
```

Correct

```python
class MyError(Exception):
    pass
```

---

### Raising Strings Instead of Exceptions

Incorrect

```python
raise "Invalid input"
```

Correct

```python
raise InvalidInputError("Invalid input")
```

---

### Using Generic Exceptions Everywhere

Avoid:

```python
raise Exception("Something went wrong.")
```

Instead, create meaningful exception classes.

---

### Ignoring Custom Exceptions

Always handle exceptions when recovery is possible.

```python
try:
    ...
except InvalidAgeError:
    ...
```

---

## 13. Best Practices

Follow these recommended practices when creating custom exceptions.

- Inherit from `Exception` or one of its subclasses.
- Give exceptions meaningful names ending with `Error`.
- Write clear and descriptive error messages.
- Use custom exceptions for business-specific errors.
- Handle exceptions where appropriate.
- Keep exception classes simple.
- Document when and why each exception is raised.
