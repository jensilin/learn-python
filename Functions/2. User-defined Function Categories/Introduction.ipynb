<h4>User-defined Function Categories</h4>

<p>
User-defined functions can be categorized based on whether they <b>accept arguments</b> and <b>return a value</b>. This classification helps programmers choose the appropriate function design based on the requirements of the program.
</p>

<h5>Categories of User-defined Functions</h5>

1. **Function without Arguments and without Return Value**
2. **Function with Arguments and without Return Value**
3. **Function without Arguments and with Return Value**
4. **Function with Arguments and with Return Value**

---

<h5>1. Function without Arguments and without Return Value</h5>

<p>
A function that <b>does not accept any arguments</b> and <b>does not return any value</b>. It performs a specific task and displays the result directly.
</p>

<h6>Syntax</h6>

```python
def function_name():
    statement(s)

function_name()
```

<h6>Example</h6>

```python
def greet():
    print("Welcome to Python")

greet()
```

<h6>Output</h6>

```text
Welcome to Python
```

---

<h5>2. Function with Arguments and without Return Value</h5>

<p>
A function that <b>accepts one or more arguments</b> but <b>does not return any value</b>. The result is displayed inside the function.
</p>

<h6>Syntax</h6>

```python
def function_name(parameters):
    statement(s)

function_name(arguments)
```

<h6>Example</h6>

```python
def greet(name):
    print(f"Welcome {name}")

greet("John")
```

<h6>Output</h6>

```text
Welcome John
```

---

<h5>3. Function without Arguments and with Return Value</h5>

<p>
A function that <b>does not accept any arguments</b> but <b>returns a value</b> to the caller using the <code>return</code> statement.
</p>

<h6>Syntax</h6>

```python
def function_name():
    statement(s)
    return value

variable = function_name()
```

<h6>Example</h6>

```python
def get_message():
    return "Welcome to Python"

message = get_message()
print(message)
```

<h6>Output</h6>

```text
Welcome to Python
```

---

<h5>4. Function with Arguments and with Return Value</h5>

<p>
A function that <b>accepts one or more arguments</b> and <b>returns a value</b> to the caller. This is the most commonly used type of user-defined function.
</p>

<h6>Syntax</h6>

```python
def function_name(parameters):
    statement(s)
    return value

variable = function_name(arguments)
```

<h6>Example</h6>

```python
def add(a, b):
    return a + b

result = add(10, 20)
print(result)
```

<h6>Output</h6>

```text
30
```

---

<h5>Comparison of User-defined Function Categories</h5>

| Category | Accepts Arguments | Returns Value |
|-----------|:-----------------:|:-------------:|
| Without Arguments & Without Return Value | ❌ | ❌ |
| With Arguments & Without Return Value | ✅ | ❌ |
| Without Arguments & With Return Value | ❌ | ✅ |
| With Arguments & With Return Value | ✅ | ✅ |

---

<h5>Note</h5>

<ul>
    <li>These are <b>categories of user-defined functions</b>, not different types of functions.</li>
    <li>The classification is based on whether the function <b>accepts arguments</b> and <b>returns a value</b>.</li>
    <li>The <code>return</code> statement sends the result back to the caller and terminates the function execution.</li>
    <li>If a function does not contain a <code>return</code> statement, it returns <code>None</code> by default.</li>
</ul>
