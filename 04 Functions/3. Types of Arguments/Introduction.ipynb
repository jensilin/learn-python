<h4>Types of Function Arguments</h4>

<p>
Function arguments are the values passed to a function when it is called. Python provides different types of arguments to make function calls more flexible and convenient.
</p>

<h5>Types of Function Arguments</h5>

1. **Positional Arguments**
2. **Keyword Arguments**
3. **Default Arguments**
4. **Variable-length Arguments (`*args`)**
5. **Keyword Variable-length Arguments (`**kwargs`)**

---

<h5>1. Positional Arguments</h5>

<p>
In positional arguments, the values are passed to the function in the <b>same order</b> as the parameters defined in the function. The position of each argument is important.
</p>

<h6>Syntax</h6>

```python
def function_name(parameter1, parameter2):
    statement(s)

function_name(value1, value2)
```

<h6>Example</h6>

```python
def student(name, age):
    print(f"Name: {name}")
    print(f"Age: {age}")

student("John", 20)
```

<h6>Output</h6>

```text
Name: John
Age: 20
```

---

<h5>2. Keyword Arguments</h5>

<p>
In keyword arguments, the values are passed using the <b>parameter names</b>. The order of the arguments does not matter.
</p>

<h6>Syntax</h6>

```python
def function_name(parameter1, parameter2):
    statement(s)

function_name(parameter2=value2, parameter1=value1)
```

<h6>Example</h6>

```python
def student(name, age):
    print(f"Name: {name}")
    print(f"Age: {age}")

student(age=20, name="John")
```

<h6>Output</h6>

```text
Name: John
Age: 20
```

---

<h5>3. Default Arguments</h5>

<p>
A default argument is a parameter that is assigned a <b>default value</b>. If no value is passed during the function call, the default value is used.
</p>

<h6>Syntax</h6>

```python
def function_name(parameter=value):
    statement(s)

function_name()
```

<h6>Example</h6>

```python
def greet(name="Guest"):
    print(f"Welcome {name}")

greet()
greet("Alice")
```

<h6>Output</h6>

```text
Welcome Guest
Welcome Alice
```

---

<h5>4. Variable-length Arguments (<code>*args</code>)</h5>

<p>
The <code>*args</code> parameter allows a function to accept <b>any number of positional arguments</b>. The received arguments are stored as a <code>tuple</code>.
</p>

<h6>Syntax</h6>

```python
def function_name(*args):
    statement(s)

function_name(value1, value2, value3)
```

<h6>Example</h6>

```python
def display(*numbers):
    print(numbers)

display(10, 20, 30, 40)
```

<h6>Output</h6>

```text
(10, 20, 30, 40)
```

---

<h5>5. Keyword Variable-length Arguments (<code>**kwargs</code>)</h5>

<p>
The <code>**kwargs</code> parameter allows a function to accept <b>any number of keyword arguments</b>. The received arguments are stored as a <code>dictionary</code>.
</p>

<h6>Syntax</h6>

```python
def function_name(**kwargs):
    statement(s)

function_name(key1=value1, key2=value2)
```

<h6>Example</h6>

```python
def student(**details):
    print(details)

student(name="John", age=20, city="Bangalore")
```

<h6>Output</h6>

```text
{'name': 'John', 'age': 20, 'city': 'Bangalore'}
```

---

<h5>Comparison of Function Arguments</h5>

| Argument Type | Description |
|---------------|-------------|
| **Positional Arguments** | Arguments are passed based on their position. |
| **Keyword Arguments** | Arguments are passed using parameter names. |
| **Default Arguments** | Parameters have predefined default values. |
| **Variable-length Arguments (`*args`)** | Accepts any number of positional arguments and stores them in a tuple. |
| **Keyword Variable-length Arguments (`**kwargs`)** | Accepts any number of keyword arguments and stores them in a dictionary. |

---

<h5>Note</h5>

<ul>
    <li>Positional arguments must be passed in the correct order.</li>
    <li>Keyword arguments can be passed in any order.</li>
    <li>Default arguments are used when no value is provided during the function call.</li>
    <li><code>*args</code> collects multiple positional arguments into a <code>tuple</code>.</li>
    <li><code>**kwargs</code> collects multiple keyword arguments into a <code>dictionary</code>.</li>
    <li>A function can use positional arguments, default arguments, <code>*args</code>, and <code>**kwargs</code> together when required.</li>
</ul>
