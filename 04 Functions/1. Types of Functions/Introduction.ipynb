<h4>Functions</h4>

<p>
A <b>function</b> is a reusable block of code that performs a specific task. Functions help improve code reusability, readability, and maintainability by allowing the same block of code to be executed multiple times without rewriting it.
</p>

<p>
In Python, a function is created using the <code>def</code> keyword.
</p>

<h5>Syntax</h5>

```python
def function_name(parameters):
    statement(s)
    return value
```

<h5>Flow of Execution</h5>

<ul>
    <li>A function is defined using the <code>def</code> keyword.</li>
    <li>The function body contains the statements to be executed.</li>
    <li>The function is executed only when it is called.</li>
    <li>Arguments can be passed to the function through parameters.</li>
    <li>The <code>return</code> statement is optional and is used to return a value to the caller.</li>
</ul>

<h5>Example</h5>

```python
def greet():
    print("Welcome to Python")

greet()
```

<h5>Output</h5>

```text
Welcome to Python
```

<h5>Advantages of Functions</h5>

<ul>
    <li>Reduces code duplication.</li>
    <li>Improves code readability.</li>
    <li>Makes debugging and maintenance easier.</li>
    <li>Promotes code reusability.</li>
</ul>

---

<h4>Types of Functions</h4>

Python functions are commonly classified into the following types:

1. **Built-in Functions**
2. **User-defined Functions**
3. **Anonymous (Lambda) Functions**
4. **Recursive Functions**

---

<h5>1. Built-in Functions</h5>

<p>
Built-in functions are predefined functions provided by Python. They can be used directly without defining them.
</p>

<h6>Syntax</h6>

```python
function_name(arguments)
```

<h6>Example</h6>

```python
numbers = [10, 20, 30]

print(len(numbers))
print(max(numbers))
print(sum(numbers))
```

<h6>Output</h6>

```text
3
30
60
```

---

<h5>2. User-defined Functions</h5>

<p>
User-defined functions are created by the programmer using the <code>def</code> keyword to perform a specific task.
</p>

<h6>Syntax</h6>

```python
def function_name(parameters):
    statement(s)
```

<h6>Example</h6>

```python
def display():
    print("Hello Python")

display()
```

<h6>Output</h6>

```text
Hello Python
```

---

<h5>3. Anonymous (Lambda) Functions</h5>

<p>
Anonymous functions are functions without a name. They are created using the <code>lambda</code> keyword and are generally used for short, single-expression operations.
</p>

<h6>Syntax</h6>

```python
lambda arguments: expression
```

<h6>Example</h6>

```python
square = lambda n: n ** 2

print(square(5))
```

<h6>Output</h6>

```text
25
```

---

<h5>4. Recursive Functions</h5>

<p>
A recursive function is a function that calls itself until a specified condition is satisfied.
</p>

<h6>Syntax</h6>

```python
def function_name(parameters):
    if condition:
        return value
    return function_name(arguments)
```

<h6>Example</h6>

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

<h6>Output</h6>

```text
120
```

---

<h5>Note</h5>

<ul>
    <li>A function is executed only when it is called.</li>
    <li>Function names should be meaningful and follow Python naming conventions.</li>
    <li>The <code>return</code> statement is optional.</li>
    <li>Functions can accept zero or more parameters.</li>
    <li>Functions may or may not return a value.</li>
</ul>
