##### Question 1: Student Grade Management System

Write a Python program to manage student records using a dictionary.

Requirements:
- Store student names as keys and marks as values.
- Allow the user to:
  - Add a student.
  - Update marks.
  - Delete a student.
  - Search for a student.
  - Display all students sorted by marks (highest to lowest).
- Display the average, highest, and lowest marks.

**Example Input**
```text
Enter number of students: 3
Enter student name: Alice
Enter marks: 85
Enter student name: Bob
Enter marks: 92
Enter student name: Charlie
Enter marks: 78
Search student: Bob
```

**Expected Output**
```text
Student Found: Bob
Marks: 92

Students (Highest to Lowest)
Bob : 92
Alice : 85
Charlie : 78

Average Marks : 85.00
Highest Marks : 92
Lowest Marks : 78
```

````python
students = {}

def add_student():
    n = int(input("Enter number of students: "))

    for i in range(n):
        student_name = input("Enter student name: ")
        student_marks = input("Enter marks: ")
        students[student_name] = student_marks

def search_student():
    s_n = input("Search student: ")
    if s_n in students.keys():
        print(s_n)

def delete_student():
    d_n = input("Student to be deleted: ")
    for i in students:
        if i == d_n:
            students.pop(i)

def sor_ahl():
    print(sorted(students))
    print((sum(students.values())/len(students))/100)
    print(max(students.values()))
    print(min(students.values()))
````
