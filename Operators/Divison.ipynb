Divison Operator 
It is used to find the values when divided by numbers

It has three types
True division
Floor division
Modulas division

True division : It is a division operator which gives the exact value when divided

Syntax : operater1/operater2

```python
#In Jupyter noter or .ipynb file, without print statement also will work

740/2
```

Floor division 

It is a division operater which gives the value by removing the decimal point.

Syntax : operater1 // operater2

```python
750//3
```

Modulus
It is a division operator which gives remainder

Syntax : operater1 % operater2

```python
37 % 4
```

NOTE : None of the division operater will support for collection data types.

NOTE : Floor division and modulas division will not support for complex

```python
#Not supported for strings

'Jensilin' / 2
```

```python
3+4j / 4+4j
```
