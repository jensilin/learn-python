<h4>Power / Exponentiation (<code>**</code>) Operator</h4>

<p>
The <code>**</code> operator is used to perform <b>exponentiation</b>. It raises the first operand (base) to the power of the second operand (exponent).
</p>

<h5>Syntax</h5>

```python
operand1 ** operand2
```

<h5>Example</h5>

```python
print(2 ** 3)
print(5 ** 2)
print(10 ** 0)
```

<h5>Output</h5>

```text
8
25
1
```

<h5>Note</h5>

<ul>
    <li>The <code>**</code> operator is supported only for <b>numeric data types</b>.</li>
    <li>It does <b>not</b> support collection data types such as <code>str</code>, <code>list</code>, <code>tuple</code>, <code>set</code>, or <code>dict</code>.</li>
</ul>
