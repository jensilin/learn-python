Power/Exponential(**)
It is a operator which is used to multiply numbers for specified number of times

It will not support for any collection data types.

Syntax : operator1 ** operator2

```python
8 ** 2
```

```python
4 ** 5
```

```python
3.2 ** 2
```

```python
3+1j ** 2
```

```python
True ** 3
```

```python
3 ** 2.3
```

```python
5 ** -2
```

```python
#Not supported for Strings
'a' ** 3
```
