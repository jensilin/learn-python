Additional Operator 

It is a operator which is used to find the summation between the operands(Individual data type)

This operator will perform concatenation in case of collection data types

Syntax : operand1 + operand2

Note : '+' operator will not support for set and dictionary

```python
a = 20
b = 17
print(a + b)
```

```python
print(5+9.8)
```

```python
#In Jupyter noter or .ipynb file, without print statement also will work
3 + (4+4j)
```

```python
3+4j + 4+8j
```

```python
True + True
```

```python
'Tom ' + 'Holland'
```

```python
[2,3,4] + [4,5,6]
```

```python
#Addition operation is not supported for different collection types
#Only supported for List1 + List2

[2,3,4] + (9,8)
```

```python
#Unsupported for sets
{1,2,3} + {4,5,6}
```

```python
(2,3,4) + (4,5,6,7)
```

```python
#Not supported for dictionary
{'a':2} + {'c':78}
```

NOTE : '+' Not supported to add different collection data types
