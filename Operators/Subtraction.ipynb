Subtraction Operator : The operator is used to find the difference between the values.

Syntax : operand1 - operand2

Note : In case of collection it supports only for set data type

In set it removes the common values and display the values present at only first set.

```python
#In Jupyter noter or .ipynb file, without print statement also will work

10 - 5
```

```python
20 - 8.765
```

```python
8+4j - 2+4j
```

```python
#Not supported for Strings
"James bond" - "bond"
```

```python
#Not supported for list

[2,3,4] - [4,5,6]
```

```python
#In set it removes the common values and display the values present at only first set.

{4,5,6,90} - {10,20,5,90}
```
