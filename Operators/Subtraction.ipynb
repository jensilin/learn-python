<h4>Subtraction (<code>-</code>) Operator</h4>

<p>
The <code>-</code> operator is used to perform <b>subtraction</b> between numeric operands. It returns the <b>difference</b> between the given values.
</p>

<h5>Syntax (Numeric Subtraction)</h5>

```python
operand1 - operand2
```

<h5>Example</h5>

```python
print(20 - 8)
print(15 - 5)
```

<h5>Output</h5>

```text
12
10
```

<h5>Set Difference</h5>

<p>
For <code>set</code> data types, the <code>-</code> operator performs the <b>set difference</b> operation. It removes the common elements from the first set and returns only the elements that are present in the first set but not in the second set.
</p>

<h5>Syntax</h5>

```python
set1 - set2
```

<h5>Example</h5>

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

print(set1 - set2)
```

<h5>Output</h5>

```text
{1, 2}
```

<h5>Note</h5>

<ul>
    <li>For collection data types, the <code>-</code> operator is supported only for <code>set</code>.</li>
    <li>It returns the elements that are present only in the <b>first set</b> after removing the common elements.</li>
</ul>
