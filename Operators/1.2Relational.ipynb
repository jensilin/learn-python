<h4>d) Less Than (<code>&lt;</code>) Operator</h4>

<p>
The <code>&lt;</code> operator is used to check whether the <b>first operand is less than the second operand</b>. It returns <code>True</code> if the condition is satisfied; otherwise, it returns <code>False</code>.
</p>

<h5>Syntax</h5>

```python
operand1 < operand2
```

<h5>Example</h5>

```python
print(10 < 20)
print(15 < 5)
```

<h5>Output</h5>

```text
True
False
```

<h5>Note</h5>

<ul>
    <li>All the rules that apply to the <code>&gt;</code> (greater than) operator also apply to the <code>&lt;</code> (less than) operator.</li>
    <li>For <code>str</code>, <code>list</code>, and <code>tuple</code>, Python performs a <b>lexicographical comparison</b>.</li>
    <li>For <code>set</code>, the <code>&lt;</code> operator returns <code>True</code> if the <b>first set is a proper subset of the second set</b>.</li>
    <li>The <code>&lt;</code> operator is not supported for <code>complex</code> numbers or <code>dict</code> data types.</li>
</ul>

````python
{5,6} < {34,67,6,5}
````

````python
[10,3,4] < [10,5,3]
````

<h4>e) Greater Than or Equal To (<code>&gt;=</code>) Operator</h4>

<p>
The <code>&gt;=</code> operator is used to check whether the <b>first operand is greater than or equal to the second operand</b>. It returns <code>True</code> if either condition is satisfied; otherwise, it returns <code>False</code>.
</p>

<h5>Syntax</h5>

```python
operand1 >= operand2
```

<h5>Example</h5>

```python
print(20 >= 10)
print(10 >= 10)
print(5 >= 15)
```

<h5>Output</h5>

```text
True
True
False
```

<h5>Note</h5>

<ul>
    <li>All the rules that apply to the <code>&gt;</code> (greater than) operator also apply to the <code>&gt;=</code> (greater than or equal to) operator.</li>
    <li>For <code>str</code>, <code>list</code>, and <code>tuple</code>, Python performs a <b>lexicographical comparison</b>.</li>
    <li>For <code>set</code>, the <code>&gt;=</code> operator returns <code>True</code> if the <b>second set is a subset of the first set</b>.</li>
    <li>The <code>&gt;=</code> operator is not supported for <code>complex</code> numbers or <code>dict</code> data types.</li>
</ul>

````python
[2,3,45] >= [2,3,45]
````

````python
{2,3,89} >= {3,2,89}
````
