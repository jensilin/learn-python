Relational Operator/Comparison Operator 
These are the operator which are used to compare between the values.

It gives output in the form of True/False

Types of relational operator
Relational equals to (==)
Relational not equals to (!=)
Relational greater than (>)
Relational lesser than (<)
Relational greater than or equals to (>=)
Relational lesser than or equals to (<=)

Relational equals to (==)
It is a operator which gives result as true, if both the values are exactly same.

Syntax : op1 == op2

```python
10 == 10
```

```python
'CSK' == 'RCB'
```

Relational not equal to (!=):
It is a operator which gives result as true if both the operand are not same.

Syntax: op1 != op2

```python
1 != 1
```

```python
'hehe' != 100
```

Relational greater than (>):
It is a operator which gives the result as true if first operand is greater than operand two.

Syntax: op1 > op2

Note:

This operator cannot be compared b/w different data types & also collection.
It will not support for complex.
It doesn't support for dictionary data type.
In case of strings it compares the ASCII values.

A = 65    a = 97    0 = 48
B = 66    b = 98    1 = 49

Val1 = "a"
Val2 = "A"

Val1 > Val2

In strings it compares the first character of first operand with first character of second operand.
If it is greater it gives result as true.
If it is lesser it gives result as false.
If both are equal then it compares 2nd character with 2nd character and moves on.
If both the operands are same it gives false.

```python
'Python' > 'Java'
```

```python
10 > 15
```

```python
22 >= 22
```
