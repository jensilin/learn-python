<h4>Multiplication (<code>*</code>) Operator</h4>

<p>
The <code>*</code> operator is used to perform <b>multiplication</b> between numeric operands. It returns the <b>product</b> of the given values.
</p>

<p>
For <b>collection data types</b>, the <code>*</code> operator performs <b>repetition</b>. It repeats the collection a specified number of times.
</p>

<h5>Syntax (Numeric Multiplication)</h5>

```python
operand1 * operand2
```

<h5>Example</h5>

```python
print(10 * 5)
print(4 * 3)
```

<h5>Output</h5>

```text
50
12
```

<h5>Syntax (Collection Repetition)</h5>

```python
collection * n
```

<p>Where <code>n</code> is a <b>positive integer</b>.</p>

<h5>Example</h5>

```python
print("Hi " * 3)
print([1, 2] * 2)
print((1, 2) * 3)
```

<h5>Output</h5>

```text
Hi Hi Hi
[1, 2, 1, 2]
(1, 2, 1, 2, 1, 2)
```

<h5>Note</h5>

<ul>
    <li>The <code>*</code> operator does <b>not</b> support <code>set</code> and <code>dict</code> data types.</li>
    <li>For collection repetition, the second operand (<code>n</code>) must be a <b>non-negative integer</b>.</li>
</ul>
