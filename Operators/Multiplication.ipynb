Multiplication Operator :
It is a operator used to find product between the values.

Syntax : operand1 * operand3

In case of collection.
Syntax : operand * n (n should be positive int)

Note : This operation will not support for set and dictionary. 

```python
6 * 6
```

```python
True * 3
```

```python
43.653 * 765
```

```python
1+2j * 2+2j
```

```python
'Zendaya' * 3
```

```python
[5,6] * 3
```

```python
(8,4) * 2
```

```python
#Not supported for st
{69,70} * 3
```

```python
#Not supported for dictionary
{'lucky':5} * 2
```

```python
#It should not contain float value

'Andrew Garfield' * 6.5
```
