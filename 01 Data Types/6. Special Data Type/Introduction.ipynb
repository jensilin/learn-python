# 🚫 Special Data Type

Python provides a special built-in data type called **`NoneType`**, which has only one possible value: **`None`**.

The value `None` is used to represent the **absence of a value**, **no result**, or a **null value**. It is commonly used as a default value for variables, function return values, optional parameters, and placeholders when no actual data is available.

The `NoneType` data type helps programmers clearly indicate that a variable or object intentionally has no value.

## 2. Characteristics of `NoneType`

Explore the important characteristics of the `NoneType` data type.

### Characteristics

- Has only one value: `None`.
- Represents the absence of a value.
- Immutable.
- Singleton object (only one instance exists).
- Frequently used as a default value.
- Commonly returned by functions that do not explicitly return a value.

**Example**

```python
value = None

print(type(value))
```

### Output

```text
<class 'NoneType'>
```
