# 🔢 Numeric Data Types

This module introduces **Python Numeric Data Types**, which are used to represent numeric values and perform mathematical operations. Python provides four built-in numeric data types, each designed for a specific purpose.

## 📂 Module Structure

### Introduction.ipynb

This notebook provides an overview of Python's numeric data types, their characteristics, and when to use each type.

**Example**

```python
a = 10
b = 3.14
c = 2 + 5j
d = True

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

---

### int.ipynb

This notebook explains the **Integer (`int`)** data type.

**Topics Covered**

- Introduction to Integers
- Creating Integer Objects
- Integer Operations
- Type Conversion
- Examples

**Example**

```python
age = 25
temperature = -10

print(age)
print(temperature)
```

---

### float.ipynb

This notebook explains the **Floating Point (`float`)** data type.

**Topics Covered**

- Introduction to Floating Point Numbers
- Creating Float Objects
- Float Operations
- Type Conversion
- Examples

**Example**

```python
price = 499.99
pi = 3.14159

print(price)
print(pi)
```

---

### complex.ipynb

This notebook explains the **Complex (`complex`)** data type.

**Topics Covered**

- Introduction to Complex Numbers
- Real and Imaginary Parts
- Creating Complex Objects
- Supported Operations
- Examples

**Example**

```python
z = 4 + 3j

print(z)
print(type(z))
```

---

### bool.ipynb

This notebook explains the **Boolean (`bool`)** data type.

**Topics Covered**

- Introduction to Boolean Values
- `True` and `False`
- Boolean Operations
- Type Conversion
- Examples

**Example**

```python
is_student = True
is_admin = False

print(is_student)
print(is_admin)
```
