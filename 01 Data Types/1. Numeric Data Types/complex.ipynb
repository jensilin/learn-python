# 🧮 Complex (`complex`) Data Type

The **Complex (`complex`)** data type is one of Python's built-in numeric data types. It is used to represent **complex numbers**, which consist of a **real part** and an **imaginary part**. In Python, the imaginary part is represented using the suffix **`j`**.

Complex numbers are immutable and support various arithmetic operations such as addition, subtraction, multiplication, and division. They are commonly used in scientific computing, electrical engineering, signal processing, and mathematical modeling.

Unlike `int` and `float`, complex numbers cannot be compared using relational operators such as `<`, `>`, `<=`, and `>=`.

## 2. Characteristics of Complex Numbers

Explore the important characteristics of the `complex` data type.

### Characteristics

- Stores real and imaginary values.
- Imaginary part is represented using `j`.
- Immutable data type.
- Supports arithmetic operations.
- Does not support relational comparisons (`<`, `>`, `<=`, `>=`).

**Example**

```python
num = 10 + 4j

print(type(num))
```

**Output**

```text
<class 'complex'>
```

## 3. Creating Complex Objects

Learn different ways to create complex objects in Python.

### Topics Covered

- Direct assignment
- Using the `complex()` constructor
- Creating complex numbers from integers
- Creating complex numbers from floats

**Example**

```python
a = 4 + 8j
b = complex(2, 6)
c = complex(3.5, 4.2)

print(a)
print(b)
print(c)
```

**Output**

```text
(4+8j)
(2+6j)
(3.5+4.2j)
```

## 5. Complex Number Operations

Learn the arithmetic operations supported by complex numbers.

### Operations Covered

- Addition
- Subtraction
- Multiplication
- Division
- Exponentiation

**Example**

```python
a = 2 + 3j
b = 1 + 2j

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** 2)
```

**Output**

```text
(3+5j)
(1+1j)
(-4+7j)
(1.6-0.2j)
(-5+12j)
```

## 7. Type Conversion

Learn how to create complex numbers using the `complex()` function.

### Topics Covered
