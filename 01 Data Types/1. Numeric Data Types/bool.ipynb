# ✅ Boolean (`bool`) Data Type

The **Boolean (`bool`)** data type is one of Python's built-in numeric data types. It is used to represent **logical values** and has only **two possible values**: `True` and `False`.

The `bool` data type is primarily used in **decision-making**, **conditional statements**, **loops**, and **logical operations**. Internally, Python treats `True` as `1` and `False` as `0`, making the Boolean data type a subclass of the `int` data type.

Boolean values are immutable and are commonly produced as the result of comparison and logical operations.

## 2. Characteristics of Boolean Values

Explore the important characteristics of the `bool` data type.

### Characteristics

- Stores only two values: `True` and `False`.
- Immutable data type.
- Subclass of the `int` data type.
- `True` is treated as `1`.
- `False` is treated as `0`.
- Commonly used in decision-making and logical expressions.

**Example**

```python
print(type(True))
print(type(False))
```

**Output**

```text
<class 'bool'>
<class 'bool'>
```
