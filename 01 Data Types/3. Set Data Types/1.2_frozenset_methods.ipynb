Python provides only **non-modifying methods** for the `frozenset` data type because a `frozenset` is **immutable**. Unlike a normal `set`, you cannot add, remove, or update elements after creating a `frozenset`.

This notebook covers the **commonly used `frozenset` methods** with their syntax, examples, and outputs.

## 1. `copy()`

Returns a shallow copy of the `frozenset`.

### Syntax

```python
frozenset_name.copy()
```

### Example

```python
numbers = frozenset({10, 20, 30})

new_numbers = numbers.copy()

print(new_numbers)
```

### Output

```text
frozenset({10, 20, 30})
```

> **Note:** Since `frozenset` is immutable, `copy()` returns the same immutable object.

---

## 2. `union()`

Returns a new `frozenset` containing all unique elements from both collections.

### Syntax

```python
frozenset1.union(frozenset2)
```

### Example

```python
a = frozenset({1, 2, 3})
b = frozenset({3, 4, 5})

print(a.union(b))
```

### Output

```text
frozenset({1, 2, 3, 4, 5})
```

---

## 3. `intersection()`

Returns the common elements present in both `frozenset` objects.

### Syntax

```python
frozenset1.intersection(frozenset2)
```

### Example

```python
a = frozenset({1, 2, 3})
b = frozenset({2, 3, 4})

print(a.intersection(b))
```

### Output

```text
frozenset({2, 3})
```

---

## 4. `difference()`

Returns the elements that are present only in the first `frozenset`.

### Syntax

```python
frozenset1.difference(frozenset2)
```

### Example

```python
a = frozenset({1, 2, 3})
b = frozenset({2, 3, 4})

print(a.difference(b))
```

### Output

```text
frozenset({1})
```

---

## 5. `symmetric_difference()`

Returns the elements that are present in either `frozenset`, but not in both.

### Syntax

```python
frozenset1.symmetric_difference(frozenset2)
```

### Example

```python
a = frozenset({1, 2, 3})
b = frozenset({2, 3, 4})

print(a.symmetric_difference(b))
```

### Output

```text
frozenset({1, 4})
```

---

## 6. `issubset()`

Checks whether one `frozenset` is a subset of another.

### Syntax

```python
frozenset1.issubset(frozenset2)
```

### Example

```python
a = frozenset({1, 2})
b = frozenset({1, 2, 3})

print(a.issubset(b))
```

### Output

```text
True
```

---

## 7. `issuperset()`

Checks whether one `frozenset` is a superset of another.

### Syntax

```python
frozenset1.issuperset(frozenset2)
```

### Example

```python
a = frozenset({1, 2, 3})
b = frozenset({1, 2})

print(a.issuperset(b))
```

### Output

```text
True
```

---

## 8. `isdisjoint()`

Checks whether two `frozenset` objects have no common elements.

### Syntax

```python
frozenset1.isdisjoint(frozenset2)
```

### Example

```python
a = frozenset({1, 2})
b = frozenset({3, 4})

print(a.isdisjoint(b))
```

### Output

```text
True
```

# đź“Š Summary

| Method | Purpose |
|---------|---------|
| `copy()` | Returns a shallow copy |
| `union()` | Combines two `frozenset` objects |
| `intersection()` | Returns common elements |
| `difference()` | Returns elements present only in the first `frozenset` |
| `symmetric_difference()` | Returns non-common elements |
| `issubset()` | Checks subset relationship |
| `issuperset()` | Checks superset relationship |
| `isdisjoint()` | Checks whether two `frozenset` objects have no common elements |
