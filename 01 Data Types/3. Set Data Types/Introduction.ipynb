# 🎯 Set Data Types

The **Set Data Types** are one of Python's built-in data type categories used to store an **unordered collection of unique elements**. Unlike sequence data types, sets do **not maintain the insertion order** (conceptually) and **do not allow duplicate values**.

Sets are primarily used when uniqueness of elements is important. They support various mathematical set operations such as **union**, **intersection**, **difference**, and **symmetric difference**, making them useful in data analysis, searching, filtering, and eliminating duplicate values.

Python provides the following built-in set data types:

- `set`
- `frozenset`

# 💻 Quick Example

```python
numbers = {10, 20, 30, 40}
languages = frozenset({"Python", "Java", "C++"})

print(numbers)
print(languages)
print(type(numbers))
print(type(languages))
```

## 2. Characteristics of Set Data Types

Explore the important characteristics of set data types.

### Characteristics

- Stores unique elements only.
- Does not allow duplicate values.
- Mutable (`set`) or immutable (`frozenset`).
- Unordered collection.
- Does not support indexing or slicing.
- Supports mathematical set operations.
- Supports membership operators (`in`, `not in`).

**Example**

```python
numbers = {10, 20, 30}

print(type(numbers))
```

### Output

```text
<class 'set'>
```

## 3. Types of Set Data Types

Python provides two built-in set data types.

### Types

- `set`
- `frozenset`

**Example**

```python
a = {1, 2, 3}
b = frozenset({4, 5, 6})

print(type(a))
print(type(b))
```

### Output

## 4. Common Set Operations

Learn the operations supported by set data types.

### Operations Covered

- Union
- Intersection
- Difference
- Symmetric Difference
- Membership Testing
- Length Calculation

**Example**

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)
print(a & b)
print(a - b)
```

### Output

```text
{1, 2, 3, 4, 5}
{3}
{1, 2}
```
