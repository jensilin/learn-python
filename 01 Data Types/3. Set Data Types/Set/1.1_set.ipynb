# 🎯 Set (`set`) Data Type

The **Set (`set`)** is one of Python's built-in **collection data types** used to store an **unordered collection of unique elements**. Unlike lists and tuples, sets automatically remove duplicate values and do not support indexing or slicing.

Sets are **mutable**, meaning elements can be added or removed after creation. They are widely used for eliminating duplicate values, performing mathematical set operations, and checking membership efficiently.

Python provides several built-in methods to manipulate sets, making them useful for data analysis, searching, filtering, and various real-world applications.

## 2. Characteristics of Sets

Explore the important characteristics of the `set` data type.

### Characteristics

- Unordered collection.
- Mutable data type.
- Stores unique elements only.
- Does not allow duplicate values.
- Does not support indexing or slicing.
- Supports mathematical set operations.
- Supports membership operators (`in`, `not in`).

**Example**

```python
data = {10, "Python", 3.14, True}

print(data)
```

### Output

```text
{10, 'Python', 3.14, True}
```

---

## 3. Creating Set Objects

Learn different ways to create set objects in Python.

### Topics Covered

- Empty set
- Direct assignment
- Using the `set()` constructor
- Creating a set from other collections

**Example**

```python
a = set()
b = {10, 20, 30}
c = set([1, 2, 3])
d = set("Python")

print(a)
print(b)
print(c)
print(d)
```

### Output

```text
set()
{10, 20, 30}
{1, 2, 3}
{'P', 'y', 't', 'h', 'o', 'n'}
```

---

## 4. Set Operations

Learn the common operations that can be performed on sets.

### Operations Covered

- Union (`|`)
- Intersection (`&`)
- Difference (`-`)
- Symmetric Difference (`^`)
- Membership (`in`, `not in`)
- Length (`len()`)

**Example**

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)
print(a & b)
print(a - b)
print(a ^ b)
```

### Output

```text
{1, 2, 3, 4, 5}
{3}
{1, 2}
{1, 2, 4, 5}
```

## 9. Limitations of Sets

Understand the limitations of the `set` data type.

### Limitations

- Does not preserve element order.
- Does not support indexing.
- Does not support slicing.
- Stores only hashable (immutable) elements.
- Duplicate values are automatically removed.

**Example**

```python
numbers = {10, 20, 30}

print(numbers)
```

### Output

```text
{10, 20, 30}
```
