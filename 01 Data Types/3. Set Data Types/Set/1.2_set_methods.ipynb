# 🎯 Set Methods

Python provides several built-in **set methods** that allow you to add, remove, update, compare, and perform mathematical operations on sets efficiently.

Since sets are **mutable**, most set methods modify the original set directly instead of creating a new one. These methods are commonly used for removing duplicates, comparing collections, filtering data, and performing set operations.

This notebook covers the **most commonly used set methods** with their syntax, examples, and outputs.

---

In this notebook, you will learn:

- What set methods are.
- Why set methods are useful.
- Syntax of each method.

- Common use cases.

---

## 1. `add()`

Adds a single element to the set.

### Syntax

```python
set_name.add(element)
```

### Example

```python
numbers = {10, 20}

numbers.add(30)

print(numbers)
```

### Output

```text
{10, 20, 30}
```

---

## 2. `update()`

Adds multiple elements from another iterable to the set.

### Syntax

```python
set_name.update(iterable)
```

### Example

```python
numbers = {10, 20}

numbers.update([30, 40])

print(numbers)
```

### Output

```text
{10, 20, 30, 40}
```

---

## 3. `remove()`

Removes the specified element from the set.

### Syntax

```python
set_name.remove(element)
```

### Example

```python
numbers = {10, 20, 30}

numbers.remove(20)

print(numbers)
```

### Output

```text
{10, 30}
```

> **Note:** Raises `KeyError` if the element is not found.

---

## 4. `discard()`

Removes the specified element if it exists.

### Syntax

```python
set_name.discard(element)
```

### Example

```python
numbers = {10, 20, 30}

numbers.discard(20)

print(numbers)
```

### Output

```text
{10, 30}
```

> **Note:** Does not raise an error if the element is not present.

---

## 5. `pop()`

Removes and returns a random element from the set.

### Syntax

```python
set_name.pop()
```

### Example

```python
numbers = {10, 20, 30}

value = numbers.pop()

print(value)
print(numbers)
```

### Possible Output

```text
20
{10, 30}
```

> **Note:** Since sets are unordered, the removed element may vary.

---

## 6. `clear()`

Removes all elements from the set.

### Syntax

```python
set_name.clear()
```

### Example

```python
numbers = {10, 20, 30}

numbers.clear()

print(numbers)
```

### Output

```text
set()
```

---

## 7. `copy()`

Creates a shallow copy of the set.

### Syntax

```python
new_set = set_name.copy()
```

### Example

```python
numbers = {10, 20, 30}

new_numbers = numbers.copy()

print(new_numbers)
```

### Output

```text
{10, 20, 30}
```

---

## 8. `union()`

Returns a new set containing all unique elements from both sets.

### Syntax

```python
set1.union(set2)
```

### Example

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
```

### Output

```text
{1, 2, 3, 4, 5}
```

---

## 9. `intersection()`

Returns the common elements present in both sets.

### Syntax

```python
set1.intersection(set2)
```

### Example

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.intersection(b))
```

### Output

```text
{2, 3}
```

---

## 10. `difference()`

Returns elements that are present only in the first set.

### Syntax

```python
set1.difference(set2)
```

### Example

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.difference(b))
```

### Output

```text
{1}
```

---

## 11. `symmetric_difference()`

Returns elements that are present in either set but not in both.

### Syntax

```python
set1.symmetric_difference(set2)
```

### Example

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.symmetric_difference(b))
```

### Output

```text
{1, 4}
```

---

## 12. `issubset()`

Checks whether one set is a subset of another.

### Syntax

```python
set1.issubset(set2)
```

### Example

```python
a = {1, 2}
b = {1, 2, 3}

print(a.issubset(b))
```

### Output

```text
True
```

---

## 13. `issuperset()`

Checks whether one set is a superset of another.

### Syntax

```python
set1.issuperset(set2)
```

### Example

```python
a = {1, 2, 3}
b = {1, 2}

print(a.issuperset(b))
```

### Output

```text
True
```

---

## 14. `isdisjoint()`

Checks whether two sets have no common elements.

### Syntax

```python
set1.isdisjoint(set2)
```

### Example

```python
a = {1, 2}
b = {3, 4}

print(a.isdisjoint(b))
```

### Output

```text
True
```

# 📊 Summary

| `update()` | Add multiple elements |

| `union()` | Combine two sets |
| `intersection()` | Return common elements |
| `difference()` | Return elements present only in the first set |

| `issuperset()` | Check superset relationship |
