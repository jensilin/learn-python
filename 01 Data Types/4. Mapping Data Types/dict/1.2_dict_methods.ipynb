# 🗂️ Dictionary Methods

Python provides several built-in **dictionary methods** that allow you to create, access, update, remove, and manipulate key-value pairs efficiently.

Since dictionaries are **mutable**, most dictionary methods modify the original dictionary directly instead of creating a new one.

This notebook covers the **most commonly used dictionary methods** with their syntax, examples, and outputs.

---

- What dictionary methods are.

- Common use cases.

## 1. `get()`

Returns the value associated with the specified key. If the key is not found, it returns `None` (or a default value if provided).

### Syntax

```python
dictionary.get(key, default_value)
```

### Example

```python
student = {"name": "John", "age": 21}

print(student.get("name"))
```

### Output

```text
John
```

---

## 2. `keys()`

Returns a view object containing all the keys in the dictionary.

### Syntax

```python
dictionary.keys()
```

### Example

```python
student = {"name": "John", "age": 21}

print(student.keys())
```

### Output

```text
dict_keys(['name', 'age'])
```

---

## 3. `values()`

Returns a view object containing all the values in the dictionary.

### Syntax

```python
dictionary.values()
```

### Example

```python
student = {"name": "John", "age": 21}

print(student.values())
```

### Output

```text
dict_values(['John', 21])
```

---

## 4. `items()`

Returns a view object containing all key-value pairs as tuples.

### Syntax

```python
dictionary.items()
```

### Example

```python
student = {"name": "John", "age": 21}

print(student.items())
```

### Output

```text
dict_items([('name', 'John'), ('age', 21)])
```

---

## 5. `update()`

Updates the dictionary by adding new key-value pairs or modifying existing ones.

### Syntax

```python
dictionary.update(other_dictionary)
```

### Example

```python
student = {"name": "John"}

student.update({"age": 21})

print(student)
```

### Output

```text
{'name': 'John', 'age': 21}
```

---

## 6. `pop()`

Removes the specified key and returns its value.

### Syntax

```python
dictionary.pop(key)
```

### Example

```python
student = {"name": "John", "age": 21}

print(student.pop("age"))
print(student)
```

### Output

```text
21
{'name': 'John'}
```

---

## 7. `popitem()`

Removes and returns the last inserted key-value pair.

### Syntax

```python
dictionary.popitem()
```

### Example

```python
student = {
    "name": "John",
    "age": 21
}

print(student.popitem())
print(student)
```

### Output

```text
('age', 21)
{'name': 'John'}
```

---

## 8. `setdefault()`

Returns the value of the specified key. If the key does not exist, it inserts the key with the given default value.

### Syntax

```python
dictionary.setdefault(key, default_value)
```

### Example

```python
student = {"name": "John"}

student.setdefault("age", 21)

print(student)
```

### Output

```text
{'name': 'John', 'age': 21}
```

---

## 9. `copy()`

Creates a shallow copy of the dictionary.

### Syntax

```python
new_dictionary = dictionary.copy()
```

### Example

```python
student = {"name": "John", "age": 21}

new_student = student.copy()

print(new_student)
```

### Output

```text
{'name': 'John', 'age': 21}
```

---

## 10. `clear()`

Removes all key-value pairs from the dictionary.

### Syntax

```python
dictionary.clear()
```

### Example

```python
student = {"name": "John", "age": 21}

student.clear()

print(student)
```

### Output

```text
{}
```

---

## 11. `fromkeys()`

Creates a new dictionary using the given keys and a common default value.

### Syntax

```python
dict.fromkeys(keys, value)
```

### Example

```python
keys = ["id", "name", "age"]

student = dict.fromkeys(keys, None)

print(student)
```

### Output

```text
{'id': None, 'name': None, 'age': None}
```

---

# 📊 Summary

| Method | Purpose |
|---------|---------|
| `get()` | Returns the value of a specified key |
| `keys()` | Returns all dictionary keys |
| `values()` | Returns all dictionary values |
| `items()` | Returns all key-value pairs |
| `update()` | Updates the dictionary |
| `pop()` | Removes a key and returns its value |
| `popitem()` | Removes the last inserted key-value pair |
| `setdefault()` | Returns or inserts a key with a default value |
| `copy()` | Creates a shallow copy |
| `clear()` | Removes all key-value pairs |
| `fromkeys()` | Creates a new dictionary from given keys |
