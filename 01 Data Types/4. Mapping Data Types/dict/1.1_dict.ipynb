# 🗂️ Dictionary (`dict`) Data Type

The **Dictionary (`dict`)** is Python's built-in **mapping data type** used to store data as **key-value pairs**. Unlike sequence data types, dictionaries access values using **keys** instead of indexes.

Each key in a dictionary must be **unique**, while values can be duplicated. Dictionaries are **mutable**, meaning you can add, update, or remove key-value pairs after creation.

The `dict` data type is widely used to represent structured data such as student records, employee details, configuration settings, JSON objects, API responses, and database records.

## 1. Introduction to the Dictionary (`dict`) Data Type

Learn what a dictionary is, why Python provides the `dict` data type, and how it stores data as key-value pairs.

```python
student = {
    "name": "Alice",
    "age": 21
}

print(student)
```

## 2. Characteristics of Dictionaries

Explore the important characteristics of the `dict` data type.

### Characteristics

- Stores data as key-value pairs.
- Mutable data type.
- Keys must be unique.
- Values can be duplicated.
- Supports heterogeneous data types.
- Preserves insertion order (Python 3.7+).
- Supports nested dictionaries.

**Example**

```python
data = {
    "id": 101,
    "name": "Python",
    "price": 499.99,
    "available": True
}

print(data)
```

### Output

```text
{'id': 101, 'name': 'Python', 'price': 499.99, 'available': True}
```

---

## 3. Creating Dictionary Objects

Learn different ways to create dictionary objects in Python.

### Topics Covered

- Empty dictionary
- Direct assignment
- Using the `dict()` constructor
- Creating dictionaries from key-value pairs
- Nested dictionaries

**Example**

```python
a = {}
b = {"name": "John", "age": 21}
c = dict(id=101, city="Bangalore")

print(a)
print(b)
print(c)
```

### Output

```text
{}
{'name': 'John', 'age': 21}
{'id': 101, 'city': 'Bangalore'}
```

---

## 4. Accessing Dictionary Elements

Learn how to access values stored in a dictionary.

### Topics Covered

- Using keys
- Using the `get()` method
- Accessing nested dictionaries

**Example**

```python
student = {
    "name": "Alice",
    "age": 21
}

print(student["name"])
print(student.get("age"))
```

### Output

```text
Alice
21
```

---

## 5. Dictionary Operations

Learn the common operations that can be performed on dictionaries.

### Operations Covered

- Accessing values
- Adding key-value pairs
- Updating values
- Removing key-value pairs
- Membership (`in`, `not in`)
- Length (`len()`)

**Example**

```python
student = {
    "name": "John"
}

student["age"] = 22

print(student)
print(len(student))
```

### Output

```text
{'name': 'John', 'age': 22}
2
```

## 6. Built-in Dictionary Methods

Learn the commonly used methods available for dictionaries.

### Methods Covered

- `clear()`
- `copy()`
- `fromkeys()`
- `get()`

- `pop()`

## 10. Limitations of Dictionaries

Understand the limitations of the `dict` data type.

### Limitations

- Keys must be unique.
- Keys must be immutable (hashable).
- Values can be of any data type.
- Cannot access values using indexes.
- Duplicate keys overwrite previous values.

**Example**

```python
student = {
    "name": "John",
    "name": "Alice"
}

print(student)
```

### Output

```text
{'name': 'Alice'}
```
