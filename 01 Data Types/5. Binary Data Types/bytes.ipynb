# 💾 Bytes (`bytes`) Data Type

The **`bytes`** data type is one of Python's built-in **binary data types** used to store a sequence of **bytes** (integers ranging from **0 to 255**).

A `bytes` object is **immutable**, which means once it is created, its contents cannot be modified. It is commonly used when working with binary files, network communication, image processing, encryption, and other low-level programming tasks.

Unlike strings, which store Unicode characters, `bytes` stores raw binary data.

## 2. Characteristics of `bytes`

Explore the important characteristics of the `bytes` data type.

### Characteristics

- Stores binary data.
- Immutable.
- Ordered collection.
- Supports indexing and slicing.
- Stores integer values from **0 to 255**.
- Memory efficient for binary data.
- Frequently used in file handling and networking.

**Example**

```python
data = b"ABC"

print(type(data))
```

### Output

## 3. Creating Bytes Objects

Learn different ways to create `bytes` objects.

### Topics Covered

- Byte literal (`b""`)
- Using the `bytes()` constructor
- From a list of integers
- From a string using encoding

**Example**

```python
a = b"Python"

b = bytes([80, 121, 116, 104, 111, 110])

c = bytes("Hello", "utf-8")

print(a)
print(b)
print(c)
```

### Output

```text
b'Python'
b'Python'
b'Hello'
```
