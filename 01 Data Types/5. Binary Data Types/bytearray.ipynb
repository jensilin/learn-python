# 💾 Bytearray (`bytearray`) Data Type

The **`bytearray`** data type is one of Python's built-in **binary data types** used to store a sequence of **bytes** (integers ranging from **0 to 255**).

Unlike the `bytes` data type, a `bytearray` is **mutable**, meaning its contents can be modified after creation. This makes it suitable for applications where binary data needs to be updated efficiently, such as file manipulation, network communication, image processing, and data buffering.

The `bytearray` data type combines the efficiency of binary data storage with the flexibility of mutable objects.

## 2. Characteristics of `bytearray`

Explore the important characteristics of the `bytearray` data type.

### Characteristics

- Stores binary data.
- Mutable.
- Ordered collection.
- Supports indexing and slicing.
- Stores integer values from **0 to 255**.
- Efficient for modifying binary data.
- Supports most bytes methods.

**Example**

```python
data = bytearray(b"ABC")

print(type(data))
```

### Output

```text
<class 'bytearray'>
```

## 3. Creating Bytearray Objects

Learn different ways to create `bytearray` objects.

### Topics Covered

- From bytes
- Using the `bytearray()` constructor
- From a list of integers
- From a string using encoding

**Example**

```python
a = bytearray(b"Python")

b = bytearray([80, 121, 116, 104, 111, 110])

c = bytearray("Hello", "utf-8")

print(a)
print(b)
print(c)
```

### Output

```text
bytearray(b'Python')
bytearray(b'Python')
bytearray(b'Hello')
```

## 5. Bytearray Operations

Learn the common operations supported by `bytearray`.

### Operations Covered

- Indexing
- Slicing
- Concatenation
- Repetition
- Membership (`in`, `not in`)
- Length (`len()`)
- Updating elements

**Example**

## 10. Limitations of `bytearray`

Understand the limitations of the `bytearray` data type.

### Limitations

- Stores only values from **0 to 255**.
- Uses more memory than immutable `bytes` objects.
- Requires encoding and decoding when working with text.
- Not suitable when immutable binary data is required.
