# 💾 Binary Data Types

The **Binary Data Types** are one of Python's built-in data type categories used to store and manipulate **binary (byte) data**. Unlike strings, which store Unicode characters, binary data types work directly with bytes and are commonly used when handling files, images, audio, videos, network communication, and low-level data processing.

Python provides three built-in binary data types:

- `bytes`
- `bytearray`
- `memoryview`

These data types are designed for efficient storage and manipulation of binary data. Depending on the data type, binary objects can be **mutable** or **immutable**.

# 💻 Quick Example

```python
data1 = b"Python"
data2 = bytearray(b"Python")
data3 = memoryview(data2)

print(data1)
print(data2)
print(data3)

print(type(data1))
print(type(data2))
print(type(data3))
```

### Output

```text
b'Python'
bytearray(b'Python')
<memory at 0x...>
<class 'bytes'>
<class 'bytearray'>
<class 'memoryview'>
```

## 5. Mutable and Immutable Binary Types

Understand which binary data types can be modified after creation.

### Mutable

- `bytearray`

### Immutable

- `bytes`
- `memoryview` (view object; mutability depends on the underlying object)

**Example**

```python
data = bytearray(b"Python")

data[0] = ord("J")

print(data)
```

### Output

```text
bytearray(b'Jython')
```
