# 💾 Memoryview (`memoryview`) Data Type

The **`memoryview`** data type is one of Python's built-in **binary data types** that provides a **memory view** of another binary object without copying its data.

Instead of creating a new copy of the data, a `memoryview` object allows direct access to the memory of objects such as `bytes`, `bytearray`, and other objects that support the **buffer protocol**. This makes it highly efficient when working with large binary data.

The `memoryview` data type is commonly used in file processing, networking, image processing, scientific computing, and applications where performance and memory efficiency are important.

## 2. Characteristics of `memoryview`

Explore the important characteristics of the `memoryview` data type.

### Characteristics

- Provides a view of existing binary data.
- Does not create a copy of the data.
- Memory efficient.
- Supports indexing and slicing.
- Useful for large binary objects.
- Changes made through the view affect the original object (if the underlying object is mutable).

**Example**

```python
data = bytearray(b"ABC")

view = memoryview(data)

print(type(view))
```

### Output

```text
<class 'memoryview'>
```

---

## 3. Creating Memoryview Objects

Learn different ways to create `memoryview` objects.

### Topics Covered

- From `bytes`
- From `bytearray`
- From other buffer-supporting objects

**Example**

```python
a = memoryview(b"Python")

b = memoryview(bytearray(b"Hello"))

print(a)
print(b)
```

## 4. Accessing Memoryview Elements

Learn how to access data using a memoryview object.

### Topics Covered

- Indexing
- Negative indexing
- Slicing

**Example**

```python
data = bytearray(b"Python")

view = memoryview(data)

print(view[0])
print(view[-1])
print(view[1:4])
```

### Output

```text
80
110
<memory at 0x...>
```

---

## 5. Memoryview Operations

Learn the common operations supported by `memoryview`.

### Operations Covered

- Indexing
- Slicing
- Iteration
- Length (`len()`)
- Converting to bytes
- Modifying mutable data

**Example**

```python
data = bytearray(b"Python")

view = memoryview(data)

view[0] = ord("J")

print(data)
```

### Output

```text
bytearray(b'Jython')
```

## 6. Built-in Memoryview Methods

Learn the commonly used methods available for memoryview objects.

### Methods Covered

- `tobytes()`
- `tolist()`
- `hex()`
- `cast()`
- `release()`

**Example**

```python
data = bytearray(b"ABC")

view = memoryview(data)

print(view.tobytes())
print(view.tolist())
```

### Output

```text
b'ABC'
[65, 66, 67]
```

---

## 7. Type Conversion

Learn how to convert a memoryview object into other binary data types.

### Topics Covered

- Memoryview to bytes
- Memoryview to list
- Memoryview to bytearray

**Example**

```python
data = bytearray(b"ABC")

view = memoryview(data)

print(bytes(view))
print(list(view))
print(bytearray(view))
```

### Output

```text
b'ABC'
[65, 66, 67]
bytearray(b'ABC')
```

## 10. Limitations of `memoryview`

Understand the limitations of the `memoryview` data type.

### Limitations

- Works only with objects that support the buffer protocol.
- Cannot be created directly from ordinary Python objects like lists or dictionaries.
- Mutability depends on the underlying object.
- Can be more difficult to understand than `bytes` or `bytearray` for beginners.
