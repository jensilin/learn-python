# 🔤 String (`str`) Data Type

The **String (`str`)** data type is one of Python's built-in **sequence data types**. It is used to represent **textual data** and consists of a sequence of **Unicode characters** enclosed within single quotes (`' '`), double quotes (`" "`), or triple quotes (`''' '''` or `""" """`).

Strings are **immutable**, which means their contents cannot be modified after creation. Python provides a rich collection of built-in functions and methods for performing various string operations such as searching, formatting, splitting, joining, replacing, and case conversion.

Strings are one of the most frequently used data types in Python and play an important role in text processing, data manipulation, file handling, web development, and many other real-world applications.

## 3. Creating String Objects

Learn different ways to create string objects in Python.

### Topics Covered

- Single Quotes

- Triple Quotes
- Using the `str()` constructor

**Example**

```python
a = 'Python'
b = "Programming"
c = """Welcome"""
d = str(100)

print(a)
print(b)
print(c)
print(d)
```

## 5. String Operations

Learn the common operations supported by strings.

### Operations Covered

- Concatenation (`+`)
- Repetition (`*`)
- Membership (`in`, `not in`)
- Comparison
- Length (`len()`)

**Example**

```python
first = "Hello"
second = "Python"

print(first + " " + second)
print(first * 2)
print("P" in second)
```

```text
Hello Python
HelloHello
True
```
