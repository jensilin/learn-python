# 📦 Tuple (`tuple`) Data Type

The **Tuple (`tuple`)** is one of Python's built-in **sequence data types**. It is used to store an **ordered collection of multiple values** in a single variable. A tuple can contain elements of the **same or different data types**, making it useful for storing related data.

Unlike lists, tuples are **immutable**, which means their elements **cannot be modified, added, or removed** after creation. Because of their immutability, tuples are faster than lists and are commonly used for storing fixed data.

Tuples support indexing, slicing, iteration, and many common sequence operations. They are widely used for returning multiple values from functions, storing records, and protecting data from accidental modification.

## 1. Introduction to the Tuple (`tuple`) Data Type

## 2. Characteristics of Tuples

Explore the important characteristics of the `tuple` data type.

### Characteristics

- Ordered collection.
- Immutable data type.
- Allows duplicate values.
- Supports heterogeneous data.
- Supports indexing and slicing.
- Faster than lists.
- Allows nested tuples.

**Example**

```python
data = (10, "Python", 3.14, True)

print(data)
```

### Output

```text
(10, 'Python', 3.14, True)
```

---

## 3. Creating Tuple Objects

Learn different ways to create tuple objects in Python.

### Topics Covered

- Empty tuple
- Single-element tuple
- Multiple-element tuple
- Using the `tuple()` constructor
- Nested tuples

**Example**

```python
a = ()
b = (100,)
c = (10, 20, 30)
d = tuple("Python")

print(a)
print(b)
print(c)
print(d)
```

### Output

```text
()
(100,)
(10, 20, 30)
('P', 'y', 't', 'h', 'o', 'n')
```

---

## 4. Indexing and Slicing

Learn how to access individual elements and sub-tuples using indexes and slices.

**Example**

```python
numbers = (10, 20, 30, 40, 50)

print(numbers[0])
print(numbers[-1])
print(numbers[1:4])
```

### Output

```text
10
50
(20, 30, 40)
```

---

## 5. Tuple Operations

Learn the common operations that can be performed on tuples.

### Operations Covered

- Concatenation (`+`)
- Repetition (`*`)
- Membership (`in`, `not in`)
- Length (`len()`)
- Iteration
- Comparison

**Example**

```python
a = (1, 2)
b = (3, 4)

print(a + b)
print(a * 2)
print(2 in a)
```

### Output

```text
(1, 2, 3, 4)
(1, 2, 1, 2)
True
```

---

## 6. Built-in Tuple Methods

Unlike lists, tuples provide only two built-in methods.

### Methods Covered

- `count()`
- `index()`

**Example**

```python
numbers = (10, 20, 10, 30)

print(numbers.count(10))
print(numbers.index(20))
```

### Output

```text
2
1
```

## 7. Type Conversion

Learn how to convert other iterable objects into tuples.

### Topics Covered

- Using `tuple()`
- String to Tuple
- List to Tuple
- Range to Tuple
- Set to Tuple

**Example**

```python
print(tuple("Python"))
print(tuple([1, 2, 3]))
```

### Output

```text
('P', 'y', 't', 'h', 'o', 'n')
(1, 2, 3)
```
