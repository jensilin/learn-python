# 📋 List Methods

Python provides several built-in **list methods** that allow you to manipulate, update, search, sort, and manage list elements efficiently. These methods simplify common operations such as adding, removing, copying, sorting, and counting elements without writing additional logic.

Since lists are **mutable**, most list methods modify the original list directly instead of creating a new one.

Understanding these methods is essential for writing efficient Python programs and is frequently tested in interviews and coding assessments.

# 1. append()

### Definition

The `append()` method is used to **add a single element to the end of the list**.

### Syntax

```python
list_name.append(element)
```

### Parameters

- `element` → The item to be added.

### Returns

- `None`
- Modifies the original list.

### Example

```python
numbers = [10, 20, 30]

numbers.append(40)

print(numbers)
```

### Output

```text
[10, 20, 30, 40]
```

### Note

- Adds only one element at a time.
- The element is always added at the end.

---

# 2. extend()

### Definition

The `extend()` method is used to add **multiple elements from another iterable** to the end of the list.

### Syntax

```python
list_name.extend(iterable)
```

### Parameters

- `iterable` → List, Tuple, Set, String, Range, etc.

### Returns

- `None`

### Example

```python
numbers = [10, 20]

numbers.extend([30, 40])

print(numbers)
```

### Output

```text
[10, 20, 30, 40]
```

### Note

- Adds every element individually.
- Accepts any iterable.

---

# 3. insert()

### Definition

The `insert()` method inserts an element at a specified position.

### Syntax

```python
list_name.insert(index, element)
```

### Parameters

- `index`
- `element`

### Returns

- `None`

### Example

```python
numbers = [10, 30]

numbers.insert(1, 20)

print(numbers)
```

### Output

```text
[10, 20, 30]
```

### Note

- Existing elements shift to the right.

---

# 4. remove()

### Definition

The `remove()` method removes the **first occurrence** of the specified value.

### Syntax

```python
list_name.remove(value)
```

### Parameters

- Value to remove.

### Returns

- `None`

### Example

```python
numbers = [10, 20, 30]

numbers.remove(20)

print(numbers)
```

### Output

```text
[10, 30]
```

### Note

- Raises `ValueError` if the value is not present.

---

# 5. pop()

### Definition

The `pop()` method removes and returns an element.

### Syntax

```python
list_name.pop(index)
```

### Parameters

- Optional index.

### Returns

- Removed element.

### Example

```python
numbers = [10, 20, 30]

value = numbers.pop()

print(value)
print(numbers)
```

### Output

```text
30
[10, 20]
```

### Note

- Default index is `-1`.

---

# 6. clear()

### Definition

Removes all elements from the list.

### Syntax

```python
list_name.clear()
```

### Example

```python
numbers = [10,20,30]

numbers.clear()

print(numbers)
```

### Output

```text
[]
```

---

# 7. index()

### Definition

Returns the index of the first occurrence of an element.

### Syntax

```python
list_name.index(value)
```

### Example

```python
numbers = [10,20,30]

print(numbers.index(20))
```

### Output

```text
1
```

---

# 8. count()

### Definition

Counts the number of occurrences of an element.

### Syntax

```python
list_name.count(value)
```

### Example

```python
numbers = [10,20,10,40]

print(numbers.count(10))
```

### Output

```text
2
```

---

# 9. sort()

### Definition

Sorts the list in ascending order by default.

### Syntax

```python
list_name.sort()
```

or

```python
list_name.sort(reverse=True)
```

### Example

```python
numbers = [30,10,20]

numbers.sort()

print(numbers)
```

### Output

```text
[10,20,30]
```

---

# 10. reverse()

### Definition

Reverses the order of the list.

### Syntax

```python
list_name.reverse()
```

### Example

```python
numbers = [10,20,30]

numbers.reverse()

print(numbers)
```

### Output

```text
[30,20,10]
```

---

# 11. copy()

### Definition

Creates a shallow copy of the list.

### Syntax

```python
new_list = list_name.copy()
```

### Example

```python
numbers = [10,20,30]

new_numbers = numbers.copy()

print(new_numbers)
```

### Output

```text
[10,20,30]
```
