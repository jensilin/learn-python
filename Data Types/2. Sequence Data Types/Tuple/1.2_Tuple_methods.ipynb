# 📦 Tuple Methods

Python provides only **two built-in methods** for tuples because tuples are **immutable**. Unlike lists, tuples cannot be modified after they are created, so methods such as `append()`, `remove()`, `insert()`, and `sort()` are not available.

This notebook explains the **two built-in tuple methods** with their syntax, examples, and outputs.

## 1. `count()`

The `count()` method returns the **number of times** a specified element appears in the tuple.

### Syntax

```python
tuple_name.count(value)
```

### Example

```python
numbers = (10, 20, 10, 30, 10)

print(numbers.count(10))
```

### Output

```text
3
```

### Note

- Returns an integer.
- Returns `0` if the element is not found.

---

## 2. `index()`

The `index()` method returns the **index position of the first occurrence** of the specified element.

### Syntax

```python
tuple_name.index(value)
```

### Example

```python
numbers = (10, 20, 30, 20)

print(numbers.index(20))
```

### Output

```text
1
```

### Note

- Returns the index of the first matching element.
- Raises a `ValueError` if the element is not present in the tuple.

---

# 📊 Summary

| Method | Purpose |
|---------|---------|
| `count()` | Returns the number of occurrences of an element. |
| `index()` | Returns the index of the first occurrence of an element. |
