# 🔤 Commonly Used String Methods

Python provides more than **45 built-in string methods** for manipulating and processing text. In this notebook, we will focus on the **most commonly used methods** that are frequently used in real-world applications such as text processing, validation, formatting, and data cleaning.

---

## 1. `upper()`

Converts all characters in the string to uppercase.

### Syntax

```python
string.upper()
```

### Example

```python
text = "python"

print(text.upper())
```

### Output

```text
PYTHON
```

---

## 2. `lower()`

Converts all characters in the string to lowercase.

### Syntax

```python
string.lower()
```

### Example

```python
text = "PYTHON"

print(text.lower())
```

### Output

```text
python
```

---

## 3. `title()`

Converts the first character of every word to uppercase.

### Syntax

```python
string.title()
```

### Example

```python
text = "python programming"

print(text.title())
```

### Output

```text
Python Programming
```

---

## 4. `capitalize()`

Converts only the first character of the string to uppercase.

### Syntax

```python
string.capitalize()
```

### Example

```python
text = "python programming"

print(text.capitalize())
```

### Output

```text
Python programming
```

---

## 5. `swapcase()`

Converts uppercase letters to lowercase and lowercase letters to uppercase.

### Syntax

```python
string.swapcase()
```

### Example

```python
text = "Python"

print(text.swapcase())
```

### Output

```text
pYTHON
```

---

## 6. `strip()`

Removes leading and trailing whitespace or specified characters.

### Syntax

```python
string.strip([characters])
```

### Example

```python
text = "   Python   "

print(text.strip())
```

### Output

```text
Python
```

---

## 7. `replace()`

Replaces a specified substring with another substring.

### Syntax

```python
string.replace(old, new)
```

### Example

```python
text = "I like Java"

print(text.replace("Java", "Python"))
```

### Output

```text
I like Python
```

---

## 8. `split()`

Splits a string into a list using a separator.

### Syntax

```python
string.split(separator)
```

### Example

```python
text = "Python Java C"

print(text.split())
```

### Output

```text
['Python', 'Java', 'C']
```

---

## 9. `join()`

Joins elements of an iterable into a single string.

### Syntax

```python
separator.join(iterable)
```

### Example

```python
languages = ["Python", "Java", "C"]

print("-".join(languages))
```

### Output

```text
Python-Java-C
```

---

## 10. `find()`

Returns the index of the first occurrence of a substring.

### Syntax

```python
string.find(substring)
```

### Example

```python
text = "Python Programming"

print(text.find("Pro"))
```

### Output

```text
7
```
