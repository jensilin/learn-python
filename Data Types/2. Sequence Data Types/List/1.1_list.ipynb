# 📋 List (`list`) Data Type

The **List (`list`)** is one of Python's built-in **sequence data types**. It is used to store an **ordered collection of multiple values** in a single variable. A list can contain elements of the **same or different data types**, making it one of the most flexible and widely used data structures in Python.

Lists are **mutable**, which means their elements can be added, removed, or modified after the list has been created. They also allow duplicate values and support indexing, slicing, iteration, and various built-in methods.

Lists are commonly used for storing collections of data, processing information, managing records, and implementing various data structures and algorithms.

## 1. Introduction to the List (`list`) Data Type

## 2. Characteristics of Lists

Explore the important characteristics of the `list` data type.

### Characteristics

- Ordered collection.
- Mutable data type.
- Allows duplicate values.
- Supports heterogeneous data.
- Supports indexing and slicing.
- Dynamic in size.
- Allows nested lists.

**Example**

```python
data = [10, "Python", 3.14, True]

print(data)
```

**Output**

```text
[10, 'Python', 3.14, True]
```

---

## 3. Creating List Objects

Learn different ways to create list objects in Python.

### Topics Covered

- Empty list
- Direct assignment
- Using the `list()` constructor
- Nested lists

**Example**

```python
a = []
b = [1, 2, 3]
c = list("Python")

print(a)
print(b)
print(c)
```

**Output**

```text
[]
[1, 2, 3]
['P', 'y', 't', 'h', 'o', 'n']
```

---

## 4. Indexing and Slicing

Learn how to access individual elements and sublists using indexes and slices.

**Example**

```python
numbers = [10, 20, 30, 40, 50]

print(numbers[0])
print(numbers[-1])
print(numbers[1:4])
```

**Output**

```text
10
50
[20, 30, 40]
```

---

## 5. List Operations

Learn the common operations that can be performed on lists.

### Operations Covered

- Concatenation (`+`)
- Repetition (`*`)
- Membership (`in`, `not in`)
- Length (`len()`)
- Iteration
- Comparison

**Example**

```python
a = [1, 2]
b = [3, 4]

print(a + b)
print(a * 2)
print(2 in a)
```

**Output**

```text
[1, 2, 3, 4]
[1, 2, 1, 2]
True
```

## 7. Type Conversion
