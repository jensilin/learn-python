# 📚 Sequence Data Types

The **Sequence Data Types** are one of Python's built-in data type categories used to store and organize **ordered collections of values**. A sequence maintains the order of its elements, allowing them to be accessed using their **index position**.

Sequence data types support operations such as **indexing**, **slicing**, **iteration**, **concatenation**, **repetition**, and **membership testing**. Depending on the data type, sequences can be **mutable** or **immutable**.

Python provides the following built-in sequence data types:

- `str`
- `list`
- `tuple`
- `range`

These data types are widely used for storing text, collections of objects, fixed data, and sequences of numbers.

## 2. Characteristics of Sequence Data Types

Explore the important characteristics of sequence data types.

## 3. Types of Sequence Data Types

Python provides four built-in sequence data types.

### Types

- String (`str`)
- List (`list`)
- Tuple (`tuple`)
- Range (`range`)

**Example**

```python
text = "Python"
numbers = [1, 2, 3]
colors = ("Red", "Blue")
values = range(5)

print(type(text))
print(type(numbers))
print(type(colors))
print(type(values))
```

## 5. Mutable and Immutable Sequences

Understand which sequence data types can be modified after creation.

### Mutable

- List (`list`)

### Immutable

- String (`str`)

**Example**

```python
numbers = [10, 20]
numbers.append(30)

print(numbers)
```
