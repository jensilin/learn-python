# 🗂️ Mapping Data Types

The **Mapping Data Types** are one of Python's built-in data type categories used to store data in the form of **key-value pairs**. Unlike sequence data types, which access elements using indexes, mapping data types allow values to be accessed using **unique keys**.

Each key in a mapping must be **unique** and is associated with a corresponding value. Mapping data types are widely used to represent structured data, making them ideal for storing records, configurations, JSON data, and many real-world applications.

Python provides the following built-in mapping data type:

- `dict`

## 2. Characteristics of Mapping Data Types

Explore the important characteristics of mapping data types.

### Characteristics

- Stores data as key-value pairs.
- Keys must be unique.
- Values can be duplicated.
- Mutable data type.
- Access values using keys.
- Supports nested dictionaries.
- Preserves insertion order (Python 3.7+).

**Example**

```python
student = {
    "name": "John",
    "age": 20
}

print(type(student))
```

### Output

```text
<class 'dict'>
```
