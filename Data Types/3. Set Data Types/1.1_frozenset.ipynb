# âť„ď¸Ź Frozen Set (`frozenset`) Data Type

The **Frozen Set (`frozenset`)** is one of Python's built-in **set data types**. It is an **immutable version of a set**, meaning that once a `frozenset` is created, its elements **cannot be added, removed, or modified**.

Like a normal set, a `frozenset` stores **unique elements**, does **not allow duplicate values**, and supports mathematical set operations such as union, intersection, difference, and symmetric difference.

Because `frozenset` is immutable, it is **hashable**, which means it can be used as a **dictionary key** or as an element of another set.

## 2. Characteristics of `frozenset`

Explore the important characteristics of the `frozenset` data type.

### Characteristics

- Stores unique elements only.
- Immutable data type.
- Does not allow duplicate values.
- Unordered collection.
- Does not support indexing or slicing.
- Supports mathematical set operations.
- Hashable and can be used as a dictionary key.

**Example**

```python
data = frozenset([10, "Python", 3.14])

print(type(data))
```

### Output

```text
<class 'frozenset'>
```

## 3. Creating `frozenset` Objects

Learn different ways to create `frozenset` objects.

### Topics Covered

- Using the `frozenset()` constructor
- Creating from a list
- Creating from a tuple
- Creating from a set
- Creating from a string

**Example**

```python
a = frozenset([1, 2, 3])
b = frozenset((4, 5, 6))
c = frozenset({7, 8, 9})
d = frozenset("Python")

print(a)
print(b)
print(c)
print(d)
```

## 4. Supported Operations

Learn the common operations supported by `frozenset`.

### Operations Covered

- Union
- Intersection
- Difference
- Symmetric Difference
- Membership (`in`, `not in`)
- Length (`len()`)

**Example**

```python
a = frozenset({1, 2, 3})
b = frozenset({3, 4, 5})

print(a | b)
print(a & b)
print(a - b)
```

### Output

```text
frozenset({1, 2, 3, 4, 5})
frozenset({3})
frozenset({1, 2})
```

## 5. Built-in `frozenset` Methods

`frozenset` supports only methods that **do not modify** the object.

### Methods Covered

- `copy()`
- `union()`
- `intersection()`
- `difference()`
- `symmetric_difference()`
- `issubset()`
- `issuperset()`
- `isdisjoint()`

**Example**

```python
a = frozenset({1, 2, 3})
b = frozenset({3, 4})

print(a.union(b))
```

### Output

```text
frozenset({1, 2, 3, 4})
```
