# 🔢 Integer (`int`) Data Type

The **Integer (`int`)** data type is one of Python's fundamental numeric data types. It is used to represent **whole numbers** without any decimal or fractional part. Integer values can be **positive**, **negative**, or **zero**, making them suitable for a wide variety of mathematical and logical operations.

Python integers are **immutable**, meaning their value cannot be modified after creation. Unlike many programming languages, Python supports integers of **unlimited precision**, allowing you to work with very large numbers without worrying about overflow (limited only by the available system memory).

Integers are commonly used in counting, indexing, mathematical computations, loop control, and many real-world applications where decimal values are not required.

## 2. Characteristics of Integers

Explore the important characteristics of the `int` data type.

### Characteristics

- Stores whole numbers only.
- Supports positive, negative, and zero values.
- Immutable data type.
- Unlimited precision (limited only by available memory).
