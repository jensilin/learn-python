# 🌊 Floating Point (`float`) Data Type

The **Floating Point (`float`)** data type is one of Python's built-in numeric data types. It is used to represent **decimal (fractional) numbers**. A float can store both positive and negative decimal values, making it suitable for mathematical calculations that require precision.

Python stores floating-point numbers using the **IEEE 754 double-precision format**, which provides approximately **15 to 17 significant decimal digits** of precision. Float values are immutable, meaning their value cannot be modified after creation.

Floating-point numbers are commonly used in scientific calculations, measurements, financial applications, statistical analysis, and any scenario where decimal values are required.
