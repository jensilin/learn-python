# ⏭️ Continue Statement

The **`continue` statement** is a jump statement in Python used to **skip the current iteration of a loop** and continue with the next iteration.

Unlike the `break` statement, which terminates the loop completely, the `continue` statement **does not stop the loop**. Instead, it skips the remaining code in the current iteration and moves directly to the next iteration.

The `continue` statement can be used with both **`for`** and **`while`** loops.

# 📖 Topics Covered

## 1. Introduction to the `continue` Statement

The `continue` statement skips the remaining statements in the current loop iteration and immediately starts the next iteration.

When `continue` executes:

- The current iteration ends.
- Remaining statements inside the loop are skipped.
- The loop continues with the next iteration.

**Example**

```python
for number in range(5):

    if number == 2:
        continue

    print(number)
```

### Output

```text
0
1
3
4
```

---

## 2. Syntax of the `continue` Statement

### Syntax

```python
continue
```

### Explanation

- `continue` is a Python keyword.
- It can only be used inside loops.
- It skips only the current iteration.
- The loop continues normally.

---

## 3. Flow of Execution

The execution of a `continue` statement follows these steps:

1. Execute the loop.
2. Check the condition.
3. If the condition becomes `True`, execute `continue`.
4. Skip the remaining statements of the current iteration.
5. Start the next iteration.

**Example**

```python
for i in range(1, 6):

    if i == 4:
        continue

    print(i)

print("Loop Finished")
```

### Output

```text
1
2
3
5
Loop Finished
```

---

## 4. Using `continue` with a `for` Loop

The `continue` statement skips selected iterations in a `for` loop.

**Example**

```python
fruits = ["Apple", "Banana", "Mango", "Orange"]

for fruit in fruits:

    if fruit == "Mango":
        continue

    print(fruit)
```

### Output

```text
Apple
Banana
Orange
```

---

## 5. Using `continue` with a `while` Loop

The `continue` statement can also be used inside a `while` loop.

**Example**

```python
count = 0

while count < 5:

    count += 1

    if count == 3:
        continue

    print(count)
```

### Output

```text
1
2
4
5
```

> **Note:** In a `while` loop, always update the loop variable **before** executing `continue` (or ensure it is updated on every iteration). Otherwise, the loop may become infinite.

## 8. Common Mistakes

Avoid these common mistakes when using the `continue` statement.

### Mistakes

- Using `continue` outside a loop.
- Forgetting to update the loop variable in a `while` loop.
- Using `continue` when `break` is actually required.
- Skipping important statements unintentionally.

**Incorrect**

```python
count = 1

while count <= 5:

    if count == 3:
        continue

    print(count)
    count += 1
```

### Output

```text
1
2
...
```

> **Why?** When `count` becomes `3`, `continue` skips `count += 1`, causing the condition to remain `True` forever and creating an infinite loop.

---

## 9. Best Practices

Follow these recommended practices when using the `continue` statement.

- Use `continue` only when skipping an iteration improves readability.
- Ensure loop variables are updated correctly.
- Keep `continue` conditions simple and meaningful.
- Avoid excessive use, as it can make loops harder to understand.
- Comment complex conditions when necessary.
