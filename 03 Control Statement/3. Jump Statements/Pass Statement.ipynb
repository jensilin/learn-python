# ⏸️ Pass Statement

The **`pass` statement** is a jump statement in Python used as a **placeholder**. It does **nothing** when executed but allows you to write syntactically correct code where a statement is required.

The `pass` statement is commonly used when:

- A block of code is not yet implemented.
- Creating empty functions or classes.
- Writing placeholder loops or conditional statements.
- Building program structure before adding logic.

# 📖 Topics Covered

## 1. Introduction to the `pass` Statement

The `pass` statement is a **null operation**.

When Python encounters `pass`, it simply ignores it and continues executing the rest of the program.

It is useful when the syntax requires a statement, but no action needs to be performed.

**Example**

```python
if True:
    pass

print("Hello Python")
```

### Output

```text
Hello Python
```

---

## 2. Syntax of the `pass` Statement

### Syntax

```python
pass
```

### Explanation

- `pass` is a Python keyword.
- It performs no operation.
- It is used wherever a statement is syntactically required.

---

## 3. Flow of Execution

The execution of a `pass` statement follows these steps:

1. Python encounters the `pass` statement.
2. No operation is performed.
3. Execution immediately continues with the next statement.

**Example**

```python
age = 20

if age >= 18:
    pass

print("Eligible")
```

### Output

```text
Eligible
```

## 5. Using `pass` with Loops

## 10. Common Mistakes

Avoid these common mistakes when using the `pass` statement.

### Mistakes

- Assuming `pass` skips an iteration like `continue`.
- Assuming `pass` exits a loop like `break`.
- Leaving `pass` statements in completed production code.
- Using `pass` where actual logic is required.

**Incorrect Understanding**

```python
for i in range(5):

    if i == 2:
        pass

    print(i)
```

### Output

```text
0
1
2
3
4
```

> **Why?** The `pass` statement does **not** skip the iteration. It simply does nothing, so the loop continues normally.

---

## 11. `pass` vs `break` vs `continue`

| Statement | Purpose |
|-----------|---------|
| `pass` | Does nothing (placeholder) |
| `break` | Exits the loop immediately |
| `continue` | Skips the current iteration and continues with the next |

**Example**

```python
for i in range(5):

    if i == 1:
        pass

    if i == 3:
        continue

    if i == 4:
        break

    print(i)
```

### Output

```text
0
1
2
```

## 12. Best Practices

Follow these recommended practices when using the `pass` statement.

- Use `pass` only as a temporary placeholder.
- Replace `pass` with actual implementation when the code is ready.
- Use descriptive comments to indicate future implementation.
- Avoid leaving unnecessary `pass` statements in production code.
- Use `pass` to improve code organization during development.
