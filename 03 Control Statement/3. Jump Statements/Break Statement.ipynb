# ⛔ Break Statement

The **`break` statement** is a jump statement in Python used to **immediately terminate the execution of a loop**. When a `break` statement is encountered, the loop stops, and the program continues executing the statement immediately after the loop.

The `break` statement can be used with both **`for`** and **`while`** loops. It is commonly used when a specific condition is met and there is no need to continue iterating.

# 📖 Topics Covered

## 1. Introduction to the `break` Statement

The `break` statement immediately terminates the nearest enclosing loop.

Once `break` executes:

- The current loop ends.
- Remaining iterations are skipped.
- Program execution continues after the loop.

**Example**

```python
for number in range(5):

    if number == 2:
        break

    print(number)
```

### Output

```text
0
1
```

---

## 2. Syntax of the `break` Statement

### Syntax

```python
break
```

### Explanation

- `break` is a Python keyword.
- It can only be used inside loops.
- It immediately exits the current loop.

---

## 3. Flow of Execution

The execution of a `break` statement follows these steps:

1. Execute the loop.
2. Check the condition.
3. If the condition becomes `True`, execute `break`.
4. Exit the loop immediately.
5. Continue with the next statement after the loop.

**Example**

```python
for i in range(1, 6):

    if i == 3:
        break

    print(i)

print("Loop Finished")
```

### Output

```text
1
2
Loop Finished
```

---

## 4. Using `break` with a `for` Loop

The `break` statement is commonly used to stop a `for` loop early.

**Example**

```python
fruits = ["Apple", "Banana", "Mango", "Orange"]

for fruit in fruits:

    if fruit == "Mango":
        break

    print(fruit)
```

### Output

```text
Apple
Banana
```

---

## 5. Using `break` with a `while` Loop

The `break` statement can also terminate a `while` loop.

**Example**

```python
count = 1

while True:

    print(count)

    if count == 3:
        break

    count += 1
```

### Output

```text
1
2
3
```

## 8. Common Mistakes

Avoid these common mistakes when using the `break` statement.

### Mistakes

- Using `break` outside a loop.
- Expecting `break` to terminate multiple nested loops.
- Forgetting that `break` exits only the nearest loop.

**Incorrect**

```python
break
```

### Output

```text
SyntaxError: 'break' outside loop
```

---

## 9. Best Practices

Follow these recommended practices when using the `break` statement.

- Use `break` only when early loop termination is required.
- Avoid excessive use of `break`, as it can reduce code readability.
- Keep the condition triggering `break` clear and meaningful.
- Prefer meaningful variable names.
- Comment complex `break` conditions if necessary.

#### Question 3: Search for a Name

#### Problem Statement
Given the following list:

```python
names = ["Alice", "Bob", "Charlie", "David", "Emma"]
```

Ask the user to enter a name to search.

If the name is found, display **"Name Found"** and stop searching using the **`break`** statement.

If the name does not exist, display **"Name Not Found"** after the loop completes.

#### Example
```text
Enter name: Charlie
Name Found
```

````python
names = ["Alice", "Bob", "Charlie", "David", "Emma"]
while True:
    user=input('Enter the Name: ')
    if user in names:
        print(f'Name found {user}')
        break
    else: print(f'Not found {user}')
````

#### Question 4: Guess the Secret Number

#### Problem Statement
Create a program where the secret number is **25**.

Ask the user to guess the number repeatedly.

If the guessed number is correct, display **"Congratulations! You guessed it."** and terminate the loop using the **`break`** statement.

#### Example
```text
Guess the number: 10
Wrong Guess!
```

````python
secret_number = 25

while True:
    a = int(input("Guess the number: "))
    if a == secret_number:
        print("Congratulations! You guessed it.")
        break
    else: print("Wrong Guess!")
````

#### Question: Find the First Duplicate Element

#### Problem Statement
Given the following list:

```python
numbers = [12, 45, 67, 89, 45, 23, 90, 12]
```

Write a Python program to find the **first duplicate element** in the list.

As soon as the first duplicate is found, display it and terminate the loop using the **`break`** statement.

#### Example Output
```text
First Duplicate Element: 45
```

````python
numbers = [12, 45, 67, 89, 45, 23, 90, 12]
dup = []

for i in numbers:
    if i not in dup:
        dup+=[i]
    else:
        print(f'First Duplicate Element: {i}')
        break
````
