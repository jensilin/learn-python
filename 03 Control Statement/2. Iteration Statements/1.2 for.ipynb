```python
#WAP to check whether the entered number is prime number or not

num = 1
for i in range(2,num):
    if i%num==0 or num==1:
        print(f'{num} is not a prime number.')
        break
else:
    print(f'{num} is a prime number.')
```

```python
#st='Abc8#*$zbE'
#B,c,no,sp=0,0,0,0
#upr,lwr,num,spc='','','',''

#{'AE': 2, 'bczb': 4, '8': 1, '#*$': 3}

st = "Abc8#*$zbE"
uc,lc,num,sp='','','',''
ucc,lcc,numc,spc=0,0,0,0
out = {}

for i in st:
    if i.isupper():
        uc+=i
        ucc+=1
    elif i.islower():
        lc+=i
        lcc+=1
    elif i.isnumeric():
        num+=i
        numc+=1
    else:
        sp+=i
        spc+=1
out[uc]=ucc
out[lc]=lcc
out[num]=numc
out[sp]=spc
print(out)
```
