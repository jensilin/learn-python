# 🔁 For Loop

The **`for` loop** is an iteration statement in Python that is used to execute a block of code **for each item in a sequence or iterable**. It is commonly used when the number of iterations is known or when iterating over collections such as strings, lists, tuples, dictionaries, sets, and ranges.

Unlike the `while` loop, which repeats based on a condition, the `for` loop automatically iterates through each element in an iterable until all elements have been processed.

# 📖 Topics Covered

## 1. Introduction to the `for` Loop

The `for` loop is used to iterate through every item in an iterable.

It automatically retrieves one element at a time until all elements have been processed.

**Example**

```python
fruits = ["Apple", "Banana", "Mango"]

for fruit in fruits:
    print(fruit)
```

### Output

```text
Apple
Banana
Mango
```

---

## 2. Syntax of the `for` Loop

### Syntax

```python
for variable in iterable:
    # Code to execute
```

### Explanation

- `for` is the Python keyword.
- `variable` stores the current element.
- `iterable` can be a list, tuple, string, dictionary, set, or range.
- Indentation defines the loop body.

---

## 3. Flow of Execution

The execution of a `for` loop follows these steps:

1. Get the first element from the iterable.
2. Execute the loop body.
3. Move to the next element.
4. Repeat until all elements are processed.
5. Continue with the next statement after the loop.

**Example**

```python
for number in [10, 20, 30]:
    print(number)

print("Loop Finished")
```

### Output

```text
10
20
30
Loop Finished
```

## 5. Iterating Over Different Data Types

A `for` loop can iterate over various iterable objects.

### Output

```text

h
o

```

#### List


### Output

```text

20
30
```

#### Tuple

```python
colors = ("Red", "Green", "Blue")

for color in colors:
    print(color)
```

### Output

```text
Red
Green
Blue
```

#### Dictionary

```python
student = {
    "name": "John",
    "age": 20
}

for key in student:
    print(key, student[key])
```

### Output

```text
name John
age 20
```

---

## 6. Nested `for` Loops

A `for` loop can be placed inside another `for` loop.

**Example**

```python
for i in range(1, 3):
    for j in range(1, 4):
        print(i, j)
```

### Output

```text
1 1
1 2
1 3
2 1
2 2
2 3
```

---

## 7. `for...else` Statement

The `else` block executes after the loop completes normally (without encountering a `break` statement).

**Example**

```python
for i in range(3):
    print(i)
else:
    print("Loop Completed")
```

### Output

```text
0
1
2
Loop Completed
```

## 10. Common Mistakes

Avoid these common mistakes when using the `for` loop.

### Mistakes

- Forgetting the colon (`:`).
- Incorrect indentation.
- Modifying the iterable while iterating over it.
- Using incorrect `range()` arguments.

**Incorrect**

```python
for i in range(5)
    print(i)
```

### Output

```text
SyntaxError
```

---

## 11. Best Practices

Follow these recommended practices when using the `for` loop.

- Use meaningful loop variable names.
- Prefer `for` when the number of iterations is known.
- Use `range()` appropriately.
- Avoid unnecessary nested loops.
- Keep the loop body simple and readable.
