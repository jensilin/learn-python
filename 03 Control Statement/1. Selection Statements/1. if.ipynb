# 🔀 If Statement

The **`if` statement** is the simplest decision-making statement in Python. It allows a program to execute a block of code **only when a specified condition is `True`**. If the condition is `False`, the block of code is skipped.

The `if` statement is widely used to make decisions based on user input, comparisons, logical conditions, and program state.

# 📖 Topics Covered

## 1. Introduction to the `if` Statement

The `if` statement executes a block of code **only if** the specified condition evaluates to `True`.

If the condition evaluates to `False`, Python skips the block and continues with the next statement.

**Example**

```python
number = 10

if number > 0:
    print("Positive Number")
```

### Output

```text
Positive Number
```

---

## 2. Syntax of the `if` Statement

### Syntax

```python
if condition:
    # Code to execute
```

### Explanation

- `if` is a Python keyword.
- `condition` is any expression that evaluates to `True` or `False`.
- A colon (`:`) marks the beginning of the code block.
- Indentation defines the statements that belong to the `if` block.

---

## 3. Flow of Execution

The execution of an `if` statement follows these steps:

1. Evaluate the condition.
2. If the condition is `True`, execute the indented block.
3. If the condition is `False`, skip the block.
4. Continue with the next statement.

**Example**

```python
marks = 80

if marks >= 50:
    print("Pass")

print("Program Finished")
```

### Output

```text
Pass
Program Finished
```

---

## 4. Using Comparison Operators

The `if` statement is commonly used with comparison operators.

### Operators

- `==`
- `!=`
- `>`
- `<`
- `>=`
- `<=`

**Example**

```python
age = 18

if age == 18:
    print("You are exactly 18 years old.")
```

### Output

```text
You are exactly 18 years old.
```

---

## 5. Using Logical Operators

Multiple conditions can be combined using logical operators.

### Operators

- `and`
- `or`
- `not`

**Example**

```python
age = 22
citizen = True

if age >= 18 and citizen:
    print("Eligible to vote")
```

### Output

```text
Eligible to vote
```

---

## 6. Nested `if` Statements

An `if` statement can be placed inside another `if` statement.

**Example**

```python
age = 20
license_available = True

if age >= 18:
    if license_available:
        print("Eligible to drive")
```

### Output

```text
Eligible to drive
```

## 9. Common Mistakes

Avoid these common mistakes when using the `if` statement.

### Mistakes

- Forgetting the colon (`:`).
- Incorrect indentation.
- Using `=` instead of `==` for comparison.
- Writing conditions that never become `True`.

**Incorrect**

```python
age = 20

if age = 20:
    print(age)
```

### Output

```text
SyntaxError
```

Follow these recommended practices when using the `if` statement.

- Keep the code inside the `if` block concise.
- Avoid unnecessary nesting.
- Use descriptive variable names.

<h4>Membership Operators</h4>

<p>
Membership operators are used to check whether a value is <b>present</b> or <b>not present</b> in a sequence or collection. They return the result in the form of <code>True</code> or <code>False</code>.
</p>

<p>
Python provides two membership operators:
</p>

1. <code>in</code> Operator
2. <code>not in</code> Operator

---

<h4>a) <code>in</code> Operator</h4>

<p>
The <code>in</code> operator is used to check whether a specified value is <b>present</b> in a sequence or collection. It returns <code>True</code> if the value is found; otherwise, it returns <code>False</code>.
</p>

<h5>Syntax</h5>

```python
value in collection
```

<h5>Example</h5>

```python
print("a" in "Python")
print(3 in [1, 2, 3, 4])
print("apple" in ("apple", "banana", "orange"))
```

<h5>Output</h5>

```text
False
True
True
```

<h5>Note</h5>

<ul>
    <li>The <code>in</code> operator can be used with <code>str</code>, <code>list</code>, <code>tuple</code>, <code>set</code>, <code>dict</code>, and other iterable objects.</li>
    <li>For dictionaries, the <code>in</code> operator checks only the <b>keys</b>, not the values.</li>
</ul>

---

<h4>b) <code>not in</code> Operator</h4>

<p>
The <code>not in</code> operator is used to check whether a specified value is <b>not present</b> in a sequence or collection. It returns <code>True</code> if the value is not found; otherwise, it returns <code>False</code>.
</p>

<h5>Syntax</h5>

```python
value not in collection
```

<h5>Example</h5>

```python
print("z" not in "Python")
print(10 not in [1, 2, 3, 4])
print("grape" not in ("apple", "banana", "orange"))
```

<h5>Output</h5>

```text
True
True
True
```

<h5>Note</h5>

<ul>
    <li>The <code>not in</code> operator is the opposite of the <code>in</code> operator.</li>
    <li>For dictionaries, the <code>not in</code> operator checks only the <b>keys</b>.</li>
</ul>

````python
#In operator example
print('i' in 'Jensilin')
````

````python
#WAP to Check Whether a Character is a Vowel
ch = input('Enter the character : ')

if ch.lower() in 'aeiou':
    print(f'{ch} is a Vowel')
````

````python
#Not in operator example

print('u' not in 'Jensilin')
````

````python
#WAP to Check Whether a Character is not a Vowel

ch = input('Enter the character : ')

if ch.lower() not in 'aeiou':
    print(f'{ch} is not a Vowel')
````
