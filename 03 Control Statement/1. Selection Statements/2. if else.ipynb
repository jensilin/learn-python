# 🔀 If Else Statement

The **`if...else` statement** is a decision-making statement in Python that allows a program to execute **one block of code if a condition is `True`** and **another block of code if the condition is `False`**.

Unlike the `if` statement, which executes code only when the condition is true, the `if...else` statement ensures that **exactly one of the two code blocks is executed**.

It is commonly used when a program needs to choose between two possible outcomes.

# 📖 Topics Covered

## 1. Introduction to the `if...else` Statement

The `if...else` statement evaluates a condition and executes one of two code blocks.

- If the condition is `True`, the `if` block executes.
- If the condition is `False`, the `else` block executes.

Only one block executes.

**Example**

```python
number = -5

if number >= 0:
    print("Positive Number")
else:
    print("Negative Number")
```

### Output

```text
Negative Number
```

---

## 2. Syntax of the `if...else` Statement

### Syntax

```python
if condition:
    # Code if condition is True
else:
    # Code if condition is False
```

### Explanation

- `if` checks the condition.
- `else` executes when the condition is `False`.
- A colon (`:`) starts each code block.
- Indentation defines the statements inside each block.

---

## 3. Flow of Execution

The execution of an `if...else` statement follows these steps:

1. Evaluate the condition.
2. If the condition is `True`, execute the `if` block.
3. Otherwise, execute the `else` block.
4. Continue with the next statement.

**Example**

```python
marks = 45

if marks >= 50:
    print("Pass")
else:
    print("Fail")

print("Result Generated")
```

### Output

```text
Fail
Result Generated
```

---

## 5. Using Logical Operators

Multiple conditions can be combined using logical operators.

### Operators

- `and`
- `or`
- `not`

**Example**

```python

if username == "admin" and password == "python":
    print("Login Successful")
else:
    print("Invalid Credentials")
```

### Output

```text

```

---

## 6. Nested `if...else` Statements

An `if...else` statement can be placed inside another `if` or `else` block.

**Example**

```python
age = 20
has_license = True

if age >= 18:
    if has_license:
        print("Can drive")
    else:
        print("License required")
else:
    print("Too young to drive")
```

### Output

```text
Can drive
```

## 9. Common Mistakes

Avoid these common mistakes when using the `if...else` statement.

### Mistakes

- Forgetting the colon (`:`).
- Incorrect indentation.
- Using `=` instead of `==`.
- Writing unnecessary `else` blocks.
- Misaligning the `else` block.

**Incorrect**

```python
age = 20

if age = 20:
    print("Adult")
else:
    print("Minor")
```

### Output

```text
SyntaxError
```
