# 🔀 If Elif Else Statement

The **`if...elif...else` statement** is a multi-way decision-making statement in Python. It allows a program to evaluate **multiple conditions** and execute the block of code associated with the **first condition that evaluates to `True`**.

Unlike the `if...else` statement, which chooses between only two outcomes, the `if...elif...else` statement can handle **multiple possible conditions**, making it ideal for decision-making involving several choices.

# 📖 Topics Covered

## 1. Introduction to the `if...elif...else` Statement

The `if...elif...else` statement evaluates multiple conditions one after another.

- If the first condition is `True`, its block executes.
- Otherwise, Python checks the next `elif` condition.
- This process continues until a condition becomes `True`.
- If none of the conditions are `True`, the `else` block executes.

Only **one** block of code is executed.

**Example**

```python
temperature = 35

if temperature >= 40:
    print("Very Hot")
elif temperature >= 30:
    print("Hot")
else:
    print("Pleasant")
```

### Output

```text
Hot
```

---

## 2. Syntax of the `if...elif...else` Statement

### Syntax

```python
if condition1:
    # Code block
elif condition2:
    # Code block
elif condition3:
    # Code block
else:
    # Code block
```

### Explanation

- `if` checks the first condition.
- `elif` checks additional conditions if previous ones are `False`.
- `else` executes when all conditions are `False`.
- Indentation defines each code block.

---

## 3. Flow of Execution

The execution of an `if...elif...else` statement follows these steps:

1. Evaluate the `if` condition.
2. If it is `True`, execute the corresponding block.
3. Otherwise, evaluate the first `elif` condition.
4. Continue checking each `elif` until one becomes `True`.
5. If none are `True`, execute the `else` block.
6. Continue with the next statement.

**Example**

```python
number = 0

if number > 0:
    print("Positive")
elif number < 0:
    print("Negative")
else:
    print("Zero")
```

### Output

```text
Zero
```

Multiple conditions can be combined using logical operators.

```text

```

## 6. Nested `if...elif...else` Statements

An `if...elif...else` statement can be placed inside another conditional block.

**Example**

```python
age = 22
citizen = True

if age >= 18:
    if citizen:
        print("Eligible to vote")
    else:
        print("Citizenship required")
else:
    print("Not eligible")
```

### Output

```text
Eligible to vote
```

## 9. Common Mistakes

Avoid these common mistakes when using the `if...elif...else` statement.

### Mistakes

- Using `if` instead of `elif` for related conditions.
- Forgetting the colon (`:`).
- Incorrect indentation.
- Placing conditions in the wrong order.
- Writing unreachable conditions.

**Incorrect**

```python
marks = 95

if marks >= 50:
    print("Pass")
elif marks >= 90:
    print("Excellent")
```

### Output

```text
Pass
```

> **Why?** The second condition is never reached because `95 >= 50` is already `True`.

---

## 10. Best Practices

Follow these recommended practices when using the `if...elif...else` statement.

- Arrange conditions from most specific to most general (or highest to lowest, depending on the logic).
- Avoid unnecessary nesting.
- Keep conditions simple and readable.
- Use meaningful variable names.
- Use `elif` instead of multiple `if` statements when only one condition should execute.
