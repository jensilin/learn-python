# 🎮 Introduction to Control Statements

Control Statements are one of the fundamental building blocks of programming. They determine the **flow of execution** of a program by allowing it to make decisions, repeat tasks, or alter the normal sequence of execution.

By default, Python executes statements one after another in a **sequential order**. However, most real-world applications require programs to make decisions, execute code repeatedly, or skip certain statements based on specific conditions. Control statements make this possible.

Python provides three major categories of control statements:

- **Selection Statements**
- **Iteration Statements**
- **Jump Statements**

# 📖 Topics Covered

## 1. What are Control Statements?

Control statements determine the order in which statements are executed inside a program.

They help programs:

- Make decisions
- Repeat tasks
- Skip unnecessary operations
- Stop or continue execution based on conditions

Without control statements, every program would simply execute one statement after another.

---

## 2. Why are Control Statements Important?

Control statements make programs dynamic and intelligent.

They allow a program to:

- Respond to user input
- Perform repetitive tasks automatically
- Execute different code for different situations
- Improve code efficiency
- Reduce duplicate code

---

## 3. Types of Control Statements

Python provides three categories of control statements.

### 1. Selection Statements

Selection statements execute different blocks of code depending on whether a condition is **True** or **False**.

Statements included:

- `if`
- `if...else`
- `if...elif...else`

**Example**

```python
marks = 75

if marks >= 50:
    print("Pass")
```

### Output

```text
Pass
```

---

### 2. Iteration Statements

Iteration statements repeatedly execute a block of code until a condition becomes false or for every element in a sequence.

Statements included:

- `while`
- `for`

**Example**

```python
for i in range(1, 4):
    print(i)
```

### Output

```text
1
2
3
```

---

### 3. Jump Statements

Jump statements modify the normal flow of loops.

Statements included:

- `break`
- `continue`
- `pass`

**Example**

```python
for i in range(5):

    if i == 3:
        break

    print(i)
```

### Output

```text
0
1
2
```

---

## 4. Control Statement Categories

| Category | Purpose | Statements |
|-----------|---------|------------|
| Selection Statements | Decision Making | `if`, `if...else`, `if...elif...else` |
| Iteration Statements | Repeating Tasks | `while`, `for` |
| Jump Statements | Alter Loop Execution | `break`, `continue`, `pass` |

## 6. Recommended Learning Order

To understand control statements effectively, study them in the following order:

1. Selection Statements
   - If Statement
   - If Else Statement
   - If Elif Else Statement

2. Iteration Statements
   - While Loop
   - For Loop

3. Jump Statements
   - Break Statement
   - Continue Statement
   - Pass Statement

---

## 7. Best Practices

When working with control statements:

- Write simple and readable conditions.
- Avoid deeply nested statements whenever possible.
- Choose the appropriate loop for the problem.
- Use `break` and `continue` only when necessary.
- Use `pass` as a temporary placeholder.
- Keep indentation consistent, as Python uses indentation to define code blocks.
