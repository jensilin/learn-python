<h4>Relational (Comparison) Operators</h4>

<p>
Relational operators, also known as <b>comparison operators</b>, are used to compare two values or expressions. They return the result in the form of a Boolean value: <code>True</code> or <code>False</code>.
</p>

<h5>Types of Relational Operators</h5>

<ul>
    <li><code>==</code> &ndash; Equal to</li>
    <li><code>!=</code> &ndash; Not equal to</li>
    <li><code>&gt;</code> &ndash; Greater than</li>
    <li><code>&lt;</code> &ndash; Less than</li>
    <li><code>&gt;=</code> &ndash; Greater than or equal to</li>
    <li><code>&lt;=</code> &ndash; Less than or equal to</li>
</ul>

<h4>a) Equal To (<code>==</code>) Operator</h4>

<p>
The <code>==</code> operator is used to check whether <b>both operands are equal</b>. It returns <code>True</code> if both operands have the same value; otherwise, it returns <code>False</code>.
</p>

<h5>Syntax</h5>

```python
operand1 == operand2
```

<h5>Example</h5>

```python
print(10 == 10)
print(15 == 20)
print("Python" == "Python")
```

<h5>Output</h5>

```text
True
False
True
```

<h5>Note</h5>

<ul>
    <li>The <code>==</code> operator compares the <b>values</b> of two operands.</li>
    <li>It returns <code>True</code> only if both operands are exactly equal.</li>
</ul>

````python
10 == 10
````

````python
'CSK' == 'RCB'
````

<h4>b) Not Equal To (<code>!=</code>) Operator</h4>

<p>
The <code>!=</code> operator is used to check whether <b>two operands are not equal</b>. It returns <code>True</code> if the operands have different values; otherwise, it returns <code>False</code>.
</p>

<h5>Syntax</h5>

```python
operand1 != operand2
```

<h5>Example</h5>

```python
print(10 != 20)
print(15 != 15)
print("Python" != "Java")
```

<h5>Output</h5>

```text
True
False
True
```

<h5>Note</h5>

<ul>
    <li>The <code>!=</code> operator compares the <b>values</b> of two operands.</li>
    <li>It returns <code>True</code> only when the operands have different values.</li>
</ul>

````python
1 != 1
````

````python
'hehe' != 100
````

<h4>c) Greater Than (<code>&gt;</code>) Operator</h4>

<p>
The <code>&gt;</code> operator is used to check whether the <b>first operand is greater than the second operand</b>. It returns <code>True</code> if the condition is satisfied; otherwise, it returns <code>False</code>.
</p>

<h5>Syntax</h5>

```python
operand1 > operand2
```

<h5>Example</h5>

```python
print(20 > 10)
print(5 > 15)
```

<h5>Output</h5>

```text
True
False
```

<h5>String Comparison</h5>

<p>
When comparing strings, Python compares the <b>Unicode (ASCII for standard English characters)</b> value of each character from left to right.
</p>

<table>
    <tr>
        <th>Character</th>
        <th>Value</th>
        <th>Character</th>
        <th>Value</th>
    </tr>
    <tr>
        <td><code>A</code></td>
        <td>65</td>
        <td><code>a</code></td>
        <td>97</td>
    </tr>
    <tr>
        <td><code>B</code></td>
        <td>66</td>
        <td><code>b</code></td>
        <td>98</td>
    </tr>
    <tr>
        <td><code>0</code></td>
        <td>48</td>
        <td><code>1</code></td>
        <td>49</td>
    </tr>
</table>

<h5>Example</h5>

```python
val1 = "a"
val2 = "A"

print(val1 > val2)
```

<h5>Output</h5>

```text
True
```

<h5>How String Comparison Works</h5>

<ul>
    <li>Python compares the characters of both strings from <b>left to right</b>.</li>
    <li>If the first pair of characters are different, their Unicode values are compared.</li>
    <li>If the first pair of characters are equal, Python compares the next pair of characters.</li>
    <li>The comparison continues until a difference is found or both strings are completely compared.</li>
    <li>If both strings are identical, the expression returns <code>False</code>.</li>
</ul>

<h5>Note</h5>

<ul>
    <li>The <code>&gt;</code> operator cannot be used to compare incompatible data types (for example, <code>int</code> and <code>str</code>).</li>
    <li>It is not supported for <code>complex</code> numbers.</li>
    <li>It cannot be used with <code>dict</code> data types.</li>
</ul>

````python
'Python' > 'Java'
````

````python
10 > 15
````

````python
22 >= 22
````

<h4><code>ord()</code> Function</h4>

<p>
The <code>ord()</code> function is used to return the <b>Unicode code point</b> of a given character. For standard English characters, this corresponds to the <b>ASCII value</b>.
</p>

<h5>Syntax</h5>

```python
ord(character)
```

<h5>Example</h5>

```python
print(ord('A'))
print(ord('a'))
print(ord('0'))
```

<h5>Output</h5>

```text
65
97
48
```

<h5>Note</h5>

<ul>
    <li>The <code>ord()</code> function accepts <b>exactly one character</b> as its argument.</li>
    <li>Passing more than one character raises a <code>TypeError</code>.</li>
    <li>For standard English characters, the returned Unicode code point is the same as the ASCII value.</li>
</ul>

````python
ord('C')
````

````python
ord('&')
````
