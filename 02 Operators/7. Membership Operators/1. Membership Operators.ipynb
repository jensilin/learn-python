# 📋 Membership Operators in Python

Membership operators are used to **check whether a value exists within a sequence or collection**. They return a Boolean value (`True` or `False`) depending on whether the specified element is present.

Python provides **two membership operators**:

- `in`
- `not in`

Membership operators work with many built-in data types, including:

- Strings
- Lists
- Tuples
- Sets
- Dictionaries
- Ranges

They are commonly used for **searching**, **validation**, **filtering**, and **decision-making**.

# 📖 Topics Covered

## 1. Introduction to Membership Operators

Membership operators check whether a value is present in a sequence or collection.

They always return:

- `True`
- `False`

Example:

```python
numbers = [10, 20, 30, 40]

print(20 in numbers)
```

### Output

```text
True
```

---

## 2. Types of Membership Operators

Python provides two membership operators.

| Operator | Name | Description |
|----------|------|-------------|
| `in` | Membership Operator | Returns `True` if the value exists in the collection |
| `not in` | Membership NOT Operator | Returns `True` if the value does not exist in the collection |

---

## 7. Membership with Dictionaries

By default, membership operators check **dictionary keys**, not values.

```python
student = {
    "name": "John",
    "age": 20
}

print("name" in student)
```

### Output

```text
True
```

Checking values:

```python
print("John" in student.values())
```

### Output

```text
True
```

Checking keys explicitly:

```python
print("age" in student.keys())
```

### Output

```text
True
```

## 11. Common Mistakes

Avoid these common mistakes while using membership operators.

### Forgetting That Dictionary Membership Checks Keys

```python
student = {
    "name": "John"
}

print("John" in student)
```

### Output

```text
False
```

Correct:

```python
print("John" in student.values())
```

### Output

```text
True
```

---

### Case Sensitivity

```python
text = "Python"

print("python" in text)
```

### Output

```text
False
```

Membership checks in strings are case-sensitive.

---

### Using Membership with Non-Iterable Objects

Incorrect

```python
print(5 in 100)
```

### Output

```text
TypeError
```

Membership operators require an iterable object.

---

### Confusing `in` with Equality

Incorrect

```python
print(5 == [1, 2, 3, 5])
```

Correct

```python
print(5 in [1, 2, 3, 5])
```

---

## 12. Best Practices

Follow these recommended practices when using membership operators.

- Use `in` and `not in` instead of writing multiple comparison conditions.
- Use sets for fast membership testing when working with large collections.
- Remember that dictionary membership checks keys by default.
- Use meaningful variable names.
- Keep membership expressions simple and readable.
- Consider case sensitivity when checking strings.
