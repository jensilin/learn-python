# 📥 Membership Operator (`in`)

The **`in` operator** is a **membership operator** in Python that checks whether a **value exists within a sequence or collection**.

It returns:

- `True` if the value is present.
- `False` if the value is not present.

The `in` operator works with various iterable data types such as:

- Strings
- Lists
- Tuples
- Sets
- Dictionaries
- Ranges

It is widely used for **searching**, **input validation**, **filtering**, and **decision-making**.

# 📖 Topics Covered

## 1. Introduction to the `in` Operator

The `in` operator checks whether a value exists in a sequence or collection.

If the value is found, it returns `True`; otherwise, it returns `False`.

Example:

```python
numbers = [10, 20, 30, 40]

print(20 in numbers)
```

### Output

```text
True
```

---

## 2. Syntax

### Syntax

```python
value in collection
```

### Explanation

- `value` → The item to search for.
- `collection` → The sequence or iterable to search.
- `in` → Returns `True` if the value exists in the collection.

---

## 7. Using `in` with Dictionaries

By default, the `in` operator checks **dictionary keys**.

```python
student = {
    "name": "John",
    "age": 20
}

print("name" in student)
```

### Output

```text
True
```

Checking values:

```python
print("John" in student.values())
```

### Output

```text
True
```

Checking keys explicitly:

```python
print("age" in student.keys())
```

### Output

```text
True
```

## 11. Common Mistakes

Avoid these common mistakes while using the `in` operator.

### Forgetting That Dictionary Membership Checks Keys

```python
student = {
    "name": "John"
}

print("John" in student)
```

### Output

```text
False
```

Correct:

```python
print("John" in student.values())
```

### Output

```text
True
```

---

### Case Sensitivity

```python
text = "Python"

print("python" in text)
```

### Output

```text
False
```

String membership is case-sensitive.

---

### Using `in` with Non-Iterable Objects

Incorrect

```python
print(5 in 100)
```

### Output

```text
TypeError
```

The `in` operator requires an iterable object.

---

### Confusing `in` with `==`

Incorrect

```python
print(5 == [1, 2, 3, 5])
```

Correct

```python
print(5 in [1, 2, 3, 5])
```

---

## 12. Best Practices

Follow these recommended practices when using the `in` operator.

- Use `in` for checking whether an element exists in a collection.
- Use sets for fast membership testing with large datasets.
- Remember that dictionary membership checks keys by default.
- Use meaningful variable names.
- Keep membership expressions simple and readable.
- Consider case sensitivity when working with strings.
