# 🚫 Membership NOT Operator (`not in`)

The **`not in` operator** is a **membership operator** in Python that checks whether a **value does not exist within a sequence or collection**.

It returns:

- `True` if the value is **not present**.
- `False` if the value **is present**.

The `not in` operator works with various iterable data types such as:

- Strings
- Lists
- Tuples
- Sets
- Dictionaries
- Ranges

It is commonly used for **validation**, **filtering**, **access control**, **error checking**, and **decision-making**.

# 📖 Topics Covered

## 1. Introduction to the `not in` Operator

The `not in` operator checks whether a value does **not exist** in a sequence or collection.

If the value is absent, it returns `True`; otherwise, it returns `False`.

Example:

```python
numbers = [10, 20, 30, 40]

print(50 not in numbers)
```

### Output

```text
True
```

---

## 2. Syntax

### Syntax

```python
value not in collection
```

### Explanation

- `value` → The item to search for.
- `collection` → The sequence or iterable to search.
- `not in` → Returns `True` if the value is not found in the collection.

## 7. Using `not in` with Dictionaries

By default, the `not in` operator checks **dictionary keys**.

```python
student = {
    "name": "John",
    "age": 20
}

print("address" not in student)
```

### Output

```text
True
```

Checking values:

```python
print("David" not in student.values())
```

```text
True
```

Checking keys explicitly:

```python
print("name" not in student.keys())
```

### Output

```text
False
```

## 11. Common Mistakes

Avoid these common mistakes while using the `not in` operator.

### Forgetting That Dictionary Membership Checks Keys

```python
student = {
    "name": "John"
}

print("John" not in student)
```

### Output

```text
True
```

Correct:

```python
print("John" not in student.values())
```

### Output

```text
False
```

---

### Case Sensitivity

```python
text = "Python"

print("python" not in text)
```

### Output

```text
True
```

Membership checks in strings are case-sensitive.

---

### Using `not in` with Non-Iterable Objects

Incorrect

```python
print(5 not in 100)
```

### Output

```text
TypeError
```

The `not in` operator requires an iterable object.

---

### Confusing `not in` with `!=`

Incorrect

```python
print(5 != [1, 2, 3, 5])
```

Correct

```python
print(5 not in [1, 2, 3, 5])
```

`!=` compares two values, whereas `not in` checks whether a value is absent from a collection.

---

## 12. Best Practices

Follow these recommended practices when using the `not in` operator.

- Use sets for efficient membership testing with large datasets.
- Remember that dictionary membership checks keys by default.
- Use meaningful variable names.
- Keep membership expressions simple and readable.
- Consider case sensitivity when working with strings.
