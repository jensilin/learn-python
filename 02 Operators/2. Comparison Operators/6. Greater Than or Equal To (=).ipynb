# ⚖️ Greater Than or Equal To Operator (`>=`)

The **Greater Than or Equal To (`>=`) operator** is a comparison operator in Python used to **check whether the value on the left is greater than or equal to the value on the right**.

If the left operand is either **greater than** or **equal to** the right operand, the operator returns **`True`**. Otherwise, it returns **`False`**.

The Greater Than or Equal To operator is commonly used in conditional statements, loops, validation, eligibility checks, grading systems, and decision-making.

# 📖 Topics Covered

## 1. Introduction to the Greater Than or Equal To Operator

The Greater Than or Equal To operator (`>=`) compares two values and checks whether the **left operand is greater than or equal to the right operand**.

If the condition is true, Python returns `True`; otherwise, it returns `False`.

**Example**

```python
print(18 >= 18)
```

### Output

```text
True
```

---

## 2. Syntax

### Syntax

```python
operand1 >= operand2
```

### Explanation

- `operand1` → Left value
- `>=` → Greater Than or Equal To operator
- `operand2` → Right value

The result is always either `True` or `False`.

---

## 3. Comparing Integers

The operator works with integer values.

**Example**

```python
x = 35
y = 25

print(x >= y)
```

### Output

```text
True
```

---

```python
print(20 >= 20)
```

### Output

```text
True
```

---

```python
print(15 >= 25)
```

### Output

```text
False
```

---

## 4. Comparing Floating-Point Numbers

The operator also works with floating-point numbers.

**Example**

```python
print(8.5 >= 8.5)
```

### Output

```text
True
```

---

```python
print(7.2 >= 9.1)
```

### Output

```text
False
```

---

## 5. Comparing Strings

Strings are compared lexicographically (dictionary order).

**Example**

```python
print("Python" >= "Programming")
```

### Output

```text
True
```

---

```python
print("Apple" >= "Banana")
```

### Output

```text
False
```

> **Note:** String comparison is **case-sensitive**.

---

## 6. Comparing Boolean Values

Boolean values can also be compared.

**Example**

```python
print(True >= False)
```

### Output

```text
True
```

---

```python
print(False >= True)
```

### Output

```text
False
```

> **Explanation:** `True` is treated as `1` and `False` as `0` during numeric comparison.

---

## 7. Comparing Different Data Types

Compatible numeric data types can be compared.

**Example**

```python
print(15 >= 14.5)
```

### Output

```text
True
```

---

Comparing incompatible data types raises an error.

```python
print(15 >= "14")
```

### Output

```text
TypeError
```

## 10. Common Mistakes

Avoid these common mistakes while using the Greater Than or Equal To operator.

### Confusing `>` with `>=`

```python
age = 18

print(age > 18)
```

### Output

```text
False
```

Correct:

```python
print(age >= 18)
```

### Output

```text
True
```

---

### Comparing Strings Numerically

```python
print("100" >= "20")
```

### Output

```text
False
```

> **Why?** Strings are compared alphabetically (lexicographically), not numerically.

---

### Comparing Incompatible Data Types

```python
print(50 >= "40")
```

### Output

```text
TypeError
```

## 11. Best Practices

Follow these recommended practices when using the Greater Than or Equal To operator.

- Convert user input to the correct data type before comparison.
- Use meaningful variable names.
- Write simple and readable comparison expressions.
