# ⚖️ Less Than Operator (`<`)

The **Less Than (`<`) operator** is a comparison operator in Python used to **check whether the value on the left is less than the value on the right**.

If the left operand is less than the right operand, the operator returns **`True`**. Otherwise, it returns **`False`**.

The Less Than operator is commonly used in conditional statements, loops, sorting, filtering, validation, and decision-making.

# 📖 Topics Covered

## 1. Introduction to the Less Than Operator

The Less Than operator (`<`) compares two values and checks whether the **left operand is less than the right operand**.

If the condition is true, Python returns `True`; otherwise, it returns `False`.

**Example**

```python
print(8 < 15)
```

### Output

```text
True
```

---

## 2. Syntax

### Syntax

```python
operand1 < operand2
```

### Explanation

- `operand1` → Left value
- `<` → Less Than operator
- `operand2` → Right value

The result is always either `True` or `False`.

---

## 3. Comparing Integers

The Less Than operator works with integer values.

**Example**

```python
x = 15
y = 25

print(x < y)
```

### Output

```text
True
```

---

```python
print(30 < 10)
```

### Output

```text
False
```

---

## 4. Comparing Floating-Point Numbers

The operator also works with floating-point numbers.

**Example**

```python
print(3.14 < 6.28)
```

### Output

```text
True
```

---

## 5. Comparing Strings

Strings are compared lexicographically (dictionary order).

**Example**

```python
print("Apple" < "Banana")
```

### Output

```text
True
```

---

```python
print("python" < "Python")
```

### Output

```text
False
```

> **Note:** String comparison is **case-sensitive**. Uppercase and lowercase letters have different Unicode values.

---

## 6. Comparing Boolean Values

Boolean values can also be compared.

**Example**

```python
print(False < True)
```

### Output

```text
True
```

> **Explanation:** `False` is treated as `0` and `True` as `1` during numeric comparison.

---

## 7. Comparing Different Data Types

Compatible numeric data types can be compared.

**Example**

```python
print(5 < 7.5)
```

### Output

```text
True
```

---

Comparing incompatible data types raises an error.

```python
print(5 < "7")
```

### Output

```text
TypeError
```

## 10. Common Mistakes

Avoid these common mistakes while using the Less Than operator.

### Reversing the Operands

**Incorrect Logic**

```python
marks = 35

print(40 < marks)
```

### Output

```text
False
```

> **Why?** Ensure the values are compared in the correct order based on your intended logic.

---

### Comparing Strings Numerically

```python
print("100" < "20")
```

### Output

```text
True
```

> **Why?** Strings are compared alphabetically (lexicographically), not based on their numeric values.

---

### Comparing Incompatible Data Types

```python
print(25 < "30")
```

### Output

```text
TypeError
```

---

## 11. Best Practices

Follow these recommended practices when using the Less Than operator.

- Use `<` only when checking whether one value is less than another.
- Compare compatible data types whenever possible.
- Convert user input to the appropriate data type before comparison.
- Use meaningful variable names.
- Keep comparison expressions simple and readable.
