# ⚖️ Greater Than Operator (`>`)

The **Greater Than (`>`) operator** is a comparison operator in Python used to **check whether the value on the left is greater than the value on the right**.

If the left operand is greater than the right operand, the operator returns **`True`**. Otherwise, it returns **`False`**.

The Greater Than operator is commonly used in conditional statements, loops, sorting, filtering, data validation, and decision-making.

# 📖 Topics Covered

## 1. Introduction to the Greater Than Operator

The Greater Than operator (`>`) compares two values and checks whether the **left operand is greater than the right operand**.

If the condition is true, Python returns `True`; otherwise, it returns `False`.

**Example**

```python
print(15 > 10)
```

### Output

```text
True
```

---

## 2. Syntax

### Syntax

```python
operand1 > operand2
```

### Explanation

- `operand1` → Left value
- `>` → Greater Than operator
- `operand2` → Right value

The result is always either `True` or `False`.

---

## 3. Comparing Integers

The Greater Than operator works with integer values.

**Example**

```python
x = 30
y = 20

print(x > y)
```

### Output

```text
True
```

---

```python
print(15 > 25)
```

### Output

```text
False
```

---

## 4. Comparing Floating-Point Numbers

The operator also works with floating-point numbers.

**Example**

```python
print(9.5 > 7.2)
```

### Output

```text
True
```

---

## 5. Comparing Strings

Strings are compared lexicographically (dictionary order).

**Example**

```python
print("Cat" > "Apple")
```

### Output

```text
True
```

---

```python
print("Python" > "python")
```

### Output

```text
False
```

> **Note:** String comparison is **case-sensitive**. Uppercase letters are compared differently from lowercase letters.

---

## 6. Comparing Boolean Values

Boolean values can also be compared.

**Example**

```python
print(True > False)
```

### Output

```text
True
```

> **Explanation:** `True` is treated as `1` and `False` as `0` during numeric comparison.

---

## 7. Comparing Different Data Types

Some different numeric data types can be compared.

**Example**

```python
print(10 > 5.5)
```

### Output

```text
True
```

---

However, comparing incompatible data types raises an error.

```python
print(10 > "5")
```

### Output

```text
TypeError
```

## 10. Common Mistakes

Avoid these common mistakes while using the Greater Than operator.

### Reversing the Operands

**Incorrect Logic**

```python
age = 18

print(18 > age)
```

### Output

```text
False
```

> **Why?** Ensure the operands are placed in the intended order.

---

### Comparing Strings Numerically

```python
print("100" > "20")
```

### Output

```text
False
```

> **Why?** Strings are compared alphabetically (lexicographically), not numerically.

---

### Comparing Incompatible Data Types

```python
print(25 > "20")
```

### Output

```text
TypeError
```

---

## 11. Best Practices

Follow these recommended practices when using the Greater Than operator.

- Use `>` only when checking whether one value is greater than another.
- Compare compatible data types whenever possible.
- Convert user input to the appropriate data type before comparison.
- Use meaningful variable names.
- Keep comparison expressions simple and readable.
