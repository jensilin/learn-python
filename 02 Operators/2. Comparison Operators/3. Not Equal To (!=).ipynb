# ⚖️ Not Equal To Operator (`!=`)

The **Not Equal To (`!=`) operator** is a comparison operator in Python used to **check whether two values or expressions are different**.

If the two operands have different values, the operator returns **`True`**. If both operands are equal, it returns **`False`**.

The Not Equal To operator is widely used in conditional statements, loops, input validation, filtering, and decision-making where actions need to be performed only when values do not match.

# 📖 Topics Covered

## 1. Introduction to the Not Equal To Operator

The Not Equal To operator (`!=`) compares two values and checks whether they are **different**.

If the values are different, Python returns `True`; otherwise, it returns `False`.

**Example**

```python
print(15 != 20)
```

### Output

```text
True
```

---

## 2. Syntax

### Syntax

```python
operand1 != operand2
```

### Explanation

- `operand1` → First value
- `!=` → Not Equal To operator
- `operand2` → Second value

The result is always either `True` or `False`.

---

## 3. Comparing Integers

The Not Equal To operator works with integer values.

**Example**

```python
x = 25
y = 30

print(x != y)
```

### Output

```text
True
```

---

```python
print(25 != 25)
```

### Output

```text
False
```

---

## 4. Comparing Floating-Point Numbers

The operator also works with floating-point numbers.

**Example**

```python
print(5.5 != 7.5)
```

### Output

```text
True
```

---

## 5. Comparing Strings

Strings are considered different if their content or letter case differs.

**Example**

```python
print("Python" != "Java")
```

### Output

```text
True
```

---

```python
print("Python" != "python")
```

### Output

```text
True
```

> **Note:** String comparison is **case-sensitive**.

---

## 6. Comparing Boolean Values

The Not Equal To operator can compare Boolean values.

**Example**

```python
print(True != False)
print(True != True)
```

### Output

```text
True
False
```

## 10. Common Mistakes

Avoid these common mistakes while using the Not Equal To operator.

### Using `!` Instead of `!=`

**Incorrect**

```python
print(!True)
```

### Output

```text
SyntaxError
```

> **Why?** Python does not use `!` as a logical NOT operator. Use `not` for logical negation and `!=` for inequality comparisons.

---

### Ignoring Case Sensitivity

```python
print("Python" != "python")
```

### Output

```text
True
```

---

### Comparing Different Data Types

```python
print(100 != "100")
```

### Output

```text
True
```

> **Note:** A number and a string are different data types, so they are not equal.
