# ⚖️ Comparison Operators

Comparison operators are used to **compare two values or expressions**. They evaluate the relationship between operands and always return a **Boolean value** (`True` or `False`).

Comparison operators are commonly used in conditional statements, loops, filtering data, searching, sorting, and many other programming tasks where decision-making is required.

# 📖 Topics Covered

## 1. Introduction to Comparison Operators

Comparison operators compare two operands and determine whether a condition is **True** or **False**.

They are frequently used with:

- `if` statements
- `while` loops
- `for` loops
- Logical operators
- Filtering data
- Decision-making

Python provides six comparison operators.

| Operator | Name | Example |
|----------|------|---------|
| `==` | Equal To | `a == b` |
| `!=` | Not Equal To | `a != b` |
| `>` | Greater Than | `a > b` |
| `<` | Less Than | `a < b` |
| `>=` | Greater Than or Equal To | `a >= b` |
| `<=` | Less Than or Equal To | `a <= b` |

---

## 2. How Comparison Operators Work

Comparison operators compare two values and return a Boolean result.

**Example**

```python
x = 15
y = 10

print(x > y)
print(x < y)
print(x == y)
```

### Output

```text
True
False
False
```

---

## 3. Comparison with Different Data Types

Comparison operators work with several compatible data types.

### Integer

```python
print(10 > 5)
```

### Output

```text
True
```

---

### Float

```python
print(5.5 <= 6.0)
```

### Output

```text
True
```

---

### String

Strings are compared lexicographically (dictionary order).

```python
print("Apple" < "Banana")
```

### Output

```text
True
```

---

### Boolean

```python
print(True == 1)
```

### Output

```text
True
```

> **Note:** In Python, `True` is treated as `1` and `False` as `0` in numeric comparisons.

---

## 4. Boolean Results

Comparison operators always return either:

- `True`
- `False`

**Example**

```python
a = 25
b = 30

result = a >= b

print(result)
```

### Output

```text
False
```

## 7. Common Mistakes

Avoid these common mistakes while using comparison operators.

### Using `=` Instead of `==`

**Incorrect**

```python
if age = 18:
    print("Eligible")
```

### Output

```text
SyntaxError
```

---

### Comparing Incompatible Data Types

**Incorrect**

```python
print(10 > "5")
```

### Output

```text
TypeError
```

---

### Assuming String Comparison Uses Length

```python
print("100" > "20")
```

### Output

```text
False
```

> **Why?** Strings are compared lexicographically (alphabetically), not by their numeric values.

---

## 8. Best Practices

Follow these recommended practices when using comparison operators.

- Use meaningful variable names.
- Ensure operands are compatible.
- Use `==` for equality comparisons.
- Convert data types when necessary.
- Use parentheses to improve readability in complex conditions.

# 📊 Comparison Operators Summary

| Operator | Purpose |
|----------|---------|
| `==` | Checks whether two values are equal |
| `!=` | Checks whether two values are not equal |
| `>` | Checks whether the left value is greater |
| `<` | Checks whether the left value is smaller |
| `>=` | Checks whether the left value is greater than or equal |
| `<=` | Checks whether the left value is less than or equal |
