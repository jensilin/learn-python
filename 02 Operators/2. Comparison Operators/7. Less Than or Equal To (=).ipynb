# ⚖️ Less Than or Equal To Operator (`<=`)

The **Less Than or Equal To (`<=`) operator** is a comparison operator in Python used to **check whether the value on the left is less than or equal to the value on the right**.

If the left operand is either **less than** or **equal to** the right operand, the operator returns **`True`**. Otherwise, it returns **`False`**.

The Less Than or Equal To operator is commonly used in conditional statements, loops, validation, eligibility checks, grading systems, and decision-making.

# 📖 Topics Covered

## 1. Introduction to the Less Than or Equal To Operator

The Less Than or Equal To operator (`<=`) compares two values and checks whether the **left operand is less than or equal to the right operand**.

If the condition is true, Python returns `True`; otherwise, it returns `False`.

**Example**

```python
print(15 <= 20)
```

### Output

```text
True
```

---

## 2. Syntax

### Syntax

```python
operand1 <= operand2
```

### Explanation

- `operand1` → Left value
- `<=` → Less Than or Equal To operator
- `operand2` → Right value

The result is always either `True` or `False`.

---

## 3. Comparing Integers

The operator works with integer values.

**Example**

```python
x = 20
y = 25

print(x <= y)
```

### Output

```text
True
```

---

```python
print(30 <= 30)
```

### Output

```text
True
```

---

```python
print(40 <= 25)
```

### Output

```text
False
```

---

## 4. Comparing Floating-Point Numbers

The operator also works with floating-point numbers.

**Example**

```python
print(7.5 <= 9.2)
```

### Output

```text
True
```

---

```python
print(10.8 <= 10.8)
```

### Output

```text
True
```

---

## 5. Comparing Strings

Strings are compared lexicographically (dictionary order).

**Example**

```python
print("Apple" <= "Banana")
```

### Output

```text
True
```

---

```python
print("Python" <= "Programming")
```

### Output

```text
False
```

> **Note:** String comparison is **case-sensitive**.

---

## 6. Comparing Boolean Values

Boolean values can also be compared.

**Example**

```python
print(False <= True)
```

### Output

```text
True
```

---

```python
print(True <= False)
```

### Output

```text
False
```

> **Explanation:** `False` is treated as `0` and `True` as `1` during numeric comparison.

---

## 7. Comparing Different Data Types

Compatible numeric data types can be compared.

**Example**

```python
print(8 <= 8.0)
```

### Output

```text
True
```

---

Comparing incompatible data types raises an error.

```python
print(10 <= "20")
```

### Output

```text
TypeError
```

## 8. Practical Examples

Apply the Less Than or Equal To operator through practical programming examples.

### Output

```text
Eligible
```

### Example 2: Passing Score Limit

```python
marks = 35

if marks <= 35:
    print("Needs Improvement")
```

### Output

```text
Needs Improvement
```

### Example 3: Speed Limit Check

```python
speed = 60

if speed <= 60:
    print("Within Speed Limit")
```

### Output

## 10. Common Mistakes

Avoid these common mistakes while using the Less Than or Equal To operator.

### Confusing `<` with `<=`

```python
age = 18

print(age < 18)
```

### Output

```text
False
```

Correct:

```python
print(age <= 18)
```

### Output

```text
True
```

---

### Comparing Strings Numerically

```python
print("100" <= "20")
```

### Output

```text
True
```

> **Why?** Strings are compared alphabetically (lexicographically), not based on their numeric values.

---

### Comparing Incompatible Data Types

```python
print(25 <= "30")
```

### Output

```text
TypeError
```

## 11. Best Practices

Follow these recommended practices when using the Less Than or Equal To operator.

- Use `<=` when equality should also satisfy the condition.
- Compare compatible data types whenever possible.
- Convert user input to the appropriate data type before comparison.
- Use meaningful variable names.
- Keep comparison expressions simple and readable.
