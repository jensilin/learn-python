# ⚖️ Equal To Operator (`==`)

The **Equal To (`==`) operator** is a comparison operator in Python used to **check whether two values or expressions are equal**.

If both operands have the same value, the operator returns **`True`**; otherwise, it returns **`False`**.

The Equal To operator is one of the most frequently used comparison operators and is commonly used in conditional statements, loops, functions, data validation, and decision-making.

# 📖 Topics Covered

## 1. Introduction to the Equal To Operator

The Equal To operator (`==`) compares two values and checks whether they are equal.

If both operands contain the same value, Python returns `True`; otherwise, it returns `False`.

**Example**

```python
print(15 == 15)
```

### Output

```text
True
```

---

## 2. Syntax

### Syntax

```python
operand1 == operand2
```

### Explanation

- `operand1` → First value
- `==` → Equal To operator
- `operand2` → Second value

The result is always either `True` or `False`.

---

## 3. Comparing Integers

The Equal To operator works with integer values.

**Example**

```python
x = 25
y = 25

print(x == y)
```

### Output

```text
True
```

---

## 4. Comparing Floating-Point Numbers

The operator also works with floating-point numbers.

**Example**

```python
print(5.5 == 5.5)
```

### Output

```text
True
```

---

## 5. Comparing Strings

Strings are equal only if both their content and case match exactly.

**Example**

```python
print("Python" == "Python")
```

### Output

```text
True
```

---

```python
print("Python" == "python")
```

### Output

```text
False
```

> **Note:** String comparison is **case-sensitive**.

---

## 6. Comparing Boolean Values

The Equal To operator can compare Boolean values.

**Example**

```python
print(True == True)
print(True == False)
```

### Output

```text
True
False
```

---

## 7. Comparing Different Data Types

Python can compare values of different data types.

**Example**

```python
print(10 == 10.0)
```

### Output

```text
True
```

---

```python
print(10 == "10")
```

### Output

```text
False
```

## 10. Common Mistakes

Avoid these common mistakes while using the Equal To operator.

### Using `=` Instead of `==`

**Incorrect**

```python
if age = 18:
    print("Eligible")
```

### Output

```text
SyntaxError
```

---

### Correct

```python
if age == 18:
    print("Eligible")
```

### Output

```text
Eligible
```

---

### Ignoring Case Sensitivity

```python
print("Python" == "python")
```

### Output

```text
False
```

---

### Comparing Different Data Types

```python
print(100 == "100")
```

### Output

```text
False
```

---

## 11. Best Practices

Follow these recommended practices when using the Equal To operator.

- Use `==` only for value comparisons.
- Use meaningful variable names.
- Ensure operands have compatible data types when appropriate.
- Be mindful of case sensitivity when comparing strings.
- Write clear and readable comparison expressions.
