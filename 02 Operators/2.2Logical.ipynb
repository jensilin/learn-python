<h3>b) <code>or</code> Operator</h2>

<p>
The <code>or</code> operator returns <b>True</b> if <b>at least one</b> of the given conditions is <b>True</b>. It returns <b>False</b> only when <b>all</b> the conditions are <b>False</b>.
</p>

<h4>Syntax</h3>

```python
condition1 or condition2
```

<h4>Truth Table</h3>

<table>
    <tr>
        <th>Condition 1</th>
        <th>Condition 2</th>
        <th>Result</th>
    </tr>
    <tr>
        <td><code>True</code></td>
        <td><code>True</code></td>
        <td><b>True</b></td>
    </tr>
    <tr>
        <td><code>True</code></td>
        <td><code>False</code></td>
        <td><b>True</b></td>
    </tr>
    <tr>
        <td><code>False</code></td>
        <td><code>True</code></td>
        <td><b>True</b></td>
    </tr>
    <tr>
        <td><code>False</code></td>
        <td><code>False</code></td>
        <td><b>False</b></td>
    </tr>
</table>

<h4>Example</h3>

```python
print(10 > 20 or 5 > 3)
print(10 > 20 or 5 < 3)
```

<h4>Output</h3>

```text
True
False
```

<div align="center">

<b>Condition 1</b> <code>or</code> <b>Condition 2</b>

</div>

<div style="border-left:4px solid #4CAF50;padding-left:12px;margin-top:10px;">

<b>Key Point:</b> The <code>or</code> operator returns <code>True</code> if <b>any one</b> of the conditions is <code>True</code>. It returns <code>False</code> only when <b>both conditions are <code>False</code></b>.

</div>

<h3>c) <code>not</code> Operator</h3>

<p>
The <code>not</code> operator is used to <b>reverse (invert)</b> the result of a condition. It converts <code>True</code> to <code>False</code> and <code>False</code> to <code>True</code>.
</p>

<p><b>Syntax</b></p>

```python
not condition
```

<p><b>Example</b></p>

```python
print(not (10 > 5))
print(not (10 < 5))
```

<p><b>Output</b></p>

```text
False
True
```

> **Key Point:** The `not` operator returns the opposite Boolean value of the given condition.
