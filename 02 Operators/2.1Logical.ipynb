<h2>Logical Operators</h1>

<p>
Logical operators are used to check <b>logical conditions</b>. Python provides
<b>three logical operators</b>:
</p>

<ul>
    <li><code>and</code></li>
    <li><code>or</code></li>
    <li><code>not</code></li>
</ul>

<hr>

<h3>a) <code>and</code> Operator</h2>

<p>
The <code>and</code> operator returns <b>True</b> only when
<b>all the conditions are True</b>. If any one of the conditions is
<b>False</b>, the result will be <b>False</b>.
</p>

<h5>Syntax</h3>

```python
condition1 and condition2
```

<h4>Truth Table</h3>

<table>
    <tr>
        <th>Condition 1</th>
        <th>Condition 2</th>
        <th>Result</th>
    </tr>
    <tr>
        <td><code>True</code></td>
        <td><code>True</code></td>
        <td><b>True</b></td>
    </tr>
    <tr>
        <td><code>True</code></td>
        <td><code>False</code></td>
        <td><b>False</b></td>
    </tr>
    <tr>
        <td><code>False</code></td>
        <td><code>True</code></td>
        <td><b>False</b></td>
    </tr>
    <tr>
        <td><code>False</code></td>
        <td><code>False</code></td>
        <td><b>False</b></td>
    </tr>
</table>

<h4>Example</h3>

```python
a = 10
b = 20

print(a < b and b > 15)
print(a > b and b > 15)
```

<h4>Output</h3>

```text
True
False
```
