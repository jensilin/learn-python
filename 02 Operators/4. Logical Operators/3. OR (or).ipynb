# đź”€ Logical OR Operator (`or`)

The **Logical OR (`or`) operator** is used to **combine two or more conditions**. It returns **`True` if at least one of the conditions is `True`**. It returns `False` only when **all conditions are `False`**.

The `or` operator is widely used in decision-making, input validation, search filters, authentication, and conditional statements.

# đź“– Topics Covered

## 1. Introduction to the OR Operator

The `or` operator checks whether **at least one condition is True**.

If one or more conditions are `True`, the result is `True`.

It returns `False` only if every condition is `False`.

Example:

```python
marks = 30

print(marks >= 35 or marks == 30)
```

### Output

```text
True
```

---

## 2. Syntax

### Syntax

```python
condition1 or condition2
```

Or

```python
condition1 or condition2 or condition3
```

### Explanation

- `condition1` â†’ First condition.
- `condition2` â†’ Second condition.
- `or` â†’ Returns `True` if at least one condition is `True`.

## 6. OR with Multiple Conditions

You can combine more than two conditions.

```python
role = "Admin"
age = 17
has_permission = False

print(role == "Admin" or age >= 18 or has_permission)
```

### Output

```text
True
```

---

## 7. Short-Circuit Evaluation

Python stops evaluating conditions as soon as it finds a `True` condition.

Example:

```python
print(True or False)
```

### Output

```text
True
```

The second condition is never evaluated because the result is already known.

Another example:

```python
x = 10

print(x > 5 or x / 0 > 1)
```

### Output

```text
True
```

No `ZeroDivisionError` occurs because Python never evaluates `x / 0`.

## 11. Common Mistakes

Avoid these common mistakes while using the `or` operator.

### Confusing `or` with `|`

Incorrect

```python
print(True | False)
```

Correct

```python
print(True or False)
```

> **Why?** `|` is a bitwise operator, while `or` is a logical operator.

---

### Expecting Both Conditions to Execute

```python
print(True or 10 / 0)
```

### Output

```text
True
```

The second expression is not evaluated because of short-circuit evaluation.

---

### Using Assignment Instead of Comparison

Incorrect

```python
if age = 18 or age = 21:
    print("Eligible")
```

Correct

```python
if age == 18 or age == 21:
    print("Eligible")
```

---

### Forgetting Parentheses in Complex Conditions

Less readable:

```python
age >= 18 or salary >= 50000 and experience >= 5
```

Better:

```python
(age >= 18 or salary >= 50000) and experience >= 5
```

---

## 12. Best Practices

Follow these recommended practices when using the `or` operator.

- Use `or` when any one condition can satisfy the requirement.
- Keep conditions simple and readable.
- Use parentheses for complex expressions.
- Understand short-circuit evaluation.
- Use meaningful variable names.
- Avoid unnecessary nested conditions.
