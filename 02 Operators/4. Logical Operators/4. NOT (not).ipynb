# 🚫 Logical NOT Operator (`not`)

The **Logical NOT (`not`) operator** is used to **reverse the Boolean value of an expression**. It converts **`True` to `False`** and **`False` to `True`**.

Unlike the `and` and `or` operators, the `not` operator works with **only one condition or expression**.

The `not` operator is commonly used to negate conditions, simplify conditional statements, validate inputs, and control program flow.

# 📖 Topics Covered

## 1. Introduction to the NOT Operator

The `not` operator reverses the Boolean value of an expression.

If the expression is:

- `True`, it becomes `False`.
- `False`, it becomes `True`.

Example:

```python
is_raining = True

print(not is_raining)
```

### Output

```text
False
```

---

## 2. Syntax

### Syntax

```python
not expression
```

### Explanation

- `expression` → Any Boolean expression or value.
- `not` → Reverses the Boolean result.

---

## 3. Truth Table

| Expression | Result |
|------------|--------|
| `not True` | False |
| `not False` | True |

---

## 4. NOT with Boolean Values

The `not` operator works directly with Boolean values.

```python
a = True

print(not a)
```

### Output

```text
False
```

---

```python
b = False

print(not b)
```

### Output

```text
True
```

---

## 5. NOT with Comparison Operators

The `not` operator is often combined with comparison operators.

```python
age = 15

print(not age >= 18)
```

### Output

```text
True
```

Another example:

```python
marks = 45

print(not (marks < 35))
```

### Output

```text
True
```

## 7. Operator Precedence

The `not` operator has **higher precedence** than `and` and `or`.

Example:

```python
print(not False and True)
```

### Output

```text
True
```

Equivalent to:

```python
print((not False) and True)
```

## 10. Common Mistakes

Avoid these common mistakes while using the `not` operator.

### Forgetting Parentheses

Less readable:

```python
not age > 18
```

Better:

```python
not (age > 18)
```

---

### Confusing `not` with `!=`

Incorrect understanding:

```python
not a == b
```

Although valid, it is often clearer to write:

```python
a != b
```

Use `!=` when checking whether two values are different.

Use `not` when reversing an entire Boolean expression.

---

### Double Negation

```python
print(not not True)
```

### Output

```text
True
```

Double negation works but usually makes code harder to read.

---

### Assuming `not` Changes Variable Values

```python
status = True

print(not status)

print(status)
```

### Output

```text
False
True
```

The `not` operator **does not modify the original variable**. It only returns the opposite Boolean value.

---

## 11. Best Practices

Follow these recommended practices when using the `not` operator.

- Use `not` only when reversing a condition improves readability.
- Use parentheses for complex expressions.
- Prefer `not variable` instead of `variable == False`.
- Avoid unnecessary double negation.
- Use meaningful variable names.
- Keep conditions simple and easy to understand.
