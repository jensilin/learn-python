# 🤝 Logical AND Operator (`and`)

The **Logical AND (`and`) operator** is used to **combine two or more conditions**. It returns **`True` only if all the conditions are `True`**. If any one of the conditions is `False`, the result is `False`.

The `and` operator is one of the most commonly used logical operators in Python for decision-making, input validation, filtering data, and controlling program flow.

# 📖 Topics Covered

## 1. Introduction to the AND Operator

The `and` operator checks whether **all conditions are True**.

If every condition is `True`, it returns `True`.

If even one condition is `False`, it returns `False`.

Example:

```python
marks = 85

print(marks >= 35 and marks <= 100)
```

### Output

```text
True
```

---

## 2. Syntax

### Syntax

```python
condition1 and condition2
```

Or

```python
condition1 and condition2 and condition3
```

### Explanation

- `condition1` → First condition.
- `condition2` → Second condition.
- `and` → Returns `True` only when all conditions are `True`.

---

## 3. Truth Table

| Condition 1 | Condition 2 | Result |
|-------------|-------------|--------|
| True | True | True |
| True | False | False |
| False | True | False |
| False | False | False |

---

## 4. AND with Boolean Values

The `and` operator works directly with Boolean values.

```python
a = True
b = False

print(a and b)
```

### Output

```text
False
```

Another example:

```python
x = True
y = True

print(x and y)
```

### Output

```text
True
```

## 6. AND with Multiple Conditions

You can combine more than two conditions.

```python
age = 30
salary = 45000
experience = 5

print(age >= 18 and salary >= 30000 and experience >= 2)
```

### Output

```text
True
```

---

## 7. Short-Circuit Evaluation

Python stops evaluating conditions as soon as it finds a `False` condition.

Example:

```python
print(False and True)
```

### Output

```text
False
```

The second condition is never evaluated because the result is already known.

Another example:

```python
x = 10

print(x > 20 and x / 0 > 1)
```

### Output

```text
False
```

No `ZeroDivisionError` occurs because Python never evaluates `x / 0`.

---

## 11. Common Mistakes

Avoid these common mistakes while using the `and` operator.

### Confusing `and` with `&`

Incorrect

```python
print(True & False)
```

Correct

```python
print(True and False)
```

> **Why?** `&` is a bitwise operator, while `and` is a logical operator.

---

### Expecting Both Conditions to Execute

```python
print(False and 10 / 0)
```

### Output

```text
False
```

The second expression is not evaluated because of short-circuit evaluation.

---

### Using Assignment Instead of Comparison

Incorrect

```python
if age = 18:
    print("Eligible")
```

Correct

```python
if age == 18:
    print("Eligible")
```

---

### Forgetting Parentheses in Complex Conditions

Less readable:

```python
age >= 18 and salary >= 30000 or experience >= 5
```

Better:

```python
(age >= 18 and salary >= 30000) or experience >= 5
```

---

## 12. Best Practices

Follow these recommended practices when using the `and` operator.

- Use `and` only when all conditions must be true.
- Keep conditions simple and readable.
- Use parentheses for complex expressions.
- Understand short-circuit evaluation.
- Use meaningful variable names.
- Avoid unnecessary nested conditions.
