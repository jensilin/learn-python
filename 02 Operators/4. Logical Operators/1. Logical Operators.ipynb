# 🧠 Logical Operators in Python

Logical operators are used to **combine multiple conditions** or **invert the result of a condition**. They are commonly used in decision-making statements such as `if`, `elif`, `while`, and conditional expressions.

Python provides **three logical operators**:

- `and`
- `or`
- `not`

These operators work with **Boolean (`True` and `False`)** values and return a Boolean result.

# 📖 Topics Covered

## 1. Introduction to Logical Operators

Logical operators are used to combine two or more Boolean expressions.

For example:

```python
age = 22
salary = 45000

print(age > 18 and salary > 30000)
```

### Output

```text
True
```

Logical operators always evaluate expressions and return either:

- `True`
- `False`

---

## 2. Types of Logical Operators

Python has three logical operators.

| Operator | Name | Description |
|----------|------|-------------|
| `and` | Logical AND | Returns `True` if both conditions are `True`. |
| `or` | Logical OR | Returns `True` if at least one condition is `True`. |
| `not` | Logical NOT | Reverses the Boolean value. |

---

## 3. Truth Table

| A | B | A and B | A or B | not A |
|---|---|----------|---------|-------|
| True | True | True | True | False |
| True | False | False | True | False |
| False | True | False | True | True |
| False | False | False | False | True |

---

## 4. Short-Circuit Evaluation

Python evaluates logical expressions using **short-circuit evaluation**, meaning it stops evaluating as soon as the final result is known.

### Example using `and`

```python
age = 15

print(age > 18 and age < 60)
```

### Output

```text
False
```

Since the first condition is `False`, Python does not evaluate the second condition.

---

### Example using `or`

```python
age = 25

print(age > 18 or age < 10)
```

### Output

```text
True
```

Since the first condition is `True`, Python does not evaluate the second condition.

---

## 5. Operator Precedence

Logical operators follow this precedence order:

| Priority | Operator |
|----------|----------|
| Highest | `not` |
| Medium | `and` |
| Lowest | `or` |

Example:

```python
print(True or False and False)
```

### Output

```text
True
```

Equivalent to:

```python
print(True or (False and False))
```

---

## 6. Logical Operators with Boolean Values

Logical operators are primarily used with Boolean values.

```python
a = True
b = False

print(a and b)
print(a or b)
print(not a)
```

### Output

```text
False
True
False
```

## 8. Logical Operators with Non-Boolean Values

Logical operators can also work with non-Boolean values. Python treats values according to their **truthiness**.

Falsy values include:

- `False`
- `0`
- `0.0`
- `''`
- `[]`
- `{}`
- `set()`
- `None`

Everything else is considered truthy.

Example:

```python
print(10 and 20)
print(0 or 50)
print(not "")
```

### Output

```text
20
50
True
```

## 11. Common Mistakes

Avoid these common mistakes while using logical operators.

### Confusing `and` with `&`

Incorrect

```python
print(True & False)
```

Although this works with Boolean values, `&` is a **bitwise operator**, not a logical operator.

Correct

```python
print(True and False)
```

---

### Confusing `or` with `|`

Incorrect

```python
print(True | False)
```

Correct

```python
print(True or False)
```

---

### Using `=` Instead of `==`

Incorrect

```python
if age = 18:
    print("Eligible")
```

Correct

```python
if age == 18:
    print("Eligible")
```

---

### Forgetting Operator Precedence

```python
print(True or False and False)
```

Always remember:

```python
and
```

is evaluated before

```python
or
```

---

## 12. Best Practices

Follow these recommended practices when using logical operators.

- Use parentheses to improve readability.
- Keep conditions simple and meaningful.
- Understand short-circuit evaluation.
- Use descriptive variable names.
- Avoid deeply nested logical expressions.
- Use logical operators only when necessary.
