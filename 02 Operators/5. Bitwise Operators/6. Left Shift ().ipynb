# ⬅️ Left Shift Operator (`<<`)

The **Left Shift (`<<`) operator** shifts the bits of an integer to the **left by a specified number of positions**. For every position shifted to the left, a **`0` is inserted on the right**.

In most cases, shifting a number left by **one position multiplies it by 2**, shifting by two positions multiplies it by **4**, and so on.

The Left Shift operator is commonly used in **bit manipulation**, **performance optimization**, **embedded systems**, **network programming**, and **low-level programming**.

# 📖 Topics Covered

## 1. Introduction to the Left Shift Operator

The Left Shift operator moves every bit of a number to the **left**.

New positions on the right are filled with `0`.

Example:

```text
5 = 00000101
```

Shift left by 2 positions:

```text
00010100
```

Result:

```text
20
```

---

## 2. Syntax

### Syntax

```python
value << positions
```

### Explanation

- `value` → Integer to shift.
- `positions` → Number of bit positions to shift left.
- `<<` → Performs the left shift operation.

---

## 3. Understanding the Binary Operation

Example:

```python
number = 5
```

Binary representation:

```text
00000101
```

After shifting left by 2:

```text
00010100
```

Result:

```python
print(number << 2)
```

### Output

```text
20
```

---

## 4. Relationship with Multiplication

Each left shift multiplies the number by **2**.

| Expression | Equivalent | Result |
|-----------|------------|-------:|
| `5 << 1` | `5 × 2` | 10 |
| `5 << 2` | `5 × 4` | 20 |
| `5 << 3` | `5 × 8` | 40 |
| `5 << 4` | `5 × 16` | 80 |

General formula:

```text
number << n = number × (2ⁿ)
```

## 10. Common Mistakes

Avoid these common mistakes while using the Left Shift operator.

### Confusing Left Shift with Arithmetic Multiplication

Although:

```python
5 << 2
```

produces:

```text
20
```

the operator works by **shifting bits**, not by performing multiplication directly.

---

### Using Floating-Point Numbers

Incorrect

```python
5.5 << 2
```

### Output

```text
TypeError
```

The `<<` operator works only with integers.

---

### Using Negative Shift Counts

Incorrect

```python
5 << -1
```

### Output

```text
ValueError
```

Shift counts must be zero or positive.

---

### Forgetting Binary Representation

The Left Shift operator manipulates binary bits, not decimal digits.
