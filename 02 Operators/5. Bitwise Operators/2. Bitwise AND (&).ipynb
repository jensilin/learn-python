# 🔗 Bitwise AND Operator (`&`)

The **Bitwise AND (`&`) operator** compares the corresponding bits of two integers. It returns **`1` only when both corresponding bits are `1`**. Otherwise, it returns `0`.

The Bitwise AND operator is commonly used for **bit masking**, **permission checking**, **flag operations**, **network programming**, and **embedded systems**.

# 📖 Topics Covered

## 1. Introduction to the Bitwise AND Operator

The Bitwise AND operator compares every corresponding bit of two integers.

A bit is set to **1 only if both bits are 1**.

Example:

```text
12 = 1100
10 = 1010
--------------
     1000
```

Result:

```text
8
```

---

## 2. Syntax

### Syntax

```python
value1 & value2
```

### Explanation

- `value1` → First integer.
- `value2` → Second integer.
- `&` → Performs a bitwise AND operation.

## 8. Practical Examples

### Example 1: Check Even or Odd

The least significant bit determines whether a number is even or odd.

```python
number = 18

print(number & 1)
```

### Output

```text
0
```

Since the result is `0`, the number is **even**.

Another example:

```python
number = 19

print(number & 1)
```

### Output

```text
1
```

Since the result is `1`, the number is **odd**.

---

### Example 2: Permission Checking

```python
permissions = 13
read_permission = 1

print(permissions & read_permission)
```

### Output

```text
1
```

---

### Example 3: Bit Masking

```python
number = 45
mask = 15

print(number & mask)
```

### Output

```text
13
```

---

### Example 4: Extract Lower Bits

```python
number = 255

print(number & 15)
```

### Output

## 10. Common Mistakes

Avoid these common mistakes while using the Bitwise AND operator.

### Confusing `&` with `and`

Incorrect for logical conditions:

```python
True & False
```

Preferred:

```python
True and False
```

> **Why?** `&` performs a bitwise operation, while `and` performs a logical operation.

---

### Using Floating-Point Numbers

Incorrect

```python
10.5 & 4
```

### Output

```text
TypeError
```

The `&` operator works only with integers.

---

### Forgetting Binary Representation

Incorrect assumption:

```python
12 & 10
```

This is **not** decimal arithmetic. The operation is performed on the binary representation of the numbers.

---

### Using Strings

Incorrect

```python
"12" & "10"
```

### Output

```text
TypeError
```

---

## 11. Best Practices

Follow these recommended practices when using the Bitwise AND operator.

- Learn binary representation before using bitwise operators.
- Use `bin()` while learning to visualize binary values.
- Use meaningful variable names such as `mask`, `flags`, and `permissions`.
- Add comments when implementing bit masking.
- Use `and` for Boolean logic and `&` for bitwise operations.
- Keep bitwise expressions simple and readable.
