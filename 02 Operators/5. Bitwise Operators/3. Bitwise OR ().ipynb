# 🔀 Bitwise OR Operator (`|`)

The **Bitwise OR (`|`) operator** compares the corresponding bits of two integers. It returns **`1` if at least one of the corresponding bits is `1`**. It returns `0` only when **both bits are `0`**.

The Bitwise OR operator is commonly used for **setting bits**, **enabling flags**, **permission management**, **network programming**, and **embedded systems**.

# 📖 Topics Covered

## 1. Introduction to the Bitwise OR Operator

The Bitwise OR operator compares every corresponding bit of two integers.

A bit is set to **1 if either or both bits are 1**.

Example:

```text
12 = 1100
10 = 1010
--------------
     1110
```

Result:

```text
14
```

---

## 2. Syntax

### Syntax

```python
value1 | value2
```

### Explanation

- `value1` → First integer.
- `value2` → Second integer.
- `|` → Performs a bitwise OR operation.

---

## 3. Truth Table

| Bit A | Bit B | A \| B |
|------|------|--------|

## 10. Common Mistakes

Avoid these common mistakes while using the Bitwise OR operator.

### Confusing `|` with `or`

Incorrect for logical conditions:

```python
True | False
```

Preferred:

```python
True or False
```

> **Why?** `|` performs a bitwise operation, while `or` performs a logical operation.

---

### Using Floating-Point Numbers

Incorrect

```python
10.5 | 4
```

### Output

```text
TypeError
```

The `|` operator works only with integers.

---

### Forgetting Binary Representation

Incorrect assumption:

```python
12 | 10
```

The operation is performed on the binary representation of the numbers, not on their decimal values.

---

### Using Strings

Incorrect

```python
"12" | "10"
```

### Output

```text
TypeError
```

---

## 11. Best Practices

Follow these recommended practices when using the Bitwise OR operator.

- Understand binary representation before using bitwise operators.
- Use `bin()` to visualize bit patterns.
- Use meaningful variable names such as `flags`, `mask`, and `permissions`.
- Add comments when manipulating bits.
- Use `or` for Boolean logic and `|` for bitwise operations.
- Keep expressions simple and readable.
