# ➡️ Right Shift Operator (`>>`)

The **Right Shift (`>>`) operator** shifts the bits of an integer to the **right by a specified number of positions**. Each shift removes the least significant bit (rightmost bit). For **positive integers**, zeros are inserted from the left.

In most cases, shifting a positive number right by **one position divides it by 2**, shifting by two positions divides it by **4**, and so on.

The Right Shift operator is commonly used in **bit manipulation**, **performance optimization**, **embedded systems**, **network programming**, and **low-level programming**.

# 📖 Topics Covered

## 1. Introduction to the Right Shift Operator

The Right Shift operator moves every bit of a number to the **right**.

For **positive integers**, new positions on the left are filled with `0`.

Example:

```text
20 = 00010100
```

Shift right by 2 positions:

```text
00000101
```

Result:

```text
5
```

---

## 2. Syntax

### Syntax

```python
value >> positions
```

### Explanation

- `value` → Integer to shift.
- `positions` → Number of bit positions to shift right.
- `>>` → Performs the right shift operation.

---

## 3. Understanding the Binary Operation

Example:

```python
number = 20
```

Binary representation:

```text
00010100
```

After shifting right by 2:

```text
00000101
```

Result:

```python
print(number >> 2)
```

### Output

```text
5
```

---

## 4. Relationship with Division

For positive integers, each right shift divides the number by **2** (discarding any remainder).

| Expression | Equivalent | Result |
|-----------|------------|-------:|
| `20 >> 1` | `20 // 2` | 10 |
| `20 >> 2` | `20 // 4` | 5 |
| `20 >> 3` | `20 // 8` | 2 |
| `64 >> 4` | `64 // 16` | 4 |

General formula (for positive integers):

```text
number >> n = number // (2ⁿ)
```

## 7. Right Shift by Zero

Shifting by zero positions leaves the number unchanged.

## 8. Right Shift with Negative Numbers

Python preserves the sign of negative numbers during a right shift (this is known as an **arithmetic right shift**).

```python
print(-8 >> 1)
```

### Output

```text
-4
```

Another example:

```python
print(-9 >> 1)
```

### Output

```text
-5
```

> **Note:** Python extends the sign bit when right-shifting negative integers, which preserves the number's sign.

## 11. Common Mistakes

Avoid these common mistakes while using the Right Shift operator.

### Confusing Right Shift with Arithmetic Division

Although:

```python
20 >> 2
```

produces:

```text
5
```

the operator works by **shifting bits**, not by performing division directly.

---

### Using Floating-Point Numbers

Incorrect

```python
20.5 >> 2
```

### Output

```text
TypeError
```

The `>>` operator works only with integers.

---

### Using Negative Shift Counts

Incorrect

```python
20 >> -1
```

### Output

```text
ValueError
```

Shift counts must be zero or positive.

---

### Forgetting Behavior with Negative Numbers

Many beginners expect the same behavior for negative numbers as for positive numbers.

Remember that Python performs an **arithmetic right shift**, preserving the sign.
