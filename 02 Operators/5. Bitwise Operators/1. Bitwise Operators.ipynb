# ⚙️ Bitwise Operators in Python

Bitwise operators are used to **perform operations directly on the binary (bit-level) representation of integers**. They manipulate individual bits, making them extremely useful in **low-level programming**, **embedded systems**, **network programming**, **cryptography**, **device drivers**, and **performance optimization**.

Unlike arithmetic operators, which work with numbers as a whole, bitwise operators work on each binary digit (bit) individually.

Python provides **six bitwise operators**:

- `&` (Bitwise AND)
- `|` (Bitwise OR)
- `^` (Bitwise XOR)
- `~` (Bitwise NOT)
- `<<` (Left Shift)
- `>>` (Right Shift)

# 📖 Topics Covered

## 1. Introduction to Bitwise Operators

Bitwise operators work on the **binary representation** of integers.

Example:

```text
Decimal      Binary

12      ->   1100
10      ->   1010
```

Using Bitwise AND:

```text
1100
1010
----
1000
```

Result:

```text
8
```

---

## 2. Why Binary Numbers?

Computers store all data as **binary digits (0 and 1)**.

Each binary digit is called a **bit**.

Example:

```text
Decimal     Binary

1          0001
2          0010
3          0011
4          0100
5          0101
6          0110
7          0111
8          1000
```

Understanding binary makes bitwise operators much easier to understand.

---

## 3. Types of Bitwise Operators

Python provides six bitwise operators.

| Operator | Name | Description |
|----------|------|-------------|
| `&` | Bitwise AND | Sets a bit to 1 only if both bits are 1 |
| `\|` | Bitwise OR | Sets a bit to 1 if either bit is 1 |
| `^` | Bitwise XOR | Sets a bit to 1 if the bits are different |
| `~` | Bitwise NOT | Inverts every bit |
| `<<` | Left Shift | Shifts bits to the left |
| `>>` | Right Shift | Shifts bits to the right |

---

## 4. Binary Representation Example

```python
a = 12
b = 10

print(bin(a))
print(bin(b))
```

### Output

```text
0b1100
0b1010
```

The `bin()` function converts an integer into its binary representation.

## 6. Bitwise Operators Work Only with Integers

Bitwise operators are designed to work with integer values.

Correct:

```python
number = 15

print(number & 7)
```

Output

```text
7
```

Incorrect:

```python
number = 15.5

print(number & 7)
```

Output

```text
TypeError
```

## 9. Common Mistakes

Avoid these common mistakes while using bitwise operators.

### Using Floating-Point Numbers

Incorrect

```python
10.5 & 5
```

Output

```text
TypeError
```

---

### Confusing Logical and Bitwise Operators

Incorrect

```python
True & False
```

Preferred for logical conditions:

```python
True and False
```

---

### Forgetting Binary Representation

Many beginners expect decimal calculations.

Always remember:

```text
Bitwise operators work on binary digits.
```

---

### Misunderstanding Bitwise NOT

```python
print(~5)
```

Output

```text
-6
```

The `~` operator uses **two's complement representation**, so the result is not simply the opposite sign.

---

## 10. Best Practices

Follow these recommended practices when using bitwise operators.

- Learn binary representation before using bitwise operators.
- Use bitwise operators only with integers.
- Add comments when performing complex bit manipulations.
- Use meaningful variable names.
- Prefer logical operators (`and`, `or`, `not`) for Boolean logic.
- Follow Python style guidelines by adding spaces around operators.
