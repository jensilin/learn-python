# ⚡ Bitwise XOR Operator (`^`)

The **Bitwise XOR (`^`) operator** compares the corresponding bits of two integers. It returns **`1` only when the corresponding bits are different**. If both bits are the same, it returns `0`.

The Bitwise XOR operator is commonly used for **toggling bits**, **data encryption**, **error detection**, **swapping values**, and **finding unique elements** in programming.

# 📖 Topics Covered

## 1. Introduction to the Bitwise XOR Operator

The Bitwise XOR operator compares every corresponding bit of two integers.

A bit is set to **1 only when the two bits are different**.

Example:

```text
12 = 1100
10 = 1010
--------------
     0110
```

Result:

```text
6
```

---

## 2. Syntax

### Syntax

```python
value1 ^ value2
```

### Explanation

- `value1` → First integer.
- `value2` → Second integer.
- `^` → Performs a bitwise XOR operation.

## 8. Important Properties of XOR

The XOR operator has several useful properties.

### Property 1: Same Numbers

```python
print(15 ^ 15)
```

### Output

```text
0
```

Since every corresponding bit is the same, the result is `0`.

---

### Property 2: XOR with Zero

```python
print(42 ^ 0)
```

### Output

```text
42
```

---

### Property 3: Reversible Operation

```python
a = 25
b = 12

result = a ^ b

print(result ^ b)
```

### Output

```text
25
```

This property makes XOR useful in encryption and decoding.

---

## 9. Practical Examples

### Example 1: Toggle Bits

```python
number = 10
mask = 4

print(number ^ mask)
```

### Output

```text
14
```

---

### Example 2: Swap Two Numbers Without a Temporary Variable

```python
a = 10
b = 20

a = a ^ b
b = a ^ b
a = a ^ b

print(a, b)
```

### Output

```text
20 10
```

> **Note:** Although this technique works, using tuple unpacking (`a, b = b, a`) is more readable and preferred in Python.

---

### Example 3: Find the Unique Number

```python
numbers = [5, 2, 5, 3, 2]

result = 0

for num in numbers:
    result ^= num

print(result)
```

### Output

```text
3
```

---

### Example 4: Simple Encryption

```python
data = 65
key = 12

encrypted = data ^ key
decrypted = encrypted ^ key

print(encrypted)
print(decrypted)
```

### Output

```text
77
65
```

## 11. Common Mistakes

Avoid these common mistakes while using the Bitwise XOR operator.

### Confusing `^` with Exponentiation

Incorrect

```python
print(2 ^ 3)
```

### Output

```text
1
```

Correct exponentiation:

```python
print(2 ** 3)
```

### Output

```text
8
```

> **Why?** `^` is the Bitwise XOR operator, while `**` is the exponentiation operator.

---

### Using Floating-Point Numbers

Incorrect

```python
10.5 ^ 4
```

### Output

```text
TypeError
```

The `^` operator works only with integers.

---

### Forgetting Binary Representation

Remember that XOR compares **binary bits**, not decimal digits.

---

### Using Strings

Incorrect

```python
"10" ^ "5"
```

### Output

```text
TypeError
```

---

## 12. Best Practices

Follow these recommended practices when using the Bitwise XOR operator.

- Understand binary representation before using XOR.
- Use `bin()` to visualize bit patterns.
- Prefer tuple unpacking (`a, b = b, a`) instead of XOR swapping in Python.
- Use meaningful variable names such as `mask`, `key`, and `flags`.
- Add comments when performing complex bit manipulations.
- Keep XOR expressions simple and readable.
