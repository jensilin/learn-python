# 🔄 Bitwise NOT Operator (`~`)

The **Bitwise NOT (`~`) operator** is a **unary operator** that inverts every bit of an integer. It changes **all `1`s to `0`s** and **all `0`s to `1`s**.

Unlike other bitwise operators that require two operands, the Bitwise NOT operator works on **only one integer**.

The Bitwise NOT operator is commonly used in **bit manipulation**, **mask generation**, **embedded systems**, **network programming**, and **low-level programming**.

# 📖 Topics Covered

## 1. Introduction to the Bitwise NOT Operator

The Bitwise NOT operator flips every bit of a number.

Example:

```text
12 = 00001100
```

After applying `~`:

```text
11110011
```

Since Python represents negative integers using **two's complement**, the result is:

```text
-13
```

---

## 2. Syntax

### Syntax

```python
~value
```

### Explanation

- `value` → Integer value.
- `~` → Inverts every bit.

---

## 3. How Bitwise NOT Works

The Bitwise NOT operator follows this mathematical rule:

```text
~x = -(x + 1)
```

Example:

```python
print(~12)
```

Calculation:

```text
~12
= -(12 + 1)
= -13
```

### Output

```text
-13
```

## 6. Viewing Binary Values

Use the `bin()` function to view binary representations.

```python
number = 12

print(bin(number))
print(bin(~number))
```

> **Note:** Python displays negative binary numbers with a minus sign (`-0b...`). Internally, negative integers are represented using **two's complement**.

## 8. Why Does Bitwise NOT Return Negative Numbers?

Many beginners expect the opposite sign.

However, Python uses **two's complement representation** for integers.

The Bitwise NOT operator follows:

```text
~x = -(x + 1)
```

Examples:

| Expression | Result |
|-----------|--------|
| `~0` | `-1` |
| `~1` | `-2` |
| `~5` | `-6` |
| `~10` | `-11` |
| `~20` | `-21` |

## 10. Common Mistakes

Avoid these common mistakes while using the Bitwise NOT operator.

### Expecting Positive Results

Incorrect expectation:

```python
~5
```

Expected:

```text
6
```

Actual:

```text
-6
```

Remember:

```text
~x = -(x + 1)
```

---

### Using Floating-Point Numbers

Incorrect

```python
~10.5
```

### Output

```text
TypeError
```

The `~` operator works only with integers.

---

### Forgetting Two's Complement

The Bitwise NOT operator works on the binary representation of integers, not on their decimal values.

---

### Using Strings

Incorrect

```python
~"10"
```

### Output

```text
TypeError
```
