# ➖ Subtract Assignment Operator (`-=`)

The **Subtract Assignment (`-=`) operator** is a compound assignment operator in Python used to **subtract a value from a variable and assign the updated result back to the same variable**.

Instead of writing `x = x - value`, you can write `x -= value`, making your code shorter, cleaner, and easier to read.

The Subtract Assignment operator works with numeric data types such as **integers (`int`)**, **floating-point numbers (`float`)**, and **complex numbers (`complex`)**.

# 📖 Topics Covered

## 1. Introduction to the Subtract Assignment Operator

The Subtract Assignment operator (`-=`) subtracts the right-hand operand from the current value of the variable and stores the result back in the same variable.

Instead of writing:

```python
balance = balance - 250
```

you can simply write:

```python
balance -= 250
```

Both statements produce the same result.

---

## 2. Syntax

### Syntax

```python
variable -= value
```

### Equivalent To

```python
variable = variable - value
```

### Explanation

- `variable` → Variable whose value will be updated.
- `-=` → Subtract Assignment operator.
- `value` → Value to be subtracted.

---

## 3. Subtracting Integers

The Subtract Assignment operator works with integers.

**Example**

```python
marks = 95

marks -= 10

print(marks)
```

### Output

```text
85
```

---

## 4. Subtracting Floating-Point Numbers

The operator also works with floating-point numbers.

**Example**

```python
temperature = 37.5

temperature -= 2.5

print(temperature)
```

### Output

```text
35.0
```

---

## 5. Subtracting Complex Numbers

The operator supports complex numbers.

**Example**

```python
num = 5 + 3j

num -= 2 + 1j

print(num)
```

### Output

```text
(3+2j)
```

## 8. Common Mistakes

Avoid these common mistakes while using the Subtract Assignment operator.

### Using an Uninitialized Variable

**Incorrect**

```python
count -= 1
```

### Output

```text
NameError
```

**Correct**

```python
count = 10

count -= 1
```

---

### Subtracting Incompatible Data Types

```python
number = 100

number -= "20"
```

### Output

```text
TypeError
```

---

### Forgetting Spaces Around Operators

Although valid,

```python
balance-=100
```

is less readable than

```python
balance -= 100
```

---

## 9. Best Practices

Follow these recommended practices when using the Subtract Assignment operator.

- Initialize variables before using `-=`.
- Use `-=` instead of `variable = variable - value` for better readability.
- Ensure operands are compatible numeric types.
- Use meaningful variable names.
- Follow Python style guidelines by adding spaces around operators.
