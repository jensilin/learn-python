# ⬆️ Exponent Assignment Operator (`**=`)

The **Exponent Assignment (`**=`) operator** is a compound assignment operator in Python used to **raise a variable to a specified power and assign the updated result back to the same variable**.

Instead of writing `x = x ** value`, you can write `x **= value`, making your code shorter, cleaner, and easier to read.

The Exponent Assignment operator works with **integers (`int`)**, **floating-point numbers (`float`)**, and **complex numbers (`complex`)**.

# 📖 Topics Covered

## 1. Introduction to the Exponent Assignment Operator

The Exponent Assignment operator (`**=`) raises the current value of a variable to a specified power and stores the result back in the same variable.

Instead of writing:

```python
number = number ** 2
```

you can simply write:

```python
number **= 2
```

Both statements produce the same result.

---

## 2. Syntax

### Syntax

```python
variable **= value
```

### Equivalent To

```python
variable = variable ** value
```

### Explanation

- `variable` → Variable whose value will be updated.
- `**=` → Exponent Assignment operator.
- `value` → Power (exponent).

---

## 3. Exponent Assignment with Integers

The operator works with integer values.

**Example**

```python
number = 4

number **= 3

print(number)
```

### Output

```text
64
```

---

## 4. Exponent Assignment with Floating-Point Numbers

The operator also works with floating-point numbers.

**Example**

```python
value = 2.5

value **= 2

print(value)
```

### Output

```text
6.25
```

---

## 5. Exponent Assignment with Complex Numbers

The operator supports complex numbers.

**Example**

```python
number = 2 + 1j

number **= 2

print(number)
```

### Output

```text
(3+4j)
```

## 9. Common Mistakes

Avoid these common mistakes while using the Exponent Assignment operator.

### Using an Uninitialized Variable

**Incorrect**

```python
number **= 2
```

### Output

```text
NameError
```

**Correct**

```python
number = 5

number **= 2
```

---

### Confusing `^=` with `**=`

```python
number = 2

number ^= 3

print(number)
```

### Output

```text
1
```

> **Why?** `^=` performs a **Bitwise XOR Assignment**, not exponentiation.

Correct:

```python
number = 2

number **= 3

print(number)
```

### Output

```text
8
```

---

### Using Very Large Exponents

```python
number = 2

number **= 100

print(number)
```

### Output

```text
1267650600228229401496703205376
```

> **Note:** Very large exponents can produce extremely large numbers and may increase computation time.

## 10. Best Practices

Follow these recommended practices when using the Exponent Assignment operator.

- Initialize variables before using `**=`.
- Use `**=` instead of `variable = variable ** value` for cleaner code.
- Be careful with very large exponents.
- Use meaningful variable names.
- Follow Python style guidelines by adding spaces around operators.
