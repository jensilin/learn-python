# 🧮 Modulus Assignment Operator (`%=`)

The **Modulus Assignment (`%=`) operator** is a compound assignment operator in Python used to **find the remainder after division and assign the result back to the same variable**.

Instead of writing `x = x % value`, you can write `x %= value`, making your code shorter, cleaner, and easier to read.

The Modulus Assignment operator works with **integers (`int`)** and **floating-point numbers (`float`)**. It is commonly used for checking divisibility, cyclic operations, time calculations, and remainder-based algorithms.

# 📖 Topics Covered

## 1. Introduction to the Modulus Assignment Operator

The Modulus Assignment operator (`%=`) finds the remainder after dividing the current value of a variable by another value and stores the result back in the same variable.

Instead of writing:

```python
number = number % 5
```

you can simply write:

```python
number %= 5
```

Both statements produce the same result.

---

## 2. Syntax

### Syntax

```python
variable %= value
```

### Equivalent To

```python
variable = variable % value
```

### Explanation

- `variable` → Variable whose value will be updated.
- `%=` → Modulus Assignment operator.
- `value` → Value used as the divisor.

---

## 3. Modulus Assignment with Integers

The operator works with integer values.

**Example**

```python
number = 29

number %= 6

print(number)
```

### Output

```text
5
```

---

## 4. Modulus Assignment with Floating-Point Numbers

The operator also works with floating-point numbers.

**Example**

```python
value = 10.5

value %= 3

print(value)
```

### Output

```text
1.5
```

---

## 5. Modulus Assignment with Negative Numbers

Python calculates the remainder so that the result has the **same sign as the divisor**.

**Example**

```python
number = -17

number %= 5

print(number)
```

### Output

```text
3
```

## 8. Common Mistakes

Avoid these common mistakes while using the Modulus Assignment operator.

### Using an Uninitialized Variable

**Incorrect**

```python
number %= 5
```

### Output

```text
NameError
```

**Correct**

```python
number = 20

number %= 5
```

---

### Dividing by Zero

```python
number = 15

number %= 0
```

### Output

```text
ZeroDivisionError
```

---

### Confusing `%=` with `/=`

```python
number = 17

number /= 5

print(number)
```

### Output

```text
3.4
```

Using Modulus Assignment:

```python
number = 17

number %= 5

print(number)
```

### Output

```text
2
```

---

### Ignoring Negative Number Behavior

```python
number = -10

number %= 3

print(number)
```

### Output

```text
2
```

> **Why?** Python returns a remainder with the same sign as the divisor.

## 9. Best Practices

Follow these recommended practices when using the Modulus Assignment operator.

- Initialize variables before using `%=`.
- Ensure the divisor is not zero.
- Use `%=` when only the remainder is required.
- Be aware of how Python handles negative numbers.
- Use meaningful variable names.
- Follow Python style guidelines by adding spaces around operators.
