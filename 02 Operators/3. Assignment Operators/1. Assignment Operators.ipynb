# 📝 Assignment Operators

Assignment operators are used to **assign values to variables** in Python. They can also perform an operation and assign the result back to the same variable.

Python provides **simple assignment** and **compound assignment operators**, making code shorter, cleaner, and easier to read.

Assignment operators are widely used in arithmetic calculations, loops, counters, accumulators, bitwise operations, and many real-world programming tasks.

# 📖 Topics Covered

## 1. Introduction to Assignment Operators

Assignment operators are used to store values in variables.

The simplest assignment operator is the **Assignment (`=`)** operator, which assigns a value to a variable.

Python also provides compound assignment operators that combine an arithmetic or bitwise operation with assignment.

---

## 2. Assignment Operators in Python

Python provides the following assignment operators.

| Operator | Name | Example | Equivalent To |
|----------|------|---------|---------------|
| `=` | Assignment | `x = 10` | Assign value |
| `+=` | Add Assignment | `x += 5` | `x = x + 5` |
| `-=` | Subtract Assignment | `x -= 5` | `x = x - 5` |
| `*=` | Multiply Assignment | `x *= 5` | `x = x * 5` |
| `/=` | Divide Assignment | `x /= 5` | `x = x / 5` |
| `//=` | Floor Divide Assignment | `x //= 5` | `x = x // 5` |
| `%=` | Modulus Assignment | `x %= 5` | `x = x % 5` |
| `**=` | Exponent Assignment | `x **= 5` | `x = x ** 5` |
| `&=` | Bitwise AND Assignment | `x &= 5` | `x = x & 5` |
| `|=` | Bitwise OR Assignment | `x |= 5` | `x = x \| 5` |
| `^=` | Bitwise XOR Assignment | `x ^= 5` | `x = x ^ 5` |
| `<<=` | Left Shift Assignment | `x <<= 2` | `x = x << 2` |
| `>>=` | Right Shift Assignment | `x >>= 2` | `x = x >> 2` |

---

## 3. How Assignment Operators Work

Assignment operators first perform the specified operation (if any) and then store the result back into the same variable.

**Example**

```python
x = 8

x *= 3

print(x)
```

### Output

```text
24
```

---

## 4. Why Use Compound Assignment?

Instead of writing:

```python
x = x + 10
```

you can simply write:

```python
x += 10
```

Both statements produce the same result.

Compound assignment operators make code:

- Shorter
- Cleaner
- Easier to read
- Easier to maintain

## 7. Common Mistakes

Avoid these common mistakes while using assignment operators.

### Using a Variable Before Assignment

```python
count += 1
```

### Output

```text
NameError
```

**Correct**

```python
count = 0
count += 1
```

---

### Confusing `=` with `==`

```python
x = 10
```

assigns a value.

```python
x == 10
```

compares values.

---

### Using Compound Assignment on Incompatible Data Types

```python
number = 10

number += "5"
```

### Output

```text
TypeError
```
