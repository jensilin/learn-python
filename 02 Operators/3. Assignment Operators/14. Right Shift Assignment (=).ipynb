# ➡️ Right Shift Assignment Operator (`>>=`)

The **Right Shift Assignment (`>>=`) operator** is a compound assignment operator in Python used to **shift the bits of an integer to the right by a specified number of positions and assign the updated result back to the same variable**.

Instead of writing `x = x >> value`, you can write `x >>= value`, making your code shorter, cleaner, and easier to read.

The Right Shift Assignment operator works only with **integer (`int`)** values because bitwise operations are performed on the binary representation of numbers.

# 📖 Topics Covered

## 1. Introduction to the Right Shift Assignment Operator

The Right Shift Assignment operator (`>>=`) shifts all bits of an integer to the **right** by a specified number of positions and stores the result back in the same variable.

Instead of writing:

```python
number = number >> 2
```

you can simply write:

```python
number >>= 2
```

Both statements produce the same result.

---

## 2. Syntax

### Syntax

```python
variable >>= value
```

### Equivalent To

```python
variable = variable >> value
```

### Explanation

- `variable` → Integer variable whose value will be updated.
- `>>=` → Right Shift Assignment operator.
- `value` → Number of bit positions to shift.

---

## 3. Understanding Right Shift

A **right shift** moves all bits to the right by the specified number of positions.

For positive integers, every right shift by **1 position** is approximately equivalent to dividing the number by **2** and discarding the remainder.

For example:

| Operation | Result |
|-----------|--------|
| `20 >> 1` | 10 |
| `20 >> 2` | 5 |
| `20 >> 3` | 2 |

---

## 4. Example with Binary Representation

```python
number = 20

number >>= 2

print(number)
```

### Binary Calculation

```text
Before Shift

20 = 00010100

Right Shift by 2

00000101
```

### Output

```text
5
```

---

## 5. Another Example

```python
value = 32

value >>= 3

print(value)
```

### Binary Calculation

```text
32 = 00100000

Shift Right by 3

00000100
```

### Output

```text
4
```

## 6. Right Shift with Negative Numbers

Python preserves the sign bit when shifting negative integers.

**Example**

```python
number = -16

number >>= 2

print(number)
```

### Output

```text
-4
```

> **Note:** Python performs an **arithmetic right shift**, preserving the sign of negative numbers.

## 9. Common Mistakes

Avoid these common mistakes while using the Right Shift Assignment operator.

### Using an Uninitialized Variable

**Incorrect**

```python
number >>= 2
```

### Output

```text
NameError
```

**Correct**

```python
number = 20

number >>= 2
```

---

### Using Floating-Point Numbers

```python
number = 20.5

number >>= 1
```

### Output

```text
TypeError
```

> **Why?** Bitwise shift operators work only with integers.

---

### Using a Negative Shift Count

```python
number = 16

number >>= -1
```

### Output

```text
ValueError: negative shift count
```

---

### Assuming It Always Means Division

Although right shifting positive integers often behaves like division by powers of 2, it actually performs a **bitwise shift**.

```python
number = 18

number >>= 2

print(number)
```

### Output

```text
4
```

The mathematical division would be:

```python
18 / 4
```

Output:

```text
4.5
```

The right shift discards fractional parts because it operates on binary integers.
