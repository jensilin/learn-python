# 📝 Assignment Operator (`=`)

The **Assignment (`=`) operator** is the most fundamental assignment operator in Python. It is used to **assign a value to a variable**.

When Python executes an assignment statement, the value on the **right-hand side** is evaluated first, and the resulting value is stored in the variable on the **left-hand side**.

The Assignment operator is used throughout every Python program to store data such as numbers, strings, lists, dictionaries, objects, and function results.

# 📖 Topics Covered

## 1. Introduction to the Assignment Operator

The Assignment operator (`=`) stores a value in a variable.

It does **not** compare two values. It simply assigns the value on the right side to the variable on the left side.

**Example**

```python
age = 25

print(age)
```

### Output

```text
25
```

---

## 2. Syntax

### Syntax

```python
variable = value
```

### Explanation

- `variable` → Name used to store data.
- `=` → Assignment operator.
- `value` → Data to be assigned.

Python first evaluates the value and then stores it in the variable.

## 8. Multiple Assignment

Python allows multiple variables to be assigned in a single statement.

**Example**

```python
x, y, z = 10, 20, 30

print(x)
print(y)
print(z)
```

### Output

```text
10
20
30
```

---

## 9. Assigning the Same Value to Multiple Variables

The same value can be assigned to multiple variables.

**Example**

```python
a = b = c = 100

print(a)
print(b)
print(c)
```

### Output

```text
100
100
100
```

## 12. Common Mistakes

Avoid these common mistakes while using the Assignment operator.

### Using `==` Instead of `=`

**Incorrect**

```python
name == "John"
```

### Output

```text
NameError
```

> **Why?** The `==` operator compares values, while `=` assigns a value to a variable.

---

### Using Invalid Variable Names

**Incorrect**

```python
1name = "Alice"
```

### Output

```text
SyntaxError
```

---

### Using Undefined Variables

```python
print(age)
```

### Output

```text
NameError
```

> **Why?** A variable must be assigned a value before it can be used.

## 13. Best Practices

Follow these recommended practices when using the Assignment operator.
