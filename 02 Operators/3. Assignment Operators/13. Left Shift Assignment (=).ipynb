# ⬅️ Left Shift Assignment Operator (`<<=`)

The **Left Shift Assignment (`<<=`) operator** is a compound assignment operator in Python used to **shift the bits of an integer to the left by a specified number of positions and assign the updated result back to the same variable**.

Instead of writing `x = x << value`, you can write `x <<= value`, making your code shorter, cleaner, and easier to read.

The Left Shift Assignment operator works only with **integer (`int`)** values because bitwise operations are performed on the binary representation of numbers.

# 📖 Topics Covered

## 1. Introduction to the Left Shift Assignment Operator

The Left Shift Assignment operator (`<<=`) shifts all bits of an integer to the **left** by a specified number of positions and stores the result back in the same variable.

Instead of writing:

```python
number = number << 2
```

you can simply write:

```python
number <<= 2
```

Both statements produce the same result.

---

## 2. Syntax

### Syntax

```python
variable <<= value
```

### Equivalent To

```python
variable = variable << value
```

### Explanation

- `variable` → Integer variable whose value will be updated.
- `<<=` → Left Shift Assignment operator.
- `value` → Number of bit positions to shift.

---

## 3. Understanding Left Shift

A **left shift** moves all bits to the left by the specified number of positions.

For every left shift by **1 position**, the value is approximately multiplied by **2**.

For example:

| Operation | Result |
|-----------|--------|
| `5 << 1` | 10 |
| `5 << 2` | 20 |
| `5 << 3` | 40 |

---

## 4. Example with Binary Representation

```python
number = 5

number <<= 2

print(number)
```

### Binary Calculation

```text
Before Shift

5 = 00000101

Left Shift by 2

00010100
```

### Output

```text
20
```

## 8. Common Mistakes

Avoid these common mistakes while using the Left Shift Assignment operator.

### Using an Uninitialized Variable

**Incorrect**

```python
number <<= 2
```

### Output

```text
NameError
```

**Correct**

```python
number = 5

number <<= 2
```

---

### Using Floating-Point Numbers

```python
number = 5.5

number <<= 1
```

### Output

```text
TypeError
```

> **Why?** Bitwise shift operators work only with integers.

---

### Using a Negative Shift Count

```python
number = 10

number <<= -1
```

### Output

```text
ValueError: negative shift count
```

---

### Assuming It Always Means Multiplication

Although left shifting by one position is equivalent to multiplying by 2 for positive integers, the operator actually performs a **bitwise shift**, not arithmetic multiplication.

```python
number = 3

number <<= 2

print(number)
```

### Output

```text
12
```

---

## 9. Best Practices

Follow these recommended practices when using the Left Shift Assignment operator.

- Use `<<=` only with integer values.
- Understand binary representation before using shift operators.
- Avoid negative shift counts.
- Use meaningful variable names.
- Add comments when performing bit manipulation.
- Follow Python style guidelines by adding spaces around operators.
