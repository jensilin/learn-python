# ✖️ Multiply Assignment Operator (`*=`)

The **Multiply Assignment (`*=`) operator** is a compound assignment operator in Python used to **multiply a variable by a value and assign the updated result back to the same variable**.

Instead of writing `x = x * value`, you can write `x *= value`, making your code shorter, cleaner, and easier to read.

The Multiply Assignment operator works with **integers (`int`)**, **floating-point numbers (`float`)**, **complex numbers (`complex`)**, and sequence types like **strings (`str`)**, **lists (`list`)**, and **tuples (`tuple`)** when multiplied by an integer.

# 📖 Topics Covered

## 1. Introduction to the Multiply Assignment Operator

The Multiply Assignment operator (`*=`) multiplies the current value of a variable by another value and stores the result back in the same variable.

Instead of writing:

```python
quantity = quantity * 5
```

you can simply write:

```python
quantity *= 5
```

Both statements produce the same result.

---

## 2. Syntax

### Syntax

```python
variable *= value
```

### Equivalent To

```python
variable = variable * value
```

### Explanation

- `variable` → Variable whose value will be updated.
- `*=` → Multiply Assignment operator.
- `value` → Value used for multiplication.

---

## 3. Multiplying Integers

The Multiply Assignment operator works with integers.

**Example**

```python
number = 12

number *= 4

print(number)
```

### Output

```text
48
```

---

## 4. Multiplying Floating-Point Numbers

The operator also works with floating-point numbers.

**Example**

```python
price = 19.5

price *= 2

print(price)
```

### Output

```text
39.0
```

---

## 5. Multiplying Complex Numbers

The operator supports complex numbers.

**Example**

```python
num = 2 + 3j

num *= 2

print(num)
```

### Output

```text
(4+6j)
```

---

## 6. Multiplying Strings

When multiplied by an integer, strings are repeated.

**Example**

```python
text = "Hi "

text *= 3

print(text)
```

### Output

```text
Hi Hi Hi
```

---

## 7. Multiplying Lists

Lists can also be repeated.

**Example**

```python
numbers = [1, 2]

numbers *= 3

print(numbers)
```

### Output

```text
[1, 2, 1, 2, 1, 2]
```

---

## 8. Multiplying Tuples

Tuples can be repeated using the Multiply Assignment operator.

**Example**

```python
colors = ("Red", "Blue")

colors *= 2

print(colors)
```

### Output

```text
('Red', 'Blue', 'Red', 'Blue')
```

## 11. Common Mistakes

Avoid these common mistakes while using the Multiply Assignment operator.

### Using an Uninitialized Variable

**Incorrect**

```python
count *= 2
```

### Output

```text
NameError
```

**Correct**

```python
count = 5

count *= 2
```

---

### Multiplying Incompatible Data Types

```python
number = 10

number *= "5"
```

### Output

```text
TypeError
```

---

### Forgetting Spaces Around Operators

Although valid,

```python
value*=4
```

is less readable than

```python
value *= 4
```

---

## 12. Best Practices

Follow these recommended practices when using the Multiply Assignment operator.

- Initialize variables before using `*=`.
- Use `*=` instead of `variable = variable * value` for cleaner code.
- Ensure operands are compatible.
- Use meaningful variable names.
- Follow Python style guidelines by adding spaces around operators.
