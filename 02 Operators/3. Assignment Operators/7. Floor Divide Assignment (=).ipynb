# ⌊⌋ Floor Divide Assignment Operator (`//=`)

The **Floor Divide Assignment (`//=`) operator** is a compound assignment operator in Python used to **perform floor division on a variable and assign the updated result back to the same variable**.

Instead of writing `x = x // value`, you can write `x //= value`, making your code shorter, cleaner, and easier to read.

The Floor Divide Assignment operator works with **integers (`int`)** and **floating-point numbers (`float`)**. It returns the **floor value** (largest whole number less than or equal to the quotient).

# 📖 Topics Covered

## 1. Introduction to the Floor Divide Assignment Operator

The Floor Divide Assignment operator (`//=`) divides the current value of a variable by another value using **floor division** and stores the result back in the same variable.

Instead of writing:

```python
number = number // 4
```

you can simply write:

```python
number //= 4
```

Both statements produce the same result.

---

## 2. Syntax

### Syntax

```python
variable //= value
```

### Equivalent To

```python
variable = variable // value
```

### Explanation

- `variable` → Variable whose value will be updated.
- `//=` → Floor Divide Assignment operator.
- `value` → Value used for floor division.

---

## 3. Floor Divide Assignment with Integers

The operator works with integer values.

**Example**

```python
count = 22

count //= 5

print(count)
```

### Output

```text
4
```

---

## 4. Floor Divide Assignment with Floating-Point Numbers

The operator also works with floating-point numbers.

**Example**

```python
price = 18.5

price //= 3

print(price)
```

### Output

```text
6.0
```

> **Note:** If either operand is a `float`, the result is also returned as a `float`.

---

## 5. Floor Divide Assignment with Negative Numbers

The operator rounds the result **down toward negative infinity**.

**Example**

```python
value = -17

value //= 5

print(value)
```

### Output

```text
-4
```

## 8. Common Mistakes

Avoid these common mistakes while using the Floor Divide Assignment operator.

### Using an Uninitialized Variable

**Incorrect**

```python
count //= 2
```

### Output

```text
NameError
```

**Correct**

```python
count = 10

count //= 2
```

---

### Dividing by Zero

```python
number = 20

number //= 0
```

### Output

```text
ZeroDivisionError
```

---

### Confusing `//=` with `/=`

```python
number = 15

number /= 4

print(number)
```

### Output

```text
3.75
```

Using floor division:

```python
number = 15

number //= 4

print(number)
```

### Output

```text
3
```

---

### Ignoring Negative Numbers

```python
value = -15

value //= 4

print(value)
```

### Output

```text
-4
```

> **Why?** Floor division always rounds **down**, not toward zero.

---

## 9. Best Practices

Follow these recommended practices when using the Floor Divide Assignment operator.

- Initialize variables before using `//=`.
- Use `//=` only when the whole-number quotient is required.
- Check that the divisor is not zero.
- Be careful when working with negative numbers.
- Use meaningful variable names.
- Follow Python style guidelines by adding spaces around operators.
