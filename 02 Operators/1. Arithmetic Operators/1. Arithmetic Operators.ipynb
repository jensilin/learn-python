# ➕ Arithmetic Operators

Arithmetic operators are used to perform **mathematical calculations** on numbers. They are among the most frequently used operators in Python and help perform operations such as addition, subtraction, multiplication, division, finding the remainder, exponentiation, and floor division.

Python provides several built-in arithmetic operators that work with integers (`int`), floating-point numbers (`float`), and complex numbers (`complex`) where applicable.

# 📖 Topics Covered

## 1. Introduction to Arithmetic Operators

Arithmetic operators perform mathematical operations on operands.

They are commonly used for:

- Calculations
- Scientific computing
- Data analysis
- Financial applications
- Games
- Automation

Python supports the following arithmetic operators.

| Operator | Name | Example |
|----------|------|---------|
| `+` | Addition | `a + b` |
| `-` | Subtraction | `a - b` |
| `*` | Multiplication | `a * b` |
| `/` | Division | `a / b` |
| `//` | Floor Division | `a // b` |
| `%` | Modulus | `a % b` |
| `**` | Exponentiation | `a ** b` |

## 2. How Arithmetic Operators Work

Arithmetic operators work on numeric operands.

## 4. Working with Different Numeric Data Types

Arithmetic operators support multiple numeric data types.

### Integer

```python
print(10 + 5)
```

### Output

```text
15
```

### Float

```python
print(10.5 + 5.2)
```

### Output

```text
15.7
```

### Complex

```python
print((2 + 3j) + (4 + 2j))
```

### Output

```text
(6+5j)
```

## 7. Common Mistakes

Avoid these common mistakes while using arithmetic operators.

### Mistakes

- Dividing by zero.
- Confusing `/` with `//`.
- Using the wrong operator precedence.
- Forgetting parentheses in complex expressions.

**Incorrect**

```python
print(10 / 0)
```

### Output

```text
ZeroDivisionError
```

## 8. Best Practices

Follow these recommended practices when using arithmetic operators.

- Use meaningful variable names.
- Use parentheses to improve readability.
- Choose the correct operator for the required calculation.
- Handle division by zero when necessary.
- Keep mathematical expressions simple.
