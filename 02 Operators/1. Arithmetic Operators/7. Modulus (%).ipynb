# 🧮 Modulus Operator (`%`)

The **Modulus (`%`) operator** is an arithmetic operator in Python used to **find the remainder** after dividing one number by another.

Instead of returning the quotient, the Modulus operator returns the **remaining value** after the division is completed. It is widely used in programming for checking even or odd numbers, cyclic operations, time calculations, hashing, and many algorithmic problems.

The Modulus operator works with **integers (`int`)** and **floating-point numbers (`float`)**.

# 📖 Topics Covered

## 1. Introduction to the Modulus Operator

The Modulus operator (`%`) returns the **remainder** after division.

It does **not** return the quotient.

**Example**

```python
print(20 % 6)
```

### Output

```text
2
```

**Explanation**

```text
20 ÷ 6 = 3 remainder 2
```

---

## 2. Syntax

### Syntax

```python
operand1 % operand2
```

### Explanation

- `operand1` → Dividend (number being divided)
- `%` → Modulus operator
- `operand2` → Divisor

The result is the remainder after division.

---

## 3. Modulus with Integers

The Modulus operator works with integer values.

**Example**

```python
a = 25
b = 7

print(a % b)
```

### Output

```text
4
```

---

## 4. Modulus with Floating-Point Numbers

The Modulus operator also works with decimal numbers.

**Example**

```python
a = 10.5
b = 3

print(a % b)
```

### Output

```text
1.5
```

---

## 5. Modulus with Negative Numbers

Python follows the rule that the remainder has the **same sign as the divisor**.

**Example**

```python
print(-17 % 5)
```

### Output

```text
3
```

**Explanation**

```text
-17 = (5 × -4) + 3
```

Another example:

```python
print(17 % -5)
```

### Output

```text
-3
```

## 7. Practical Examples

Apply the Modulus operator through practical programming examples.

### Example 1: Check Even or Odd

```python
number = 14

if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

### Output

```text
Even
```

---

### Example 2: Check Divisibility

```python
number = 25

if number % 5 == 0:
    print("Divisible by 5")
```

### Output

```text
Divisible by 5
```

---

### Example 3: Convert Minutes to Hours

```python
minutes = 135

remaining_minutes = minutes % 60

print(remaining_minutes)
```

### Output

```text
15
```

## 9. Common Mistakes

Avoid these common mistakes while using the Modulus operator.

### Dividing by Zero

**Incorrect**

```python
print(15 % 0)
```

### Output

```text
ZeroDivisionError
```

---

### Confusing Modulus with Division

**Example**

```python
print(10 / 3)
print(10 % 3)
```

### Output

```text
3.3333333333333335
1
```

---

### Ignoring Negative Number Behavior

**Example**

```python
print(-10 % 3)
```

### Output

```text
2
```

> **Why?** Python calculates the remainder so that the result has the same sign as the divisor.
