# ➗ Division Operator (`/`)

The **Division (`/`) operator** is an arithmetic operator in Python used to **divide one number by another**. It returns the quotient of the division.

Unlike some programming languages, the Division operator in Python **always returns a floating-point (`float`) value**, even when the result is a whole number.

The Division operator works with **integers (`int`)**, **floating-point numbers (`float`)**, and **complex numbers (`complex`)** (except division by zero).

# 📖 Topics Covered

## 1. Introduction to the Division Operator

The Division operator (`/`) divides the first operand by the second operand.

The result is **always a floating-point number**, regardless of whether the operands are integers or floats.

**Example**

```python
x = 15
y = 3

print(x / y)
```

### Output

```text
5.0
```

---

## 2. Syntax

### Syntax

```python
operand1 / operand2
```

### Explanation

- `operand1` → Dividend (number being divided).
- `/` → Division operator.
- `operand2` → Divisor (number used for division).

The result is the quotient.

---

## 3. Division with Integers

Integer division using `/` always produces a float.

**Example**

```python
a = 18
b = 4

print(a / b)
```

### Output

```text
4.5
```

---

## 4. Division with Floating-Point Numbers

The Division operator works with decimal values.

**Example**

```python
a = 15.5
b = 2.5

print(a / b)
```

### Output

```text
6.2
```

---

## 5. Division with Complex Numbers

The Division operator also supports complex numbers.

**Example**

```python
a = 4 + 2j
b = 1 + 1j

print(a / b)
```

### Output

```text
(3-1j)
```

---

## 6. Division by Zero

Division by zero is not allowed in Python.

Attempting to divide by zero raises a **ZeroDivisionError**.

**Example**

```python
print(10 / 0)
```

### Output

```text
ZeroDivisionError: division by zero
```

<h4>b) Floor Division (<code>//</code>)</h4>

<p>
The <code>//</code> operator performs <b>floor division</b>. It returns the quotient after <b>discarding the fractional (decimal) part</b> of the result.
</p>

<h5>Syntax</h5>

```python
operand1 // operand2
```

<h5>Example</h5>

```python
print(10 // 2)
print(7 // 2)
print(5 // 2)
print(-7 // 2)
```

<h5>Output</h5>

```text
5
3
2
-4
```

<h5>Note</h5>

<ul>
    <li>The <code>//</code> operator returns the <b>floor value</b> (the greatest integer less than or equal to the actual quotient).</li>
    <li>If either operand is a <code>float</code>, the result is also returned as a <code>float</code>.</li>
    <li>Division by <code>0</code> raises a <code>ZeroDivisionError</code>.</li>
</ul>

````python
750//3
````

<h4>c) Modulus (<code>%</code>) Operator</h4>

<p>
The <code>%</code> operator performs the <b>modulus</b> operation. It returns the <b>remainder</b> after dividing one operand by another.
</p>

<h5>Syntax</h5>

```python
operand1 % operand2
```

<h5>Example</h5>

```python
print(10 % 3)
print(20 % 6)
print(15 % 5)
```

<h5>Output</h5>

```text
1
2
0
```

<h5>Note</h5>

<ul>
    <li>The <code>%</code> operator returns the <b>remainder</b> of the division.</li>
    <li>If the dividend is exactly divisible by the divisor, the result is <code>0</code>.</li>
    <li>Division by <code>0</code> raises a <code>ZeroDivisionError</code>.</li>
</ul>

````python
37 % 4
````

<h5>Note</h5>

<ul>
    <li>None of the division operators (<code>/</code>, <code>//</code>, and <code>%</code>) support <b>collection data types</b> such as strings, lists, tuples, sets, or dictionaries.</li>
    <li>The <code>//</code> (floor division) and <code>%</code> (modulus) operators do <b>not</b> support <code>complex</code> numbers.</li>
    <li>The <code>/</code> (true division) operator supports <code>complex</code> numbers.</li>
</ul>
