# ✖️ Multiplication Operator (`*`)

The **Multiplication (`*`) operator** is an arithmetic operator in Python used to **multiply two operands**. It calculates the product of two numeric values and is commonly used in mathematical calculations, finance, engineering, scientific computing, and data analysis.

In addition to numeric multiplication, the `*` operator can also be used to **repeat sequences** such as strings, lists, and tuples.

# 📖 Topics Covered

## 1. Introduction to the Multiplication Operator

The Multiplication operator (`*`) is used to calculate the **product** of two operands.

It is one of the most frequently used arithmetic operators in Python.

The operator can also repeat sequences such as strings, lists, and tuples.

---

## 2. Syntax

### Syntax

```python
operand1 * operand2
```

### Explanation

- `operand1` → First operand.
- `*` → Multiplication operator.
- `operand2` → Second operand.

The result is the product of the two operands.

---

## 3. Multiplication with Integers

The operator multiplies integer values.

**Example**

```python
a = 15
b = 6

print(a * b)
```

### Output

```text
90
```

---

## 4. Multiplication with Floating-Point Numbers

The Multiplication operator also works with decimal numbers.

**Example**

```python
a = 12.5
b = 4

print(a * b)
```

### Output

```text
50.0
```

---

## 5. Multiplication with Complex Numbers

Complex numbers can also be multiplied.

**Example**

```python
a = 2 + 3j
b = 4 + 5j

print(a * b)
```

### Output

```text
(-7+22j)
```

---

## 6. Sequence Repetition

The `*` operator can repeat sequences.

### String Repetition

```python
text = "Hi "

print(text * 3)
```

### Output

```text
Hi Hi Hi
```

### List Repetition

```python
numbers = [1, 2]

print(numbers * 3)
```

### Output

```text
[1, 2, 1, 2, 1, 2]
```

### Tuple Repetition

```python
colors = ("Red", "Blue")

print(colors * 2)
```

### Output

```text
('Red', 'Blue', 'Red', 'Blue')
```

## 9. Common Mistakes

Avoid these common mistakes while using the Multiplication operator.

### Multiplying Incompatible Data Types

**Incorrect**

```python

```

### Output

```text
TypeError
```

### Correct

```python
print("10" * 2)
```

### Output

```text
1010
```
