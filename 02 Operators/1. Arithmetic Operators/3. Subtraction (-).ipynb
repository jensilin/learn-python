# ➖ Subtraction Operator (`-`)

The **Subtraction (`-`) operator** is an arithmetic operator in Python used to **subtract one numeric value from another**. It calculates the difference between two operands and is commonly used in mathematical calculations, finance, statistics, scientific computing, and everyday programming.

The subtraction operator works with **integers (`int`)**, **floating-point numbers (`float`)**, and **complex numbers (`complex`)**.

# 📖 Topics Covered

## 1. Introduction to the Subtraction Operator

The Subtraction operator (`-`) is used to find the **difference** between two numbers.

It subtracts the second operand from the first operand.

**Example**

```python
x = 50
y = 15

print(x - y)
```

### Output

```text
35
```

---

## 2. Syntax

### Syntax

```python
operand1 - operand2
```

### Explanation

- `operand1` → The value from which another value is subtracted.
- `-` → Subtraction operator.
- `operand2` → The value to subtract.

The result is the difference between the two operands.

---

## 3. Subtraction with Integers

The Subtraction operator works with integer values.

**Example**

```python
a = 100
b = 35

print(a - b)
```

### Output

```text
65
```

---

## 4. Subtraction with Floating-Point Numbers

The operator also works with decimal numbers.

**Example**

```python
a = 25.8
b = 10.4

print(a - b)
```

### Output

```text
15.4
```

---

## 5. Subtraction with Complex Numbers

Complex numbers can also be subtracted.

**Example**

```python
a = 8 + 5j
b = 3 + 2j

print(a - b)
```

### Output

```text
(5+3j)
```

---

## 6. Unary Minus Operator

The minus (`-`) symbol can also be used as a **unary operator** to change the sign of a number.

**Example**

```python
number = 15

print(-number)
```

### Output

```text
-15
```

## 9. Common Mistakes

Avoid these common mistakes while using the Subtraction operator.

### Using Strings Instead of Numbers

**Incorrect**

```python
print("20" - "10")
```

### Output

```text
TypeError
```

### Correct

```python
print(int("20") - int("10"))
```

### Output

```text
10
```
