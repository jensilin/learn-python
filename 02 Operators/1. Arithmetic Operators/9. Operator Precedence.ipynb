# 📊 Operator Precedence

**Operator Precedence** determines the **order in which Python evaluates operators in an expression**.

When an expression contains multiple operators, Python follows a predefined order of precedence to decide which operation should be performed first. Understanding operator precedence helps you write correct expressions and avoid unexpected results.

When operators have the same precedence, Python usually evaluates them **from left to right** (except for exponentiation, which is evaluated from **right to left**).

# 📖 Topics Covered

## 1. Introduction to Operator Precedence

When an expression contains multiple operators, Python does **not** evaluate them from left to right automatically.

Instead, it follows predefined precedence rules.

**Example**

```python
print(10 + 5 * 2)
```

### Output

```text
20
```

**Explanation**

```text
10 + (5 × 2)
= 10 + 10
= 20
```

---

## 2. Why Operator Precedence Matters

Without understanding precedence, expressions may produce unexpected results.

**Example**

```python
print(20 - 6 / 2)
```

### Output

```text
17.0
```

**Explanation**

```text
20 - (6 / 2)
= 20 - 3
= 17.0
```

---

## 3. Operator Precedence Table

The following table shows common Python operators from **highest** to **lowest** precedence.

| Precedence | Operator | Description |
|------------|----------|-------------|
| Highest | `()` | Parentheses |
| | `**` | Exponentiation |
| | `+x`, `-x`, `~x` | Unary operators |
| | `*`, `/`, `//`, `%` | Multiplication, Division, Floor Division, Modulus |
| | `+`, `-` | Addition, Subtraction |
| | `<<`, `>>` | Bitwise Shift |
| | `&` | Bitwise AND |
| | `^` | Bitwise XOR |
| | `\|` | Bitwise OR |
| | `==`, `!=`, `>`, `<`, `>=`, `<=`, `is`, `is not`, `in`, `not in` | Comparison, Identity, Membership |
| | `not` | Logical NOT |
| | `and` | Logical AND |
| Lowest | `or` | Logical OR |

## 4. Operator Associativity

Associativity determines the evaluation order when operators have the same precedence.

### Left-to-Right Associativity

Most operators follow left-to-right evaluation.

**Example**

```python
print(20 - 5 - 3)
```

### Output

```text
12
```

**Explanation**

```text
(20 - 5) - 3
= 15 - 3
= 12
```

---

### Right-to-Left Associativity

The Exponentiation operator (`**`) is evaluated from right to left.

**Example**

```python
print(2 ** 3 ** 2)
```

### Output

```text
512
```

**Explanation**

```text
2 ** (3 ** 2)
= 2 ** 9
= 512
```

---

## 5. Using Parentheses

Parentheses have the **highest precedence** and make expressions easier to understand.

**Example**

```python
print((5 + 2) * 3)
```

### Output

```text
21
```

Without parentheses:

```python
print(5 + 2 * 3)
```

### Output

```text
11
```

## 8. Common Mistakes

Avoid these common mistakes when working with operator precedence.

### Assuming Left-to-Right Evaluation

**Incorrect Assumption**

```python
print(5 + 10 * 2)
```

Expected:

```text
30
```

Actual Output

```text
25
```

---

### Forgetting Parentheses

**Example**

```python
print(100 / 5 * 2)
```

### Output

```text
40.0
```

If you intended:

```python
print(100 / (5 * 2))
```

### Output

```text
10.0
```

---

### Confusing `^` with `**`

```python
print(2 ^ 3)
```

### Output

```text
1
```

The correct exponentiation operator is:

```python
print(2 ** 3)
```

### Output

```text
8
```

---

## 9. Best Practices

Follow these recommended practices when writing expressions.

- Use parentheses to improve readability.
- Do not rely entirely on operator precedence for complex expressions.
- Break long expressions into smaller parts.
- Use meaningful variable names.
- Remember that `**` is evaluated from right to left.
