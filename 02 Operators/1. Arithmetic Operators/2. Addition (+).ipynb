# ➕ Addition Operator (`+`)

The **Addition (`+`) operator** is an arithmetic operator in Python used to **add two or more values**. It is one of the most commonly used operators and works with various data types, including numbers and strings.

For numeric data types, it performs mathematical addition. For strings, it performs concatenation (joining strings).

# 📖 Topics Covered

## 1. Introduction to the Addition Operator

The Addition operator (`+`) is used to combine two operands.

Depending on the data type:

- Adds numeric values.
- Concatenates strings.
- Combines lists.

---

## 2. Syntax

### Syntax

```python
operand1 + operand2
```

### Explanation

- `operand1` → First value
- `+` → Addition operator
- `operand2` → Second value

The result depends on the type of operands.

---

## 3. Addition with Integers

The Addition operator adds integer values.

**Example**

```python
a = 25
b = 15

print(a + b)
```

### Output

```text
40
```

---

## 4. Addition with Floating-Point Numbers

The operator also works with decimal numbers.

**Example**

```python
a = 10.5
b = 4.2

print(a + b)
```

### Output

```text
14.7
```

---

## 5. Addition with Complex Numbers

Complex numbers can also be added.

**Example**

```python
a = 2 + 3j
b = 4 + 5j

print(a + b)
```

### Output

```text
(6+8j)
```

---

## 6. String Concatenation

When both operands are strings, the `+` operator joins them together.

**Example**

```python
first = "Hello"
second = " Python"

print(first + second)
```

### Output

```text
Hello Python
```

---

## 7. List Concatenation

The `+` operator combines two lists.

**Example**

```python
list1 = [1, 2]
list2 = [3, 4]

print(list1 + list2)
```

### Output

```text
[1, 2, 3, 4]
```

## 10. Common Mistakes

Avoid these common mistakes while using the Addition operator.

### Mixing Different Data Types

**Incorrect**

```python
print(10 + "20")
```

### Output

```text
TypeError
```

### Correct

```python
print(10 + int("20"))
```

### Output

```text
30
```

<h5>Note</h5>

<ul>
    <li>The <code>+</code> operator does <b>not</b> support <code>set</code> and <code>dictionary</code> data types.</li>
    <li>The <code>+</code> operator cannot be used to concatenate <b>different collection data types</b>. Both operands must belong to the <b>same collection type</b>.</li>
</ul>

<h5>Example</h5>

```python

```
