# ⌊⌋ Floor Division Operator (`//`)

The **Floor Division (`//`) operator** is an arithmetic operator in Python used to **divide one number by another and return the largest whole number less than or equal to the result**.

Unlike the Division (`/`) operator, which returns a floating-point result, the Floor Division operator **discards the fractional part** of the quotient by rounding the result down (towards negative infinity).

The Floor Division operator works with **integers (`int`)** and **floating-point numbers (`float`)**.

# 📖 Topics Covered

## 1. Introduction to the Floor Division Operator

The Floor Division operator (`//`) divides two operands and returns the **floor value** of the quotient.

Instead of returning the decimal part, Python rounds the result **down** to the nearest whole number.

**Example**

```python
print(20 // 6)
```

### Output

```text
3
```

---

## 2. Syntax

### Syntax

```python
operand1 // operand2
```

### Explanation

- `operand1` → Dividend (number being divided).
- `//` → Floor Division operator.
- `operand2` → Divisor (number used for division).

The result is the floor value of the quotient.

---

## 3. Floor Division with Integers

The operator works with integer values.

**Example**

```python
a = 22
b = 4

print(a // b)
```

### Output

```text
5
```

---

## 4. Floor Division with Floating-Point Numbers

The operator also works with decimal numbers.

**Example**

```python
a = 15.5
b = 2

print(a // b)
```

### Output

```text
7.0
```

> **Note:** If either operand is a `float`, the result is also returned as a `float`.

---

## 5. Floor Division with Negative Numbers

For negative numbers, Python rounds the result **down toward negative infinity**, not toward zero.

**Example**

```python
print(-17 // 5)
```

### Output

```text
-4
```

**Explanation**

```text
-17 / 5 = -3.4

Floor Value = -4
```

---

## 6. Division vs Floor Division

Understanding the difference between `/` and `//` is important.

| Expression | Result |
|------------|--------|
| `17 / 5` | `3.4` |
| `17 // 5` | `3` |
| `17.0 // 5` | `3.0` |
| `-17 // 5` | `-4` |

## 8. Common Use Cases

The Floor Division operator is commonly used in:

- Inventory management
- Resource allocation
- Page numbering
- Batch processing
- Ticket booking systems
- Manufacturing software
- Data processing

**Example**

```python
students = 47
students_per_bus = 10

buses = students // students_per_bus

print(buses)
```

### Output

```text
4
```

> **Note:** This calculates the number of **completely filled buses**. One additional bus would still be required for the remaining students.

## 9. Common Mistakes

Avoid these common mistakes while using the Floor Division operator.

### Confusing `//` with `/`

**Example**

```python
print(10 / 3)
print(10 // 3)
```

### Output

```text
3.3333333333333335
3
```

---

### Ignoring Negative Numbers

**Example**

```python
print(-10 // 3)
```

### Output

```text
-4
```

> **Why?** Python rounds the result **down** (toward negative infinity), not toward zero.

---

### Dividing by Zero

**Incorrect**

```python
print(15 // 0)
```

### Output

```text
ZeroDivisionError: integer division or modulo by zero
```

---

## 10. Best Practices

Follow these recommended practices when using the Floor Division operator.

- Use `//` when only the whole-number quotient is required.
- Use `/` when decimal precision is needed.
- Be careful when working with negative numbers.
- Always check that the divisor is not zero.
- Use meaningful variable names for readability.
