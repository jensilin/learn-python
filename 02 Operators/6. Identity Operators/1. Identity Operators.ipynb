# 🆔 Identity Operators in Python

Identity operators are used to **compare the identity of two objects**. Instead of comparing the values stored in variables, they check whether two variables **refer to the same object in memory**.

Python provides **two identity operators**:

- `is`
- `is not`

Identity operators are commonly used when working with **objects**, **lists**, **dictionaries**, **classes**, and checking whether a variable is `None`.

# 📖 Topics Covered

## 1. Introduction to Identity Operators

Identity operators compare whether **two variables point to the same object in memory**.

They do **not** compare the values of the objects.

Example:

```python
a = [10, 20]
b = a

print(a is b)
```

### Output

```text
True
```

Both variables reference the **same list object**.

---

## 2. Identity vs Equality

Many beginners confuse identity operators with comparison operators.

### Equality (`==`)

Checks whether the values are equal.

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)
```

### Output

```text
True
```

---

### Identity (`is`)

Checks whether both variables refer to the same object.

```python
print(a is b)
```

### Output

```text
False
```

Although both lists contain the same values, they are stored as **different objects** in memory.

---

## 3. Types of Identity Operators

Python provides two identity operators.

| Operator | Name | Description |
|----------|------|-------------|
| `is` | Identity Operator | Returns `True` if both variables refer to the same object |
| `is not` | Identity Not Operator | Returns `True` if both variables refer to different objects |

---

## 4. Memory Representation

Example:

```python
a = [1, 2]
b = a
c = [1, 2]
```

Memory illustration:

```text
a ─────┐
       │
b ─────┘──► [1, 2]

c ─────────► [1, 2]
```

Here:

- `a` and `b` refer to the same object.
- `c` refers to a different object.

---

## 5. Using the `id()` Function

The `id()` function returns the memory identity of an object.

```python
a = [1, 2, 3]
b = a

print(id(a))
print(id(b))
```

### Sample Output

```text
140243451600128
140243451600128
```

Both objects have the same identity.

---

## 6. Identity Operators with Immutable Objects

Python may reuse immutable objects such as small integers and short strings.

```python
a = 10
b = 10

print(a is b)
```

### Output

```text
True
```

Another example:

```python
x = "Python"
y = "Python"

print(x is y)
```

### Output

```text
True
```

> **Note:** Python may internally reuse immutable objects (known as interning). Do not rely on this behavior when comparing values. Use `==` for value comparison.

## 7. Practical Examples

### Example 1: Shared List

```python
list1 = [1, 2, 3]
list2 = list1

print(list1 is list2)
```

### Output

```text
True
```

---

### Example 2: Separate Lists

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]

print(list1 is list2)
```

### Output

```text
False
```

---

### Example 3: Compare Values

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]

print(list1 == list2)
```

### Output

```text
True
```

---

### Example 4: Checking for None

```python
name = None

print(name is None)
```

### Output

```text
True
```

This is the recommended way to check for `None`.

Identity operators are commonly used in:

- Memory management

## 9. Common Mistakes

Avoid these common mistakes while using identity operators.

### Using `is` Instead of `==`

Incorrect

```python
a = [1, 2]
b = [1, 2]

print(a is b)
```

Output

```text
False
```

Correct

```python
print(a == b)
```

Output

```text
True
```

Use:

- `==` → Compare values.
- `is` → Compare object identities.

---

### Comparing Strings with `is`

Incorrect

```python
name = "Python"

if name is "Python":
    print("Matched")
```

Preferred

```python
if name == "Python":
    print("Matched")
```

---

### Relying on Object Interning

Sometimes this works:

```python
a = 100
b = 100

print(a is b)
```

But for larger or dynamically created objects, the result may differ depending on Python's implementation.

Always use `==` for value comparison.

---

## 10. Best Practices

Follow these recommended practices when using identity operators.

- Use `is` and `is not` only to compare object identity.
- Use `==` when comparing values.
- Always use `is None` and `is not None` for checking `None`.
- Avoid relying on object interning.
- Use `id()` when learning about object identities.
- Write clear and readable comparisons.
