# 🆔 Identity Operator (`is`)

The **`is` operator** is an **identity operator** in Python that checks whether **two variables refer to the same object in memory**.

Unlike the equality operator (`==`), which compares values, the `is` operator compares the **identity (memory reference)** of two objects.

The `is` operator is commonly used when working with **objects**, **lists**, **classes**, and for checking whether a variable is **`None`**.

# 📖 Topics Covered

## 1. Introduction to the `is` Operator

The `is` operator checks whether two variables point to the **same object** in memory.

It returns:

- `True` if both variables reference the same object.
- `False` otherwise.

Example:

```python
a = [10, 20]
b = a

print(a is b)
```

### Output

```text
True
```

Both variables reference the same list object.

---

## 2. Syntax

### Syntax

```python
object1 is object2
```

### Explanation

- `object1` → First object.
- `object2` → Second object.
- `is` → Returns `True` if both objects are identical.

---

## 3. Difference Between `is` and `==`

The `is` operator compares **object identity**, while the `==` operator compares **values**.

Example:

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)
print(a is b)
```

### Output

```text
True
False
```

Explanation:

- `==` compares the contents of the lists.
- `is` compares whether both variables refer to the same object.

---

## 4. Using the `id()` Function

The `id()` function returns the unique identity (memory address) of an object during its lifetime.

```python
a = [1, 2]
b = a

print(id(a))
print(id(b))
```

### Sample Output

```text
140495124560512
140495124560512
```

Since both IDs are the same, `a is b` returns `True`.

---

## 5. Working with Different Objects

```python
a = [1, 2]
b = [1, 2]

print(a is b)
```

### Output

```text
False
```

Although both lists contain the same values, they are different objects in memory.

---

## 6. Working with Immutable Objects

Python may reuse immutable objects such as small integers and short strings.

```python
a = 100
b = 100

print(a is b)
```

### Output

```text
True
```

Another example:

```python
x = "Python"
y = "Python"

print(x is y)
```

### Output

```text
True
```

> **Note:** Python may internally reuse immutable objects (known as interning). Do not rely on this behavior for value comparisons.

---

## 7. Checking for `None`

The recommended way to check whether a variable is `None` is by using the `is` operator.

```python
name = None

print(name is None)
```

### Output

```text
True
```

Another example:

```python
value = 10

print(value is None)
```

### Output

```text
False
```

---

## 8. Practical Examples

### Example 1: Shared List

```python
list1 = [10, 20, 30]
list2 = list1

print(list1 is list2)
```

### Output

```text
True
```

---

### Example 2: Separate Lists

```python
list1 = [10, 20, 30]
list2 = [10, 20, 30]

print(list1 is list2)
```

### Output

```text
False
```

---

### Example 3: Shared Dictionary

```python
student1 = {"name": "John"}
student2 = student1

print(student1 is student2)
```

### Output

```text
True
```

---

### Example 4: Checking for `None`

```python
result = None

if result is None:
    print("No result found")
```

### Output

```text
No result found
```

## 9. Common Use Cases

The `is` operator is commonly used in:

- Checking whether a variable is `None`
- Comparing object references
- Singleton objects
- Object-oriented programming
- Debugging object identity
- Memory reference checking
- Class instance comparisons

---

## 10. Common Mistakes

Avoid these common mistakes while using the `is` operator.

### Using `is` Instead of `==`

Incorrect

```python
a = [1, 2]
b = [1, 2]

print(a is b)
```

### Output

```text
False
```

Correct

```python
print(a == b)
```

### Output

```text
True
```

Use:

- `is` → Compare object identity.
- `==` → Compare values.

---

### Comparing Strings with `is`

Incorrect

```python
language = "Python"

if language is "Python":
    print("Matched")
```

Preferred

```python
if language == "Python":
    print("Matched")
```

---

### Relying on Object Interning

Sometimes this works:

```python
a = 50
b = 50

print(a is b)
```

However, for larger or dynamically created objects, the result may differ depending on Python's implementation.

Always use `==` when comparing values.

---

### Assuming `id()` Values Are Permanent

The value returned by `id()` is unique only during the object's lifetime. It should not be stored or relied upon across different program executions.

---

## 11. Best Practices

Follow these recommended practices when using the `is` operator.

- Use `is` only for comparing object identity.
- Use `==` for comparing values.
- Always use `is None` instead of `== None`.
- Avoid relying on Python's object interning.
- Use `id()` for learning and debugging object identity.
- Keep identity comparisons clear and readable.
