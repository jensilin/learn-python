# 🚫 Identity NOT Operator (`is not`)

The **`is not` operator** is an **identity operator** in Python that checks whether **two variables refer to different objects in memory**.

Unlike the inequality operator (`!=`), which compares values, the `is not` operator compares the **identity (memory reference)** of two objects.

The `is not` operator is commonly used to verify that an object is **not `None`**, compare object references, and work with Python objects.

# 📖 Topics Covered

## 1. Introduction to the `is not` Operator

The `is not` operator checks whether two variables refer to **different objects** in memory.

It returns:

- `True` if both variables refer to different objects.
- `False` if both variables refer to the same object.

Example:

```python
a = [10, 20]
b = [10, 20]

print(a is not b)
```

### Output

```text
True
```

Although both lists contain the same values, they are different objects.

---

## 2. Syntax

### Syntax

```python
object1 is not object2
```

### Explanation

- `object1` → First object.
- `object2` → Second object.
- `is not` → Returns `True` if both objects are different.

---

## 3. Difference Between `is not` and `!=`

The `is not` operator compares **object identity**, while the `!=` operator compares **values**.

Example:

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a != b)
print(a is not b)
```

### Output

```text
False
True
```

Explanation:

- `!=` compares the contents of the lists.
- `is not` compares whether both variables refer to different objects.

---

## 4. Using the `id()` Function

The `id()` function returns the unique identity (memory reference) of an object during its lifetime.

```python
a = [1, 2]
b = [1, 2]

print(id(a))
print(id(b))
```

### Sample Output

```text
140495124560512
140495124560768
```

Since both IDs are different, `a is not b` returns `True`.

---

## 5. Working with the Same Object

```python
a = [1, 2]
b = a

print(a is not b)
```

### Output

```text
False
```

Both variables reference the same object.

---

## 6. Working with Immutable Objects

Python may internally reuse immutable objects such as small integers and short strings.

```python
a = 100
b = 100

print(a is not b)
```

### Output

```text
False
```

Another example:

```python
x = "Python"
y = "Python"

print(x is not y)
```

### Output

```text
False
```

> **Note:** Python may internally reuse immutable objects (known as interning). Do not rely on this behavior for value comparisons.

## 7. Checking for `None`

The recommended way to check that a variable is **not `None`** is by using the `is not` operator.

```python
name = "Alice"

print(name is not None)
```

### Output

```text
True
```

Another example:

```python
value = None

print(value is not None)
```

### Output

```text
False
```

## 9. Common Use Cases

The `is not` operator is commonly used in:

- Comparing object references
- Singleton objects
- Object-oriented programming
- Debugging object identity
- Memory reference checking
- Conditional validation

## 10. Common Mistakes

Avoid these common mistakes while using the `is not` operator.

### Using `is not` Instead of `!=`

Incorrect

```python
a = [1, 2]
b = [3, 4]

print(a is not b)
```

This checks whether the objects are different, **not** whether their values are different.

Correct

```python
print(a != b)
```

Use:

- `is not` → Compare object identity.
- `!=` → Compare values.

---

### Comparing Strings with `is not`

Incorrect

```python
language = "Java"

if language is not "Python":
    print("Different")
```

Preferred

```python
if language != "Python":
    print("Different")
```

---

### Relying on Object Interning

Sometimes this works:

```python
a = 50
b = 50

print(a is not b)
```

However, for larger or dynamically created objects, the result may differ depending on Python's implementation.

Always use `!=` when comparing values.

---

### Using `!= None`

Less preferred

```python
if value != None:
    print("Available")
```

Recommended

```python
if value is not None:
    print("Available")
```

---

## 11. Best Practices

Follow these recommended practices when using the `is not` operator.

- Use `is not` only for comparing object identity.
- Use `!=` for comparing values.
- Always use `is not None` instead of `!= None`.
- Avoid relying on Python's object interning.
- Use `id()` for learning and debugging object identity.
- Keep identity comparisons simple and readable.
