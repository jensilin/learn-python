<h2>Control Statements</h4>

<p>
Control statements are used to <b>control the flow of execution</b> of a program. They determine the order in which statements are executed based on conditions, repeated execution, or by transferring control within the program.
</p>

<p>
Python provides the following types of control statements:
</p>

<h4>Types of Control Statements</h5>

1. **Selection (Decision-Making) Statements**
   - `if` Statement
   - `if...else` Statement
   - `if...elif...else` Statement
   - `Nested if` Statement

2. **Iterative (Looping) Statements**
   - `while` Loop
   - `for` Loop
   - `Nested Loops`

3. **Jump (Branching) Statements**
   - `break` Statement
   - `continue` Statement
   - `pass` Statement
