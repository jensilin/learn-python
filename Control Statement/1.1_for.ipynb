<h4><code>for</code> Loop</h4>

<p>
The <code>for</code> loop is an <b>iterative (looping)</b> statement used to iterate over the elements of a sequence or any iterable object. It executes a block of code once for each element in the iterable.
</p>

<h5>Types of <code>for</code> Loop</h5>

1. **Simple `for` Loop**
2. **Nested `for` Loop**
3. **`for...else` Loop**

---

<h5>1. Simple <code>for</code> Loop</h5>

<p>
A simple <code>for</code> loop iterates through each element of an iterable and executes the loop body once for every element.
</p>

<h6>Syntax</h6>

```python
for variable in iterable:
    statement(s)
```

<h6>Example</h6>

```python
for ch in "Python":
    print(f"{ch}")
```

<h6>Output</h6>

```text
P
y
t
h
o
n
```

---

<h5>2. Nested <code>for</code> Loop</h5>

<p>
A nested <code>for</code> loop is a <code>for</code> loop placed inside another <code>for</code> loop.
</p>

<h6>Syntax</h6>

```python
for variable1 in iterable1:
    for variable2 in iterable2:
        statement(s)
```

<h6>Example</h6>

```python
for i in range(1, 3):
    for j in range(1, 4):
        print(f"i = {i}, j = {j}")
```

<h6>Output</h6>

```text
i = 1, j = 1
i = 1, j = 2
i = 1, j = 3
i = 2, j = 1
i = 2, j = 2
i = 2, j = 3
```

---

<h5>3. <code>for...else</code> Loop</h5>

<p>
The <code>else</code> block is executed only if the <code>for</code> loop completes normally without encountering a <code>break</code> statement.
</p>

<h6>Syntax</h6>

```python
for variable in iterable:
    statement(s)
else:
    statement(s)
```

<h6>Example</h6>

```python
for i in range(1, 4):
    print(f"{i}")
else:
    print("Loop completed successfully.")
```

<h6>Output</h6>

```text
1
2
3
Loop completed successfully.
```

---

<h5>Note</h5>

<ul>
    <li>The <code>for</code> loop is generally used when the number of iterations is known.</li>
    <li>It can iterate over strings, lists, tuples, sets, dictionaries, ranges, and other iterable objects.</li>
    <li>The <code>else</code> block is executed only if the loop completes normally and is skipped if a <code>break</code> statement is encountered.</li>
    <li>The <code>continue</code> statement does not prevent the <code>else</code> block from executing.</li>
    <li>Proper indentation is mandatory for all loop blocks.</li>
</ul>
