<h4><code>if...else</code> Statement</h4>

<p>
The <code>if...else</code> statement is a <b>selection (decision-making)</b> statement used to execute one block of code when the condition is <code>True</code> and another block of code when the condition is <code>False</code>.
</p>

<h5>Syntax</h5>

```python
if condition:
    statement(s)
else:
    statement(s)
```

<h5>Flow of Execution</h5>

<ul>
    <li>The condition is evaluated first.</li>
    <li>If the condition is <code>True</code>, the statements inside the <code>if</code> block are executed.</li>
    <li>If the condition is <code>False</code>, the statements inside the <code>else</code> block are executed.</li>
    <li>Only one of the two blocks is executed during program execution.</li>
</ul>

<h5>Example</h5>

```python
age = 16

if age >= 18:
    print(f"You are eligible to vote.")
else:
    print(f"You are not eligible to vote.")
```

<h5>Output</h5>

```text
You are not eligible to vote.
```

<h5>Note</h5>

<ul>
    <li>The <code>else</code> block does not have a condition.</li>
    <li>The <code>else</code> block is executed only when the <code>if</code> condition evaluates to <code>False</code>.</li>
    <li>A colon (<code>:</code>) is mandatory after both <code>if</code> and <code>else</code>.</li>
    <li>Proper indentation is required for both blocks.</li>
</ul>

````python
#WAP to check whether the number is greater or lesser
fn = int(input("Enter the first number: "))
sn = int(input("Enter the second number: "))

if fn>sn :
    print(f'{fn} is greater than {sn}.')
else:
    print(f'{fn} is lesser than {sn}.')
````

````python
#WAP to check whether a Number is Even or Odd
num = int(input("Enter the number: "))

if num%2 == 0:
    print(f'{num} is an Even number.')
else:
    print(f'{num} is an Odd number.')
````

````python
#WAP to Check Whether a Character is a Vowel or not
ch = input('Enter the character : ')

if ch.lower() in 'aeiou':
    print(f'{ch} is a Vowel')
else:
    print(f'{ch} is not a Vowel')
````
