<h4><code>while</code> Loop</h4>

<p>
The <code>while</code> loop is an <b>iterative (looping)</b> statement used to execute a block of code <b>repeatedly</b> as long as the specified condition evaluates to <code>True</code>. The loop terminates when the condition becomes <code>False</code>.
</p>

<h5>Syntax</h5>

```python
while condition:
    statement(s)
```

<h5>Flow of Execution</h5>

<ul>
    <li>The condition is evaluated before each iteration.</li>
    <li>If the condition is <code>True</code>, the statements inside the <code>while</code> loop are executed.</li>
    <li>After executing the loop body, the condition is evaluated again.</li>
    <li>The loop continues until the condition becomes <code>False</code>.</li>
    <li>Once the condition is <code>False</code>, control moves to the next statement after the loop.</li>
</ul>

<h5>Example</h5>

```python
count = 1

while count <= 5:
    print(f"Count = {count}")
    count += 1
```

<h5>Output</h5>

```text
Count = 1
Count = 2
Count = 3
Count = 4
Count = 5
```

<h5>Note</h5>

<ul>
    <li>The <code>while</code> loop is an <b>entry-controlled loop</b> because the condition is checked before executing the loop body.</li>
    <li>If the condition is initially <code>False</code>, the loop body is <b>not executed</b>.</li>
    <li>The loop condition should be updated inside the loop whenever required to avoid an <b>infinite loop</b>.</li>
    <li>Proper indentation is mandatory for the statements inside the <code>while</code> loop.</li>
</ul>

````python
#While loop example
i = 0
while i<=10:
    print(i)
    i+=1
````
