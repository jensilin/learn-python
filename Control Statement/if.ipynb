<h4><code>if</code> Statement</h4>

<p>
The <code>if</code> statement is a <b>selection (decision-making)</b> statement used to execute a block of code only when a specified condition evaluates to <code>True</code>. If the condition is <code>False</code>, the block of code is skipped.
</p>

<h5>Syntax</h5>

```python
if condition:
    statement(s)
```

<h5>Flow of Execution</h5>

<ul>
    <li>The condition is evaluated first.</li>
    <li>If the condition is <code>True</code>, the statements inside the <code>if</code> block are executed.</li>
    <li>If the condition is <code>False</code>, the <code>if</code> block is skipped, and execution continues with the next statement after the block.</li>
</ul>

<h5>Example</h5>

```python
age = 18

if age >= 18:
    print("Eligible to vote")
```

<h5>Output</h5>

```text
Eligible to vote
```

<h5>Note</h5>

<ul>
    <li>The condition must evaluate to either <code>True</code> or <code>False</code>.</li>
    <li>A colon (<code>:</code>) is mandatory after the condition.</li>
    <li>The statements inside the <code>if</code> block must be properly indented.</li>
</ul>

````python
#WAP to find the gretest of two numbers

a = 10
b = 5

if a>b :
    print("a is greater than b")
````

<h4>Membership Operators</h4>

<p>
Membership operators are used to check whether a value is <b>present</b> or <b>not present</b> in a sequence or collection. They return the result in the form of <code>True</code> or <code>False</code>.
</p>

<p>
Python provides two membership operators:
</p>

1. <code>in</code> Operator
2. <code>not in</code> Operator

---

<h4>a) <code>in</code> Operator</h4>

<p>
The <code>in</code> operator is used to check whether a specified value is <b>present</b> in a sequence or collection. It returns <code>True</code> if the value is found; otherwise, it returns <code>False</code>.
</p>

<h5>Syntax</h5>

```python
value in collection
```

<h5>Example</h5>

```python
print("a" in "Python")
print(3 in [1, 2, 3, 4])
print("apple" in ("apple", "banana", "orange"))
```

<h5>Output</h5>

```text
False
True
True
```

<h5>Note</h5>

<ul>
    <li>The <code>in</code> operator can be used with <code>str</code>, <code>list</code>, <code>tuple</code>, <code>set</code>, <code>dict</code>, and other iterable objects.</li>
    <li>For dictionaries, the <code>in</code> operator checks only the <b>keys</b>, not the values.</li>
</ul>

---

<h4>b) <code>not in</code> Operator</h4>

<p>
The <code>not in</code> operator is used to check whether a specified value is <b>not present</b> in a sequence or collection. It returns <code>True</code> if the value is not found; otherwise, it returns <code>False</code>.
</p>

<h5>Syntax</h5>

```python
value not in collection
```

<h5>Example</h5>

```python
print("z" not in "Python")
print(10 not in [1, 2, 3, 4])
print("grape" not in ("apple", "banana", "orange"))
```

<h5>Output</h5>

```text
True
True
True
```

<h5>Note</h5>

<ul>
    <li>The <code>not in</code> operator is the opposite of the <code>in</code> operator.</li>
    <li>For dictionaries, the <code>not in</code> operator checks only the <b>keys</b>.</li>
</ul>

````python
#In operator example
print('i' in 'Jensilin')
````

````python
#WAP to Check Whether a Character is a Vowel
ch = input('Enter the character : ')

if ch.lower() in 'aeiou':
    print(f'{ch} is a Vowel')
````

````python
#Not in operator example

print('u' not in 'Jensilin')
````

````python
#WAP to Check Whether a Character is not a Vowel

ch = input('Enter the character : ')

if ch.lower() not in 'aeiou':
    print(f'{ch} is not a Vowel')
````
