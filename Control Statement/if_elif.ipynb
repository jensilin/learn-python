<h4><code>if...elif...else</code> Statement</h4>

<p>
The <code>if...elif...else</code> statement is a <b>selection (decision-making)</b> statement used to evaluate <b>multiple conditions</b>. It executes the block of code corresponding to the <b>first condition</b> that evaluates to <code>True</code>. If none of the conditions are <code>True</code>, the <code>else</code> block is executed.
</p>

<h5>Syntax</h5>

```python
if condition1:
    statement(s)
elif condition2:
    statement(s)
elif condition3:
    statement(s)
else:
    statement(s)
```

<h5>Flow of Execution</h5>

<ul>
    <li>The <code>if</code> condition is evaluated first.</li>
    <li>If the <code>if</code> condition is <code>True</code>, its corresponding block is executed, and the remaining conditions are skipped.</li>
    <li>If the <code>if</code> condition is <code>False</code>, the <code>elif</code> conditions are evaluated one by one.</li>
    <li>The block corresponding to the <b>first</b> <code>True</code> condition is executed.</li>
    <li>If none of the conditions evaluate to <code>True</code>, the <code>else</code> block is executed.</li>
</ul>

<h5>Example</h5>

```python
marks = 78

if marks >= 90:
    print(f"Grade A")
elif marks >= 75:
    print(f"Grade B")
elif marks >= 50:
    print(f"Grade C")
else:
    print(f"Grade F")
```

<h5>Output</h5>

```text
Grade B
```

<h5>Note</h5>

<ul>
    <li>There can be only <b>one</b> <code>if</code> statement in an <code>if...elif...else</code> structure.</li>
    <li>There can be <b>multiple</b> <code>elif</code> statements.</li>
    <li>The <code>else</code> block is optional and is executed only if none of the conditions evaluate to <code>True</code>.</li>
    <li>Only <b>one</b> block of code is executed during program execution.</li>
    <li>A colon (<code>:</code>) is mandatory after <code>if</code>, <code>elif</code>, and <code>else</code>.</li>
    <li>Proper indentation must be maintained for all blocks.</li>
</ul>
