<h4><code>while</code> Loop</h4>

<p>
The <code>while</code> loop is an <b>iterative (looping)</b> statement used to execute a block of code <b>repeatedly</b> as long as the specified condition evaluates to <code>True</code>. The loop terminates when the condition becomes <code>False</code>.
</p>

<h5>Types of <code>while</code> Loop</h5>

1. **Simple `while` Loop**
2. **Nested `while` Loop**
3. **`while...else` Loop**

---

<h5>1. Simple <code>while</code> Loop</h5>

<p>
A simple <code>while</code> loop executes a block of code repeatedly until the specified condition becomes <code>False</code>.
</p>

<h6>Syntax</h6>

```python
while condition:
    statement(s)
```

<h6>Example</h6>

```python
count = 1

while count <= 5:
    print(f"Count = {count}")
    count += 1
```

<h6>Output</h6>

```text
Count = 1
Count = 2
Count = 3
Count = 4
Count = 5
```

---

<h5>2. Nested <code>while</code> Loop</h5>

<p>
A nested <code>while</code> loop is a <code>while</code> loop inside another <code>while</code> loop.
</p>

<h6>Syntax</h6>

```python
while condition1:
    while condition2:
        statement(s)
```

<h6>Example</h6>

```python
i = 1

while i <= 2:
    j = 1
    while j <= 3:
        print(f"i = {i}, j = {j}")
        j += 1
    i += 1
```

<h6>Output</h6>

```text
i = 1, j = 1
i = 1, j = 2
i = 1, j = 3
i = 2, j = 1
i = 2, j = 2
i = 2, j = 3
```

---

<h5>3. <code>while...else</code> Loop</h5>

<p>
The <code>else</code> block is executed only if the <code>while</code> loop terminates normally without encountering a <code>break</code> statement.
</p>

<h6>Syntax</h6>

```python
while condition:
    statement(s)
else:
    statement(s)
```

<h6>Example</h6>

```python
count = 1

while count <= 3:
    print(f"{count}")
    count += 1
else:
    print("Loop completed successfully.")
```

<h6>Output</h6>

```text
1
2
3
Loop completed successfully.
```

---

<h5>Note</h5>

<ul>
    <li>The <code>while</code> loop is an <b>entry-controlled loop</b> because the condition is checked before each iteration.</li>
    <li>If the condition is initially <code>False</code>, the loop body is <b>not executed</b>.</li>
    <li>The loop condition should be updated inside the loop to avoid an <b>infinite loop</b>.</li>
    <li>The <code>else</code> block executes only when the loop completes normally and is skipped if a <code>break</code> statement is encountered.</li>
    <li>Proper indentation is mandatory for all loop blocks.</li>
</ul>

````python
#WAP to check the string is palindrome or not without slicing

a=input("Enter the String: ")
n=len(a)-1
i=0
rev=""

while n>=i:
    rev+=a[n-i]
    i+=1

if rev==a:
    print(f'{a} is a palindrome.')
else:
    print(f'{a} is not a palindrome.')
````

````python
#WAP to extract all the characters at even index and odd index into two different output collection without slicing

a=input("Enter the String: ")
i=0
even_index=""
odd_index=""

while i<len(a):
    if i%2==0:
        even_index+=a[i]
    else:
        odd_index+=a[i]
    i+=1

print(f'The even index characters are {even_index}.')
print(f'The odd index characters are {odd_index}.')
````
