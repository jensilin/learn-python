# String

## What is a String?

A **String** is a built-in Python data type used to store **text or a sequence of characters**. A string can contain letters, numbers, symbols, spaces, and special characters. In Python, strings are enclosed within **single quotes (`' '`), double quotes (`" "`), or triple quotes (`''' '''` or `""" """`)**.

## Syntax

```python
variable_name = "string_value"
```

or

```python
variable_name = 'string_value'
```

For multi-line strings:

```python
variable_name = """This is
a multi-line
string."""
```


````python
#Anything written inside double quotes or single quotes are String
print("Every expert was once a beginner. Keep coding!")
````

What is \n in Python?

\n is a newline escape character in Python. It is used to move the cursor to the next line, allowing you to print text on multiple lines with a single print() statement.

````python
#Basically I can print in multiple lines using single print statement
print("Think.\nCode.")
````

String Concatenation
What is Concatenation?

Concatenation is the process of joining two or more strings together to form a single string. In Python, the + (plus) operator is used to concatenate strings.

Syntax
string1 + string2


Note: When concatenating strings, Python combines them exactly as written. If you want a space between the words, include a space inside one of the strings (e.g., "Hello " + "World").

````python
print("hello" + "World")
````

````python
#In the above output you can see there is no space between them
#In below ways you can add space

print("Hello "+"Guys")

print("Hello"+" Guys")

print("Hello"+" "+"Guys")
````
